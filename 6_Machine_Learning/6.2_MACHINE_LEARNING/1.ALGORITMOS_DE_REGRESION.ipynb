## **M茅tricas para evaluar modelos de Regresi贸n**

Para evaluar modelos de regresi贸n, existen varias m茅tricas comunes que se utilizan para medir su desempe帽o. A continuaci贸n, te mostrar茅 cuatro m茅tricas fundamentales para evaluar modelos de regresi贸n: el error cuadr谩tico medio (MSE), el error absoluto medio (MAE), el coeficiente de determinaci贸n (R-cuadrado) y el gr谩fico de dispersi贸n de los valores reales frente a los valores predichos.

1. **Error cuadr谩tico medio (MSE):**

El MSE calcula la media de los errores cuadr谩ticos entre los valores reales y los valores predichos. Cuanto menor sea el MSE, mejor ser谩 el modelo.

```python
from sklearn.metrics import mean_squared_error

# Valores reales
y_true = [3, 5, 7, 9]

# Valores predichos
y_pred = [2.5, 5.1, 6.8, 8.9]

mse = mean_squared_error(y_true, y_pred)
print("MSE:", mse)
```

2. **Error absoluto medio (MAE):**

El MAE calcula la media de los errores absolutos entre los valores reales y los valores predichos. Al igual que el MSE, se busca minimizar este valor.

**Ejemplo:**

Puedes calcular el MAE utilizando la funci贸n `mean_absolute_error` de scikit-learn:

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_true, y_pred)
print("MAE:", mae)
```

3. **Coeficiente de determinaci贸n (R-cuadrado):**

El coeficiente de determinaci贸n, com煤nmente conocido como R-cuadrado, mide la proporci贸n de la varianza en los valores de la variable dependiente que es predecible a partir de las variables independientes. El R-cuadrado var铆a entre 0 y 1, donde 1 indica un ajuste perfecto del modelo.

**Ejemplo:**

Puedes calcular el R-cuadrado utilizando la funci贸n `r2_score de scikit-learn`:

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_true, y_pred)
print("R-cuadrado:", r2)
```

4. **Gr谩fico de dispersi贸n:**

El gr谩fico de dispersi贸n es una forma visual de evaluar el rendimiento del modelo de regresi贸n. Consiste en trazar los valores reales en el eje `y` y los valores predichos en el eje `x`. Un modelo de regresi贸n preciso mostrar谩 una l铆nea cercana a la diagonal, lo que indica una buena correspondencia entre los valores reales y los valores predichos.

**Ejemplo:**

Puedes generar el gr谩fico de dispersi贸n utilizando bibliotecas como matplotlib o seaborn:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que ya tienes definidos los arrays y_true y y_pred

# Personalizar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Tama帽o y estilo del gr谩fico
plt.figure(figsize=(10, 8))

# Crear el gr谩fico de dispersi贸n
scatter_plot = sns.scatterplot(x=y_true, y=y_pred, color='blue', marker='o', s=80, alpha=0.7)

# A帽adir una l铆nea de regresi贸n (opcional)
sns.regplot(x=y_true, y=y_pred, scatter=False, ax=scatter_plot, color='red')

# Etiquetas y t铆tulo
plt.xlabel("Valores reales", fontsize=14)
plt.ylabel("Valores predichos", fontsize=14)
plt.title("Gr谩fico de dispersi贸n", fontsize=16)

# A帽adir leyenda (opcional)
plt.legend(["Regresi贸n lineal"])

# Personalizar la cuadr铆cula (opcional)
plt.grid(True, linestyle='--', alpha=0.5)

# A帽adir l铆neas gu铆a (opcional)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Mostrar el gr谩fico
plt.show()
```

Estas m茅tricas y gr谩ficos te permiten evaluar el rendimiento de un modelo de regresi贸n y determinar su capacidad para hacer predicciones precisas.

## **Regresi贸n Lineal simple y m煤ltiple**

La regresi贸n lineal es un m茅todo de aprendizaje autom谩tico supervisado utilizado para predecir una variable continua basada en una o m谩s variables independientes. Existen dos tipos comunes de regresi贸n lineal: la regresi贸n lineal simple y la regresi贸n lineal m煤ltiple.

1. **Regresi贸n Lineal Simple:**

La regresi贸n lineal simple se utiliza cuando solo hay una variable independiente que se utiliza para predecir una variable dependiente continua. La relaci贸n entre las variables se modela como una l铆nea recta.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que contiene el salario de los empleados (`y`) en funci贸n de su experiencia laboral (`x`). Podemos usar la regresi贸n lineal simple para predecir el salario de un empleado en funci贸n de su experiencia laboral.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5])  # Experiencia laboral
y = np.array([30000, 40000, 50000, 60000, 70000])  # Salario

# Crear el modelo de regresi贸n lineal
regression = LinearRegression()

# Entrenar el modelo
regression.fit(x.reshape(-1, 1), y)

# Realizar predicciones
y_pred = regression.predict(x.reshape(-1, 1))

# Configuraci贸n de la gr谩fica
plt.figure(figsize=(8, 6))

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', label='Regresi贸n lineal')

# A帽adir etiquetas y t铆tulo
plt.xlabel('Experiencia laboral')
plt.ylabel('Salario')
plt.title('Regresi贸n lineal simple')

# A帽adir una cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.7)

# A帽adir leyenda
plt.legend()

# A帽adir texto adicional
plt.text(2.5, 60000, f'R虏 = {regression.score(x.reshape(-1, 1), y):.2f}', fontsize=12, ha='center')

# A帽adir l铆nea de identidad
plt.plot(x, x, '--', color='gray', label='L铆nea de identidad', alpha=0.5)

# Ajustar la disposici贸n
plt.tight_layout()

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, utilizamos la biblioteca scikit-learn para crear un modelo de regresi贸n lineal (`LinearRegression`). Entrenamos el modelo con los datos de experiencia laboral (`x`) y salario (`y`) y luego realizamos predicciones (`y_pred`) basadas en el modelo entrenado. Finalmente, graficamos los datos de muestra y la l铆nea de regresi贸n.

2. **Regresi贸n Lineal M煤ltiple:**

La regresi贸n lineal m煤ltiple se utiliza cuando hay m谩s de una variable independiente que se utiliza para predecir una variable dependiente continua. La relaci贸n entre las variables se modela como un plano o una hiperplano en espacios de dimensiones superiores.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que contiene el precio de las casas (`y`) en funci贸n de su tama帽o (`x1`) y el n煤mero de habitaciones (`x2`). Podemos usar la regresi贸n lineal m煤ltiple para predecir el precio de una casa en funci贸n de su tama帽o y n煤mero de habitaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from matplotlib.lines import Line2D

# Datos de ejemplo
x1 = np.array([1200, 1500, 1700, 2000, 2200])  # Tama帽o de la casa en pies cuadrados
x2 = np.array([2, 3, 4, 3, 4])  # N煤mero de habitaciones
y = np.array([200000, 250000, 300000, 280000, 350000])  # Precio de la casa

# Crear el modelo de regresi贸n lineal m煤ltiple
regression = LinearRegression()

# Combinar las variables independientes en una matriz
X = np.column_stack((x1, x2))

# Entrenar el modelo
regression.fit(X, y)

# Realizar predicciones
y_pred = regression.predict(X)

# Graficar los datos y la superficie de regresi贸n
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x1, x2, y, color='blue', label='Datos')
surface = ax.plot_trisurf(x1, x2, y_pred, color='red', alpha=0.5, label='Regresi贸n lineal')

# Configuraci贸n de la leyenda con un manejador personalizado
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Datos'),
                   Line2D([0], [0], color='w', markerfacecolor='red', markersize=10, label='Regresi贸n lineal')]
ax.legend(handles=legend_elements)

# Configuraci贸n de la gr谩fica
ax.set_xlabel('Tama帽o')
ax.set_ylabel('N煤mero de habitaciones')
ax.set_zlabel('Precio')
ax.set_title('Regresi贸n lineal m煤ltiple')

# A帽adir etiquetas a puntos espec铆ficos
for i, txt in enumerate(['Punto 1', 'Punto 2', 'Punto 3', 'Punto 4', 'Punto 5']):
    ax.text(x1[i], x2[i], y[i], txt)

# Ajustar la disposici贸n
fig.tight_layout()

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, utilizamos la biblioteca scikit-learn para crear un modelo de regresi贸n lineal (`LinearRegression`). Combinamos las variables independientes (`x1` y `x2`) en una matriz `X`. Entrenamos el modelo con los datos de tama帽o de la casa, n煤mero de habitaciones y precio de la casa, y luego realizamos predicciones (`y_pred`) basadas en el modelo entrenado. Finalmente, graficamos los datos de muestra y la superficie de regresi贸n.

Estos ejemplos te proporcionan una idea de c贸mo implementar la regresi贸n lineal simple y m煤ltiple en Python utilizando scikit-learn. Recuerda que los conjuntos de datos y las caracter铆sticas utilizadas en los ejemplos son solo ilustrativos, y puedes adaptar el c贸digo a tus propios conjuntos de datos y caracter铆sticas espec铆ficas.

## **Modelos Lasso y ridge**

Los modelos Lasso y Ridge son dos t茅cnicas de regularizaci贸n utilizadas en la regresi贸n lineal para evitar el sobreajuste y mejorar la capacidad de generalizaci贸n del modelo. Ambas t茅cnicas introducen una penalizaci贸n en la funci贸n de costo para controlar la complejidad del modelo.

El modelo Lasso utiliza la regularizaci贸n `L1`, que penaliza los coeficientes de las caracter铆sticas menos importantes, reduci茅ndolos a cero si es posible. Esto permite realizar selecci贸n autom谩tica de caracter铆sticas, ya que algunas caracter铆sticas pueden ser descartadas completamente. Por otro lado, el modelo Ridge utiliza la regularizaci贸n `L2`, que penaliza los coeficientes de las caracter铆sticas, pero no los reduce a cero. En cambio, los coeficientes se reducen proporcionalmente.

A continuaci贸n, te mostrar茅 un ejemplo en c贸digo Python para cada modelo y c贸mo se pueden utilizar junto con gr谩ficas.

**Ejemplo de modelo Lasso:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

# Datos de ejemplo
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 2, 3, 4])

# Crear el modelo Lasso
lasso = Lasso(alpha=0.1)

# Ajustar el modelo
lasso.fit(X, y)

# Coeficientes del modelo
coef = lasso.coef_

# Realizar predicciones
y_pred = lasso.predict(X)

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresi贸n Lasso', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar t铆tulo
plt.title('Regresi贸n Lasso', fontsize=16)

# A帽adir leyenda
plt.legend(fontsize=12)

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, se utiliza la clase Lasso de la biblioteca scikit-learn para crear un modelo Lasso. Se ajusta el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, se obtienen los coeficientes del modelo (`coef`) y se realizan predicciones (`y_pred`). Finalmente, se grafican los datos de entrada junto con la l铆nea de regresi贸n obtenida por el modelo Lasso.

**Ejemplo de modelo Ridge:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

# Datos de ejemplo
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 2, 3, 4])

# Crear el modelo Ridge
ridge = Ridge(alpha=0.1)

# Ajustar el modelo
ridge.fit(X, y)

# Coeficientes del modelo
coef = ridge.coef_

# Realizar predicciones
y_pred = ridge.predict(X)

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresi贸n Ridge', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar t铆tulo
plt.title('Regresi贸n Ridge', fontsize=16)

# A帽adir leyenda
plt.legend(fontsize=12)

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, se utiliza la clase `Ridge` de scikit-learn para crear un modelo Ridge. El proceso es similar al ejemplo anterior con Lasso. Se ajusta el modelo a los datos, se obtienen los coeficientes del modelo, se realizan predicciones y se grafican los datos de entrada junto con la l铆nea de regresi贸n del modelo Ridge.

Es importante tener en cuenta que los valores de los par谩metros `alpha` en ambos ejemplos controlan la intensidad de la regularizaci贸n. Un valor m谩s alto de `alpha` aumentar谩 la regularizaci贸n y conducir谩 a coeficientes m谩s peque帽os y m谩s cercanos a cero.

Recuerda que estos son ejemplos simples para demostrar el uso de los modelos Lasso y Ridge. En casos reales, se deben realizar tareas adicionales como la divisi贸n de datos en conjuntos de entrenamiento y prueba, la validaci贸n cruzada y la optimizaci贸n de hiperpar谩metros para obtener un modelo m谩s robusto y preciso.

## **Regresi贸n con Vectores de soporte**

La regresi贸n con vectores de soporte (Support Vector Regression, SVR) es una t茅cnica de aprendizaje autom谩tico que se utiliza para realizar tareas de regresi贸n. A diferencia de la regresi贸n lineal, que busca encontrar una l铆nea recta que mejor se ajuste a los datos, la SVR busca encontrar una funci贸n no lineal que se ajuste a los datos de manera 贸ptima.

La idea principal de la SVR es mapear los datos de entrada a un espacio de mayor dimensionalidad utilizando una funci贸n de kernel, donde se busca encontrar un hiperplano en este espacio que se ajuste a los datos. El objetivo es encontrar el hiperplano que tenga el margen m谩s amplio posible con respecto a los puntos de datos. Los puntos de datos que est谩n dentro del margen o que violan el margen se denominan vectores de soporte y son los puntos m谩s influyentes en el modelo.

A continuaci贸n, te mostrar茅 un ejemplo en c贸digo Python de c贸mo realizar una regresi贸n con vectores de soporte utilizando la biblioteca scikit-learn y c贸mo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

# Datos de ejemplo
X = np.array([[0, 0], [2, 2], [4, 4], [6, 6], [8, 8]])
y = np.array([0, 2, 4, 6, 8])

# Crear el modelo SVR
svr = SVR(kernel='rbf')

# Ajustar el modelo
svr.fit(X, y)

# Realizar predicciones
y_pred = svr.predict(X)

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresi贸n SVR', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar t铆tulo
plt.title('Regresi贸n con Vectores de Soporte (SVR)', fontsize=16)

# A帽adir leyenda
plt.legend(fontsize=12)

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, utilizamos la clase SVR de scikit-learn para crear un modelo de regresi贸n con vectores de soporte. Configuramos el par谩metro kernel en '`rbf`', que corresponde a una funci贸n de kernel radial (o gaussiano), que se utiliza com煤nmente en SVR. A continuaci贸n, ajustamos el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, realizamos predicciones (`y_pred`) utilizando el modelo ajustado.

Finalmente, graficamos los datos de entrada junto con la l铆nea de regresi贸n obtenida por la SVR.

Es importante destacar que la elecci贸n del kernel y los hiperpar谩metros de la SVR puede tener un impacto significativo en los resultados. Es posible experimentar con diferentes kernels (lineal, polinomial, sigmoidal, entre otros) y ajustar los hiperpar谩metros para obtener un modelo m谩s preciso y adecuado a los datos espec铆ficos.

## **Regresi贸n con rboles de Decisi贸n**

La regresi贸n con 谩rboles de decisi贸n es un m茅todo de aprendizaje autom谩tico utilizado para realizar tareas de regresi贸n. A diferencia de los modelos lineales, la regresi贸n con 谩rboles de decisi贸n es capaz de capturar relaciones no lineales entre las variables de entrada y la variable de salida.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Datos de ejemplo
X = np.array([[0], [2], [4], [6], [8]])
y = np.array([1, 2, 4, 6, 8])

# Crear el modelo de regresi贸n con 谩rboles de decisi贸n
regressor = DecisionTreeRegressor()

# Ajustar el modelo
regressor.fit(X, y)

# Realizar predicciones
y_pred = regressor.predict(X)

# Ordenar los datos para graficar la l铆nea de regresi贸n
X_sorted = np.sort(X, axis=0)
y_pred_sorted = regressor.predict(X_sorted)

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(X, y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regresi贸n con rbol de Decisi贸n', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar t铆tulo
plt.title('Regresi贸n con rbol de Decisi贸n', fontsize=16)

# A帽adir leyenda
plt.legend(fontsize=12)

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, utilizamos la clase `DecisionTreeRegressor` de scikit-learn para crear un modelo de regresi贸n con 谩rboles de decisi贸n. A continuaci贸n, ajustamos el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, realizamos predicciones (`y_pred`) utilizando el modelo ajustado.

Es importante destacar que los 谩rboles de decisi贸n pueden ser propensos al sobreajuste (`overfitting`) si no se controla su complejidad. Es posible ajustar hiperpar谩metros como la profundidad m谩xima del 谩rbol o el n煤mero m铆nimo de muestras requeridas para dividir un nodo para obtener un modelo m谩s equilibrado y generalizable. Adem谩s, se pueden utilizar t茅cnicas de ensamblado de 谩rboles, como el bosque aleatorio (`Random Forest`) o el aumento de gradiente (`Gradient Boosting`), para mejorar a煤n m谩s el rendimiento de la regresi贸n con 谩rboles de decisi贸n.

## **Regresi贸n con Random Forest**

La regresi贸n con Random Forest es un m茅todo de aprendizaje autom谩tico que utiliza un conjunto de 谩rboles de decisi贸n para realizar tareas de regresi贸n. Random Forest combina m煤ltiples 谩rboles de decisi贸n independientes y promedia sus predicciones para obtener una predicci贸n final m谩s precisa y robusta.

A diferencia de un 谩rbol de decisi贸n 煤nico, que puede ser propenso al sobreajuste, Random Forest reduce el riesgo de sobreajuste promediando las predicciones de varios 谩rboles. Cada 谩rbol se entrena con una muestra aleatoria del conjunto de datos y, adem谩s, utiliza una selecci贸n aleatoria de caracter铆sticas en cada divisi贸n. Esto introduce variabilidad en los 谩rboles y ayuda a reducir la correlaci贸n entre ellos, mejorando as铆 la capacidad de generalizaci贸n y la precisi贸n del modelo.

## **Regresi贸n con XGBOOST**

La regresi贸n con XGBoost (Extreme Gradient Boosting) es un m茅todo de aprendizaje autom谩tico que utiliza un conjunto de modelos de 谩rboles de decisi贸n para realizar tareas de regresi贸n. XGBoost es una implementaci贸n optimizada y altamente eficiente del algoritmo de Gradient Boosting.

El algoritmo de XGBoost funciona mediante la creaci贸n de una serie de 谩rboles de decisi贸n secuenciales. Cada 谩rbol se entrena para corregir los errores residuales del 谩rbol anterior. Esto significa que cada 谩rbol se ajusta al gradiente negativo de la funci贸n de p茅rdida, tratando de reducir los errores cometidos por los 谩rboles anteriores.

A continuaci贸n, te mostrar茅 un ejemplo en c贸digo Python de c贸mo realizar una regresi贸n con XGBoost utilizando la biblioteca `XGBoost` y c贸mo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb

# Datos de ejemplo
X = np.array([[0], [2], [4], [6], [8]])
y = np.array([1, 2, 4, 6, 8])

# Crear el objeto DMatrix para los datos de entrenamiento
dtrain = xgb.DMatrix(X, label=y)

# Definir los par谩metros del modelo
params = {
    'objective': 'reg:squarederror',  # Funci贸n de p茅rdida para regresi贸n
    'max_depth': 3,                   # Profundidad m谩xima de los 谩rboles
    'eta': 0.1,                       # Tasa de aprendizaje
    'n_estimators': 100               # N煤mero de 谩rboles
}

# Entrenar el modelo
model = xgb.train(params, dtrain)

# Realizar predicciones
y_pred = model.predict(dtrain)

# Ordenar los datos para graficar la l铆nea de regresi贸n
X_sorted = np.sort(X, axis=0)
dtest = xgb.DMatrix(X_sorted)
y_pred_sorted = model.predict(dtest)

# Graficar los datos y la l铆nea de regresi贸n
plt.scatter(X, y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regresi贸n con XGBoost', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar t铆tulo
plt.title('Regresi贸n con XGBoost', fontsize=16)

# A帽adir leyenda
plt.legend(fontsize=12)

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, utilizamos la biblioteca `XGBoost` para realizar la regresi贸n con XGBoost. Primero, creamos un objeto `DMatrix` que contiene los datos de entrenamiento `X` y las etiquetas `y`.

A continuaci贸n, definimos los par谩metros del modelo, como la funci贸n de p茅rdida, la profundidad m谩xima de los 谩rboles, la tasa de aprendizaje y el n煤mero de 谩rboles. Estos par谩metros se pasan al m茅todo `xgb.train()` junto con el objeto `DMatrix` de entrenamiento para entrenar el modelo.

Despu茅s de entrenar el modelo, realizamos predicciones utilizando el modelo entrenado y el objeto `DMatrix` de entrenamiento. Luego, ordenamos los datos de entrada (`X_sorted`) y utilizamos el modelo para predecir los valores correspondientes (`y_pred_sorted`).

Finalmente, graficamos los datos de entrada junto con la l铆nea de regresi贸n obtenida por XGBoost.

XGBoost ofrece una amplia gama de hiperpar谩metros que se pueden ajustar para controlar la complejidad del modelo y mejorar su rendimiento, como la profundidad m谩xima de los 谩rboles, la tasa de aprendizaje, el n煤mero de 谩rboles, la regularizaci贸n y la funci贸n de p茅rdida. La selecci贸n adecuada de estos hiperpar谩metros puede influir en la capacidad de generalizaci贸n y la precisi贸n del modelo.
