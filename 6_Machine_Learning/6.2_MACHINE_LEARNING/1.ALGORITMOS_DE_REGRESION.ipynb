## **Métricas para evaluar modelos de Regresión**

Para evaluar modelos de regresión, existen varias métricas comunes que se utilizan para medir su desempeño. A continuación, te mostraré cuatro métricas fundamentales para evaluar modelos de regresión: el error cuadrático medio (MSE), el error absoluto medio (MAE), el coeficiente de determinación (R-cuadrado) y el gráfico de dispersión de los valores reales frente a los valores predichos.

1. **Error cuadrático medio (MSE):**

El MSE calcula la media de los errores cuadráticos entre los valores reales y los valores predichos. Cuanto menor sea el MSE, mejor será el modelo.

```python
from sklearn.metrics import mean_squared_error

# Valores reales
y_true = [3, 5, 7, 9]

# Valores predichos
y_pred = [2.5, 5.1, 6.8, 8.9]

mse = mean_squared_error(y_true, y_pred)
print("MSE:", mse)
```

2. **Error absoluto medio (MAE):**

El MAE calcula la media de los errores absolutos entre los valores reales y los valores predichos. Al igual que el MSE, se busca minimizar este valor.

**Ejemplo:**

Puedes calcular el MAE utilizando la función `mean_absolute_error` de scikit-learn:

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_true, y_pred)
print("MAE:", mae)
```

3. **Coeficiente de determinación (R-cuadrado):**

El coeficiente de determinación, comúnmente conocido como R-cuadrado, mide la proporción de la varianza en los valores de la variable dependiente que es predecible a partir de las variables independientes. El R-cuadrado varía entre 0 y 1, donde 1 indica un ajuste perfecto del modelo.

**Ejemplo:**

Puedes calcular el R-cuadrado utilizando la función `r2_score de scikit-learn`:

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_true, y_pred)
print("R-cuadrado:", r2)
```

4. **Gráfico de dispersión:**

El gráfico de dispersión es una forma visual de evaluar el rendimiento del modelo de regresión. Consiste en trazar los valores reales en el eje `y` y los valores predichos en el eje `x`. Un modelo de regresión preciso mostrará una línea cercana a la diagonal, lo que indica una buena correspondencia entre los valores reales y los valores predichos.

**Ejemplo:**

Puedes generar el gráfico de dispersión utilizando bibliotecas como matplotlib o seaborn:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que ya tienes definidos los arrays y_true y y_pred

# Personalizar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Tamaño y estilo del gráfico
plt.figure(figsize=(10, 8))

# Crear el gráfico de dispersión
scatter_plot = sns.scatterplot(x=y_true, y=y_pred, color='blue', marker='o', s=80, alpha=0.7)

# Añadir una línea de regresión (opcional)
sns.regplot(x=y_true, y=y_pred, scatter=False, ax=scatter_plot, color='red')

# Etiquetas y título
plt.xlabel("Valores reales", fontsize=14)
plt.ylabel("Valores predichos", fontsize=14)
plt.title("Gráfico de dispersión", fontsize=16)

# Añadir leyenda (opcional)
plt.legend(["Regresión lineal"])

# Personalizar la cuadrícula (opcional)
plt.grid(True, linestyle='--', alpha=0.5)

# Añadir líneas guía (opcional)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Mostrar el gráfico
plt.show()
```

Estas métricas y gráficos te permiten evaluar el rendimiento de un modelo de regresión y determinar su capacidad para hacer predicciones precisas.

## **Regresión Lineal simple y múltiple**

La regresión lineal es un método de aprendizaje automático supervisado utilizado para predecir una variable continua basada en una o más variables independientes. Existen dos tipos comunes de regresión lineal: la regresión lineal simple y la regresión lineal múltiple.

1. **Regresión Lineal Simple:**

La regresión lineal simple se utiliza cuando solo hay una variable independiente que se utiliza para predecir una variable dependiente continua. La relación entre las variables se modela como una línea recta.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que contiene el salario de los empleados (`y`) en función de su experiencia laboral (`x`). Podemos usar la regresión lineal simple para predecir el salario de un empleado en función de su experiencia laboral.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5])  # Experiencia laboral
y = np.array([30000, 40000, 50000, 60000, 70000])  # Salario

# Crear el modelo de regresión lineal
regression = LinearRegression()

# Entrenar el modelo
regression.fit(x.reshape(-1, 1), y)

# Realizar predicciones
y_pred = regression.predict(x.reshape(-1, 1))

# Configuración de la gráfica
plt.figure(figsize=(8, 6))

# Graficar los datos y la línea de regresión
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', label='Regresión lineal')

# Añadir etiquetas y título
plt.xlabel('Experiencia laboral')
plt.ylabel('Salario')
plt.title('Regresión lineal simple')

# Añadir una cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir leyenda
plt.legend()

# Añadir texto adicional
plt.text(2.5, 60000, f'R² = {regression.score(x.reshape(-1, 1), y):.2f}', fontsize=12, ha='center')

# Añadir línea de identidad
plt.plot(x, x, '--', color='gray', label='Línea de identidad', alpha=0.5)

# Ajustar la disposición
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

En este ejemplo, utilizamos la biblioteca scikit-learn para crear un modelo de regresión lineal (`LinearRegression`). Entrenamos el modelo con los datos de experiencia laboral (`x`) y salario (`y`) y luego realizamos predicciones (`y_pred`) basadas en el modelo entrenado. Finalmente, graficamos los datos de muestra y la línea de regresión.

2. **Regresión Lineal Múltiple:**

La regresión lineal múltiple se utiliza cuando hay más de una variable independiente que se utiliza para predecir una variable dependiente continua. La relación entre las variables se modela como un plano o una hiperplano en espacios de dimensiones superiores.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que contiene el precio de las casas (`y`) en función de su tamaño (`x1`) y el número de habitaciones (`x2`). Podemos usar la regresión lineal múltiple para predecir el precio de una casa en función de su tamaño y número de habitaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from matplotlib.lines import Line2D

# Datos de ejemplo
x1 = np.array([1200, 1500, 1700, 2000, 2200])  # Tamaño de la casa en pies cuadrados
x2 = np.array([2, 3, 4, 3, 4])  # Número de habitaciones
y = np.array([200000, 250000, 300000, 280000, 350000])  # Precio de la casa

# Crear el modelo de regresión lineal múltiple
regression = LinearRegression()

# Combinar las variables independientes en una matriz
X = np.column_stack((x1, x2))

# Entrenar el modelo
regression.fit(X, y)

# Realizar predicciones
y_pred = regression.predict(X)

# Graficar los datos y la superficie de regresión
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x1, x2, y, color='blue', label='Datos')
surface = ax.plot_trisurf(x1, x2, y_pred, color='red', alpha=0.5, label='Regresión lineal')

# Configuración de la leyenda con un manejador personalizado
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Datos'),
                   Line2D([0], [0], color='w', markerfacecolor='red', markersize=10, label='Regresión lineal')]
ax.legend(handles=legend_elements)

# Configuración de la gráfica
ax.set_xlabel('Tamaño')
ax.set_ylabel('Número de habitaciones')
ax.set_zlabel('Precio')
ax.set_title('Regresión lineal múltiple')

# Añadir etiquetas a puntos específicos
for i, txt in enumerate(['Punto 1', 'Punto 2', 'Punto 3', 'Punto 4', 'Punto 5']):
    ax.text(x1[i], x2[i], y[i], txt)

# Ajustar la disposición
fig.tight_layout()

# Mostrar la gráfica
plt.show()
```

En este ejemplo, utilizamos la biblioteca scikit-learn para crear un modelo de regresión lineal (`LinearRegression`). Combinamos las variables independientes (`x1` y `x2`) en una matriz `X`. Entrenamos el modelo con los datos de tamaño de la casa, número de habitaciones y precio de la casa, y luego realizamos predicciones (`y_pred`) basadas en el modelo entrenado. Finalmente, graficamos los datos de muestra y la superficie de regresión.

Estos ejemplos te proporcionan una idea de cómo implementar la regresión lineal simple y múltiple en Python utilizando scikit-learn. Recuerda que los conjuntos de datos y las características utilizadas en los ejemplos son solo ilustrativos, y puedes adaptar el código a tus propios conjuntos de datos y características específicas.

## **Modelos Lasso y ridge**

Los modelos Lasso y Ridge son dos técnicas de regularización utilizadas en la regresión lineal para evitar el sobreajuste y mejorar la capacidad de generalización del modelo. Ambas técnicas introducen una penalización en la función de costo para controlar la complejidad del modelo.

El modelo Lasso utiliza la regularización `L1`, que penaliza los coeficientes de las características menos importantes, reduciéndolos a cero si es posible. Esto permite realizar selección automática de características, ya que algunas características pueden ser descartadas completamente. Por otro lado, el modelo Ridge utiliza la regularización `L2`, que penaliza los coeficientes de las características, pero no los reduce a cero. En cambio, los coeficientes se reducen proporcionalmente.

A continuación, te mostraré un ejemplo en código Python para cada modelo y cómo se pueden utilizar junto con gráficas.

**Ejemplo de modelo Lasso:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

# Datos de ejemplo
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 2, 3, 4])

# Crear el modelo Lasso
lasso = Lasso(alpha=0.1)

# Ajustar el modelo
lasso.fit(X, y)

# Coeficientes del modelo
coef = lasso.coef_

# Realizar predicciones
y_pred = lasso.predict(X)

# Graficar los datos y la línea de regresión
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresión Lasso', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión Lasso', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, se utiliza la clase Lasso de la biblioteca scikit-learn para crear un modelo Lasso. Se ajusta el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, se obtienen los coeficientes del modelo (`coef`) y se realizan predicciones (`y_pred`). Finalmente, se grafican los datos de entrada junto con la línea de regresión obtenida por el modelo Lasso.

**Ejemplo de modelo Ridge:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

# Datos de ejemplo
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 2, 3, 4])

# Crear el modelo Ridge
ridge = Ridge(alpha=0.1)

# Ajustar el modelo
ridge.fit(X, y)

# Coeficientes del modelo
coef = ridge.coef_

# Realizar predicciones
y_pred = ridge.predict(X)

# Graficar los datos y la línea de regresión
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresión Ridge', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión Ridge', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, se utiliza la clase `Ridge` de scikit-learn para crear un modelo Ridge. El proceso es similar al ejemplo anterior con Lasso. Se ajusta el modelo a los datos, se obtienen los coeficientes del modelo, se realizan predicciones y se grafican los datos de entrada junto con la línea de regresión del modelo Ridge.

Es importante tener en cuenta que los valores de los parámetros `alpha` en ambos ejemplos controlan la intensidad de la regularización. Un valor más alto de `alpha` aumentará la regularización y conducirá a coeficientes más pequeños y más cercanos a cero.

Recuerda que estos son ejemplos simples para demostrar el uso de los modelos Lasso y Ridge. En casos reales, se deben realizar tareas adicionales como la división de datos en conjuntos de entrenamiento y prueba, la validación cruzada y la optimización de hiperparámetros para obtener un modelo más robusto y preciso.

## **Regresión con Vectores de soporte**

La regresión con vectores de soporte (Support Vector Regression, SVR) es una técnica de aprendizaje automático que se utiliza para realizar tareas de regresión. A diferencia de la regresión lineal, que busca encontrar una línea recta que mejor se ajuste a los datos, la SVR busca encontrar una función no lineal que se ajuste a los datos de manera óptima.

La idea principal de la SVR es mapear los datos de entrada a un espacio de mayor dimensionalidad utilizando una función de kernel, donde se busca encontrar un hiperplano en este espacio que se ajuste a los datos. El objetivo es encontrar el hiperplano que tenga el margen más amplio posible con respecto a los puntos de datos. Los puntos de datos que están dentro del margen o que violan el margen se denominan vectores de soporte y son los puntos más influyentes en el modelo.

A continuación, te mostraré un ejemplo en código Python de cómo realizar una regresión con vectores de soporte utilizando la biblioteca scikit-learn y cómo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

# Datos de ejemplo
X = np.array([[0, 0], [2, 2], [4, 4], [6, 6], [8, 8]])
y = np.array([0, 2, 4, 6, 8])

# Crear el modelo SVR
svr = SVR(kernel='rbf')

# Ajustar el modelo
svr.fit(X, y)

# Realizar predicciones
y_pred = svr.predict(X)

# Graficar los datos y la línea de regresión
plt.scatter(X[:, 0], y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X[:, 0], y_pred, color='red', label='Regresión SVR', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión con Vectores de Soporte (SVR)', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, utilizamos la clase SVR de scikit-learn para crear un modelo de regresión con vectores de soporte. Configuramos el parámetro kernel en '`rbf`', que corresponde a una función de kernel radial (o gaussiano), que se utiliza comúnmente en SVR. A continuación, ajustamos el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, realizamos predicciones (`y_pred`) utilizando el modelo ajustado.

Finalmente, graficamos los datos de entrada junto con la línea de regresión obtenida por la SVR.

Es importante destacar que la elección del kernel y los hiperparámetros de la SVR puede tener un impacto significativo en los resultados. Es posible experimentar con diferentes kernels (lineal, polinomial, sigmoidal, entre otros) y ajustar los hiperparámetros para obtener un modelo más preciso y adecuado a los datos específicos.

## **Regresión con Árboles de Decisión**

La regresión con árboles de decisión es un método de aprendizaje automático utilizado para realizar tareas de regresión. A diferencia de los modelos lineales, la regresión con árboles de decisión es capaz de capturar relaciones no lineales entre las variables de entrada y la variable de salida.

En un árbol de decisión, se divide el espacio de entrada en regiones rectangulares y se asigna un valor constante a cada región. Para hacer una predicción, se determina la región a la que pertenece un punto de datos de entrada y se utiliza el valor constante asociado a esa región como predicción.

A continuación, te mostraré un ejemplo en código Python de cómo realizar una regresión con árboles de decisión utilizando la biblioteca scikit-learn y cómo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Datos de ejemplo
X = np.array([[0], [2], [4], [6], [8]])
y = np.array([1, 2, 4, 6, 8])

# Crear el modelo de regresión con árboles de decisión
regressor = DecisionTreeRegressor()

# Ajustar el modelo
regressor.fit(X, y)

# Realizar predicciones
y_pred = regressor.predict(X)

# Ordenar los datos para graficar la línea de regresión
X_sorted = np.sort(X, axis=0)
y_pred_sorted = regressor.predict(X_sorted)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regresión con Árbol de Decisión', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión con Árbol de Decisión', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, utilizamos la clase `DecisionTreeRegressor` de scikit-learn para crear un modelo de regresión con árboles de decisión. A continuación, ajustamos el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, realizamos predicciones (`y_pred`) utilizando el modelo ajustado.

Para graficar la línea de regresión, ordenamos los datos de entrada (`X_sorted`) y utilizamos el modelo para predecir los valores correspondientes (`y_pred_sorted`).

Finalmente, graficamos los datos de entrada junto con la línea de regresión obtenida por el árbol de decisión.

Es importante destacar que los árboles de decisión pueden ser propensos al sobreajuste (`overfitting`) si no se controla su complejidad. Es posible ajustar hiperparámetros como la profundidad máxima del árbol o el número mínimo de muestras requeridas para dividir un nodo para obtener un modelo más equilibrado y generalizable. Además, se pueden utilizar técnicas de ensamblado de árboles, como el bosque aleatorio (`Random Forest`) o el aumento de gradiente (`Gradient Boosting`), para mejorar aún más el rendimiento de la regresión con árboles de decisión.

## **Regresión con Random Forest**

La regresión con Random Forest es un método de aprendizaje automático que utiliza un conjunto de árboles de decisión para realizar tareas de regresión. Random Forest combina múltiples árboles de decisión independientes y promedia sus predicciones para obtener una predicción final más precisa y robusta.

A diferencia de un árbol de decisión único, que puede ser propenso al sobreajuste, Random Forest reduce el riesgo de sobreajuste promediando las predicciones de varios árboles. Cada árbol se entrena con una muestra aleatoria del conjunto de datos y, además, utiliza una selección aleatoria de características en cada división. Esto introduce variabilidad en los árboles y ayuda a reducir la correlación entre ellos, mejorando así la capacidad de generalización y la precisión del modelo.

A continuación, te mostraré un ejemplo en código Python de cómo realizar una regresión con Random Forest utilizando la biblioteca scikit-learn y cómo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Datos de ejemplo
X = np.array([[0], [2], [4], [6], [8]])
y = np.array([1, 2, 4, 6, 8])

# Crear el modelo de regresión con Random Forest
regressor = RandomForestRegressor(n_estimators=10)

# Ajustar el modelo
regressor.fit(X, y)

# Realizar predicciones
y_pred = regressor.predict(X)

# Ordenar los datos para graficar la línea de regresión
X_sorted = np.sort(X, axis=0)
y_pred_sorted = regressor.predict(X_sorted)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regresión con Random Forest', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión con Random Forest', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, utilizamos la clase `RandomForestRegressor` de scikit-learn para crear un modelo de regresión con Random Forest. Especificamos el número de árboles en el bosque utilizando el parámetro `n_estimators`.

A continuación, ajustamos el modelo a los datos de entrada (`X`) y las etiquetas (`y`). Luego, realizamos predicciones (`y_pred`) utilizando el modelo ajustado.

Para graficar la línea de regresión, ordenamos los datos de entrada (`X_sorted`) y utilizamos el modelo para predecir los valores correspondientes (`y_pred_sorted`).

Finalmente, graficamos los datos de entrada junto con la línea de regresión obtenida por el Random Forest.

Es importante destacar que Random Forest ofrece varios hiperparámetros que se pueden ajustar para controlar la complejidad del modelo y mejorar su rendimiento, como la profundidad máxima de los árboles, el número mínimo de muestras requeridas para dividir un nodo y el número máximo de características a considerar en cada división. La selección adecuada de estos hiperparámetros puede mejorar la capacidad de generalización y la precisión del modelo.

## **Regresión con XGBOOST**

La regresión con XGBoost (Extreme Gradient Boosting) es un método de aprendizaje automático que utiliza un conjunto de modelos de árboles de decisión para realizar tareas de regresión. XGBoost es una implementación optimizada y altamente eficiente del algoritmo de Gradient Boosting.

El algoritmo de XGBoost funciona mediante la creación de una serie de árboles de decisión secuenciales. Cada árbol se entrena para corregir los errores residuales del árbol anterior. Esto significa que cada árbol se ajusta al gradiente negativo de la función de pérdida, tratando de reducir los errores cometidos por los árboles anteriores.

A continuación, te mostraré un ejemplo en código Python de cómo realizar una regresión con XGBoost utilizando la biblioteca `XGBoost` y cómo graficar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb

# Datos de ejemplo
X = np.array([[0], [2], [4], [6], [8]])
y = np.array([1, 2, 4, 6, 8])

# Crear el objeto DMatrix para los datos de entrenamiento
dtrain = xgb.DMatrix(X, label=y)

# Definir los parámetros del modelo
params = {
    'objective': 'reg:squarederror',  # Función de pérdida para regresión
    'max_depth': 3,                   # Profundidad máxima de los árboles
    'eta': 0.1,                       # Tasa de aprendizaje
    'n_estimators': 100               # Número de árboles
}

# Entrenar el modelo
model = xgb.train(params, dtrain)

# Realizar predicciones
y_pred = model.predict(dtrain)

# Ordenar los datos para graficar la línea de regresión
X_sorted = np.sort(X, axis=0)
dtest = xgb.DMatrix(X_sorted)
y_pred_sorted = model.predict(dtest)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue', label='Datos', marker='o', s=80, alpha=0.7)
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regresión con XGBoost', linewidth=2, linestyle='--')

# Personalizar ejes
plt.xlabel('X', fontsize=14)
plt.ylabel('y', fontsize=14)

# Personalizar título
plt.title('Regresión con XGBoost', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, utilizamos la biblioteca `XGBoost` para realizar la regresión con XGBoost. Primero, creamos un objeto `DMatrix` que contiene los datos de entrenamiento `X` y las etiquetas `y`.

A continuación, definimos los parámetros del modelo, como la función de pérdida, la profundidad máxima de los árboles, la tasa de aprendizaje y el número de árboles. Estos parámetros se pasan al método `xgb.train()` junto con el objeto `DMatrix` de entrenamiento para entrenar el modelo.

Después de entrenar el modelo, realizamos predicciones utilizando el modelo entrenado y el objeto `DMatrix` de entrenamiento. Luego, ordenamos los datos de entrada (`X_sorted`) y utilizamos el modelo para predecir los valores correspondientes (`y_pred_sorted`).

Finalmente, graficamos los datos de entrada junto con la línea de regresión obtenida por XGBoost.

XGBoost ofrece una amplia gama de hiperparámetros que se pueden ajustar para controlar la complejidad del modelo y mejorar su rendimiento, como la profundidad máxima de los árboles, la tasa de aprendizaje, el número de árboles, la regularización y la función de pérdida. La selección adecuada de estos hiperparámetros puede influir en la capacidad de generalización y la precisión del modelo.
