### **M茅todos de bisecci贸n y Newton-Raphson**

**M茅todo de Bisecci贸n:**

El m茅todo de bisecci贸n es un algoritmo de b煤squeda de ra铆ces que opera en intervalos cerrados y utiliza el principio de los intervalos intermedios. Supongamos que tenemos una funci贸n continua $f(x)$ en un intervalo $[a, b]$ tal que $( f(a) \cdot f(b) < 0 )$ (es decir, la funci贸n tiene valores opuestos en los extremos del intervalo). Entonces, el m茅todo de bisecci贸n divide iterativamente el intervalo por la mitad y selecciona el subintervalo donde la ra铆z se encuentra, hasta alcanzar una precisi贸n deseada.

**F贸rmulas:**

Dado un intervalo $[a, b]$ y un punto medio $(c = \frac{a + b}{2})$, el m茅todo de bisecci贸n actualiza el intervalo de la siguiente manera:

$\text{Si } f(a) \cdot f(c) < 0, \text{ entonces el nuevo intervalo es } [a, c]$

$\text{Si } f(b) \cdot f(c) < 0, \text{ entonces el nuevo intervalo es } [c, b]$

El proceso se repite hasta que se alcanza una tolerancia o un n煤mero m谩ximo de iteraciones.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 6*x**2 + 11*x - 6

def bisection_method(a, b, tol, max_iter):
    iterations = 0
    while (b - a) / 2 > tol and iterations < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return c, iterations
```


Vamos a completar el ejemplo del m茅todo de bisecci贸n con gr谩ficos en Python:

```python
# Graficar la funci贸n
x_values = np.linspace(0, 5, 100)
y_values = f(x_values)

plt.plot(x_values, y_values, label='$f(x) = x^3 - 6x^2 + 11x - 6$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funci贸n y Ra铆z Buscada')
plt.legend()
plt.grid(True)
plt.show()

# Aplicar el m茅todo de bisecci贸n en el intervalo [0, 5]
a, b = 0, 5
tolerance = 1e-5
max_iterations = 100

root, iterations = bisection_method(a, b, tolerance, max_iterations)

# Imprimir resultados
print(f'Ra铆z encontrada: {root}')
print(f'Iteraciones realizadas: {iterations}')
```

En este ejemplo, la funci贸n $f(x) = x^3 - 6x^2 + 11x - 6$ tiene una ra铆z en el intervalo $[0, 5]$. El m茅todo de bisecci贸n se aplica para encontrar esta ra铆z, y los resultados se imprimen. Tambi茅n se muestra la gr谩fica de la funci贸n para visualizar la posici贸n de la ra铆z.

**M茅todo de Newton-Raphson:**

El m茅todo de Newton-Raphson es otro algoritmo para encontrar ra铆ces de una funci贸n. Dado un punto inicial $(x_0)$, el m茅todo utiliza la tangente a la curva en ese punto para encontrar una mejor aproximaci贸n de la ra铆z. La iteraci贸n se realiza con la f贸rmula:

$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$

Donde $f'(x_n)$ es la derivada de $f(x)$ evaluada en $(x_n)$.

**Ejemplo con Gr谩fico en Python:**

```python
def f_prime(x):
    return 3*x**2 - 12*x + 11

def newton_raphson_method(x_0, tol, max_iter):
    iterations = 0
    while abs(f(x_0)) > tol and iterations < max_iter:
        x_0 = x_0 - f(x_0) / f_prime(x_0)
        iterations += 1
    return x_0, iterations

# Graficar la funci贸n y la tangente en un punto inicial
x_0 = 3.5
tangent_line = f_prime(x_0) * (x_values - x_0) + f(x_0)

plt.plot(x_values, y_values, label='$f(x) = x^3 - 6x^2 + 11x - 6$')
plt.plot(x_values, tangent_line, label='Tangente en $x_0$', linestyle='--')
plt.scatter(x_0, f(x_0), color='red', label='Punto Inicial $(x_0)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funci贸n y Tangente con Punto Inicial')
plt.legend()
plt.grid(True)
plt.show()

# Aplicar el m茅todo de Newton-Raphson con el punto inicial x_0
root_newton, iterations_newton = newton_raphson_method(x_0, tolerance, max_iterations)

# Imprimir resultados
print(f'Ra铆z encontrada con Newton-Raphson: {root_newton}')
print(f'Iteraciones realizadas con Newton-Raphson: {iterations_newton}')
```

En este ejemplo, se muestra la funci贸n $f(x)$ junto con la tangente en un punto inicial $(x_0)$. Luego, se aplica el m茅todo de Newton-Raphson para encontrar la ra铆z.

### **Reglas del trapecio, Simpson**

**Regla del Trapecio:**

La regla del trapecio es una t茅cnica de integraci贸n num茅rica que aproxima el valor de una integral definida dividiendo el 谩rea bajo la curva en m煤ltiples trapecios. La f贸rmula para la regla del trapecio es:

$\int_{a}^{b} f(x) \,dx \approx \frac{h}{2} \left[ f(a) + 2f(x_1) + 2f(x_2) + \ldots + 2f(x_{n-1}) + f(b) \right]$

Donde $(h)$ es la longitud de cada subintervalo, $(a)$ y $(b)$ son los l铆mites de integraci贸n, y $(x_1, x_2, \ldots, x_{n-1})$ son los puntos dentro de cada subintervalo.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def trapezoidal_rule(a, b, n):
    h = (b - a) / n
    x_values = np.linspace(a, b, n + 1)
    y_values = f(x_values)
    return h * (np.sum(y_values) - 0.5 * (y_values[0] + y_values[-1]))

# Graficar la funci贸n
x_values = np.linspace(0, 2, 100)
y_values = f(x_values)

plt.plot(x_values, y_values, label='$f(x) = x^2$')
plt.fill_between(x_values, y_values, alpha=0.1, color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funci贸n y rea bajo la Curva')
plt.legend()
plt.grid(True)
plt.show()

# Aplicar la regla del trapecio
a, b = 0, 2
n = 4  # N煤mero de subintervalos
area_trapezoidal = trapezoidal_rule(a, b, n)

# Imprimir resultado
print(f'rea bajo la curva (Regla del Trapecio): {area_trapezoidal}')
```

Este c贸digo utiliza la regla del trapecio para aproximar el 谩rea bajo la curva de la funci贸n $f(x) = x^2$ en el intervalo $[0, 2]$. La gr谩fica muestra la funci贸n y el 谩rea aproximada en azul.

**Regla de Simpson:**

La regla de Simpson es otra t茅cnica de integraci贸n num茅rica que utiliza polinomios de segundo grado (par谩bolas) para aproximar el valor de la integral definida. La f贸rmula para la regla de Simpson es:

$\int_{a}^{b} f(x) \,dx \approx \frac{h}{3} \left[ f(a) + 4f(x_1) + 2f(x_2) + \ldots + 2f(x_{n-2}) + 4f(x_{n-1}) + f(b) \right]$

donde $(h)$ es la longitud de cada subintervalo, $(a)$ y $(b)$ son los l铆mites de integraci贸n, y $(x_1, x_2, \ldots, x_{n-1})$ son los puntos dentro de cada subintervalo.

**Ejemplo con Gr谩fico en Python:**

```python
def simpson_rule(a, b, n):
    h = (b - a) / n
    x_values = np.linspace(a, b, n + 1)
    y_values = f(x_values)
    return h / 3 * np.sum(y_values[0:-1:2] + 4*y_values[1::2] + y_values[2::2])

# Graficar la funci贸n
plt.plot(x_values, y_values, label='$f(x) = x^2$')
plt.fill_between(x_values, y_values, alpha=0.1, color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funci贸n y rea bajo la Curva')
plt.legend()
plt.grid(True)
plt.show()

# Aplicar la regla de Simpson
area_simpson = simpson_rule(a, b, n)

# Imprimir resultado
print(f'rea bajo la curva (Regla de Simpson): {area_simpson}')
```

Este c贸digo utiliza la regla de Simpson para aproximar el 谩rea bajo la curva de la funci贸n $f(x) = x^2$ en el intervalo $[0, 2]$. La gr谩fica muestra la funci贸n y el 谩rea aproximada en azul. Ambos m茅todos (trapecio y Simpson) son utilizados com煤nmente para realizar integraci贸n num茅rica cuando la funci贸n no tiene una forma anal铆tica f谩cil de integrar.

### **M茅todos de Monte Carlo**

El m茅todo de Monte Carlo es una t茅cnica estad铆stica y computacional que utiliza n煤meros aleatorios para obtener resultados num茅ricos. Es particularmente 煤til para aproximar soluciones a problemas matem谩ticos complejos y realizar integraci贸n num茅rica. La idea b谩sica es usar la aleatoriedad para simular el comportamiento de un sistema y obtener estad铆sticas a partir de m煤ltiples simulaciones.

**Aplicaci贸n a la Aproximaci贸n de reas:**

Una aplicaci贸n com煤n del m茅todo de Monte Carlo es la aproximaci贸n del 谩rea de una regi贸n. Supongamos que queremos encontrar el 谩rea de una regi贸n $(D)$ en un plano. Podemos inscribir esta regi贸n en un rect谩ngulo m谩s grande $(R)$ y generar puntos aleatorios uniformemente distribuidos en $(R)$. Luego, calculamos la fracci贸n de puntos que caen dentro de la regi贸n $(D)$ y la multiplicamos por el 谩rea total de $(R)$ para obtener una aproximaci贸n del 谩rea de $(D)$.

**F贸rmula:**

$\text{rea}(D) \approx \frac{\text{N煤mero de puntos en } D}{\text{N煤mero total de puntos en } R} \times \text{rea}(R)$

**Ejemplo con Gr谩fico en Python:**

Supongamos que queremos aproximar el 谩rea del c铆rculo unitario $(D)$ inscrito en un cuadrado $(R)$ de lado 2. El 谩rea del c铆rculo unitario es $(\pi)$, y el 谩rea del cuadrado es $(4)$. La relaci贸n de 谩reas es $(\frac{\pi}{4})$.

```python
import numpy as np
import matplotlib.pyplot as plt

# N煤mero total de puntos generados
num_points = 1000

# Generar puntos aleatorios en el cuadrado [-1, 1] x [-1, 1]
points = np.random.rand(num_points, 2) * 2 - 1

# Calcular la distancia desde el origen para cada punto
distances = np.linalg.norm(points, axis=1)

# Verificar si cada punto est谩 dentro del c铆rculo unitario
points_inside_circle = distances <= 1

# Graficar puntos dentro y fuera del c铆rculo
plt.scatter(points[:, 0], points[:, 1], c=points_inside_circle, cmap='viridis')
circle = plt.Circle((0, 0), 1, edgecolor='red', facecolor='none', linewidth=2)
plt.gca().add_patch(circle)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.title('Aproximaci贸n del rea del C铆rculo con Monte Carlo')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Calcular el 谩rea aproximada del c铆rculo
area_ratio = np.sum(points_inside_circle) / num_points
approximated_area = area_ratio * 4  # rea del cuadrado

# Imprimir resultado
print(f'rea aproximada del c铆rculo: {approximated_area}')
```

En este ejemplo, generamos puntos aleatorios en el cuadrado $([-1, 1] \times [-1, 1])$ y utilizamos la condici贸n de que un punto est谩 dentro del c铆rculo unitario si su distancia al origen es menor o igual a 1. La relaci贸n de 谩reas obtenida de la fracci贸n de puntos dentro del c铆rculo nos proporciona una aproximaci贸n de $(\pi)$.
