### **3.1.2 Derivada de una funci贸n simple**

**Concepto B谩sico de la Derivada**

La derivada de una funci贸n $f(x)$ en un punto $(x)$ se define como el l铆mite del cociente diferencial cuando el incremento en $(x)$ tiende a cero. **Definici贸n Formal**:

**Ejemplo de Derivada: Funci贸n Lineal**

Considere la funci贸n lineal $f(x) = 3x + 5$. Vamos a calcular su derivada.

**Visualizaci贸n y C贸digo en Python**

Podemos visualizar la funci贸n y su derivada utilizando Python. Aqu铆 est谩 el c贸digo utilizando Matplotlib y NumPy:

### **3.1.5 Comprobaci贸n del concepto de derivada**

La **comprobaci贸n del concepto de derivada** implica verificar que la definici贸n formal de la derivada corresponde con el comportamiento esperado de la funci贸n bajo an谩lisis, y que el proceso de c谩lculo utilizado para encontrar la derivada es correcto. Esto se hace a menudo utilizando la definici贸n de derivada como l铆mite y comparando los resultados obtenidos mediante este m茅todo con los que proporcionan las reglas de derivaci贸n m谩s comunes o con resultados conocidos.

**Concepto de Derivada**

La derivada de una funci贸n en un punto proporciona la tasa instant谩nea de cambio de la funci贸n en ese punto. Matem谩ticamente, se define como el l铆mite del cociente diferencial:

$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$

**Ejemplo de Comprobaci贸n del Concepto de Derivada**

Vamos a considerar la funci贸n $f(x) = x^2$ y verificar su derivada en el punto $(x = 3)$ utilizando tanto la definici贸n de derivada como la regla de potencia, que dice que la derivada de $(x^n)$ es $(nx^{n-1})$.

**Derivada mediante la Definici贸n de L铆mite**

Para $f(x) = x^2$:

$f'(x) = \lim_{h \to 0} \frac{(x+h)^2 - x^2}{h} = \lim_{h \to 0} \frac{x^2 + 2xh + h^2 - x^2}{h} = \lim_{h \to 0} \frac{2xh + h^2}{h} = \lim_{h \to 0} (2x + h) = 2x$

En el punto $(x = 3)$:

$f'(3) = 2 \times 3 = 6$

**Visualizaci贸n y C贸digo en Python**

Utilizaremos Python para visualizar la funci贸n y su derivada, y para confirmar el c谩lculo de la derivada en el punto espec铆fico $(x = 3)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n y su derivada
def f(x):
    return x**2

def df(x):
    return 2*x

# Rango de valores para x
x_values = np.linspace(0, 5, 400)
y_values = f(x_values)
dy_values = df(x_values)

# Calcular el valor de la derivada en x = 3
x_point = 3
y_point = f(x_point)
slope = df(x_point)  # Pendiente de la tangente en x = 3

# Punto de tangencia y l铆nea tangente
tangent_line = slope * (x_values - x_point) + y_point

# Crear gr谩fico
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = x^2$', color='blue')
plt.plot(x_values, dy_values, label="$f'(x) = 2x$", color='red', linestyle='--')
plt.plot(x_values, tangent_line, label='Tangente en x = 3', color='green', linestyle=':')
plt.scatter([x_point], [y_point], color='black', marker='o', label='Punto de Tangencia')  # Marcar el punto de tangencia
plt.title('Funci贸n Cuadr谩tica y su Derivada', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axvline(x=x_point, color='grey', linestyle='--')
plt.axhline(y=y_point, color='grey', linestyle='--')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

**Interpretaci贸n**
