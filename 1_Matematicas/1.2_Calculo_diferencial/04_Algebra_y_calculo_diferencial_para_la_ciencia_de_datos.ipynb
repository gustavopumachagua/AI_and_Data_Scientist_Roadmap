## **4.1 Regla del producto y de la cadena**

Utilice las reglas del producto y de la cadena para calcular las derivadas de funciones m谩s complicadas.

### **4.1.1 Regla del producto**

La **Regla del Producto** en c谩lculo es una t茅cnica fundamental utilizada para encontrar la derivada de un producto de dos funciones. Esta regla es esencial cuando se manejan funciones donde cada t茅rmino es una funci贸n de $(x)$ que necesita ser diferenciada. La regla del producto permite simplificar la diferenciaci贸n de estas expresiones complejas.

**Concepto de la Regla del Producto**

Si tienes dos funciones diferenciables, $f(x)$ y $g(x)$, la derivada de su producto $f(x)g(x)$ se calcula como sigue:

$(fg)'(x) = f'(x)g(x) + f(x)g'(x)$

Esta f贸rmula indica que la derivada del producto de dos funciones es el producto de la derivada de la primera funci贸n por la segunda funci贸n, m谩s el producto de la primera funci贸n por la derivada de la segunda funci贸n.

**Ejemplo Pr谩ctico: Derivada del Producto de dos Funciones**

**Funciones**:

- $f(x) = x^2$

- $g(x) = e^x$

**Derivada usando la Regla del Producto**

Para estas funciones, aplicamos la regla del producto:

- Derivadas individuales:

  $f'(x) = 2x, \quad g'(x) = e^x$

- Aplicando la regla del producto:

  $(fg)'(x) = f'(x)g(x) + f(x)g'(x) = 2x \cdot e^x + x^2 \cdot e^x = (2x + x^2) e^x$

**Visualizaci贸n con Python**

Vamos a visualizar las funciones y su derivada de producto usando Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones y su derivada de producto
def f(x):
    return x**2

def g(x):
    return np.exp(x)

def f_prime(x):
    return 2*x

def g_prime(x):
    return np.exp(x)

def product_derivative(x):
    return (f_prime(x) * g(x) + f(x) * g_prime(x))

# Generar valores de x
x_values = np.linspace(-1, 2, 400)
f_values = f(x_values)
g_values = g(x_values)
product_derivative_values = product_derivative(x_values)

# Crear gr谩fico
plt.figure(figsize=(12, 8))
plt.plot(x_values, f_values, label='$f(x) = x^2$', color='blue', linewidth=2)
plt.plot(x_values, g_values, label='$g(x) = e^x$', color='green', linewidth=2)
plt.plot(x_values, product_derivative_values, label="Product Derivative $(2x + x^2)e^x$", color='red', linestyle='--', linewidth=2)
plt.title('Funciones y Derivada del Producto', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Valor', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.xlim(-1, 2)  # Limitar los valores de x
plt.ylim(-1, 10)  # Limitar los valores de y
plt.show()
```

**Interpretaci贸n**

- **Gr谩ficos Azul y Verde**: Representan las funciones $f(x) = x^2$ y $g(x) = e^x$ respectivamente.

- **Gr谩fico Rojo**: Muestra la derivada del producto $(2x + x^2) e^x$. Esta curva refleja c贸mo la tasa de cambio del producto de estas dos funciones depende tanto de sus valores individuales como de sus tasas de cambio individuales.

Este ejemplo ilustra c贸mo la regla del producto simplifica la tarea de diferenciar el producto de dos funciones complejas, mostrando visualmente la aplicaci贸n y verificaci贸n de esta regla esencial del c谩lculo.

### **4.1.2 Regla del cociente**

La **Regla del Cociente** es otra t茅cnica fundamental en c谩lculo diferencial, utilizada para derivar el cociente de dos funciones. Es especialmente 煤til cuando se necesita encontrar la derivada de una divisi贸n de dos funciones, cada una de las cuales depende de la misma variable.

**Concepto de la Regla del Cociente**

La Regla del Cociente establece que si tienes dos funciones diferenciables $f(x)$ y $g(x)$, y necesitas encontrar la derivada de su cociente $(\frac{f(x)}{g(x)})$, puedes usar la siguiente f贸rmula:

$\left(\frac{f}{g}\right)'(x) = \frac{f'(x)g(x) - f(x)g'(x)}{[g(x)]^2}$

Esta f贸rmula dice que la derivada del cociente es el denominador (la funci贸n en el denominador de la fracci贸n) multiplicado por la derivada del numerador, menos el numerador multiplicado por la derivada del denominador, todo dividido por el cuadrado del denominador.

**Ejemplo Pr谩ctico: Derivada de un Cociente de Funciones**

**Funciones**:

- $f(x) = x^2$ (Numerador)

- $g(x) = e^x$ (Denominador)

**Derivada Usando la Regla del Cociente**

Para estas funciones, primero calculamos las derivadas individuales:

- $f'(x) = 2x$

- $g'(x) = e^x$

Aplicando la regla del cociente:

$\left(\frac{f}{g}\right)'(x) = \frac{2x \cdot e^x - x^2 \cdot e^x}{(e^x)^2} = \frac{(2x - x^2)e^x}{e^{2x}} = \frac{2x - x^2}{e^x}$

**Visualizaci贸n con Python**

Vamos a graficar las funciones y la derivada de su cociente usando Python para visualizar estos c谩lculos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones y su derivada de cociente
def f(x):
    return x**2

def g(x):
    return np.exp(x)

def df(x):
    return 2*x

def dg(x):
    return np.exp(x)

def quotient_derivative(x):
    return (df(x) * g(x) - f(x) * dg(x)) / (g(x)**2)

# Generar valores de x
x_values = np.linspace(-2, 2, 400)
f_values = f(x_values)
g_values = g(x_values)
quotient_derivative_values = quotient_derivative(x_values)

# Crear gr谩fico
plt.figure(figsize=(12, 8))
plt.plot(x_values, f_values, label='$f(x) = x^2$', color='blue', linewidth=2)
plt.plot(x_values, g_values, label='$g(x) = e^x$', color='green', linewidth=2)
plt.plot(x_values, quotient_derivative_values, label="Quotient Derivative $(2x - x^2)/e^x$", color='red', linestyle='--', linewidth=2)
plt.title('Funciones y Derivada de su Cociente', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Valor', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.xlim(-2, 2)  # Limitar los valores de x
plt.ylim(-1, 10)  # Limitar los valores de y
plt.show()
```

**Interpretaci贸n**

- **Gr谩fico Azul**: Muestra la funci贸n $f(x) = x^2$, que es el numerador.

- **Gr谩fico Verde**: Representa la funci贸n $g(x) = e^x$, que es el denominador.

- **Gr谩fico Rojo**: Muestra la derivada del cociente $\frac{2x - x^2}{e^x}$. Este gr谩fico ayuda a visualizar c贸mo la tasa de cambio del cociente de estas dos funciones se comporta a lo largo del dominio considerado.

Esta visualizaci贸n y ejemplo matem谩tico demuestran c贸mo la Regla del Cociente simplifica el proceso de encontrar la derivada de un cociente de dos funciones, permitiendo entender c贸mo var铆a la relaci贸n entre dos cantidades cambiantes.

### **4.1.3 Regla de la cadena**

La **Regla de la Cadena** es una t茅cnica esencial en c谩lculo diferencial que se utiliza para derivar composiciones de funciones. Es decir, cuando una funci贸n est谩 compuesta por otras funciones, la Regla de la Cadena permite calcular la derivada de esta composici贸n compleja de una manera sistem谩tica y eficiente.

**Concepto de la Regla de la Cadena**

Si tienes dos funciones, $f(x)$ y $g(x)$, y formas una funci贸n compuesta $h(x) = f(g(x))$, la Regla de la Cadena establece que la derivada de $(h)$ respecto a $(x)$ es el producto de la derivada de $(f)$ respecto a $g(x)$ y la derivada de $(g)$ respecto a $(x)$. Matem谩ticamente, se expresa como:

$\frac{dh}{dx} = \frac{df}{dg} \cdot \frac{dg}{dx}$

O, usando una notaci贸n m谩s funcional:

$h'(x) = f'(g(x)) \cdot g'(x)$

**Ejemplo Pr谩ctico: Uso de la Regla de la Cadena**

Supongamos que queremos derivar la funci贸n $h(x) = \sin(x^2)$.

**Funciones involucradas**:

- $g(x) = x^2$

- $f(g) = \sin(g)$

**C谩lculo de Derivadas**

- Derivada de $g(x)$:

  $g'(x) = 2x$

- Derivada de $f(g)$:

  $f'(g) = \cos(g)$

Aplicando la Regla de la Cadena:

$h'(x) = \cos(x^2) \cdot 2x$

**Visualizaci贸n y C贸digo en Python**

Utilicemos Python para visualizar la funci贸n $\sin(x^2)$ y su derivada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n compuesta y su derivada usando la regla de la cadena
def h(x):
    return np.sin(x**2)

def dh(x):
    return 2 * x * np.cos(x**2)

# Rango de valores para x
x_values = np.linspace(-2*np.pi, 2*np.pi, 400)
h_values = h(x_values)
dh_values = dh(x_values)

# Crear gr谩fico
plt.figure(figsize=(12, 8))
plt.plot(x_values, h_values, label='$h(x) = \sin(x^2)$', color='blue', linewidth=2)
plt.plot(x_values, dh_values, label="$h'(x) = 2x \cos(x^2)$", color='red', linestyle='--', linewidth=2)
plt.title('Funci贸n Compuesta y su Derivada (Regla de la Cadena)', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('h(x), h\'(x)', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.xlim(-2*np.pi, 2*np.pi)  # Limitar los valores de x
plt.show()
```

**Interpretaci贸n**

- **Gr谩fico Azul**: Representa la funci贸n $h(x) = \sin(x^2)$. Esta gr谩fica muestra las oscilaciones del seno conforme el argumento $(x^2)$ aumenta m谩s r谩pidamente que $(x)$.

- **Gr谩fico Rojo**: Muestra la derivada de $h(x)$ calculada con la Regla de la Cadena, $h'(x) = 2x \cos(x^2)$. Esta curva indica c贸mo la tasa de cambio de la funci贸n compuesta var铆a a lo largo de $(x)$.

Este ejemplo demuestra la utilidad de la Regla de la Cadena para manejar derivadas de funciones compuestas, facilitando el c谩lculo de tasas de cambio en contextos m谩s complejos.
