### **1.1.3 Funciones a trozos**

Las **funciones a trozos** son funciones matem谩ticas que tienen diferentes expresiones o reglas dependiendo del intervalo del dominio al que pertenece el valor de entrada. Este tipo de funciones son muy 煤tiles para describir situaciones en las que el comportamiento de una funci贸n cambia bajo diferentes condiciones o en distintos rangos de valores.

**Concepto de Funciones a Trozos**

Una funci贸n a trozos se define mediante varias subfunciones, cada una aplicable a una parte espec铆fica del dominio de la funci贸n total. La notaci贸n general para una funci贸n a trozos es la siguiente:

$f(x) =

\vdots \\
f_n(x) & \text{si } x \in A_n
\end{cases}$

donde $(A_1, A_2, \ldots, A_n)$ son subconjuntos del dominio, usualmente intervalos.

**Ejemplo Pr谩ctico de una Funci贸n a Trozos**

Consideremos la funci贸n:

$f(x) =
\begin{cases}

\end{cases}$

Esta funci贸n usa una regla $(x^2)$ para valores de $(x)$ que son menores o iguales a cero y otra regla $(x + 2)$ para valores de $(x)$ que son positivos.

**Visualizaci贸n con Python**

Vamos a visualizar esta funci贸n usando Python y la biblioteca Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n a trozos
def f(x):
    return np.where(x <= 0, x**2, x + 2)

# Generar valores de x
x = np.linspace(-3, 3, 400)

# Evaluar la funci贸n
y = f(x)

# Crear la gr谩fica
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="f(x) = $x^2$ si $x \leq 0$\n      $x + 2$ si $x > 0$", color='blue', linewidth=2)
plt.plot(x[x <= 0], x[x <= 0]**2, label="$x^2$ si $x \leq 0$", linestyle='--', color='red')  # Segmento para x <= 0
plt.plot(x[x > 0], x[x > 0] + 2, label="$x + 2$ si $x > 0$", linestyle='-.', color='green')  # Segmento para x > 0
plt.title('Gr谩fica de la Funci贸n a Trozos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

**Interpretaci贸n**

- **Gr谩fica**: La gr谩fica muestra claramente dos comportamientos diferentes dependiendo del valor de $(x)$. Para $(x \leq 0)$, la funci贸n sigue la par谩bola $y = x^2$, mientras que para $(x > 0)$, la funci贸n se convierte en una l铆nea recta que corresponde a $y = x + 2$.

- **Aplicaciones**: Las funciones a trozos son 煤tiles en ingenier铆a y f铆sica para modelar fen贸menos que cambian sus reglas de comportamiento bajo diferentes condiciones, como los sistemas de control, la econom铆a (modelos de precios con descuentos por volumen), y m谩s.

Las funciones a trozos ofrecen una manera flexible de representar matem谩ticamente situaciones complejas en las que una 煤nica regla no puede describir adecuadamente todo el comportamiento de una funci贸n a lo largo de su dominio. ### **1.1.5 Multiplicaci贸n de binomios**

La **multiplicaci贸n de binomios** es un tema fundamental en 谩lgebra que implica el producto de expresiones algebraicas que tienen dos t茅rminos, conocidos como binomios. **Concepto de la Multiplicaci贸n de Binomios**

**F贸rmula y M茅todo**

La multiplicaci贸n de dos binomios generalmente sigue la regla del producto notable conocida como "producto de suma por suma", que se puede expresar como:

**Ejemplo Pr谩ctico**

**Visualizaci贸n con Python**

A continuaci贸n, proporcionar茅 un ejemplo en Python que ilustra c贸mo se puede realizar y visualizar la multiplicaci贸n de binomios utilizando el paquete `sympy`, el cual es 煤til para c谩lculos simb贸licos.

**C贸digo en Python para Multiplicar Binomios**

### **1.1.6 Racionalizar denominadores**

**Racionalizar denominadores** es un proceso algebraico utilizado para eliminar ra铆ces, como ra铆ces cuadradas o ra铆ces c煤bicas, del denominador de una fracci贸n. Este proceso es 煤til porque simplifica la manipulaci贸n y la evaluaci贸n de expresiones algebraicas, y es preferido para la presentaci贸n final de respuestas en matem谩ticas.

**Por qu茅 Racionalizar**

Cuando una fracci贸n contiene una ra铆z cuadrada o cualquier ra铆z en el denominador, puede ser dif铆cil trabajar con ella, especialmente si necesitas realizar sumas, restas o comparaciones. Racionalizar el denominador convierte la fracci贸n en una forma m谩s manejable.

**M茅todo B谩sico para Racionalizar Denominadores**

El m茅todo m谩s com煤n para racionalizar un denominador que contiene una ra铆z cuadrada involucra multiplicar tanto el numerador como el denominador por la ra铆z cuadrada que se desea eliminar. Esto se basa en la propiedad que dice que la ra铆z cuadrada de un n煤mero, multiplicada por s铆 misma, es igual al n煤mero original sin ra铆z.

**Ejemplo 1: Racionalizando un Denominador Simple**

Considera la fracci贸n:

Para racionalizar el denominador, multiplica tanto el numerador como el denominador por $(\sqrt{2})$:

$\frac{5}{\sqrt{2}} \times \frac{\sqrt{2}}{\sqrt{2}} = \frac{5\sqrt{2}}{2}$

**Ejemplo 2: Racionalizando un Denominador con una Suma de Ra铆ces**

Considera la fracci贸n:

$\frac{1}{1+\sqrt{3}}$

Multiplica por el conjugado del denominador para eliminar la ra铆z:

$\frac{1}{1+\sqrt{3}} \times \frac{1-\sqrt{3}}{1-\sqrt{3}} = \frac{1-\sqrt{3}}{1^2-(\sqrt{3})^2} = \frac{1-\sqrt{3}}{1-3} = \frac{1-\sqrt{3}}{-2}$

**Implementaci贸n en Python**

Podemos utilizar Python para ilustrar c贸mo racionalizar denominadores, mostrando el proceso y el resultado final con gr谩ficas cuando sea aplicable.

```python
import sympy as sp

# Definir las variables
x = sp.symbols('x')

# Ejemplo 1: Racionalizando un denominador simple
expr1 = 5 / sp.sqrt(2)
racionalizado1 = expr1 * sp.sqrt(2) / sp.sqrt(2)

# Ejemplo 2: Racionalizando una suma de ra铆ces
expr2 = 1 / (1 + sp.sqrt(3))
racionalizado2 = expr2 * (1 - sp.sqrt(3)) / (1 - sp.sqrt(3))

# Simplificar las expresiones racionalizadas
racionalizado1_simpl = sp.simplify(racionalizado1)
racionalizado2_simpl = sp.simplify(racionalizado2)

# Imprimir las expresiones racionalizadas y simplificadas
print("Ejemplo 1 - Racionalizado:")
sp.pprint(racionalizado1_simpl)
print("\nEjemplo 2 - Racionalizado:")
sp.pprint(racionalizado2_simpl)
```

### **1.1.7 Reglas de los exponentes**

Las **reglas de los exponentes**, tambi茅n conocidas como **leyes de los exponentes**, son principios fundamentales en 谩lgebra que facilitan la manipulaci贸n de expresiones que contienen potencias o exponentes. Estas reglas son esenciales para simplificar expresiones algebraicas, resolver ecuaciones, y realizar c谩lculos m谩s complejos en matem谩ticas avanzadas.

**Reglas B谩sicas de los Exponentes**

Aqu铆 describir茅 las reglas m谩s comunes de los exponentes con ejemplos para cada una:

**Producto de Potencias con la Misma Base**:

   $a^m \times a^n = a^{m+n}$

   Ejemplo: $2^3 \times 2^2 = 2^{3+2} = 2^5 = 32$.

**Cociente de Potencias con la Misma Base**:

   $\frac{a^m}{a^n} = a^{m-n}$

   Ejemplo: $\frac{2^5}{2^3} = 2^{5-3} = 2^2 = 4$.

**Potencia de una Potencia**:

   $(a^m)^n = a^{m \times n}$

   Ejemplo: $(2^3)^2 = 2^{3 \times 2} = 2^6 = 64$.

**Potencia de un Producto**:

   $(ab)^n = a^n \times b^n$

   Ejemplo: $(2 \times 3)^2 = 2^2 \times 3^2 = 4 \times 9 = 36$.

**Potencia de un Cociente**:

   $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$

   Ejemplo: $\left(\frac{2}{3}\right)^3 = \frac{2^3}{3^3} = \frac{8}{27}$.

**Exponente Cero**:

   $a^0 = 1 \text{ (siempre que } a \neq 0\text{)}$

   Ejemplo: $(2^0 = 1)$.

**Exponente Negativo**:

   $a^{-n} = \frac{1}{a^n} \text{ (siempre que } a \neq 0\text{)}$

   Ejemplo: $2^{-3} = \frac{1}{2^3} = \frac{1}{8}$.

**Visualizaci贸n y Ejemplo en Python**

**Definici贸n de Logaritmo**
