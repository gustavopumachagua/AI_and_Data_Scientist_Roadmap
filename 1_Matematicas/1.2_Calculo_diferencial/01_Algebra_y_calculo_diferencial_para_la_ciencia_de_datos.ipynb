### **1.1.1 ¿Qué es una función? (Algebraica y gráficamente)**

Una **función** es un concepto fundamental en matemáticas que describe una relación particular entre dos conjuntos de información. Cada entrada del conjunto inicial, llamado el dominio, está conectada a una salida en el conjunto final, llamado el codominio, de tal manera que a cada entrada le corresponde una única salida. Esta relación se puede expresar de varias maneras: verbalmente, numéricamente, visualmente y algebráicamente.

**Descripción Algebraica de una Función**

Desde el punto de vista algebraico, una función se suele escribir como:

$f: X \rightarrow Y$

donde $(X)$ es el dominio y $(Y)$ es el codominio. La función $(f)$ asigna cada elemento $(x)$ en $(X)$ a un elemento $f(x)$ en $(Y)$. Por ejemplo, la función

$f(x) = x^2$

define una regla donde cada número $(x)$ se mapea a su cuadrado $(x^2)$. Aquí, $(X)$ y $(Y)$ pueden ser el conjunto de todos los números reales $(\mathbb{R})$.

**Ejemplo Algebraico**

Considere la función:

$f(x) = 2x + 1$

Esta función toma cualquier número real $(x)$, lo duplica, y le suma uno para obtener la salida. Por ejemplo:

$f(3) = 2 \times 3 + 1 = 7$

**Descripción Gráfica de una Función**

Gráficamente, una función se representa mediante su gráfica en un sistema de coordenadas, donde el eje horizontal (eje $(x)$) representa los valores del dominio y el eje vertical (eje $(y)$) los valores del codominio. Cada par $(x, f(x))$ define un punto en esta gráfica.

**Ejemplo Gráfico**

Para visualizar la función $f(x) = 2x + 1$, podemos usar Python y la biblioteca Matplotlib para crear su gráfica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return 2 * x + 1

# Generar valores x
x = np.linspace(-10, 10, 400)

# Calcular valores y
y = f(x)

# Crear la gráfica
plt.figure(figsize=(10, 6))  # Aumentar el tamaño de la figura
plt.plot(x, y, label='f(x) = 2x + 1', color='blue', linewidth=2)  # Aumentar el grosor de la línea
plt.title('Gráfica de la función f(x) = 2x + 1', fontsize=16)  # Aumentar el tamaño de fuente del título
plt.xlabel('x', fontsize=14)  # Aumentar el tamaño de fuente del eje x
plt.ylabel('f(x)', fontsize=14)  # Aumentar el tamaño de fuente del eje y
plt.axhline(0, color='black',linewidth=0.5)  # Añadir línea horizontal en y=0
plt.axvline(0, color='black',linewidth=0.5)  # Añadir línea vertical en x=0
plt.grid(True, linestyle='--', alpha=0.5)  # Cambiar el estilo de la cuadrícula y la opacidad
plt.legend(fontsize=12)  # Aumentar el tamaño de fuente de la leyenda
plt.show()
```

**Interpretación Gráfica**

La gráfica muestra una línea recta que intercepta el eje $(y)$ en 1 (el término constante de la función) y tiene una pendiente de 2 (el coeficiente de $(x)$). Esto significa que por cada unidad que $(x)$ se incrementa, $f(x)$ aumenta en 2 unidades.

**Conclusiones**

Una función es una herramienta matemática esencial que permite relacionar de manera única elementos de un conjunto con elementos de otro. Puede describirse de muchas maneras, pero las representaciones algebraicas y gráficas son particularmente útiles para entender la relación entre variables y para prever cómo cambiará una variable en respuesta a otra. Estas representaciones son utilizadas en ciencias, ingeniería, economía y muchas otras áreas para modelar y resolver problemas reales.

### **1.1.2 Dominio y alcance**

El **dominio** y el **alcance** (o rango) son conceptos fundamentales en el estudio de funciones en matemáticas. Estos términos describen respectivamente el conjunto de todas las entradas posibles para una función y el conjunto de todas las salidas posibles que resultan de esas entradas.

**Dominio de una Función**

El dominio de una función es el conjunto de todos los valores para los cuales la función está definida. En otras palabras, son todos los posibles valores de entrada que puedes poner en la función sin causar errores matemáticos como la división por cero o la raíz cuadrada de un número negativo.

**Ejemplo 1: Función Lineal**

Considera la función $f(x) = 2x + 3$. Esta es una función lineal y no tiene restricciones en los valores de $(x)$ que pueden ser usados. Por lo tanto, el dominio de esta función es todos los números reales, denotado matemáticamente como:

$\text{Dominio} = \mathbb{R}$

**Ejemplo 2: Función con Raíz Cuadrada**

Considera la función $g(x) = \sqrt{x - 2}$. Esta función solo está definida cuando $(x - 2 \geq 0)$, lo que implica que $(x \geq 2)$. Por lo tanto, el dominio de $(g)$ es:

$\text{Dominio} = [2, \infty)$

**Alcance de una Función**

El alcance (o rango) de una función es el conjunto de todos los posibles valores de salida que la función puede producir al tomar valores dentro de su dominio.

**Ejemplo 1: Función Lineal**

Para la función $f(x) = 2x + 3$, cualquier número real puede ser alcanzado ajustando $(x)$ adecuadamente. Por lo tanto, el alcance de $(f)$ es:

$\text{Rango} = \mathbb{R}$

**Ejemplo 2: Función con Raíz Cuadrada**

Para $g(x) = \sqrt{x - 2}$, el menor valor de $(x)$ es 2, lo que hace que $g(x)$ sea 0. A medida que $(x)$ aumenta, $(\sqrt{x - 2})$ también aumenta sin límite superior. Por lo tanto, el alcance de $(g)$ es:

$\text{Rango} = [0, \infty) $

**Visualización con Python**

A continuación, te proporciono un código en Python que utiliza Matplotlib para visualizar las funciones mencionadas y demostrar gráficamente sus dominios y rangos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Función lineal f(x) = 2x + 3
x1 = np.linspace(-10, 10, 400)
y1 = 2 * x1 + 3

# Función con raíz cuadrada g(x) = sqrt(x - 2)
x2 = np.linspace(2, 10, 400)
y2 = np.sqrt(x2 - 2)

# Crear gráficos
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, label='f(x) = 2x + 3', color='blue', linewidth=2)  # Aumentar el grosor de la línea
plt.title('Función Lineal', fontsize=16)  # Aumentar el tamaño de fuente del título
plt.xlabel('x', fontsize=14)  # Aumentar el tamaño de fuente del eje x
plt.ylabel('f(x)', fontsize=14)  # Aumentar el tamaño de fuente del eje y
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)  # Cambiar el estilo de la cuadrícula y la opacidad
plt.legend(fontsize=12)  # Aumentar el tamaño de fuente de la leyenda

plt.subplot(1, 2, 2)
plt.plot(x2, y2, label='$g(x) = \sqrt{x - 2}$', color='red', linewidth=2)  # Aumentar el grosor de la línea
plt.title('Función con Raíz Cuadrada', fontsize=16)  # Aumentar el tamaño de fuente del título
plt.xlabel('x', fontsize=14)  # Aumentar el tamaño de fuente del eje x
plt.ylabel('g(x)', fontsize=14)  # Aumentar el tamaño de fuente del eje y
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)  # Cambiar el estilo de la cuadrícula y la opacidad
plt.legend(fontsize=12)  # Aumentar el tamaño de fuente de la leyenda

plt.tight_layout()  # Ajustar automáticamente el diseño de los subgráficos
plt.show()
```

**Conclusión**

El dominio y el rango son conceptos esenciales para entender cómo funcionan las funciones matemáticas y son cruciales para resolver problemas en ciencias, ingeniería y matemáticas. Comprender estos conceptos permite a los estudiantes y profesionales predecir el comportamiento de las funciones y determinar sus limitaciones prácticas.

### **1.1.3 Funciones a trozos**

Las **funciones a trozos** son funciones matemáticas que tienen diferentes expresiones o reglas dependiendo del intervalo del dominio al que pertenece el valor de entrada. Este tipo de funciones son muy útiles para describir situaciones en las que el comportamiento de una función cambia bajo diferentes condiciones o en distintos rangos de valores.

**Concepto de Funciones a Trozos**

Una función a trozos se define mediante varias subfunciones, cada una aplicable a una parte específica del dominio de la función total. La notación general para una función a trozos es la siguiente:

$f(x) =
\begin{cases}
f_1(x) & \text{si } x \in A_1 \\
f_2(x) & \text{si } x \in A_2 \\
\vdots \\
f_n(x) & \text{si } x \in A_n
\end{cases}$

donde $(A_1, A_2, \ldots, A_n)$ son subconjuntos del dominio, usualmente intervalos.

**Ejemplo Práctico de una Función a Trozos**

Consideremos la función:

$f(x) =
\begin{cases}
x^2 & \text{si } x \leq 0 \\
x + 2 & \text{si } x > 0
\end{cases}$

Esta función usa una regla $(x^2)$ para valores de $(x)$ que son menores o iguales a cero y otra regla $(x + 2)$ para valores de $(x)$ que son positivos.

**Visualización con Python**

Vamos a visualizar esta función usando Python y la biblioteca Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a trozos
def f(x):
    return np.where(x <= 0, x**2, x + 2)

# Generar valores de x
x = np.linspace(-3, 3, 400)

# Evaluar la función
y = f(x)

# Crear la gráfica
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="f(x) = $x^2$ si $x \leq 0$\n      $x + 2$ si $x > 0$", color='blue', linewidth=2)
plt.plot(x[x <= 0], x[x <= 0]**2, label="$x^2$ si $x \leq 0$", linestyle='--', color='red')  # Segmento para x <= 0
plt.plot(x[x > 0], x[x > 0] + 2, label="$x + 2$ si $x > 0$", linestyle='-.', color='green')  # Segmento para x > 0
plt.title('Gráfica de la Función a Trozos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

**Interpretación**

- **Gráfica**: La gráfica muestra claramente dos comportamientos diferentes dependiendo del valor de $(x)$. Para $(x \leq 0)$, la función sigue la parábola $y = x^2$, mientras que para $(x > 0)$, la función se convierte en una línea recta que corresponde a $y = x + 2$.

- **Aplicaciones**: Las funciones a trozos son útiles en ingeniería y física para modelar fenómenos que cambian sus reglas de comportamiento bajo diferentes condiciones, como los sistemas de control, la economía (modelos de precios con descuentos por volumen), y más.

Las funciones a trozos ofrecen una manera flexible de representar matemáticamente situaciones complejas en las que una única regla no puede describir adecuadamente todo el comportamiento de una función a lo largo de su dominio. Este tipo de funciones es crucial en campos donde los modelos matemáticos deben adaptarse a múltiples condiciones o regímenes operativos.

### **1.1.4 ¿Qué se le puede hacer a una función?**

Cuando trabajamos con funciones en matemáticas, existen diversas operaciones y transformaciones que podemos aplicar para explorar sus propiedades, analizar su comportamiento o adaptarlas para aplicaciones específicas. Aquí, exploraremos algunas de las modificaciones y análisis más comunes que se pueden realizar en funciones.

**Operaciones y Transformaciones Comunes en Funciones**

**Evaluación**: Calcular el valor de la función para un valor específico de entrada.

**Simplificación**: Reducir la función a una forma más simple que sea equivalente.

**Expansión**: Expresar la función en una forma extendida, descomponiendo términos cuando sea posible.

**Diferenciación**: Encontrar la derivada de la función, que mide cómo cambia la salida de la función respecto a un cambio en su entrada.

**Integración**: Determinar la integral de la función, que puede interpretarse como el área bajo la curva de la función en un intervalo dado.

**Transformación**: Cambiar la función aplicando escalado, traslación u otras modificaciones a su gráfica.

**Inversión**: Encontrar la función inversa, que invierte la relación entre las entradas y las salidas.

**Composición**: Combinar dos o más funciones en una sola función que representa la aplicación secuencial de estas.

**Limitación**: Determinar el comportamiento de la función conforme la entrada se aproxima a un valor específico.

**Ejemplo Detallado**

Consideremos la función $f(x) = x^2$. Podemos aplicar varias de las operaciones mencionadas para analizar y modificar esta función.

**Visualización con Python**

Usaremos Python para demostrar algunas transformaciones básicas sobre $f(x) = x^2$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original
def f(x):
    return x**2

# Generar valores de x
x = np.linspace(-10, 10, 400)

# Evaluar la función original
y = f(x)

# Crear gráficas de transformaciones
plt.figure(figsize=(12, 8))

# Gráfica original
plt.subplot(2, 2, 1)
plt.plot(x, y, label='Original $f(x) = x^2$')
plt.title('Función Original')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Transformación por traslación
plt.subplot(2, 2, 2)
plt.plot(x, f(x - 2), label='Traslación $f(x-2)$')
plt.title('Traslación Horizontal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Transformación por escalado
plt.subplot(2, 2, 3)
plt.plot(x, f(x) * 2, label='Escala $2f(x)$')
plt.title('Escala Vertical')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Función inversa
plt.subplot(2, 2, 4)
y_inversa = np.sqrt(x[x >= 0])  # Solo valores no negativos para la raíz cuadrada
plt.plot(x[x >= 0], y_inversa, label='Inversa $f^{-1}(x) = \sqrt{x}$')
plt.title('Función Inversa')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.legend()
plt.tight_layout()
plt.show()
```

**Explicación de las Transformaciones**

- **Función Original**: La curva parabólica típica de $f(x) = x^2$.

- **Traslación Horizontal**: La función $f(x-2)$ representa una traslación hacia la derecha de la parábola por 2 unidades.

- **Escala Vertical**: Multiplicar la función por 2, $2f(x)$, estira la parábola verticalmente, duplicando sus valores.

- **Función Inversa**: La inversa de $(x^2)$ para $(x \geq 0)4$ es $(\sqrt{x})$, mostrando cómo se "desdobla" la función cuadrática.

Cada una de estas operaciones tiene aplicaciones específicas en contextos matemáticos, físicos y de ingeniería, facilitando el análisis de fenómenos complejos o la adaptación de modelos teóricos a condiciones experimentales.

### **1.1.5 Multiplicación de binomios**

La **multiplicación de binomios** es un tema fundamental en álgebra que implica el producto de expresiones algebraicas que tienen dos términos, conocidos como binomios. A través de la multiplicación de binomios, se puede expandir y simplificar expresiones que son esenciales para resolver ecuaciones y otros problemas algebraicos.

**Concepto de la Multiplicación de Binomios**

Un binomio es una expresión algebraica que contiene dos términos, como $(a + b)$. Cuando multiplicamos dos binomios, por ejemplo, $(a + b)$ y $(c + d)$, aplicamos la propiedad distributiva para obtener cada término del producto.

**Fórmula y Método**

La multiplicación de dos binomios generalmente sigue la regla del producto notable conocida como "producto de suma por suma", que se puede expresar como:

$(a + b)(c + d) = ac + ad + bc + bd$

**Ejemplo Práctico**

Considera los binomios $(x + 3)$ y $(x - 2)$. Para multiplicar estos dos binomios, aplicamos la fórmula:

$(x + 3)(x - 2) = x \cdot x + x \cdot (-2) + 3 \cdot x + 3 \cdot (-2) = x^2 - 2x + 3x - 6$

Simplificando, obtenemos:

$x^2 + x - 6$

**Visualización con Python**

A continuación, proporcionaré un ejemplo en Python que ilustra cómo se puede realizar y visualizar la multiplicación de binomios utilizando el paquete `sympy`, el cual es útil para cálculos simbólicos.

**Código en Python para Multiplicar Binomios**

```python
import sympy as sp

# Definir las variables
x = sp.symbols('x')

# Definir los binomios
binomio1 = x + 3
binomio2 = x - 2

# Multiplicar los binomios
producto = sp.expand(binomio1 * binomio2)

# Mostrar el producto de los binomios
print("El producto de los binomios (x + 3) y (x - 2) es:", producto)
```

Este código realiza la multiplicación de los binomios y muestra el resultado simplificado. Para visualizar la gráfica de la función resultante, puedes usar el siguiente fragmento de código:

**Visualización Gráfica**

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir una función para el polinomio resultante
x = sp.symbols('x')
producto = x**3 - 2*x**2 + 3*x - 4
f = sp.lambdify(x, producto, 'numpy')

# Generar datos para x
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

# Crear la gráfica
plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_vals, label='f(x) = $x^3 - 2x^2 + 3x - 4$', color='blue', linestyle='-', marker='o', markersize=3)
plt.title('Gráfica del Polinomio Resultante', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend(fontsize=12)
plt.show()
```

**Interpretación**

- **Polinomio Resultante**: La expresión $(x^2 + x - 6)$ es un polinomio cuadrático que se deriva de la multiplicación de los dos binomios.

- **Visualización Gráfica**: La gráfica muestra el comportamiento del polinomio, incluyendo sus raíces y su forma parabólica, típica de las funciones cuadráticas.

La multiplicación de binomios es una herramienta esencial en álgebra que permite desarrollar y simplificar expresiones complejas. Es fundamental para estudiantes y profesionales que trabajan en matemáticas, ingeniería y ciencias aplicadas.

### **1.1.6 Racionalizar denominadores**

**Racionalizar denominadores** es un proceso algebraico utilizado para eliminar raíces, como raíces cuadradas o raíces cúbicas, del denominador de una fracción. Este proceso es útil porque simplifica la manipulación y la evaluación de expresiones algebraicas, y es preferido para la presentación final de respuestas en matemáticas.

**Por qué Racionalizar**

Cuando una fracción contiene una raíz cuadrada o cualquier raíz en el denominador, puede ser difícil trabajar con ella, especialmente si necesitas realizar sumas, restas o comparaciones. Racionalizar el denominador convierte la fracción en una forma más manejable.

**Método Básico para Racionalizar Denominadores**

El método más común para racionalizar un denominador que contiene una raíz cuadrada involucra multiplicar tanto el numerador como el denominador por la raíz cuadrada que se desea eliminar. Esto se basa en la propiedad que dice que la raíz cuadrada de un número, multiplicada por sí misma, es igual al número original sin raíz.

**Ejemplo 1: Racionalizando un Denominador Simple**

Considera la fracción:

$\frac{5}{\sqrt{2}}$

Para racionalizar el denominador, multiplica tanto el numerador como el denominador por $(\sqrt{2})$:

$\frac{5}{\sqrt{2}} \times \frac{\sqrt{2}}{\sqrt{2}} = \frac{5\sqrt{2}}{2}$

**Ejemplo 2: Racionalizando un Denominador con una Suma de Raíces**

Considera la fracción:

$\frac{1}{1+\sqrt{3}}$

Multiplica por el conjugado del denominador para eliminar la raíz:

$\frac{1}{1+\sqrt{3}} \times \frac{1-\sqrt{3}}{1-\sqrt{3}} = \frac{1-\sqrt{3}}{1^2-(\sqrt{3})^2} = \frac{1-\sqrt{3}}{1-3} = \frac{1-\sqrt{3}}{-2}$

**Implementación en Python**

Podemos utilizar Python para ilustrar cómo racionalizar denominadores, mostrando el proceso y el resultado final con gráficas cuando sea aplicable.

```python
import sympy as sp

# Definir las variables
x = sp.symbols('x')

# Ejemplo 1: Racionalizando un denominador simple
expr1 = 5 / sp.sqrt(2)
racionalizado1 = expr1 * sp.sqrt(2) / sp.sqrt(2)

# Ejemplo 2: Racionalizando una suma de raíces
expr2 = 1 / (1 + sp.sqrt(3))
racionalizado2 = expr2 * (1 - sp.sqrt(3)) / (1 - sp.sqrt(3))

# Simplificar las expresiones racionalizadas
racionalizado1_simpl = sp.simplify(racionalizado1)
racionalizado2_simpl = sp.simplify(racionalizado2)

# Imprimir las expresiones racionalizadas y simplificadas
print("Ejemplo 1 - Racionalizado:")
sp.pprint(racionalizado1_simpl)
print("\nEjemplo 2 - Racionalizado:")
sp.pprint(racionalizado2_simpl)
```

**Visualización**

Podemos visualizar la simplificación de las expresiones, aunque en este caso específico, las gráficas no son la herramienta más ilustrativa, dado que estamos tratando con operaciones algebraicas puras más que con funciones variables. Sin embargo, `sympy` permite manipular y simplificar estas expresiones fácilmente.

**Conclusión**

Racionalizar denominadores es una técnica importante en matemáticas que facilita otras operaciones matemáticas y proporciona una forma estandarizada de presentar respuestas. Es especialmente relevante en educación y en cualquier contexto donde la claridad y la precisión son críticas.

### **1.1.7 Reglas de los exponentes**

Las **reglas de los exponentes**, también conocidas como **leyes de los exponentes**, son principios fundamentales en álgebra que facilitan la manipulación de expresiones que contienen potencias o exponentes. Estas reglas son esenciales para simplificar expresiones algebraicas, resolver ecuaciones, y realizar cálculos más complejos en matemáticas avanzadas.

**Reglas Básicas de los Exponentes**

Aquí describiré las reglas más comunes de los exponentes con ejemplos para cada una:

**Producto de Potencias con la Misma Base**:

   $a^m \times a^n = a^{m+n}$

   Ejemplo: $2^3 \times 2^2 = 2^{3+2} = 2^5 = 32$.

**Cociente de Potencias con la Misma Base**:

   $\frac{a^m}{a^n} = a^{m-n}$

   Ejemplo: $\frac{2^5}{2^3} = 2^{5-3} = 2^2 = 4$.

**Potencia de una Potencia**:

   $(a^m)^n = a^{m \times n}$

   Ejemplo: $(2^3)^2 = 2^{3 \times 2} = 2^6 = 64$.

**Potencia de un Producto**:

   $(ab)^n = a^n \times b^n$

   Ejemplo: $(2 \times 3)^2 = 2^2 \times 3^2 = 4 \times 9 = 36$.

**Potencia de un Cociente**:

   $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$

   Ejemplo: $\left(\frac{2}{3}\right)^3 = \frac{2^3}{3^3} = \frac{8}{27}$.

**Exponente Cero**:

   $a^0 = 1 \text{ (siempre que } a \neq 0\text{)}$

   Ejemplo: $(2^0 = 1)$.

**Exponente Negativo**:

   $a^{-n} = \frac{1}{a^n} \text{ (siempre que } a \neq 0\text{)}$

   Ejemplo: $2^{-3} = \frac{1}{2^3} = \frac{1}{8}$.

**Visualización y Ejemplo en Python**

Podemos usar Python para visualizar el efecto de algunas de estas reglas en gráficos. A continuación, proporciono un código que muestra cómo se pueden representar gráficamente estas reglas usando la biblioteca Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos de valores para x (evitando el cero para el caso de los exponentes negativos)
x = np.linspace(0.1, 3, 400)
y1 = x**2  # x al cuadrado
y2 = x**3  # x al cubo
y3 = x**(-1)  # x al exponente -1

# Crear gráficos
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='$x^2$', color='blue', linestyle='-', linewidth=2)
plt.title('$x^2$', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.5)

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='$x^3$', color='red', linestyle='--', linewidth=2)
plt.title('$x^3$', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.5)

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='$x^{-1}$', color='green', linestyle='-.', linewidth=2)
plt.title('$x^{-1}$', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

**Explicación del Código**

- El código genera tres subplots para visualizar $(x^2)$, $(x^3)$, y $(x^{-1})$.

- Estas gráficas ayudan a entender visualmente cómo se comportan las funciones con diferentes exponentes en términos de su crecimiento o decrecimiento.

Las reglas de los exponentes son herramientas cruciales en el álgebra que permiten simplificar y manipular expresiones y ecuaciones que involucran potencias. ### **1.1.8 Logaritmos**

Los **logaritmos** son una herramienta fundamental en matemáticas, utilizada para resolver ecuaciones en las que la incógnita aparece como exponente de alguna base. Son la operación inversa de la exponenciación, lo que significa que el logaritmo de un número es el exponente al que otro número fijo, la base, debe ser elevado para producir ese número.

**Definición de Logaritmo**

El logaritmo de un número $(y)$ en base $(a)$ se define como el exponente $(x)$ al que debe ser elevada la base $(a)$ para obtener $(y)$. Matemáticamente, esto se expresa como:

$\log_a y = x \quad \text{si y sólo si} \quad a^x = y$

donde:

- $(a)$ es la base del logaritmo y debe ser un número positivo diferente de 1.

- $(y)$ es el número del cual estamos tomando el logaritmo, y debe ser positivo.

**Tipos de Logaritmos**

**Logaritmo Natural (ln)**: Tiene como base el número de Euler $(e \approx 2.71828)$. Se denota como $(\ln(y))$ y es ampliamente utilizado en ciencias e ingeniería.

**Logaritmo Común (log)**: Tiene como base el número 10. Se usa comúnmente en ciencias e ingeniería, y a menudo se denota como $(\log(y))$.

**Logaritmo Binario**: Tiene como base el número 2 y es muy utilizado en informática.

**Propiedades de los Logaritmos**

Los logaritmos tienen varias propiedades que facilitan el cálculo con ellos:

**Producto**:

   $\log_a(xy) = \log_a x + \log_a y$

**Cociente**:

   $\log_a\left(\frac{x}{y}\right) = \log_a x - \log_a y$

**Potencia**:

   $\log_a(x^b) = b \cdot \log_a x$

**Cambio de Base**:

   $\log_b x = \frac{\log_a x}{\log_a b}$

**Ejemplos con Cálculos**

Supongamos que queremos calcular el logaritmo en base 10 de 100.

**Cálculo Analítico**:

$\log_{10} 100 = x \quad \text{significa que} \quad 10^x = 100$

Es fácil ver que $(10^2 = 100)$, entonces:

$\log_{10} 100 = 2$

**Visualización y Ejemplo en Python**

Podemos usar Python para ilustrar cómo los logaritmos varían con diferentes bases y cómo se aplican las propiedades mencionadas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una serie de puntos para x
x = np.linspace(0.1, 10, 400)
y_ln = np.log(x)  # Logaritmo natural
y_log10 = np.log10(x)  # Logaritmo base 10
y_log2 = np.log2(x)  # Logaritmo base 2

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y_ln, label='ln(x)', color='green', linestyle='-', linewidth=2)
plt.plot(x, y_log10, label='log10(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, y_log2, label='log2(x)', color='red', linestyle='-.', linewidth=2)
plt.title('Logaritmos Naturales, Base 10 y Base 2', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('log(x)', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(1, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```



**Interpretación**

- **Gráfica**: La gráfica muestra cómo el logaritmo natural, el logaritmo común y el logaritmo binario cambian con $(x)$. Cada logaritmo aumenta a una tasa diferente debido a su base diferente.

- **Uso**: Los logaritmos son útiles en diversos campos para modelar crecimiento exponencial o decaimiento, ajustar escalas en gráficos y resolver ecuaciones que involucran exponentes.

Los logaritmos no solo facilitan la manipulación algebraica sino que también proporcionan una forma de entender cambios y patrones en datos que varían exponencialmente.

### **1.1.9 Aplicaciones de los logaritmos**

Los logaritmos son una herramienta matemática versátil con numerosas aplicaciones en ciencia, ingeniería, economía, y otras disciplinas. Estas aplicaciones aprovechan las propiedades de los logaritmos para simplificar el cálculo y proporcionar una mejor comprensión de los fenómenos que involucran crecimiento exponencial o relaciones multiplicativas.

**Aplicaciones Principales de los Logaritmos**

**Crecimiento Exponencial y Decaimiento**:

   - **Biología**: El crecimiento de poblaciones puede modelarse con ecuaciones que utilizan exponenciales y logaritmos, como en modelos logísticos o de crecimiento bacteriano.

   - **Química**: Los logaritmos se usan para calcular el pH, que es una medida logarítmica de la concentración de iones de hidrógeno en una solución.

**Ciencias de la Tierra**:

   - **Geología**: Los logaritmos ayudan a determinar la magnitud de los terremotos a través de la Escala de Richter, que es logarítmica.

   - **Astronomía**: Las escalas de magnitudes estelares también son logarítmicas, utilizadas para medir el brillo de las estrellas.

**Matemáticas Financieras**:

   - **Interés Compuesto**: Los logaritmos permiten calcular el tiempo necesario para alcanzar cierta cantidad de dinero con una tasa de interés compuesto.

   - **Amortización de Hipotecas**: Los cálculos para determinar pagos mensuales o la duración de un préstamo a menudo involucran logaritmos.

**Ingeniería**:

   - **Teoría de Control y Circuitos Eléctricos**: Los logaritmos se usan en la Ley de Ohm y en el análisis de circuitos RC y RL para describir cambios en corrientes y voltajes.

   - **Acústica**: El decibelio, una unidad logarítmica, mide la intensidad del sonido.

**Informática**:

   - **Análisis de Algoritmos**: La complejidad de los algoritmos frecuentemente se expresa en términos de funciones logarítmicas, como en los tiempos de ejecución de algoritmos de búsqueda y ordenamiento.

**Ejemplo Práctico: Cálculo de Interés Compuesto**

Supongamos que queremos saber cuánto tiempo tardará en duplicarse una inversión con un interés compuesto anual del 5%. La fórmula para el valor futuro $(A)$ de una inversión inicial $(P)$ después de $(t)$ años a una tasa de interés $(r)$ es:

$A = P \times (1 + r)^t$

Para encontrar el tiempo necesario para duplicar la inversión, $(A = 2P)$:

$2 = (1 + 0.05)^t$

Aplicando logaritmos para resolver para $(t)$:

$\log(2) = t \log(1.05) \implies t = \frac{\log(2)}{\log(1.05)}$

**Implementación en Python**

Podemos calcular esto en Python y visualizar cómo cambia la inversión con el tiempo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la tasa de interés y el cálculo de t
interest_rate = 0.05
t_doubling = np.log(2) / np.log(1 + interest_rate)

# Crear valores de tiempo y calcular crecimiento de la inversión
years = np.linspace(0, 20, 400)
investment_growth = 2**(years / t_doubling)

# Visualización
plt.figure(figsize=(10, 5))
plt.plot(years, investment_growth, label='Crecimiento de la Inversión', color='blue', linestyle='-')
plt.axhline(2, color='red', linestyle='--', label='Doblar la Inversión', linewidth=1)
plt.title('Crecimiento de la Inversión con Interés Compuesto', fontsize=16)
plt.xlabel('Años', fontsize=14)
plt.ylabel('Multiplicador de la Inversión Inicial', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

print("Tiempo necesario para duplicar la inversión:", t_doubling, "años")
```

**Interpretación**

- **Gráfico**: Muestra cómo la inversión inicial aumenta con el tiempo bajo una tasa de interés compuesto del 5%.

- **Tiempo de Duplicación**: El cálculo muestra que se necesitan aproximadamente 14.2 años para duplicar la inversión.

Los logaritmos simplifican el entendimiento y la manipulación de problemas que involucran relaciones exponenciales y multiplicativas, haciendo posible realizar cálculos que de otro modo serían muy complejos o imposibles de resolver analíticamente.
