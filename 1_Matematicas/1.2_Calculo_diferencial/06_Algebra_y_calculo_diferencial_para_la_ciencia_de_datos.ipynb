# **6. √Ålgebra y c√°lculo diferencial para la ciencia de datos**

## **6.1 Encontrar m√°ximo y m√≠nimos**

Utilice las derivadas para hallar los valores m√°ximo y m√≠nimo de las funciones.

### **6.1.1 M√°ximos y m√≠nimos de funciones**

Los m√°ximos y m√≠nimos de funciones son puntos cr√≠ticos donde una funci√≥n alcanza su mayor o menor valor, respectivamente, dentro de un intervalo dado. Estos puntos son esenciales para la optimizaci√≥n y se utilizan en diversas aplicaciones pr√°cticas, desde la f√≠sica hasta la econom√≠a.

**Tipos de M√°ximos y M√≠nimos**

**M√°ximos y M√≠nimos Locales**

- **M√°ximo Local**: Un punto $(x = c)$ es un m√°ximo local de $f(x)$ si $f(c)$ es mayor que los valores de $f(x)$ en todos los puntos cercanos a $(c)$.

- **M√≠nimo Local**: Un punto $(x = c)$ es un m√≠nimo local de $f(x)$ si $f(c)$ es menor que los valores de $f(x)$ en todos los puntos cercanos a $(c)$.

**M√°ximos y M√≠nimos Absolutos (Globales)**

- **M√°ximo Absoluto**: Un punto $(x = c)$ es un m√°ximo absoluto de $f(x)$ en un intervalo si $f(c)$ es el mayor valor que $f(x)$ alcanza en ese intervalo.

- **M√≠nimo Absoluto**: Un punto $(x = c)$ es un m√≠nimo absoluto de $f(x)$ en un intervalo si $f(c)$ es el menor valor que $f(x)$ alcanza en ese intervalo.

**M√©todos para Encontrar M√°ximos y M√≠nimos**

**Test de la Primera Derivada**

   - Si $f'(x)$ cambia de positivo a negativo en $(x = c)$, entonces $f(c)$ es un m√°ximo local.

   - Si $f'(x)$ cambia de negativo a positivo en $(x = c)$, entonces $f(c)$ es un m√≠nimo local.

**Test de la Segunda Derivada**

   - Si $f''(c) > 0$, entonces $f(c)$ es un m√≠nimo local (concavidad hacia arriba).

   - Si $f''(c) < 0$, entonces $f(c)$ es un m√°ximo local (concavidad hacia abajo).

**Ejemplo Pr√°ctico: An√°lisis de $f(x) = x^3 - 3x^2 + 4$**

Vamos a determinar los m√°ximos y m√≠nimos de esta funci√≥n, tanto locales como absolutos.

**Funci√≥n**:

$f(x) = x^3 - 3x^2 + 4$

**Derivadas**:

- **Primera Derivada**:

  $f'(x) = 3x^2 - 6x$

- **Segunda Derivada**:

  $f''(x) = 6x - 6$

**Encontrar Puntos Cr√≠ticos**:

$f'(x) = 0 \rightarrow 3x^2 - 6x = 0 \rightarrow x(x-2) = 0 \rightarrow x = 0, \; x = 2$

**Visualizaci√≥n con Python**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n y sus derivadas
def f(x):
    return x**3 - 3*x**2 + 4

def df(x):
    return 3*x**2 - 6*x

def ddf(x):
    return 6*x - 6

# Generar valores de x
x_values = np.linspace(-1, 3, 400)
y_values = f(x_values)
df_values = df(x_values)
ddf_values = ddf(x_values)

# Crear gr√°fico
plt.figure(figsize=(12, 7))
plt.plot(x_values, y_values, label='f(x) = $x^3 - 3x^2 + 4$', color='navy', linestyle='-', linewidth=2)
plt.plot(x_values, df_values, label="f'(x) = $3x^2 - 6x$", color='darkred', linestyle='--', linewidth=2)
plt.plot(x_values, ddf_values, label="f''(x) = $6x - 6$", color='darkgreen', linestyle=':', linewidth=2)
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.title('An√°lisis de M√°ximos, M√≠nimos y Puntos de Inflexi√≥n', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(frameon=True, loc='best', shadow=True)
plt.show()
```

**Interpretaci√≥n**

- **Funci√≥n $f(x)$**: Muestra el comportamiento de la funci√≥n. Observamos posibles m√°ximos y m√≠nimos en $(x = 0)$ y $(x = 2)$.

- **Primera Derivada $f'(x)$**: Las ra√≠ces de $f'(x)$ indican los puntos donde la funci√≥n puede tener m√°ximos o m√≠nimos. La derivada tambi√©n muestra d√≥nde la funci√≥n es creciente o decreciente.

- **Segunda Derivada $f''(x)$**: Ayuda a confirmar si los puntos cr√≠ticos son m√°ximos o m√≠nimos seg√∫n la concavidad indicada por la segunda derivada.

Este an√°lisis proporciona una visi√≥n completa de c√≥mo los m√©todos de derivadas pueden ser utilizados para identificar caracter√≠sticas cr√≠ticas de la funci√≥n, facilitando la optimizaci√≥n y otros an√°lisis importantes en diversas aplicaciones.

### **6.1.2 Aplicaciones de m√°ximos y m√≠nimos**

Las aplicaciones de m√°ximos y m√≠nimos en matem√°ticas son vastas y cruciales en una multitud de campos como la econom√≠a, ingenier√≠a, f√≠sica, y m√°s. Estos conceptos se utilizan para optimizar soluciones y resolver problemas pr√°cticos en los que se busca maximizar o minimizar alguna cantidad. Aqu√≠, exploraremos c√≥mo se utilizan los m√°ximos y m√≠nimos para resolver problemas pr√°cticos con un enfoque en un ejemplo espec√≠fico.

**Ejemplo Pr√°ctico: Maximizaci√≥n del √Årea**

Supongamos que necesitas construir un corral rectangular que est√© pegado a un edificio, utilizando el edificio como uno de los lados del corral. Dispones de 100 metros de valla para los otros tres lados del corral. ¬øCu√°l es la mayor √°rea que puedes encerrar?

**Descripci√≥n del problema**:

- **Per√≠metro disponible para tres lados**: 100 metros.

- **Variables**:

  - $(x)$: los dos lados del corral que son paralelos al edificio.

  - $(y)$: el lado del corral perpendicual al edificio (solo se necesita una valla para este lado).

**Relaci√≥n de Per√≠metro**:

$2x + y = 100$

**√Årea a Maximizar**:

$A = x \times y$

**Formulaci√≥n Matem√°tica**

Primero, expresamos $(y)$ en t√©rminos de $(x)$ usando la restricci√≥n del per√≠metro:

$y = 100 - 2x$

Sustituir en la f√≥rmula del √°rea:

$A(x) = x(100 - 2x) = 100x - 2x^2$

**Derivaci√≥n para Encontrar M√°ximos**

Para maximizar el √°rea, derivamos $A(x)$ respecto a $(x)$ y encontramos los puntos cr√≠ticos:

$A'(x) = 100 - 4x$

Igualamos a cero para encontrar los puntos cr√≠ticos:

$100 - 4x = 0$

$x = 25 \, \text{metros}$

Calculamos $(y)$ cuando $(x = 25)$:

$y = 100 - 2(25) = 50 \, \text{metros}$

**Segunda Derivada para Confirmar M√°ximo**

$A''(x) = -4$

Dado que $A''(x) < 0$, hay un m√°ximo local en $(x = 25)$.

**Visualizaci√≥n con Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def A(x):
    return 100*x - 2*x**2

x_values = np.linspace(0, 50, 400)
A_values = A(x_values)

# Punto m√°ximo
x_max = 25
A_max = A(x_max)

plt.figure(figsize=(14, 8))
plt.plot(x_values, A_values, label='√Årea $A(x) = 100x - 2x^2$', color='darkblue', linewidth=2)
plt.scatter(x_max, A_max, color='red', s=50, zorder=5)  # M√°ximo m√°s visible
plt.title('Maximizaci√≥n del √Årea del Corral', fontsize=16, fontweight='bold')
plt.xlabel('Longitud del lado x (m)', fontsize=14)
plt.ylabel('√Årea (m¬≤)', fontsize=14)
plt.axvline(x=x_max, color='grey', linestyle='--', label=f'M√°ximo en x={x_max}m')
plt.axhline(y=A_max, color='grey', linestyle='--', label=f'√Årea m√°xima={A_max} m¬≤')
plt.annotate('Punto m√°ximo', xy=(x_max, A_max), xytext=(30, 100),
             textcoords='offset points', ha='center',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='black'))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', frameon=True, shadow=True, fontsize=12)
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9)  # Ajustar manualmente los m√°rgenes
plt.show()
```

**Interpretaci√≥n**

- **Gr√°fico Azul**: Representa c√≥mo cambia el √°rea del corral en funci√≥n de la longitud de $(x)$, los lados paralelos al edificio.

- **Punto Rojo**: Indica el valor de $(x)$ que maximiza el √°rea, que es 25 metros, dando un √°rea m√°xima de 1250 metros cuadrados.

Este ejemplo ilustra c√≥mo se pueden utilizar conceptos de c√°lculo para resolver problemas pr√°cticos de optimizaci√≥n. Los m√°ximos y m√≠nimos son fundamentales para dise√±ar soluciones eficientes y econ√≥micas en situaciones reales.

### **6.1.3 Graficar utilizando m√°ximos y m√≠nimos**

Graficar una funci√≥n utilizando m√°ximos y m√≠nimos es una t√©cnica muy √∫til para visualizar y entender el comportamiento de la funci√≥n a lo largo de su dominio. Identificar los puntos donde una funci√≥n alcanza sus valores m√°ximos y m√≠nimos locales y absolutos nos ayuda a interpretar gr√°ficamente la funci√≥n de manera efectiva.

**Conceptos Clave**

Antes de graficar, es fundamental identificar estos elementos:

**M√°ximos y M√≠nimos Locales**: Puntos donde la funci√≥n alcanza valores locales extremos.

**Puntos Cr√≠ticos**: Puntos donde la primera derivada $f'(x)$ es cero o no est√° definida. Estos puntos son candidatos para ser m√°ximos o m√≠nimos locales.

**Concavidad y Puntos de Inflexi√≥n**: Utilizar la segunda derivada $f''(x)$ para determinar la concavidad (hacia arriba o hacia abajo) y localizar puntos de inflexi√≥n.

**Ejemplo Pr√°ctico: Funci√≥n Cuadr√°tica**

Vamos a tomar una funci√≥n cuadr√°tica simple para ilustrar c√≥mo graficarla utilizando su m√°ximo local.

**Funci√≥n**:

$f(x) = -x^2 + 4x - 3$

**Derivadas y An√°lisis**

- **Primera Derivada** (para encontrar puntos cr√≠ticos):

  $f'(x) = -2x + 4$

  Igualando a cero para encontrar puntos cr√≠ticos:

  $-2x + 4 = 0 \rightarrow x = 2$

- **Segunda Derivada** (para determinar la concavidad):

  $f''(x) = -2$

  Ya que $f''(x) < 0$ para todo $(x)$, la funci√≥n es c√≥ncava hacia abajo en todo su dominio, lo que implica que el punto cr√≠tico $(x = 2)$ es un m√°ximo local.

**Valor del M√°ximo**

$f(2) = -(2)^2 + 4(2) - 3 = 1$

El m√°ximo local est√° en $(2, 1)$.

**Visualizaci√≥n con Python**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n
def f(x):
    return -x**2 + 4*x - 3

# Generar valores de x
x_values = np.linspace(0, 4, 400)
y_values = f(x_values)

# Punto de m√°ximo local
x_max = 2
y_max = f(x_max)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = -x^2 + 4x - 3$', color='navy', linewidth=2)
plt.scatter(x_max, y_max, color='darkred', s=50, zorder=5, label='M√°ximo Local $(2, 1)$')  # M√°ximo m√°s visible
plt.title('Gr√°fica de la Funci√≥n Cuadr√°tica', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.axvline(x=x_max, color='grey', linestyle='--', linewidth=1.5, label=f'x={x_max}')
plt.axhline(y=y_max, color='grey', linestyle='--', linewidth=1.5, label=f'f(x)={y_max}')
plt.annotate('Punto m√°ximo', xy=(x_max, y_max), xytext=(3, 5),
             textcoords='offset points', ha='center',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='black'))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', frameon=True, shadow=True, fontsize=12)
plt.tight_layout()
plt.show()
```

**Interpretaci√≥n de la Gr√°fica**

- **Curva Azul**: Representa la funci√≥n cuadr√°tica $f(x)$. Esta curva muestra claramente el comportamiento de la funci√≥n y su punto m√°ximo.

- **Punto Rojo**: Indica el m√°ximo local en $(2, 1)$, que es el punto m√°s alto en la gr√°fica debido a que la funci√≥n es c√≥ncava hacia abajo en todo su dominio.

Este tipo de visualizaci√≥n es fundamental para entender completamente el comportamiento de funciones matem√°ticas complejas y para aplicaciones pr√°cticas en optimizaci√≥n, ingenier√≠a y ciencias econ√≥micas.

### **6.1.4 Tasa de cambio**

La **tasa de cambio** es un concepto fundamental en c√°lculo y otras √°reas de las matem√°ticas, que describe c√≥mo var√≠a una cantidad en relaci√≥n con otra. En el contexto del c√°lculo diferencial, la tasa de cambio es esencialmente la derivada de una funci√≥n y proporciona informaci√≥n crucial sobre la velocidad a la que una variable cambia con respecto a otra.

**Conceptos de Tasa de Cambio**

**Tasa de Cambio Promedio**: Es la tasa a la que cambia una cantidad en relaci√≥n con otra a lo largo de un intervalo. Se calcula como el cociente de los cambios en las variables.

   $\text{Tasa de cambio promedio} = \frac{\Delta y}{\Delta x} = \frac{f(b) - f(a)}{b - a}$

   donde $f(x)$ es la funci√≥n, y $(a)$ y $(b)$ son dos puntos en el dominio de $(f)$.

**Tasa de Cambio Instant√°nea**: Es la velocidad instant√°nea con la que cambia una variable respecto a otra en un punto espec√≠fico, y se define como la derivada de la funci√≥n en ese punto.

   $\text{Tasa de cambio instant√°nea en } x = a = f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}$

**Ejemplo Pr√°ctico: Velocidad de un Autom√≥vil**

Supongamos que la posici√≥n $s(t)$ de un autom√≥vil que se mueve a lo largo de una carretera recta est√° dada por la funci√≥n $s(t) = t^2 + 2t$ donde $(s)$ est√° en metros y $(t)$ en segundos.

**Tasa de Cambio Promedio entre 2 segundos y 5 segundos**

$\text{Tasa de cambio promedio} = \frac{s(5) - s(2)}{5 - 2}$

Calculamos $s(5)$ y $s(2)$:

$s(5) = 5^2 + 2 \times 5 = 25 + 10 = 35 \, \text{m}$

$s(2) = 2^2 + 2 \times 2 = 4 + 4 = 8 \, \text{m}$

$\text{Tasa de cambio promedio} = \frac{35 - 8}{3} = \frac{27}{3} = 9 \, \text{m/s}$

**Tasa de Cambio Instant√°nea en $(t = 3)$ segundos**

Calculamos la derivada $s'(t)$:

$s'(t) = 2t + 2$

$s'(3) = 2 \times 3 + 2 = 8 \, \text{m/s}$

**Visualizaci√≥n con Python**

Vamos a visualizar la funci√≥n de posici√≥n y su tasa de cambio (velocidad).

```python
import numpy as np
import matplotlib.pyplot as plt

def s(t):
    return t**2 + 2*t

def ds(t):
    return 2*t + 2

# Generar valores de t
t_values = np.linspace(0, 6, 400)
s_values = s(t_values)
ds_values = ds(t_values)

plt.figure(figsize=(12, 8))

# Configurar el primer subplot para posici√≥n
plt.subplot(2, 1, 1)
plt.plot(t_values, s_values, label='Posici√≥n $s(t) = t^2 + 2t$', color='navy', linestyle='-', linewidth=2)
plt.title('An√°lisis de Movimiento del Autom√≥vil', fontsize=16, fontweight='bold')
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('Posici√≥n (m)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left', frameon=True, shadow=True, fontsize=10)

# Configurar el segundo subplot para velocidad
plt.subplot(2, 1, 2)
plt.plot(t_values, ds_values, label='Velocidad $s\'(t) = 2t + 2$', color='crimson', linestyle='--', linewidth=2)
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('Velocidad (m/s)', fontsize=12)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(loc='upper left', frameon=True, shadow=True, fontsize=10)

plt.tight_layout(pad=3.0)
plt.show()
```

**Interpretaci√≥n**

- **Gr√°fico Superior (Azul)**: Representa la posici√≥n del autom√≥vil en funci√≥n del tiempo, mostrando c√≥mo la distancia aumenta con el tiempo.

- **Gr√°fico Inferior (Rojo)**: Muestra la velocidad del autom√≥vil, que es la tasa de cambio instant√°nea de su posici√≥n, tambi√©n en funci√≥n del tiempo.

Este enfoque nos permite comprender visualmente c√≥mo cambia la posici√≥n y la velocidad del autom√≥vil a lo largo del tiempo, ilustrando la aplicaci√≥n de las tasas de cambio en problemas pr√°cticos de movimiento.

### **6.1.5 Encontrar m√°ximos y m√≠nimos**

Encontrar m√°ximos y m√≠nimos de una funci√≥n es un proceso crucial en c√°lculo y aplicaciones matem√°ticas, especialmente cuando se trata de optimizaci√≥n. Los m√°ximos y m√≠nimos pueden ser locales, es decir, los m√°s altos o bajos en una peque√±a regi√≥n, o globales, los m√°s altos o bajos en todo el dominio de la funci√≥n.

**Conceptos Clave**

**M√°ximo Local**: Un punto $(x=c)$ es un m√°ximo local de $f(x)$ si $f(c)$ es mayor que todos los valores de $f(x)$ cerca de $(c)$.

**M√≠nimo Local**: Un punto $(x=c)$ es un m√≠nimo local de $f(x)$ si $f(c)$ es menor que todos los valores de $f(x)$ cerca de $(c)$.

**M√°ximo Absoluto**: El valor m√°s alto que $f(x)$ alcanza en su dominio.

**M√≠nimo Absoluto**: El valor m√°s bajo que $f(x)$ alcanza en su dominio.

**M√©todo para Encontrar M√°ximos y M√≠nimos**

**Primera Derivada**:

   - Encuentra la primera derivada $f'(x)$.

   - Resuelve $f'(x) = 0$ para encontrar los puntos cr√≠ticos.

   - Estos puntos son candidatos para ser m√°ximos o m√≠nimos locales.

**Segunda Derivada (Test de Concavidad)**:

   - Calcula la segunda derivada $f''(x)$.

   - Si $f''(x) > 0$ en un punto cr√≠tico, entonces es un m√≠nimo local (c√≥ncava hacia arriba).

   - Si $f''(x) < 0$ en un punto cr√≠tico, entonces es un m√°ximo local (c√≥ncava hacia abajo).

**Ejemplo: An√°lisis de $f(x) = x^3 - 6x^2 + 9x + 1$**

**C√°lculo de Derivadas**

- **Primera Derivada**:

  $f'(x) = 3x^2 - 12x + 9$

- **Segunda Derivada**:

  $f''(x) = 6x - 12$

**Encontrar Puntos Cr√≠ticos**

- Resolver $f'(x) = 0$:

  $3x^2 - 12x + 9 = 0$

  $x = 1, \; x = 3 \; \text{(usando la f√≥rmula cuadr√°tica)}$

**Test de Concavidad**

- Evaluar $f''(x)$ en $(x = 1)$ y $(x = 3)$:

  $f''(1) = 6(1) - 12 = -6 \; (\text{m√°ximo local})$

  $f''(3) = 6(3) - 12 = 6 \; (\text{m√≠nimo local})$

**Visualizaci√≥n con Python**

Vamos a graficar la funci√≥n y sus puntos cr√≠ticos para visualizar estos m√°ximos y m√≠nimos.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 6*x**2 + 9*x + 1

def df(x):
    return 3*x**2 - 12*x + 9

def ddf(x):
    return 6*x - 12

x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)

# Puntos cr√≠ticos
critical_points = [1, 3]
critical_values = [f(x) for x in critical_points]

plt.figure(figsize=(12, 7))
plt.plot(x_values, y_values, label='$f(x) = x^3 - 6x^2 + 9x + 1$', color='navy', linewidth=2)
plt.scatter(critical_points, critical_values, color='darkred', s=50, label='Puntos Cr√≠ticos')
plt.title('An√°lisis de M√°ximos y M√≠nimos en Funci√≥n Polin√≥mica', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', frameon=True, shadow=True)
for x, y in zip(critical_points, critical_values):
    plt.annotate(f'({x}, {y})', xy=(x, y), xytext=(10, -20), textcoords='offset points')
plt.tight_layout()
plt.show()
```

**Interpretaci√≥n**

- **Gr√°fico Azul**: Muestra la funci√≥n con su forma polin√≥mica t√≠pica, ascendiendo y descendiendo en el dominio dado.

- **Puntos Rojos**: Indican los m√°ximos y m√≠nimos locales encontrados a trav√©s del an√°lisis de derivadas. Aqu√≠, $(x = 1)$ es un m√°ximo local debido a que la funci√≥n es c√≥ncava hacia abajo, y $(x = 3)$ es un m√≠nimo local porque la funci√≥n es c√≥ncava hacia arriba en esos puntos.

Este enfoque detallado de encontrar m√°ximos y m√≠nimos es esencial para entender el comportamiento de funciones en ciencias, ingenier√≠a, econom√≠a, y m√°s.
