# **2. Álgebra y cálculo diferencial para la ciencia de datos**

## **2.1 Prueba de inducción, límites y continuidad**

Pruebas sencillas de inducción y límites en el infinito para funciones

### **2.1.1 Prueba por inducción**

La **prueba por inducción matemática** es un método poderoso utilizado para demostrar que una afirmación o proposición es verdadera para todos los números naturales. Este tipo de prueba es fundamental cuando se trata de proposiciones que dependen de un número entero $(n)$.

**Concepto de Prueba por Inducción**

La inducción matemática se basa en dos pasos principales:

**Paso Base**: Demuestra que la afirmación es verdadera para el primer valor natural dentro del alcance de la proposición, generalmente $(n = 1)$.

**Paso Inductivo**: Demuestra que si la afirmación es verdadera para un número entero arbitrario $(n)$, entonces también lo será para el siguiente número entero, $(n+1)$. Esto se hace asumiendo que la afirmación es verdadera para $(n)$ (hipótesis inductiva) y luego usando esta hipótesis para probar que la afirmación es verdadera para $(n+1)$.

**Ejemplo de Prueba por Inducción**

**Proposición**: Para todo número natural $(n)$, la suma de los primeros $(n)$ números naturales es:

$\sum_{k=1}^n k = \frac{n(n+1)}{2}$

**Paso Base**

Verifica la validez de la afirmación para $(n = 1)$:

$\sum_{k=1}^1 k = 1 \quad \text{y} \quad \frac{1(1+1)}{2} = 1$

Ambas expresiones dan 1, por lo que la afirmación es cierta para $(n = 1)$.

**Paso Inductivo**

**Hipótesis Inductiva**: Asumimos que la afirmación es verdadera para $(n)$, es decir:

$\sum_{k=1}^n k = \frac{n(n+1)}{2}$

**Objetivo**: Demostrar que la afirmación es verdadera para $(n+1)$, es decir:

$\sum_{k=1}^{n+1} k = \frac{(n+1)(n+2)}{2}$

**Demostración**:

$\sum_{k=1}^{n+1} k = \sum_{k=1}^n k + (n+1) = \frac{n(n+1)}{2} + (n+1)$

Factorizando $(n+1)$ del lado derecho de la igualdad:

$\frac{n(n+1)}{2} + (n+1) = (n+1) \left(\frac{n}{2} + 1\right) = (n+1) \frac{n+2}{2} = \frac{(n+1)(n+2)}{2}$

Por lo tanto, la afirmación es cierta para $(n+1)$ si es cierta para $(n)$.

**Visualización y Código en Python**

Podemos demostrar numéricamente la proposición para algunos valores de $(n)$ y visualizar la relación entre $(\sum k)$ y $(\frac{n(n+1)}{2})$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de términos
n_terms = 100

# Generar valores de n
n = np.arange(1, n_terms + 1)

# Calcular la suma de los primeros n números naturales
sum_n = n * (n + 1) / 2

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(n, sum_n, label='Suma de los primeros n números', marker='o', linestyle='-', color='blue', markersize=5)
plt.title('Verificación de la Fórmula de la Suma de los Primeros n Números Naturales', fontsize=16)
plt.xlabel('n', fontsize=14)
plt.ylabel('Suma de los primeros n números', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```

**Conclusión**

La prueba por inducción es un método esencial en matemáticas que permite demostrar afirmaciones para todos los números naturales, asegurando que la lógica aplicada a un caso particular se extiende de manera ordenada y lógica a todos los casos generales. Este tipo de pruebas es crucial en teoría de números, informática y más.

### **2.1.2 Límites en el infinito**

Los **límites en el infinito** son un concepto fundamental en cálculo que se utiliza para describir el comportamiento de una función a medida que sus argumentos se acercan a valores extremadamente grandes (infinito) o extremadamente pequeños (infinito negativo). Este concepto es crucial para entender cómo se comportan las funciones en sus extremos y es fundamental para el análisis de tendencias asintóticas, optimización y en el cálculo de integrales impropias, entre otros.

**Concepto de Límites en el Infinito**

Un límite en el infinito examina el valor al que se acerca una función $f(x)$ cuando $(x)$ tiende a $(\infty)$ o $(-\infty)$. Matemáticamente, esto se representa como:

- $(\lim_{x \to \infty} f(x))$ y

- $(\lim_{x \to -\infty} f(x))$.

Estos límites describen, respectivamente, el comportamiento de $f(x)$ a medida que $(x)$ se hace muy grande o muy pequeño en términos negativos.

**Ejemplos de Límites en el Infinito**

**Ejemplo 1: Función Lineal**

Consideremos la función lineal $f(x) = 2x + 3$. El límite de $f(x)$ cuando $(x)$ tiende a infinito es:

$\lim_{x \to \infty} (2x + 3) = \infty$

Esto significa que a medida que $(x)$ se incrementa indefinidamente, $f(x)$ también crece sin límites.

**Ejemplo 2: Función Racional**

Consideremos ahora una función racional $g(x) = \frac{1}{x}$.

- Cuando $(x)$ tiende a infinito:

  $\lim_{x \to \infty} \frac{1}{x} = 0$

  Esto indica que a medida que $(x)$ se hace más grande, $g(x)$ se acerca a cero.

- Cuando $(x)$ tiende a infinito negativo:

  $\lim_{x \to -\infty} \frac{1}{x} = 0$

**Visualización y Código en Python**

Podemos visualizar estos conceptos utilizando Python y la biblioteca Matplotlib para graficar las funciones mencionadas y observar sus comportamientos en los límites.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos de x
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, -0.1, 400)
x3 = np.linspace(0.1, 10, 400)

# Calcular y para las funciones
y1 = 2 * x1 + 3
y2 = 1 / x2
y3 = 1 / x3

# Crear gráficos
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, label='f(x) = 2x + 3', color='green', linestyle='-')
plt.title('Función Lineal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x2, y2, label='g(x) = 1/x (x < 0)', color='red', linestyle='-', linewidth=2)
plt.plot(x3, y3, label='g(x) = 1/x (x > 0)', color='blue', linestyle='-', linewidth=2)
plt.title('Función Racional')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.ylim(-10, 10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

**Interpretación**

- **Función Lineal**: La gráfica muestra que a medida que $(x)$ se incrementa, $f(x)$ también se incrementa linealmente hacia el infinito.

- **Función Racional**: La gráfica de $g(x) = \frac{1}{x}$ muestra que mientras más grande o más pequeño se hace $(x)$, $g(x)$ se acerca a cero desde ambos lados del eje $(x)$.

Los límites en el infinito proporcionan información valiosa sobre el comportamiento a largo plazo de las funciones, lo que es crucial en campos como la ingeniería, la física teórica y las matemáticas aplicadas.

### **2.1.3 Límites en un punto específico**

Los **límites en un punto específico** se refieren al comportamiento de una función $f(x)$ a medida que $(x)$ se acerca a un valor concreto $(c)$ desde ambos lados (izquierda y derecha). Este concepto es fundamental en cálculo para entender cómo se comportan las funciones cerca de puntos específicos, especialmente en puntos donde la función puede no estar definida o tener discontinuidades.

**Concepto de Límites en un Punto Específico**

El límite de $f(x)$ cuando $(x)$ tiende a $(c)$, denotado como $(\lim_{x \to c} f(x))$, es el valor al que se acercan los valores de $f(x)$ conforme $(x)$ se aproxima a $(c)$. Este límite existe si, y solo si, los límites desde la izquierda y la derecha existen y son iguales.

**Notación y Definición**

- **Límite desde la izquierda** (límite por la izquierda de $(c)$): $(\lim_{x \to c^-} f(x))$

- **Límite desde la derecha** (límite por la derecha de $(c)$): $(\lim_{x \to c^+} f(x))$

Para que el límite $(\lim_{x \to c} f(x))$ exista, es necesario que:

$\lim_{x \to c^-} f(x) = \lim_{x \to c^+} f(x)$

**Ejemplo 1: Función Continua**

Consideremos la función $f(x) = 2x + 1$. Calcularemos el límite cuando $(x)$ tiende a 3.

$\lim_{x \to 3} (2x + 1) = 2(3) + 1 = 7$

Esta es una función lineal y es continua en todos sus puntos, incluido $(x = 3)$.

**Ejemplo 2: Función con una Discontinuidad**

Consideremos la función definida a trozos:

$
f(x) =
\begin{cases}
x^2 & \text{si } x < 2 \\
5 & \text{si } x = 2 \\
x + 3 & \text{si } x > 2
\end{cases}$

Calcularemos el límite en $(x = 2)$.

$\lim_{x \to 2^-} f(x) = \lim_{x \to 2^-} x^2 = 4$

$\lim_{x \to 2^+} f(x) = \lim_{x \to 2^+} (x + 3) = 5$

Dado que $(\lim_{x \to 2^-} f(x) \neq \lim_{x \to 2^+} f(x))$, $(\lim_{x \to 2} f(x))$ no existe.

**Visualización y Código en Python**

Vamos a graficar estas funciones utilizando Python y la biblioteca Matplotlib para visualizar sus comportamientos en torno a los puntos de interés.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos de x y las funciones
x1 = np.linspace(1, 5, 400)
f1 = 2 * x1 + 1

x2_left = np.linspace(1, 2, 200, endpoint=False)
x2_right = np.linspace(2, 3, 200)
f2_left = x2_left**2
f2_right = x2_right + 3

# Crear gráficos
plt.figure(figsize=(12, 6))

# Gráfico para la función continua
plt.subplot(1, 2, 1)
plt.plot(x1, f1, label='f(x) = 2x + 1', color='blue', linestyle='-')
plt.scatter([3], [7], color='red', marker='o', label='Punto de interés')
plt.title('Función Lineal Continua')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

# Gráfico para la función con discontinuidad
plt.subplot(1, 2, 2)
plt.plot(x2_left, f2_left, label='$x^2 (x < 2)$', color='green', linestyle='--')
plt.scatter([2], [4], color='green', marker='o', label='Límite desde la izquierda')
plt.plot(x2_right, f2_right, label='x + 3 (x > 2)', color='red', linestyle='-')
plt.scatter([2], [5], color='red', marker='o', label='Límite desde la derecha')
plt.title('Función Discontinua en x=2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

**Conclusión**

Los límites en puntos específicos son esenciales para entender el comportamiento local de las funciones, particularmente en puntos donde podrían no estar definidas o podrían cambiar de comportamiento. Este conocimiento es crucial para el análisis de funciones en cálculo, tanto para aplicaciones teóricas como prácticas en ingeniería, física y otras ciencias.

### **2.1.4 Continuidad de una función**

La **continuidad de una función** en matemáticas describe una propiedad fundamental donde la función no tiene interrupciones, saltos, o asimetrías en su gráfica. Una función continua permite que su gráfica se dibuje sin levantar el lápiz del papel. Este concepto es crucial para entender el comportamiento de las funciones en cálculo, análisis real, y muchas áreas aplicadas.

**Definición de Continuidad**

Una función $f(x)$ es continua en un punto $(c)$ de su dominio si se cumplen las siguientes tres condiciones:

**$f(c)$ está definida**: El valor de $f(c)$ existe.

**$(\lim_{x \to c} f(x))$ existe**: El límite de $f(x)$ cuando $(x)$ tiende a $(c)$ existe.

**$(\lim_{x \to c} f(x) = f(c))$**: El límite de la función cuando $(x)$ se aproxima a $(c)$ es igual al valor de la función en $(c)$.

Si una función cumple estas condiciones en todos los puntos de un intervalo, se dice que es continua en ese intervalo.

**Ejemplos de Continuidad**

**Ejemplo 1: Función Continua**

La función $f(x) = x^2$ es un ejemplo clásico de una función continua en todos los reales $(\mathbb{R})$. Cumple con las tres condiciones de continuidad para cualquier valor de $(x)$.

**Ejemplo 2: Función Discontinua**

Consideremos la función definida a trozos:

$f(x) =
\begin{cases}
x^2 & \text{si } x < 3 \\
6 & \text{si } x = 3 \\
x + 1 & \text{si } x > 3
\end{cases}$

Esta función no es continua en $(x = 3)$ porque $(\lim_{x \to 3^-} f(x) = 9)$ y $(\lim_{x \to 3^+} f(x) = 4)$, pero $f(3) = 6$. Los límites laterales no coinciden, y tampoco coinciden con el valor de la función en ese punto.

**Visualización con Python**

Podemos usar Python y Matplotlib para visualizar estas funciones y demostrar gráficamente sus propiedades de continuidad o discontinuidad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos de x y las funciones
x_continuous = np.linspace(-10, 10, 400)
y_continuous = x_continuous**2

x_discontinuous_left = np.linspace(-10, 3, 200, endpoint=False)
x_discontinuous_right = np.linspace(3, 10, 200)
y_discontinuous_left = x_discontinuous_left**2
y_discontinuous_right = x_discontinuous_right + 1

# Crear gráficos
plt.figure(figsize=(12, 6))

# Gráfico para la función continua
plt.subplot(1, 2, 1)
plt.plot(x_continuous, y_continuous, label='$f(x) = x^2$', color='blue', linestyle='-')
plt.title('Función Continua $f(x) = x^2$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10)

# Gráfico para la función discontinua
plt.subplot(1, 2, 2)
plt.plot(x_discontinuous_left, y_discontinuous_left, label='$x^2$ para $x < 3$', color='green', linestyle='--')
plt.plot(x_discontinuous_right, y_discontinuous_right, label='$x + 1$ para $x > 3$', color='red', linestyle='-.')
plt.scatter([3], [6], color='purple', zorder=5, label='Punto discontinuo')  # Punto discontinuo
plt.title('Función Discontinua en $x = 3$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()
```

**Conclusión**

La continuidad es una propiedad crítica que afecta muchas áreas del análisis matemático, incluyendo la integración y la diferenciación. Entender y poder identificar si una función es continua en un punto o en un intervalo es esencial para trabajar con funciones en prácticamente todos los campos de la ciencia y la ingeniería.

### **2.1.5 Prueba de límites en un punto concreto**

La **prueba de límites en un punto concreto** es un proceso matemático utilizado para determinar el valor al que se acerca una función cuando la variable independiente se aproxima a un valor específico. Este concepto es fundamental en cálculo para entender el comportamiento de las funciones cerca de ciertos puntos, especialmente donde las funciones pueden no estar definidas o donde su comportamiento cambia dramáticamente.

**Concepto de Límites**

El límite de una función $f(x)$ en un punto $(x = c)$ intenta encontrar el valor al que se acerca $f(x)$ a medida que $(x)$ se acerca a $(c)$, denotado como:

$\lim_{x \to c} f(x) = L$

Este límite existe si, y solo si, los valores de $f(x)$ se pueden hacer arbitrariamente cercanos a $(L)$ haciendo $(x)$ suficientemente cerca de $(c)$, independientemente de cómo $(x)$ se acerque a $(c)$ (ya sea desde la izquierda o la derecha).

**Ejemplo Práctico: Límite en un Punto**

**Función**:

$f(x) = \frac{x^2 - 4}{x - 2}$

**Punto de Interés**: $(x = 2)$

A primera vista, parece que $f(x)$ no está definida en $(x = 2)$ porque si sustituimos directamente, obtenemos una forma indeterminada $(\frac{0}{0})$. Sin embargo, podemos simplificar la expresión:

$\frac{x^2 - 4}{x - 2} = \frac{(x-2)(x+2)}{x-2}$

Simplificando, $f(x)$ se reduce a $(x + 2)$ siempre que $(x \neq 2)$. Ahora, calculemos el límite cuando $(x)$ tiende a 2:

$\lim_{x \to 2} f(x) = \lim_{x \to 2} (x + 2) = 4$

Este resultado muestra que aunque $f(x)$ no está definida en $(x = 2)$, el límite cuando $(x)$ se aproxima a 2 es 4.

**Visualización y Código en Python**

Vamos a visualizar esta función y su comportamiento cerca de $(x = 2)$ utilizando Python y Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return (x**2 - 4) / (x - 2)

# Rango de x cerca de 2, evitando la división por cero
x = np.linspace(1, 3, 400)
x = x[x != 2]  # Eliminar x = 2 para evitar división por cero
y = f(x)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='$f(x) = (x^2 - 4)/(x - 2)$', color='blue', linestyle='-', linewidth=2)
plt.scatter(2, 4, color='red', zorder=5, label='Límite en x = 2', marker='o')
plt.title('Gráfica de f(x) cerca de x = 2', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.axvline(x=2, color='gray', linestyle='--', label='x = 2')
plt.axhline(y=4, color='gray', linestyle='--', label='Límite = 4')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(1.5, 2.5)
plt.ylim(3.5, 4.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

**Interpretación**

- **Gráfica**: La gráfica muestra cómo $f(x)$ se comporta cerca de $(x = 2)$. La función parece acercarse a 4, aunque no está definida exactamente en ese punto.

- **Punto Rojo**: Indica el límite de la función cuando $(x)$ tiende a 2, que hemos calculado como 4.

Este ejemplo ilustra cómo se puede utilizar el concepto de límites para comprender el comportamiento de una función cerca de puntos donde no está directamente definida o tiene discontinuidades. Los límites son esenciales en el estudio del cálculo y proporcionan la base para el desarrollo posterior de conceptos como la continuidad y la derivabilidad.
