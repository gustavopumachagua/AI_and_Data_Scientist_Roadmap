**Concepto de Prueba por Inducci贸n**

La inducci贸n matem谩tica se basa en dos pasos principales:

**Ejemplo de Prueba por Inducci贸n**

**Proposici贸n**: Para todo n煤mero natural $(n)$, la suma de los primeros $(n)$ n煤meros naturales es:

$\sum_{k=1}^n k = \frac{n(n+1)}{2}$

**Paso Base**

Verifica la validez de la afirmaci贸n para $(n = 1)$:

$\sum_{k=1}^1 k = 1 \quad \text{y} \quad \frac{1(1+1)}{2} = 1$

Ambas expresiones dan 1, por lo que la afirmaci贸n es cierta para $(n = 1)$.

**Paso Inductivo**

**Hip贸tesis Inductiva**: Asumimos que la afirmaci贸n es verdadera para $(n)$, es decir:

$\sum_{k=1}^n k = \frac{n(n+1)}{2}$

**Objetivo**: Demostrar que la afirmaci贸n es verdadera para $(n+1)$, es decir:

$\sum_{k=1}^{n+1} k = \frac{(n+1)(n+2)}{2}$

**Demostraci贸n**:

$\sum_{k=1}^{n+1} k = \sum_{k=1}^n k + (n+1) = \frac{n(n+1)}{2} + (n+1)$

Factorizando $(n+1)$ del lado derecho de la igualdad:

$\frac{n(n+1)}{2} + (n+1) = (n+1) \left(\frac{n}{2} + 1\right) = (n+1) \frac{n+2}{2} = \frac{(n+1)(n+2)}{2}$

Por lo tanto, la afirmaci贸n es cierta para $(n+1)$ si es cierta para $(n)$.

**Visualizaci贸n y C贸digo en Python**

Podemos visualizar estos conceptos utilizando Python y la biblioteca Matplotlib para graficar las funciones mencionadas y observar sus comportamientos en los l铆mites.

**Interpretaci贸n**

### **2.1.3 L铆mites en un punto espec铆fico**

Los **l铆mites en un punto espec铆fico** se refieren al comportamiento de una funci贸n $f(x)$ a medida que $(x)$ se acerca a un valor concreto $(c)$ desde ambos lados (izquierda y derecha). **Concepto de L铆mites en un Punto Espec铆fico**

**Notaci贸n y Definici贸n**

Para que el l铆mite $(\lim_{x \to c} f(x))$ exista, es necesario que:

**Ejemplo 1: Funci贸n Continua**

Consideremos la funci贸n $f(x) = 2x + 1$. Calcularemos el l铆mite cuando $(x)$ tiende a 3.

**Ejemplo 2: Funci贸n con una Discontinuidad**

Consideremos la funci贸n definida a trozos:

Calcularemos el l铆mite en $(x = 2)$.

**Visualizaci贸n y C贸digo en Python**

Vamos a graficar estas funciones utilizando Python y la biblioteca Matplotlib para visualizar sus comportamientos en torno a los puntos de inter茅s.

### **2.1.4 Continuidad de una funci贸n**

La **continuidad de una funci贸n** en matem谩ticas describe una propiedad fundamental donde la funci贸n no tiene interrupciones, saltos, o asimetr铆as en su gr谩fica. **Definici贸n de Continuidad**

**$(\lim_{x \to c} f(x) = f(c))$**: El l铆mite de la funci贸n cuando $(x)$ se aproxima a $(c)$ es igual al valor de la funci贸n en $(c)$.

Si una funci贸n cumple estas condiciones en todos los puntos de un intervalo, se dice que es continua en ese intervalo.

**Ejemplos de Continuidad**

**Ejemplo 1: Funci贸n Continua**

La funci贸n $f(x) = x^2$ es un ejemplo cl谩sico de una funci贸n continua en todos los reales $(\mathbb{R})$. Cumple con las tres condiciones de continuidad para cualquier valor de $(x)$.

**Ejemplo 2: Funci贸n Discontinua**

Consideremos la funci贸n definida a trozos:

Esta funci贸n no es continua en $(x = 3)$ porque $(\lim_{x \to 3^-} f(x) = 9)$ y $(\lim_{x \to 3^+} f(x) = 4)$, pero $f(3) = 6$. Los l铆mites laterales no coinciden, y tampoco coinciden con el valor de la funci贸n en ese punto.

**Visualizaci贸n con Python**

Podemos usar Python y Matplotlib para visualizar estas funciones y demostrar gr谩ficamente sus propiedades de continuidad o discontinuidad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos de x y las funciones
x_continuous = np.linspace(-10, 10, 400)
y_continuous = x_continuous**2

x_discontinuous_left = np.linspace(-10, 3, 200, endpoint=False)
x_discontinuous_right = np.linspace(3, 10, 200)
y_discontinuous_left = x_discontinuous_left**2
y_discontinuous_right = x_discontinuous_right + 1

# Crear gr谩ficos
plt.figure(figsize=(12, 6))

# Gr谩fico para la funci贸n continua
plt.subplot(1, 2, 1)
plt.plot(x_continuous, y_continuous, label='$f(x) = x^2$', color='blue', linestyle='-')
plt.title('Funci贸n Continua $f(x) = x^2$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10)

# Gr谩fico para la funci贸n discontinua
plt.subplot(1, 2, 2)
plt.plot(x_discontinuous_left, y_discontinuous_left, label='$x^2$ para $x < 3$', color='green', linestyle='--')
plt.plot(x_discontinuous_right, y_discontinuous_right, label='$x + 1$ para $x > 3$', color='red', linestyle='-.')
plt.scatter([3], [6], color='purple', zorder=5, label='Punto discontinuo')  # Punto discontinuo
plt.title('Funci贸n Discontinua en $x = 3$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()
```

**Conclusi贸n**

La continuidad es una propiedad cr铆tica que afecta muchas 谩reas del an谩lisis matem谩tico, incluyendo la integraci贸n y la diferenciaci贸n.
