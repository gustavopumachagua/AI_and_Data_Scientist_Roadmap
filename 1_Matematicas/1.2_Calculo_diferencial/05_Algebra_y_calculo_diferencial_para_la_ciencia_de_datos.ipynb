# **5. Álgebra y cálculo diferencial para la ciencia de datos**

## **5.1 Uso de derivadas para representar gráficamente funciones**

Utilizar cuando las derivadas son positivas y negativas para ayudar a representar gráficamente una función.

### **5.1.1 Uso de la derivada para representar gráficamente funciones**

El uso de la derivada para representar gráficamente funciones es una herramienta poderosa en matemáticas y ciencias, ya que las derivadas proporcionan información crítica sobre el comportamiento de una función. Esto incluye determinar dónde una función aumenta o disminuye, dónde alcanza valores máximos o mínimos locales, y dónde tiene puntos de inflexión.

**Conceptos Básicos del Uso de Derivadas en Gráficos de Funciones**

**Crecimiento y Decrecimiento**

- Una función $f(x)$ está **aumentando** en un intervalo si $f'(x) > 0$ en ese intervalo.
- Está **disminuyendo** si $f'(x) < 0$ en ese intervalo.

**Máximos y Mínimos Locales**

- Un **máximo local** ocurre en $(x = c)$ si $f'(c) = 0$ y el signo de $f'(x)$ cambia de positivo a negativo en $(c)$.

- Un **mínimo local** ocurre en $(x = c)$ si $f'(c) = 0$ y el signo de $f'(x)$ cambia de negativo a positivo en $(c)$.

**Puntos de Inflexión**

- Un **punto de inflexión** es donde la curvatura de la función cambia, y se encuentra evaluando la segunda derivada $f''(x)$:

  - Si $f''(x) > 0$, la función es cóncava hacia arriba.

  - Si $f''(x) < 0$, la función es cóncava hacia abajo.

  - Cambios de concavidad implican un punto de inflexión.

**Ejemplo Práctico: Análisis y Gráfico de $f(x) = x^3 - 6x^2 + 9x + 1$**

**Función**: $f(x) = x^3 - 6x^2 + 9x + 1$

**Cálculo de Derivadas**

**Primera derivada** para encontrar máximos, mínimos y puntos donde la función no crece ni decrece:

$f'(x) = 3x^2 - 12x + 9$

**Segunda derivada** para encontrar puntos de inflexión y determinar concavidad:

$f''(x) = 6x - 12$

**Visualización con Python**

Utilizaremos Python para visualizar la función y sus propiedades derivadas:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y sus derivadas
def f(x):
    return x**3 - 6*x**2 + 9*x + 1

def df(x):
    return 3*x**2 - 12*x + 9

def ddf(x):
    return 6*x - 12

# Generar valores de x
x_values = np.linspace(-1, 4, 400)
y_values = f(x_values)
df_values = df(x_values)
ddf_values = ddf(x_values)

# Puntos críticos y de inflexión
critical_points = np.roots([3, -12, 9])  # raíces de la primera derivada
inflexion_points = np.roots([6, -12])  # raíces de la segunda derivada

# Crear gráfico
plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, label='$f(x)$', color='blue', linewidth=2)
plt.plot(x_values, df_values, label="$f'(x)$", color='red', linestyle='--', linewidth=2)
plt.plot(x_values, ddf_values, label="$f''(x)$", color='green', linestyle=':', linewidth=2)
plt.scatter(critical_points, f(critical_points), color='purple', zorder=5, label='Critical Points', marker='o', s=100)
plt.scatter(inflexion_points, f(inflexion_points), color='orange', zorder=5, label='Inflexion Points', marker='x', s=100)
plt.title('Función Polinómica y Derivadas', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

**Interpretación**

- **Gráfico Azul**: Muestra la función polinómica $f(x)$.

- **Gráfico Rojo**: Representa la primera derivada $f'(x)$, indicando dónde la función tiene máximos y mínimos locales (cruces con el eje $(x)$).

- **Gráfico Verde**: Muestra la segunda derivada $f''(x)$, indicando los puntos de inflexión (cruces con el eje $(x)$) y las regiones de concavidad hacia arriba o hacia abajo.

El análisis y la visualización de una función mediante sus derivadas proporcionan una comprensión profunda de su comportamiento, lo cual es crucial para muchas aplicaciones prácticas en ciencia e ingeniería.

### **5.1.2 Encontrar la concavidad con la segunda derivada**

La **concavidad de una función** y el uso de la segunda derivada para determinarla son conceptos fundamentales en cálculo diferencial. Estos conceptos nos ayudan a comprender cómo se curva la gráfica de una función y a identificar puntos de inflexión, donde cambia la concavidad de la curva.

**Concepto de Concavidad**

La concavidad de una función describe la orientación de la curvatura de su gráfica:

- Una función es **cóncava hacia arriba** (o convexa) en un intervalo si la gráfica de la función se curva hacia arriba en ese intervalo, es decir, si la tangente trazada en cualquier punto del intervalo queda por debajo de la gráfica.
- Una función es **cóncava hacia abajo** en un intervalo si se curva hacia abajo, y la tangente en cualquier punto del intervalo queda por encima de la gráfica.

**Uso de la Segunda Derivada para Determinar la Concavidad**

La segunda derivada de una función, $f''(x)$, proporciona una manera directa de determinar la concavidad de la función:

- Si $f''(x) > 0$ en un intervalo, entonces $f(x)$ es cóncava hacia arriba en ese intervalo.

- Si $f''(x) < 0$ en un intervalo, entonces $f(x)$ es cóncava hacia abajo en ese intervalo.

- Si $f''(x) = 0$ en un punto, ese punto puede ser un **punto de inflexión**, que es donde la concavidad cambia.

**Ejemplo Práctico: Función y su Segunda Derivada**

Consideremos la función:

$f(x) = x^3 - 6x^2 + 12x - 5$

**Derivada primera y segunda**:

- Primera derivada: $f'(x) = 3x^2 - 12x + 12$

- Segunda derivada: $f''(x) = 6x - 12$

**Determinación de la Concavidad**

Analizamos la segunda derivada para determinar la concavidad:

- $f''(x) = 6x - 12$

- Para encontrar cambios en la concavidad, buscamos cuando $f''(x) = 0$:

  $6x - 12 = 0 \Rightarrow x = 2$

  Este es un punto de inflexión.

**Intervalos de Concavidad**

- $f''(x) > 0$ cuando $x > 2$: Cóncava hacia arriba.

- $f''(x) < 0$ cuando $x < 2$: Cóncava hacia abajo.

**Visualización con Python**

Aquí está el código para visualizar la función y entender su concavidad:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función, su primera y segunda derivadas
def f(x):
    return x**3 - 6*x**2 + 12*x - 5

def ddf(x):
    return 6*x - 12

# Rango de valores para x
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
ddf_values = ddf(x_values)

# Crear gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = x^3 - 6x^2 + 12x - 5$', color='blue', linewidth=2)
plt.plot(x_values, ddf_values, label="$f''(x) = 6x - 12$", color='red', linestyle='--', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(2, color='purple', linestyle='--', label='Punto de Inflexión en x=2', linewidth=2)
plt.title('Función y Segunda Derivada (Concavidad)', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

**Interpretación**

- **Gráfico Azul**: Representa la función $f(x)$.

- **Gráfico Rojo**: Muestra la segunda derivada, que ayuda a identificar dónde la función es cóncava hacia arriba o hacia abajo. Los cambios de signo en  $f''(x)$ cerca de $(x = 2)$ indican un punto de inflexión.

- **Línea Púrpura**: Marca el punto de inflexión $(x = 2)$, donde la concavidad cambia de cóncava hacia abajo a cóncava hacia arriba.

Este análisis y visualización demuestran cómo la segunda derivada es una herramienta poderosa para analizar la concavidad de una función y localizar puntos de inflexión.

### **5.1.3 Comparación de las gráficas de $f(x)$ y $f'(x)$**

Comparar las gráficas de una función $f(x)$ y su derivada $f'(x)$ es una forma efectiva de entender visualmente la relación entre la función y su tasa de cambio. Esta comparación puede revelar mucho sobre el comportamiento de la función original, incluyendo dónde es creciente o decreciente, y dónde alcanza máximos y mínimos locales.

**Conceptos Básicos**

**Crecimiento y Decrecimiento**

   - **Creciente**: Si $f'(x) > 0$, entonces $f(x)$ está creciendo en ese intervalo.

   - **Decreciente**: Si $f'(x) < 0$, entonces $f(x)$ está decreciendo.

**Máximos y Mínimos Locales**

   - **Máximos Locales**: Ocurren donde $f'(x)$ cambia de positivo a negativo, y $f'(x) = 0$.

   - **Mínimos Locales**: Ocurren donde $f'(x)$ cambia de negativo a positivo, y $f'(x) = 0$.

**Puntos de Inflexión y Concavidad**

   - **Concavidad Hacia Arriba**: Si $f''(x) > 0$ (donde $f'(x)$ está creciendo).

   - **Concavidad Hacia Abajo**: Si $f''(x) < 0$ (donde $f'(x)$ está decreciendo).

   - **Puntos de Inflexión**: Donde $f''(x) = 0$ y cambia el signo de $f''(x)$.

**Ejemplo: Análisis y Comparación de Gráficas**

Supongamos que queremos analizar y comparar las gráficas de la función $f(x) = x^3 - 3x^2 + 2$ y su derivada.

**Función**:

$f(x) = x^3 - 3x^2 + 2$

**Primera Derivada** (para determinar el crecimiento, decrecimiento y puntos críticos):

$f'(x) = 3x^2 - 6x$

**Segunda Derivada** (para determinar la concavidad y puntos de inflexión):

$f''(x) = 6x - 6$

**Código Python para Visualización**

Usaremos Python para visualizar $f(x)$, $f'(x)$ y $f''(x)$ y comparar sus gráficas:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función, su primera y segunda derivadas
def f(x):
    return x**3 - 3*x**2 + 2

def df(x):
    return 3*x**2 - 6*x

def ddf(x):
    return 6*x - 6

# Generar valores de x
x_values = np.linspace(-1, 4, 400)
f_values = f(x_values)
df_values = df(x_values)
ddf_values = ddf(x_values)

# Crear gráfico
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(x_values, f_values, label='f(x)', color='blue', linewidth=2, linestyle='-')
plt.title('Función $f(x)$', fontsize=14)
plt.grid(True)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)

plt.subplot(3, 1, 2)
plt.plot(x_values, df_values, label="f'(x)", color='red', linewidth=2, linestyle='-')
plt.title("Derivada Primera $f'(x)$", fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel('x', fontsize=12)
plt.ylabel("f'(x)", fontsize=12)

plt.subplot(3, 1, 3)
plt.plot(x_values, ddf_values, label="f''(x)", color='green', linewidth=2, linestyle='-')
plt.title("Derivada Segunda $f''(x)$", fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel('x', fontsize=12)
plt.ylabel("f''(x)", fontsize=12)

plt.tight_layout()
plt.show()
```

**Interpretación de las Gráficas**

- **Gráfico Superior (Azul)**: Muestra la función original $f(x)$. Las raíces de $f'(x)$ indican los puntos donde $f(x)$ tiene máximos y mínimos locales.

- **Gráfico Medio (Rojo)**: Muestra $f'(x)$, cuyos ceros indican los puntos donde $f(x)$ podría tener máximos y mínimos locales. La signatura de $f'(x)$ (positiva o negativa) indica si $f(x)$ está creciendo o decreciendo.

- **Gráfico Inferior (Verde)**: Muestra $f''(x)$, donde los ceros podrían ser puntos de inflexión de $f(x)$. La signatura de $f''(x)$ indica la concavidad de $f(x)$.

Esta comparación visual y analítica ayuda a entender profundamente cómo la función se comporta, cómo cambia su tasa de cambio y cómo varía su curvatura.

### **5.1.4 Funciones gráficas - Un ejemplo más complicado**

Explorar ejemplos más complicados de funciones gráficas puede ayudar a comprender mejor cómo interactúan diversas características matemáticas en una función. A continuación, abordaremos una función polinómica de grado superior que incluye términos que producen oscilaciones, puntos críticos múltiples, y puntos de inflexión. Este tipo de función permite una exploración detallada de máximos, mínimos, y cambios de concavidad, utilizando tanto análisis gráfico como analítico.

**Función de Ejemplo**

Consideremos la función:

$f(x) = x^4 - 4x^3 + 4x^2$

Esta función es interesante porque es un polinomio de cuarto grado con múltiples características:

- **Máximos y mínimos locales**: Dado que es un polinomio de cuarto grado, puede tener hasta tres puntos críticos (máximos y mínimos).

- **Puntos de inflexión**: Puede tener hasta dos puntos de inflexión.

**Análisis Derivativo**

Primero, calculemos las primeras y segundas derivadas para identificar los puntos críticos y los puntos de inflexión.

**Primera Derivada**:

$f'(x) = 4x^3 - 12x^2 + 8x$

**Segunda Derivada**:

$f''(x) = 12x^2 - 24x + 8$

**Encontrar Puntos Críticos y Puntos de Inflexión**

Resolvamos $f'(x) = 0$ y $f''(x) = 0$ para encontrar los puntos críticos y los puntos de inflexión, respectivamente.

**Puntos Críticos**:

$4x^3 - 12x^2 + 8x = 0$

$x(4x^2 - 12x + 8) = 0$

$x = 0, \; 4x^2 - 12x + 8 = 0$

$x = \frac{12 \pm \sqrt{144 - 128}}{8} = \frac{12 \pm 4}{8}$

$x = 2, \; 1$

**Puntos de Inflexión**:

$12x^2 - 24x + 8 = 0$

$3x^2 - 6x + 2 = 0$

$x = \frac{6 \pm \sqrt{36 - 24}}{6} = \frac{6 \pm \sqrt{12}}{6}$

$x = 1 \pm \frac{\sqrt{3}}{3}$

**Visualización con Python**

Ahora, visualizaremos esta función y sus derivadas para observar gráficamente sus propiedades.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y sus derivadas
def f(x):
    return x**4 - 4*x**3 + 4*x**2

def df(x):
    return 4*x**3 - 12*x**2 + 8*x

def ddf(x):
    return 12*x**2 - 24*x + 8

# Rango de valores para x
x_values = np.linspace(-1, 3, 400)
y_values = f(x_values)
df_values = df(x_values)
ddf_values = ddf(x_values)

# Crear gráfico
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(x_values, y_values, label='f(x)', color='blue')
plt.title('Función Polinómica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x_values, df_values, label="f'(x)", color='red')
plt.title("Primera Derivada (Puntos Críticos)")
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(x_values, ddf_values, label="f''(x)", color='green')
plt.title("Segunda Derivada (Puntos de Inflexión)")
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Interpretación de las Gráficas**

- **Función Polinómica $f(x)$**: El gráfico superior muestra cómo la función se comporta a lo largo de su dominio, incluyendo regiones donde es cóncava hacia arriba o hacia abajo y sus puntos críticos.

- **Primera Derivada $f'(x)$**: El gráfico del medio ayuda a identificar los máximos y mínimos locales, mostrando dónde la derivada cruza el eje x.

- **Segunda Derivada $f''(x)$**: El gráfico inferior muestra los puntos de inflexión, donde la concavidad de $f(x)$ cambia. Este análisis y visualización proporcionan una comprensión integral de cómo la función y sus derivadas describen el comportamiento gráfico y las propiedades de la función. Este método es útil en campos que van desde la física hasta la economía, donde tales funciones se utilizan para modelar situaciones reales.

### **5.1.5 Utilizar la derivada para graficar funciones**

Utilizar la derivada para graficar funciones es una técnica crucial en cálculo diferencial. La derivada proporciona información detallada sobre la tasa de cambio de una función, lo que ayuda a entender su comportamiento gráfico, incluyendo la identificación de máximos y mínimos locales, puntos de inflexión, y regiones donde la función es creciente o decreciente.

**Conceptos Clave al Graficar Usando Derivadas**

**Crecimiento y Decrecimiento**

   - Una función es **creciente** en intervalos donde su derivada primera, $f'(x)$, es positiva.

  - Es **decreciente** en intervalos donde $f'(x)$ es negativa.

**Máximos y Mínimos Locales**

   - Un **máximo local** es un punto donde $f'(x) = 0$ y el signo de $f'(x)$ cambia de positivo a negativo.

   - Un **mínimo local** es un punto donde $f'(x) = 0$ y el signo de $f'(x)$ cambia de negativo a positivo.

**Puntos de Inflexión y Concavidad**

   - Si $f''(x) > 0$ (segunda derivada positiva), la función es **cóncava hacia arriba**.

   - Si $f''(x) < 0$ (segunda derivada negativa), la función es **cóncava hacia abajo**.

   - Un **punto de inflexión** ocurre donde $f''(x) = 0$ y hay un cambio en el signo de $f''(x)$.

**Ejemplo: Análisis y Gráfico de $f(x) = x^3 - 3x^2 + 2$**

**Función**:

$f(x) = x^3 - 3x^2 + 2$

**Primera Derivada (para crecimiento y puntos críticos)**:

$f'(x) = 3x^2 - 6x$

**Segunda Derivada (para concavidad y puntos de inflexión)**:

$f''(x) = 6x - 6$

**Determinación de Intervalos y Puntos Especiales**

**Puntos Críticos** (soluciones de $f'(x) = 0$):

$x = 0, \, x = 2$ (candidatos para máximos y mínimos)

**Concavidad**:

- **Cóncava hacia arriba**: $f''(x) > 0$ cuando $x > 1$

- **Cóncava hacia abajo**: $f''(x) < 0$ cuando $x < 1$

- **Punto de inflexión**: $x = 1$

**Visualización con Python**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y sus derivadas
def f(x):
    return x**3 - 3*x**2 + 2

def df(x):
    return 3*x**2 - 6*x

def ddf(x):
    return 6*x - 6

# Generar valores de x
x_values = np.linspace(-1, 4, 400)
y_values = f(x_values)
df_values = df(x_values)
ddf_values = ddf(x_values)

# Crear gráfico
plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, label='f(x) = $x^3 - 3x^2 + 2$', color='blue', linestyle='-')
plt.plot(x_values, df_values, label="f'(x) = $3x^2 - 6x$", color='red', linestyle='--')
plt.plot(x_values, ddf_values, label="f''(x) = $6x - 6$", color='green', linestyle=':')

# Añadir líneas horizontales y verticales para los ejes
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)

# Título y etiquetas de los ejes
plt.title('Análisis de la Función y sus Derivadas')
plt.xlabel('Valor de x')
plt.ylabel('Valor de y')

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Añadir leyenda
plt.legend(loc='upper right', frameon=True, shadow=True)

# Anotaciones para máximos y mínimos
plt.annotate('Máximo local', xy=(0, 2), xytext=(0.5, 10), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Mínimo local', xy=(2, -2), xytext=(2.5, -10), arrowprops=dict(facecolor='black', arrowstyle='->'))

# Mostrar gráfico
plt.show()
```

**Interpretación de las Gráficas**

- **Función $f(x)$**: Muestra el comportamiento general de la función, incluyendo sus curvas y puntos donde toca el eje x.

- **Primera Derivada $f'(x)$**: Indica dónde la función es creciente o decreciente, y muestra la ubicación de los puntos críticos (máximos y mínimos).

- **Segunda Derivada $f''(x)$**: Revela los puntos de inflexión y las regiones de concavidad hacia arriba o hacia abajo.

Este enfoque combinado proporciona una visión completa del comportamiento de una función, facilitando la comprensión y la predicción de su comportamiento en diferentes intervalos.

### **5.1.6 Encontrar la concavidad con la segunda derivada**

La **concavidad** de una función describe la curvatura de su gráfica. Determinar la concavidad es crucial para comprender cómo se comporta la función a lo largo de su dominio, especialmente en términos de cómo se curva hacia arriba o hacia abajo. La segunda derivada de una función, $f''(x)$, es la herramienta primordial para evaluar esta característica.

**Conceptos Básicos de Concavidad**

**Cóncava hacia Arriba**: Una función es cóncava hacia arriba en un intervalo si la gráfica de la función forma una "cuenca", es decir, si cualquier línea tangente a la gráfica en ese intervalo yace debajo de la gráfica. Matemáticamente, esto ocurre si $f''(x) > 0$ en ese intervalo.

**Cóncava hacia Abajo**: Una función es cóncava hacia abajo en un intervalo si la gráfica de la función forma una "cúpula", es decir, si cualquier línea tangente a la gráfica en ese intervalo yace por encima de la gráfica. Esto ocurre si $f''(x) < 0$ en ese intervalo.

**Puntos de Inflexión**: Un punto de inflexión es un punto en la gráfica de la función donde cambia la concavidad. En términos de cálculo, un punto de inflexión se encuentra donde $f''(x) = 0$ y cambia de signo.

**Ejemplo: Análisis de Concavidad y Puntos de Inflexión**

Consideremos la función polinómica $f(x) = x^3 - 6x^2 + 9x + 1$ para ilustrar cómo encontrar su concavidad utilizando la segunda derivada.

**Cálculo de Derivadas**

Primero, calculemos la primera y segunda derivadas de $f(x)$:

- **Primera Derivada**:

  $f'(x) = 3x^2 - 12x + 9$

- **Segunda Derivada**:

  $f''(x) = 6x - 12$

**Análisis de Concavidad**

Para determinar la concavidad de $f(x)$, examinamos el signo de $f''(x)$:

- **Cóncava hacia Abajo** cuando $f''(x) < 0$: $6x - 12 < 0$ implica $x < 2$.

- **Cóncava hacia Arriba** cuando $f''(x) > 0$: $6x - 12 > 0$ implica $x > 2$.

El punto $x = 2$ es un candidato a punto de inflexión, ya que es donde $f''(x) = 0$ y la concavidad podría cambiar.

**Visualización con Python**

Implementemos este análisis en Python para visualizar la función y su concavidad.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y sus derivadas
def f(x):
    return x**3 - 6*x**2 + 9*x + 1

def df(x):
    return 3*x**2 - 12*x + 9

def ddf(x):
    return 6*x - 12

# Rango de valores para x
x_values = np.linspace(-1, 4, 400)
y_values = f(x_values)
ddf_values = ddf(x_values)

# Identificar zonas de concavidad y puntos de inflexión
inflexion_point = 2
ddf_signs = ddf_values > 0

# Crear gráfico
plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, label='$f(x)$', color='royalblue', linewidth=2)
plt.plot(x_values, ddf_values, label="$f''(x)$", color='crimson', linestyle='--', linewidth=1.5)
plt.axvline(x=inflexion_point, color='purple', linestyle='--', label='Punto de Inflexión en x=2', linewidth=1.5)
plt.fill_between(x_values, y_values, color='lime', where=ddf_signs, alpha=0.3, label='Cóncava hacia arriba')
plt.fill_between(x_values, y_values, color='gold', where=~ddf_signs, alpha=0.3, label='Cóncava hacia abajo')
plt.title('Análisis de Concavidad de una Función Polinómica')
plt.xlabel('Valor de x')
plt.ylabel('Valor de f(x)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(loc='upper right', frameon=True, shadow=True)
plt.show()
```

**Interpretación de las Gráficas**

- **Función $f(x)$**: Muestra el comportamiento general y las curvas de la función.

- **Gráfico de $f''(x)$**: Indica la concavidad, con la línea roja mostrando cómo cambia la segunda derivada. El cambio de signo cerca de $(x = 2)$ sugiere un punto de inflexión.

- **Áreas coloreadas**: Indican visualmente donde la función es cóncava hacia arriba (verde) y cóncava hacia abajo (naranja).

Este análisis gráfico, reforzado por el cálculo de derivadas, ofrece una comprensión profunda de cómo la función se comporta, facilitando la identificación de características importantes como la concavidad y puntos de inflexión.
