### **5.1.1 Uso de la derivada para representar gr谩ficamente funciones**

El an谩lisis y la visualizaci贸n de una funci贸n mediante sus derivadas proporcionan una comprensi贸n profunda de su comportamiento, lo cual es crucial para muchas aplicaciones pr谩cticas en ciencia e ingenier铆a.

### **5.1.2 Encontrar la concavidad con la segunda derivada**

**Intervalos de Concavidad**

- $f''(x) > 0$ cuando $x > 2$: C贸ncava hacia arriba.

**Visualizaci贸n con Python**

Aqu铆 est谩 el c贸digo para visualizar la funci贸n y entender su concavidad:

Determinar la concavidad es crucial para comprender c贸mo se comporta la funci贸n a lo largo de su dominio, especialmente en t茅rminos de c贸mo se curva hacia arriba o hacia abajo. La segunda derivada de una funci贸n, $f''(x)$, es la herramienta primordial para evaluar esta caracter铆stica.

**Conceptos B谩sicos de Concavidad**

**C贸ncava hacia Arriba**: Una funci贸n es c贸ncava hacia arriba en un intervalo si la gr谩fica de la funci贸n forma una "cuenca", es decir, si cualquier l铆nea tangente a la gr谩fica en ese intervalo yace debajo de la gr谩fica. Matem谩ticamente, esto ocurre si $f''(x) > 0$ en ese intervalo.

**Puntos de Inflexi贸n**: Un punto de inflexi贸n es un punto en la gr谩fica de la funci贸n donde cambia la concavidad. En t茅rminos de c谩lculo, un punto de inflexi贸n se encuentra donde $f''(x) = 0$ y cambia de signo.

**Ejemplo: An谩lisis de Concavidad y Puntos de Inflexi贸n**

Consideremos la funci贸n polin贸mica $f(x) = x^3 - 6x^2 + 9x + 1$ para ilustrar c贸mo encontrar su concavidad utilizando la segunda derivada.

**C谩lculo de Derivadas**

Primero, calculemos la primera y segunda derivadas de $f(x)$:

- **Primera Derivada**:

  $f'(x) = 3x^2 - 12x + 9$

- **Segunda Derivada**:

  $f''(x) = 6x - 12$

**An谩lisis de Concavidad**

Para determinar la concavidad de $f(x)$, examinamos el signo de $f''(x)$:

- **C贸ncava hacia Abajo** cuando $f''(x) < 0$: $6x - 12 < 0$ implica $x < 2$.

- **C贸ncava hacia Arriba** cuando $f''(x) > 0$: $6x - 12 > 0$ implica $x > 2$.

El punto $x = 2$ es un candidato a punto de inflexi贸n, ya que es donde $f''(x) = 0$ y la concavidad podr铆a cambiar.

**Visualizaci贸n con Python**

Implementemos este an谩lisis en Python para visualizar la funci贸n y su concavidad.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n y sus derivadas
def f(x):
    return x**3 - 6*x**2 + 9*x + 1

def df(x):
    return 3*x**2 - 12*x + 9

def ddf(x):
    return 6*x - 12

# Rango de valores para x
x_values = np.linspace(-1, 4, 400)
y_values = f(x_values)
ddf_values = ddf(x_values)

# Identificar zonas de concavidad y puntos de inflexi贸n
inflexion_point = 2
ddf_signs = ddf_values > 0

# Crear gr谩fico
plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, label='$f(x)$', color='royalblue', linewidth=2)
plt.plot(x_values, ddf_values, label="$f''(x)$", color='crimson', linestyle='--', linewidth=1.5)
plt.axvline(x=inflexion_point, color='purple', linestyle='--', label='Punto de Inflexi贸n en x=2', linewidth=1.5)
plt.fill_between(x_values, y_values, color='lime', where=ddf_signs, alpha=0.3, label='C贸ncava hacia arriba')
plt.fill_between(x_values, y_values, color='gold', where=~ddf_signs, alpha=0.3, label='C贸ncava hacia abajo')
plt.title('An谩lisis de Concavidad de una Funci贸n Polin贸mica')
plt.xlabel('Valor de x')
plt.ylabel('Valor de f(x)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(loc='upper right', frameon=True, shadow=True)
plt.show()
```

- **reas coloreadas**: Indican visualmente donde la funci贸n es c贸ncava hacia arriba (verde) y c贸ncava hacia abajo (naranja).
