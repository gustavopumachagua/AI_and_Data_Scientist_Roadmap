En matem谩ticas, una funci贸n compuesta es una funci贸n que se compone de dos o m谩s funciones, donde la salida de una funci贸n se utiliza como entrada para otra funci贸n. La funci贸n compuesta se denota como $(f  g)(x)$, que significa que primero se aplica la funci贸n $g(x)$ y luego se aplica la funci贸n $f(x)$ al resultado de la funci贸n $g(x)$.

Por ejemplo, si tenemos dos funciones $f(x) = x^2$ y $g(x) = x + 1$, podemos crear una funci贸n compuesta $(f  g)(x)$ de la siguiente manera:

$(f  g)(x) = f(g(x)) = f(x + 1) = (x + 1)^2$

Esto significa que primero se aplica la funci贸n $g(x)$ para sumar `1` a $(x)$, y luego se aplica la funci贸n $f(x)$ para elevar al cuadrado el resultado.

En Python, podemos definir funciones compuestas utilizando la sintaxis de funciones regulares y la composici贸n de funciones utilizando la sintaxis de $(f  g)(x)$. Por ejemplo, podemos definir las funciones $f(x)$ y $g(x)$ de la siguiente manera:

```python
def f(x):
    return x**2

def g(x):
    return x + 1
```

Luego, podemos definir la funci贸n compuesta $(f  g)(x)$ de la siguiente manera:

```python
def compuesta(x):
    return f(g(x))
```

Podemos usar esta funci贸n para calcular el valor de $(f  g)(x)$ para cualquier valor de $(x)$. Por ejemplo, si queremos calcular $(f  g)(2)$, podemos escribir:

En matem谩ticas, la derivada es una medida de la tasa de cambio de una funci贸n en un punto determinado. En t茅rminos simples, la derivada nos dice qu茅 tan r谩pido cambia una funci贸n en un punto espec铆fico. La derivada de una funci贸n $f(x)$ en un punto $x=a$ se denota por $f'(a)$ y se define como el l铆mite de la tasa de cambio de la funci贸n cuando $(x)$ se acerca a $(a)$. En otras palabras, la derivada es la pendiente de la recta tangente a la funci贸n en el punto $(a)$.

La f贸rmula para la derivada de una funci贸n $f(x)$ es:

$f'(x) = \lim_{h\rightarrow 0}\frac{f(x+h) - f(x)}{h}$

donde $h$ es una cantidad infinitesimal que se acerca a cero.

A continuaci贸n se muestra un ejemplo de c贸mo calcular la derivada de una funci贸n utilizando Python:

Supongamos que queremos encontrar la derivada de la funci贸n $f(x) = x^2$ en el punto $x=2$. Podemos hacerlo utilizando la siguiente funci贸n de Python:

```python
def f(x):
    return x**2

def derivada(f, a, h=0.0001):
    return (f(a+h) - f(a))/h

f_prima = derivada(f, 2)
print(f_prima)
```

### **Programaci贸n Lineal**

La programaci贸n lineal es una t茅cnica matem谩tica utilizada para resolver problemas de optimizaci贸n en los que se busca maximizar o minimizar una funci贸n lineal sujeta a un conjunto de restricciones lineales. La funci贸n lineal se representa mediante una ecuaci贸n o funci贸n matem谩tica que es lineal en las variables de decisi贸n, es decir, variables que se pueden ajustar para optimizar la funci贸n.

Supongamos que una empresa produce dos tipos de productos: $(A)$ y $(B)$. El producto $(A)$ requiere `1` hora de trabajo y `2` unidades de materiales para ser fabricado, mientras que el producto $(B)$ requiere `2` horas de trabajo y `1` unidad de materiales. La empresa tiene `100` horas de trabajo y `80` unidades de materiales disponibles por semana. El beneficio por cada unidad de $(A)$ es de `$300` y el beneficio por cada unidad de $(B)$ es de `$200`. La empresa quiere maximizar su beneficio semanal.

Para modelar este problema de programaci贸n lineal, podemos utilizar las siguientes variables de decisi贸n:

$(x_1)$: n煤mero de unidades de $(A)$ producidas por semana

$(x_2)$: n煤mero de unidades de $(B)$ producidas por semana

Luego, podemos formular el problema de la siguiente manera:

**Maximizar $300x_1 + 200x_2$**

sujeto a:

$1x_1 + 2x2 <= 100$ (restricci贸n de horas de trabajo)

$2x_1 + 1x_2 <= 80$ (restricci贸n de materiales)

$x_1 >= 0$ (no se pueden producir unidades negativas de $(A)$)

$x_2 >= 0$ (no se pueden producir unidades negativas de $(B)$)

En Python, podemos resolver este problema utilizando la biblioteca `PuLP` de la siguiente manera:

```python
from pulp import *
import matplotlib.pyplot as plt
import numpy as np

# Crear el problema de maximizaci贸n
prob = LpProblem("Maximizacion_de_beneficios", LpMaximize)

# Definir las variables de decisi贸n
A = LpVariable("Unidades_de_A", lowBound=0)
B = LpVariable("Unidades_de_B", lowBound=0)

# Definir la funci贸n objetivo
prob += 300 * A + 200 * B, "Beneficio"

# Definir las restricciones
prob += A + 2 * B <= 100, "Horas_de_trabajo"
prob += 2 * A + B <= 80, "Unidades_de_materiales"

# Resolver el problema
prob.solve()

# Graficar la funci贸n y el m铆nimo
x = np.linspace(0, 50, 1000)
y1 = (100 - x) / 2
y2 = 80 - 2 * x

plt.figure(figsize=(10, 8))
plt.plot(x, y1, label="Horas de trabajo", linewidth=2, color='dodgerblue')
plt.plot(x, y2, label="Unidades de materiales", linewidth=2, color='crimson')

# Graficar la regi贸n factible
plt.fill_between(x, 0, np.minimum(y1, y2), color='gray', alpha=0.3, label='Regi贸n factible')

# Resaltar la soluci贸n 贸ptima
plt.scatter(value(A), value(B), color='gold', zorder=5, s=100, label='Soluci贸n 贸ptima')
plt.annotate(f'Soluci贸n 贸ptima\n(A={value(A):.2f}, B={value(B):.2f})', xy=(value(A), value(B)), xytext=(20, 10),
             textcoords="offset points", bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5"))

# Configuraci贸n final
plt.xlabel("Unidades de A", fontsize=14, color='navy')
plt.ylabel("Unidades de B", fontsize=14, color='navy')
plt.title("Programaci贸n Lineal - Maximizaci贸n de Beneficios", fontsize=16, fontweight='bold', color='green')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(0, 50)
plt.ylim(0, 50)

# Mostrar la gr谩fica
plt.show()
```

Esto significa que la empresa debe producir `20` unidades de `A` y `40` unidades de `B` por semana para maximizar su beneficio semanal, lo que resulta en un beneficio m谩ximo semanal de `$14000`.
