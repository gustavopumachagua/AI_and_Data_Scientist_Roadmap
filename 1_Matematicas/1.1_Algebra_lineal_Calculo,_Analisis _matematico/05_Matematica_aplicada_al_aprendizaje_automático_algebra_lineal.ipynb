### **5.1.1 ¬øQu√© son los valores propios y los vectores propios?**

Los valores propios y los vectores propios son conceptos centrales en √°lgebra lineal que tienen importantes aplicaciones en √°reas como mec√°nica cu√°ntica, teor√≠a de sistemas, procesamiento de im√°genes, econom√≠a, y m√°s. Son fundamentales para entender las propiedades fundamentales de las matrices y las transformaciones lineales que describen.

**Definici√≥n de Valores Propios y Vectores Propios**

Un **vector propio** de una matriz cuadrada $(A)$ es un vector no nulo $(\mathbf{v})$ tal que, cuando se multiplica por $(A)$, el resultado es un m√∫ltiplo escalar de $(\mathbf{v})$. El escalar por el cual se multiplica $(\mathbf{v})$ se llama **valor propio** asociado con ese vector propio.

Matem√°ticamente, esto se define como:

$A \mathbf{v} = \lambda \mathbf{v}$

donde:

- $(A)$ es una matriz cuadrada.

- $(\mathbf{v})$ es un vector propio de $(A)$.

- $(\lambda)$ es el valor propio correspondiente a $(\mathbf{v})$.

**C√≥mo Encontrar Valores Propios y Vectores Propios**

Para encontrar los valores propios de una matriz $(A)$, debemos resolver la ecuaci√≥n caracter√≠stica:

$\det(A - \lambda I) = 0$

donde $(I)$ es la matriz identidad del mismo tama√±o que $(A)$. La soluci√≥n de esta ecuaci√≥n proporciona los valores propios $(\lambda)$. Una vez conocidos los valores propios, los vectores propios correspondientes se pueden encontrar resolviendo el sistema de ecuaciones lineales resultante de $(A - \lambda I) \mathbf{v} = 0$.

**Ejemplo: Matriz 2x2**

Consideremos la matriz:

$A = \begin{pmatrix}
2 & 1 \\
1 & 2
\end{pmatrix}$

**Encontrar los Valores Propios**

La ecuaci√≥n caracter√≠stica es:

$\det(A - \lambda I) = \det \begin{pmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{pmatrix} = (2 - \lambda)^2 - 1 = \lambda^2 - 4\lambda + 3 = 0$

Resolviendo esta ecuaci√≥n cuadr√°tica:

$\lambda^2 - 4\lambda + 3 = 0 \implies \lambda = 1, \lambda = 3$

**Encontrar los Vectores Propios**

Para $(\lambda = 1)$:

$(A - I) \mathbf{v} = \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix} \mathbf{v} = 0 \implies \text{tomar } \mathbf{v} = \begin{pmatrix}
1 \\
-1
\end{pmatrix}$

Para $(\lambda = 3)$:

$(A - 3I) \mathbf{v} = \begin{pmatrix}
-1 & 1 \\
1 & -1
\end{pmatrix} \mathbf{v} = 0 \implies \text{tomar } \mathbf{v} = \begin{pmatrix}
1 \\
1
\end{pmatrix}$

**C√≥digo en Python para Encontrar Valores y Vectores Propios**

```python
import numpy as np

# Definir la matriz A
A = np.array([[2, 1], [1, 2]])

# Calcular valores y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios (columnas de la matriz siguiente):\n", vectores_propios)
```

**Conclusi√≥n**

Los valores y vectores propios proporcionan una visi√≥n profunda de la estructura de una matriz, revelando c√≥mo transforma el espacio vectorial. Son cruciales para simplificar problemas, especialmente aquellos involucrados con sistemas din√°micos y transformaciones lineales, y son ampliamente utilizados para descomposiciones matriciales, an√°lisis de estabilidad, y en la teor√≠a espectral de grafos.

### **5.1.2 Casos propios especiales**

En matem√°ticas, especialmente en √°lgebra lineal, los "casos propios especiales" suelen referirse a situaciones particulares que involucran valores propios y vectores propios de matrices que tienen caracter√≠sticas √∫nicas o que requieren un tratamiento espec√≠fico. Estos casos pueden incluir matrices con propiedades como simetr√≠a, ser ortogonales, tener determinante cero, o poseer multiplicidad algebraica y geom√©trica que no coinciden. A continuaci√≥n, exploraremos algunos de estos casos especiales con ejemplos detallados y su implementaci√≥n en Python.

Las matrices sim√©tricas son particularmente interesantes en el contexto de los valores propios y vectores propios porque siempre son diagonalizables, y sus valores propios son reales.

**Ejemplo: Matriz Sim√©trica**

Consideremos la matriz sim√©trica $(A)$:

$A = \begin{pmatrix}
4 & 1 \\
1 & 3
\end{pmatrix}$

**Propiedades**

- Todos los valores propios son reales.

- Los vectores propios asociados con diferentes valores propios son ortogonales entre s√≠.

**C√≥digo en Python**

```python
import numpy as np

# Definir la matriz sim√©trica A
A = np.array([[4, 1], [1, 3]])

# Calcular valores y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios:\n", vectores_propios)
```

**Ejemplo: Matriz Singular**

Consideremos la matriz singular $(B)$:

$B = \begin{pmatrix}

- No es invertible.

**C√≥digo en Python**

```python
# Definir la matriz B
B = np.array([[2, 4], [1, 2]])

# Calcular valores y vectores propios
valores_propios_b, vectores_propios_b = np.linalg.eig(B)

print("Valores propios de B:", valores_propios_b)
print("Vectores propios de B:\n", vectores_propios_b)
```

**Matrices con Multiplicidad Algebraica y Geom√©trica Diferente**

La multiplicidad algebraica de un valor propio es el n√∫mero de veces que este valor aparece como ra√≠z del polinomio caracter√≠stico, mientras que la multiplicidad geom√©trica es la dimensi√≥n del espacio generado por sus vectores propios. Cuando estas multiplicidades no coinciden, indica la presencia de una matriz defectuosa, lo que afecta a su diagonalizaci√≥n.

**Ejemplo: Multiplicidad Diferente**

Consideremos una matriz $(C)$:

$C = \begin{pmatrix}

**Propiedades**

- Tiene un valor propio $(\lambda = 2)$ con multiplicidad algebraica 2, pero su multiplicidad geom√©trica es 1.

**C√≥digo en Python**

```python
# Definir la matriz C
C = np.array([[2, 1], [0, 2]])

# Calcular valores y vectores propios
valores_propios_c, vectores_propios_c = np.linalg.eig(C)

print("Valores propios de C:", valores_propios_c)
print("Vectores propios de C:\n", vectores_propios_c)
```

### **5.1.3 C√°lculo de los vectores propios**

El c√°lculo de los vectores propios es un aspecto central del √°lgebra lineal y tiene aplicaciones en diversas √°reas como mec√°nica cu√°ntica, an√°lisis de sistemas din√°micos, reducci√≥n de dimensionalidad en estad√≠stica y m√°s. Un vector propio de una matriz es un vector no nulo que, cuando se multiplica por la matriz, da como resultado un m√∫ltiplo escalar del mismo vector. Este escalar es conocido como el valor propio asociado al vector propio.

**¬øC√≥mo se Calculan los Vectores Propios?**

Para calcular los vectores propios de una matriz $(A)$, se deben seguir varios pasos que involucran tanto la determinaci√≥n de los valores propios como la soluci√≥n de sistemas de ecuaciones lineales.

**Determinar los Valores Propios**

Los valores propios $(\lambda)$ se obtienen resolviendo la ecuaci√≥n caracter√≠stica de la matriz:

$\det(A - \lambda I) = 0$

Donde:

- $(A)$ es la matriz de la cual queremos encontrar los vectores propios.

- $(I)$ es la matriz identidad del mismo tama√±o que $(A)$.

- $(\lambda)$ representa los valores propios.

Esta ecuaci√≥n resulta en un polinomio conocido como el polinomio caracter√≠stico, cuyas ra√≠ces son los valores propios de $(A)$.

**Calcular los Vectores Propios para Cada Valor Propio**

Una vez conocidos los valores propios, para cada $(\lambda)$, se resuelve el sistema lineal:

$(A - \lambda I)\mathbf{v} = 0$

Donde $(\mathbf{v})$ es el vector propio asociado a $(\lambda)$. Este sistema generalmente se resuelve mediante m√©todos de √°lgebra lineal como eliminaci√≥n gaussiana o descomposici√≥n LU.

**Ejemplo Pr√°ctico: Matriz 2x2**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix}
4 & 1 \\
2 & 3
\end{pmatrix}$

**Paso 1: Calcular los Valores Propios**

Calculamos el determinante de $(A - \lambda I)$:

$\det(A - \lambda I) = \det\left(\begin{pmatrix}
4 - \lambda & 1 \\
2 & 3 - \lambda
\end{pmatrix}\right) = (4 - \lambda)(3 - \lambda) - 2 \cdot 1 = \lambda^2 - 7\lambda + 10$

Resolvemos $(\lambda^2 - 7\lambda + 10 = 0)$:

$\lambda = \frac{7 \pm \sqrt{49 - 40}}{2} = \frac{7 \pm 3}{2} \rightarrow \lambda_1 = 5, \lambda_2 = 2$

**Paso 2: Encontrar los Vectores Propios**

Para $(\lambda_1 = 5)$:

$(A - 5I)\mathbf{v} = \begin{pmatrix}
-1 & 1 \\
2 & -2
\end{pmatrix}\mathbf{v} = 0 \rightarrow \text{una soluci√≥n es } \mathbf{v} = \begin{pmatrix}
1 \\
1
\end{pmatrix}$

Para $(\lambda_2 = 2)$:

$(A - 2I)\mathbf{v} = \begin{pmatrix}
2 & 1 \\
2 & 1
\end{pmatrix}\mathbf{v} = 0 \rightarrow \text{una soluci√≥n es } \mathbf{v} = \begin{pmatrix}
-1 \\
2
\end{pmatrix}$

**C√≥digo en Python para Calcular Valores y Vectores Propios**

```python
import numpy as np

A = np.array([[4, 1], [2, 3]])

# Calcular valores y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios (cada columna es un vector propio):\n", vectores_propios)
```

**Conclusi√≥n**

Calcular los vectores propios de una matriz es fundamental para muchas aplicaciones en ciencia y ingenier√≠a, incluyendo an√°lisis de vibraciones, teor√≠a de estabilidad, y m√°s. Este proceso no solo ayuda a entender las propiedades intr√≠nsecas de las transformaciones lineales representadas por las matrices, sino que tambi√©n es crucial para t√©cnicas como la descomposici√≥n de matrices y la diagonalizaci√≥n.

### **5.1.4 Cambio a la base propia**

El cambio a la base propia de una matriz es un concepto fundamental en √°lgebra lineal que permite simplificar muchas operaciones matem√°ticas y anal√≠ticas al trabajar con matrices y transformaciones lineales. Este proceso implica cambiar la base de un espacio vectorial de su base est√°ndar a una base formada por vectores propios de una matriz.

**¬øQu√© es una Base Propia?**

Una **base propia** de una matriz es un conjunto de vectores propios que forman una base para el espacio vectorial en el que opera la matriz. Utilizar una base propia simplifica la forma de la matriz, convirti√©ndola en una matriz diagonal en el caso ideal, donde los elementos diagonales son los valores propios de la matriz.

**Proceso de Cambio a la Base Propia**

El cambio a la base propia implica varios pasos clave:

**Encontrar los vectores propios** de la matriz. Estos vectores servir√°n como la nueva base para el espacio vectorial.

**Formar la matriz de cambio de base** $(P)$, cuyas columnas son los vectores propios de la matriz original $(A)$.

**Transformar la matriz original** $(A)$ a la nueva base utilizando la matriz de cambio de base:

   $A' = P^{-1}AP$

   donde $(A')$ es la representaci√≥n de $(A)$ en la nueva base y es una matriz diagonal si $(A)$ es diagonalizable.

**Ejemplo: Cambio a la Base Propia**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix}
4 & 1 \\
1 & 4
\end{pmatrix}$

**Paso 1: Encontrar los Vectores Propios**

Calculamos los valores y vectores propios de $(A)$:

```python
import numpy as np

A = np.array([[4, 1], [1, 4]])
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios:\n", vectores_propios)
```

**Paso 2: Formar la Matriz de Cambio de Base**

La matriz de cambio de base $(P)$ se forma con los vectores propios como columnas:

$P = \text{vectores\_propios}$

**Paso 3: Transformar la Matriz Original**

Calculamos $(A')$ usando la f√≥rmula $(A' = P^{-1}AP)$:

```python
P = vectores_propios
P_inv = np.linalg.inv(P)
A_prime = P_inv @ A @ P

print("A' en la base propia:\n", A_prime)
```

**Visualizaci√≥n de la Transformaci√≥n**

Podemos visualizar el efecto del cambio de base en los vectores unitarios est√°ndar, mostrando c√≥mo se alinean con los vectores propios despu√©s de la transformaci√≥n.

**C√≥digo para Visualizaci√≥n**

```python
import numpy as np
import matplotlib.pyplot as plt

# Supongamos una matriz de proyecci√≥n P como ejemplo
P = np.array([[1, 2], [2, 1]])  # Debe ser una matriz realista en tu contexto

# Vectores unitarios est√°ndar
I = np.identity(2)

# Aplicar transformaci√≥n a los vectores unitarios
transformed_I = P @ I

# Funci√≥n para graficar vectores
def plot_vector(v, origin=[0, 0], color='r', label=None):
    plt.quiver(origin[0], origin[1], v[0], v[1], scale=1, scale_units='xy', angles='xy', color=color, label=label, width=0.012, headwidth=3, headlength=4)

# Crear gr√°fico
plt.figure(figsize=(6, 6))
plot_vector(I[:, 0], color='blue', label='$I_x$ (Original)')
plot_vector(I[:, 1], color='green', label='$I_y$ (Original)')
plot_vector(transformed_I[:, 0], color='red', label='Primero Propio')
plot_vector(transformed_I[:, 1], color='purple', label='Segundo Propio')

# Configuraci√≥n adicional para mejorar la visualizaci√≥n
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid(True, linestyle='--', alpha=0.7)  # Grilla mejorada para visibilidad
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend(loc='upper right', frameon=True, framealpha=0.9, shadow=True)
plt.title('Efecto del Cambio a la Base Propia', fontsize=14, fontweight='bold')

# Mostrar el gr√°fico
plt.show()
```

**Conclusi√≥n**

El cambio a la base propia es una t√©cnica poderosa en √°lgebra lineal que simplifica la representaci√≥n de matrices y facilita la soluci√≥n de problemas complejos al alinear la base con las direcciones naturales de estiramiento y compresi√≥n de la transformaci√≥n lineal. Este proceso es crucial en disciplinas que incluyen f√≠sica, ingenier√≠a, estad√≠stica y m√°s, donde las propiedades estructurales de los sistemas son esenciales para su an√°lisis y comprensi√≥n.

### **5.1.5 Ejemplo de base propia**

Cuando hablamos de una **base propia**, nos referimos a un conjunto de vectores propios que forma una base para un espacio vectorial. Este conjunto es especialmente √∫til porque cualquier vector en el espacio se puede expresar de manera √∫nica como una combinaci√≥n lineal de los vectores propios, y las transformaciones lineales representadas por la matriz asociada a estos vectores se simplifican considerablemente.

**Definici√≥n y Propiedades**

Una base propia de una matriz $(A)$ es un conjunto de vectores propios que no solo son linealmente independientes, sino que tambi√©n cubren el espacio vectorial en el que $(A)$ opera. Si una matriz es diagonalizable, entonces es posible transformarla en una matriz diagonal en esta base, donde los elementos diagonales son los valores propios correspondientes a cada vector propio de la base.

**Propiedades de la Base Propia**

- **Simplificaci√≥n de la Matriz**: En la base propia, la matriz de transformaci√≥n se convierte en una matriz diagonal, lo que facilita operaciones como la elevaci√≥n a potencias, c√°lculo de exponentes de matrices, y otras operaciones algebraicas.

- **Descomposici√≥n Espectral**: La matriz puede escribirse como una suma de proyectores ortogonales escalados por sus valores propios.

**Ejemplo Pr√°ctico: Matriz Diagonalizable**

Consideremos la matriz $(A)$ definida por:

$A = \begin{pmatrix}
2 & 1 \\
1 & 2
\end{pmatrix}$

**Paso 1: Determinar Valores Propios**

Calculamos el polinomio caracter√≠stico y los valores propios:

$\text{det}(A - \lambda I) = \text{det} \left( \begin{pmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{pmatrix} \right) = (2 - \lambda)^2 - 1 = \lambda^2 - 4\lambda + 3 = 0$

Resolviendo $(\lambda^2 - 4\lambda + 3 = 0)$:

$\lambda_1 = 3, \quad \lambda_2 = 1$

**Paso 2: Encontrar los Vectores Propios**

Para $(\lambda_1 = 3)$:

$(A - 3I) = \begin{pmatrix}
-1 & 1 \\
1 & -1
\end{pmatrix} \rightarrow \text{El vector propio es } \begin{pmatrix}
1 \\
1
\end{pmatrix}$

Para $(\lambda_2 = 1)$:

$(A - I) = \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix} \rightarrow \text{El vector propio es } \begin{pmatrix}
-1 \\
1
\end{pmatrix}$

**Paso 3: Formar la Matriz de Cambio de Base y Diagonalizar**

La matriz de cambio de base $(P)$ que tiene los vectores propios como columnas es:

$P = \begin{pmatrix}
1 & -1 \\
1 & 1
\end{pmatrix}$

La matriz diagonal $(D)$ correspondiente en la base propia es:

$D = P^{-1}AP = \begin{pmatrix}
3 & 0 \\
0 & 1
\end{pmatrix}$

**C√≥digo en Python**

```python
import numpy as np

# Definir la matriz
A = np.array([[2, 1], [1, 2]])

# Calcular valores propios y vectores propios
eigenvalues, eigenvectors = np.linalg.eig(A)

# Matriz de cambio de base
P = eigenvectors

# Diagonalizar A
D = np.linalg.inv(P) @ A @ P

print("Valores propios:", eigenvalues)
print("Vectores propios:\n", eigenvectors)
print("Matriz diagonalizada:\n", D)
```

### **5.1.6 Introducci√≥n al PageRank**

**PageRank** es un algoritmo desarrollado por Larry Page y Sergey Brin, los fundadores de Google, en la Universidad de Stanford. Este algoritmo fue dise√±ado para medir la importancia de las p√°ginas web en los resultados de b√∫squeda. La idea central del PageRank es que la importancia de una p√°gina web puede ser juzgada por el n√∫mero de p√°ginas que enlazan a ella, as√≠ como por la importancia de esas mismas p√°ginas.

**Concepto B√°sico de PageRank**

PageRank trabaja bajo la premisa de que las p√°ginas importantes son m√°s propensas a recibir m√°s enlaces de otras p√°ginas. En esencia, el algoritmo cuenta el n√∫mero y la calidad de los enlaces para determinar una estimaci√≥n aproximada de la importancia del sitio. Cuanto m√°s importante es el sitio que enlaza a otro, mayor es el "peso" de ese enlace en la puntuaci√≥n de PageRank de la p√°gina enlazada.

**F√≥rmula de PageRank**

La f√≥rmula original de PageRank que se utilizaba en Google se puede representar de la siguiente manera:

$PR(p_i) = \frac{1-d}{N} + d \sum_{p_j \in M(p_i)} \frac{PR(p_j)}{L(p_j)}$

donde:

- $PR(p_i)$ es el PageRank de la p√°gina $(p_i)$.

- $M(p_i)$ es el conjunto de p√°ginas que enlazan a $(p_i)$.

- $L(p_j)$ es el n√∫mero de enlaces salientes de la p√°gina $(p_j)$.

- $(d)$ es un factor de amortiguaci√≥n, generalmente establecido en 0.85.

- $(N)$ es el n√∫mero total de p√°ginas.

**Ejemplo Ilustrativo**

Imaginemos un peque√±o conjunto de p√°ginas web con la siguiente estructura de enlaces:

- P√°gina A enlaza a las p√°ginas B y C.

- P√°gina B enlaza a la p√°gina C y A.

- P√°gina C enlaza solo a la p√°gina A.

- P√°gina D enlaza a todas las p√°ginas A, B y C.

**Representaci√≥n Matem√°tica**

Podemos representar esta red como una matriz de transici√≥n $(P)$, donde cada columna representa una p√°gina y cada fila indica a qu√© p√°ginas enlaza. Si normalizamos esta matriz por columnas, obtenemos la matriz de probabilidad de transici√≥n usada en el algoritmo de PageRank.

**C√≥digo en Python para Calcular PageRank**

```python
import numpy as np

# Matriz de adyacencia
A = np.array([
    [0, 1, 1, 1],  # A recibe enlaces de B, C y D
    [1, 0, 0, 1],  # B recibe enlaces de A y D
    [1, 1, 0, 1],  # C recibe enlaces de A, B y D
    [0, 0, 0, 0]   # D no recibe enlaces
])

# N√∫mero total de p√°ginas
N = A.shape[0]

# Factor de amortiguaci√≥n
d = 0.85

# Matriz de transici√≥n P
P = A / A.sum(axis=0, keepdims=True)
P[:, 3] = 1 / N  # Corregir la columna de D que tiene solo ceros

# Vector de PageRank inicial (suposici√≥n uniforme)
pr = np.ones(N) / N

# Iterar hasta convergencia
for _ in range(100):
    pr = (1 - d) / N + d * P @ pr

print("PageRank de las p√°ginas A, B, C, y D:")
print(pr)
```

### **5.1.7 Selecci√≥n de vectores propios por inspecci√≥n**

La selecci√≥n de vectores propios por inspecci√≥n es un enfoque pr√°ctico para identificar vectores propios y valores propios de una matriz cuando su estructura es simple y los patrones en sus elementos lo permiten. Este m√©todo es especialmente √∫til para matrices peque√±as o matrices con una estructura clara que facilita la identificaci√≥n directa de sus propiedades espectrales sin c√°lculos extensos.

**Concepto de Vectores Propios y Valores Propios**

Un **vector propio** de una matriz $(A)$ es un vector no nulo $(\mathbf{v})$ que, cuando se multiplica por $(A)$, resulta en un m√∫ltiplo escalar de $( \mathbf{v})$, es decir, $A \mathbf{v} = \lambda \mathbf{v}$, donde $(\lambda)$ es el **valor propio** correspondiente.

**Proceso de Selecci√≥n por Inspecci√≥n**

Seleccionar vectores propios por inspecci√≥n generalmente involucra los siguientes pasos:

**Observar la Matriz**: Examinar la matriz buscando patrones como filas o columnas sim√©tricas, filas o columnas de ceros, y otros indicativos que puedan sugerir relaciones simples entre los elementos.

**Identificar Candidatos a Vectores Propios**: Basado en la observaci√≥n, identificar vectores que probablemente se mantengan invariantes o se transformen de manera predecible cuando se multiplica por la matriz.

**Verificar Matem√°ticamente**: Realizar c√°lculos para confirmar que los vectores candidatos son efectivamente vectores propios y determinar sus correspondientes valores propios.

**Ejemplo Pr√°ctico: Matriz Diagonal y Matriz con Simetr√≠as**

Consideremos dos matrices, una diagonal y otra con un patr√≥n sim√©trico claro:

**Matriz Diagonal $(D)$**:

   $D = \begin{pmatrix}
   3 & 0 & 0 \\
   0 & 2 & 0 \\
   0 & 0 & 5
   \end{pmatrix}$

   Aqu√≠, por inspecci√≥n, podemos ver directamente que los vectores propios son los vectores de la base est√°ndar $(\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3)$ con valores propios $(3, 2, 5)$ respectivamente.

**Matriz Sim√©trica $(S)$**:

   $S = \begin{pmatrix}
   2 & -2 & 4 \\
   -2 & 3 & -6 \\
   4 & -6 & 8
   \end{pmatrix}$

   Aunque no tan inmediata como una matriz diagonal, una matriz sim√©trica sugiere que podemos probar vectores simples como $\mathbf{v} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}$ para ver si revela un patr√≥n. La simetr√≠a a menudo implica que los valores propios ser√°n reales.

**C√≥digo en Python para Verificaci√≥n**

Vamos a calcular y verificar los vectores propios para la matriz sim√©trica $(S)$:

```python
import numpy as np

# Definir la matriz sim√©trica S
S = np.array([[2, -2, 4], [-2, 3, -6], [4, -6, 8]])

# Calcular valores y vectores propios
valores_propios, vectores_propios = np.linalg.eig(S)

print("Valores propios de S:", valores_propios)
print("Vectores propios de S (cada columna es un vector propio):\n", vectores_propios)
```

### **5.1.8 Polinomios caracter√≠sticos, valores propios y vectores propios**

Los polinomios caracter√≠sticos, valores propios y vectores propios son conceptos centrales en √°lgebra lineal que tienen aplicaciones en muchas √°reas de la ciencia y la ingenier√≠a, incluyendo an√°lisis estructural, mec√°nica cu√°ntica, an√°lisis de redes, procesamiento de im√°genes y m√°s.

**Polinomio Caracter√≠stico**

El **polinomio caracter√≠stico** de una matriz $(A)$ es un polinomio que se obtiene de calcular el determinante de $(A - \lambda I)$, donde $(\lambda)$ es una variable escalar e $(I)$ es la matriz identidad del mismo tama√±o que $(A)$. Este polinomio tiene la forma general:

$p(\lambda) = \det(A - \lambda I)$

La soluci√≥n de la ecuaci√≥n $p(\lambda) = 0$, conocida como la **ecuaci√≥n caracter√≠stica**, da como resultado los **valores propios** de la matriz.

**Valores Propios y Vectores Propios**

Un **valor propio** $(\lambda)$ de una matriz $(A)$ es un escalar para el cual existe un vector no nulo $(\mathbf{v})$, llamado **vector propio**, tal que:

$A \mathbf{v} = \lambda \mathbf{v}$

Los vectores propios asociados a un valor propio particular forman un espacio llamado **espacio propio**.

**Ejemplo Pr√°ctico: Matriz 2x2**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix}
2 & 1 \\
1 & 2
\end{pmatrix}$

**Paso 1: Calcular el Polinomio Caracter√≠stico**

Para encontrar el polinomio caracter√≠stico de $(A)$, calculamos el determinante de $(A - \lambda I)$:

$A - \lambda I = \begin{pmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{pmatrix}$

$p(\lambda) = \det(A - \lambda I) = (2-\lambda)(2-\lambda) - 1 = \lambda^2 - 4\lambda + 3$

**Paso 2: Resolver la Ecuaci√≥n Caracter√≠stica**

Resolvemos $p(\lambda) = 0$:

$\lambda^2 - 4\lambda + 3 = 0$

Usando la f√≥rmula cuadr√°tica:

$\lambda = \frac{4 \pm \sqrt{(4)^2 - 4 \times 1 \times 3}}{2 \times 1} = \frac{4 \pm \sqrt{16 - 12}}{2} = \frac{4 \pm 2}{2}$

$\lambda_1 = 3, \quad \lambda_2 = 1$

**Paso 3: Encontrar los Vectores Propios**

Para $\lambda_1 = 3$:

$(A - 3I)\mathbf{v} = \begin{pmatrix}
-1 & 1 \\
1 & -1
\end{pmatrix}\mathbf{v} = 0 \rightarrow \text{soluci√≥n: } \mathbf{v}_1 = \begin{pmatrix}
1 \\
1
\end{pmatrix}$

Para $\lambda_2 = 1$:

$(A - 1I)\mathbf{v} = \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix}\mathbf{v} = 0 \rightarrow \text{soluci√≥n: } \mathbf{v}_2 = \begin{pmatrix}
-1 \\
1
\end{pmatrix}$

**C√≥digo en Python para C√°lculo y Visualizaci√≥n**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz
A = np.array([[2, 1], [1, 2]])

# Calcular valores y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios:\n", vectores_propios)

# Configurar la gr√°fica
fig, ax = plt.subplots()
colors = ['red', 'blue']  # Colores para cada vector propio

# Graficar vectores propios
for i in range(len(valores_propios)):
    ax.quiver(0, 0, vectores_propios[0, i], vectores_propios[1, i], angles='xy', scale_units='xy', scale=1,
              color=colors[i], label=f'Œª={valores_propios[i]:.2f}', width=0.012, headwidth=3, headlength=4)

# Configuraci√≥n del gr√°fico
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.axhline(0, color='black', linewidth=1)  # Hacer la l√≠nea del eje m√°s visible
ax.axvline(0, color='black', linewidth=1)
ax.grid(True, linestyle='--', alpha=0.7)  # Grilla m√°s sutil
ax.set_aspect('equal')  # Asegurar que las proporciones sean iguales

# A√±adir leyenda y t√≠tulo
ax.legend(loc='upper right', frameon=True, framealpha=0.9, shadow=True)
plt.title('Vectores Propios de la Matriz A', fontsize=14, fontweight='bold')

# Mostrar el gr√°fico
plt.show()
```

**Conclusi√≥n**

Entender el polinomio caracter√≠stico, los valores propios y los vectores propios es crucial para muchos campos de la ciencia y la tecnolog√≠a, permitiendo desde la simplificaci√≥n de sistemas de ecuaciones hasta el an√°lisis de estabilidad de sistemas din√°micos. El uso de software como Python facilita el c√°lculo y la visualizaci√≥n de estas propiedades matem√°ticas importantes.

### **5.1.9 Diagonalizaci√≥n y aplicaciones**

La diagonalizaci√≥n es un proceso fundamental en √°lgebra lineal que permite simplificar el c√°lculo de potencias de matrices, la soluci√≥n de ecuaciones diferenciales lineales, y la transformaci√≥n de sistemas en formas m√°s manejables. Es especialmente √∫til en √°reas como mec√°nica cu√°ntica, teor√≠a de control, econom√≠a, y procesamiento de se√±ales.

**¬øQu√© es la Diagonalizaci√≥n?**

Una matriz cuadrada $(A)$ se dice que es **diagonalizable** si existe una matriz $(P)$ y una matriz diagonal $(D)$ tales que:

$P^{-1} A P = D$

donde:

- $(P)$ es una matriz cuyas columnas son los vectores propios de $(A)$.

- $(D)$ es una matriz diagonal cuyos elementos diagonales son los valores propios correspondientes a los vectores propios en $(P)$.

**Propiedades de las Matrices Diagonalizables**

- **Simplificaci√≥n de operaciones**: Las operaciones como elevar la matriz a una potencia pueden simplificarse enormemente, ya que $(A^n = P D^n P^{-1})$, donde $(D^n)$ es simplemente elevar cada elemento diagonal a $(n)$.

- **An√°lisis de sistemas**: En sistemas din√°micos, la diagonalizaci√≥n permite estudiar el sistema en t√©rminos de sus modos naturales de comportamiento, facilitando la comprensi√≥n de la estabilidad y la din√°mica del sistema.

**Ejemplo Pr√°ctico: Diagonalizaci√≥n de una Matriz**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix}
4 & 1 \\
1 & 4
\end{pmatrix}$

**Paso 1: Calcular los Valores y Vectores Propios**

Primero, encontramos los valores propios $(\lambda)$ de $(A)$ resolviendo el determinante de $(A - \lambda I)$:

$\det(A - \lambda I) = \det\left(\begin{pmatrix}
4 - \lambda & 1 \\
1 & 4 - \lambda
\end{pmatrix}\right) = (4-\lambda)^2 - 1 = \lambda^2 - 8\lambda + 15 = 0$

Resolviendo esta ecuaci√≥n cuadr√°tica obtenemos:

$\lambda_1 = 5, \quad \lambda_2 = 3$

Los vectores propios correspondientes son encontrados resolviendo $(A - \lambda_i I)\mathbf{v} = 0$ para cada $(\lambda_i)$:

Para $(\lambda_1 = 5)$:

$\mathbf{v}_1 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$

Para $(\lambda_2 = 3)$:

$\mathbf{v}_2 = \begin{pmatrix} -1 \\ 1 \end{pmatrix}$

**Paso 2: Formar $(P)$ y $(D)$**

$P = \begin{pmatrix}
1 & -1 \\
1 & 1
\end{pmatrix}, \quad D = \begin{pmatrix}
5 & 0 \\
0 & 3
\end{pmatrix}$

**Paso 3: Verificaci√≥n**

Verificamos que $P^{-1} A P = D$:

```python
import numpy as np

A = np.array([[4, 1], [1, 4]])
P = np.array([[1, -1], [1, 1]])
D = np.array([[5, 0], [0, 3]])

# Calcular P inversa
P_inv = np.linalg.inv(P)

# Verificar la diagonalizaci√≥n
A_prime = P_inv @ A @ P
print(A_prime)
```

**Aplicaciones de la Diagonalizaci√≥n**

**Potencias de Matrices**: Calcular $(A^n)$ se simplifica enormemente.

**Sistemas de Ecuaciones Diferenciales**: Resolver $(\dot{\mathbf{x}} = A\mathbf{x})$ se facilita al desacoplar el sistema en la base de vectores propios.

**Econom√≠a y Finanzas**: An√°lisis de modelos de Markov y otros modelos econ√≥micos.

**Mec√°nica Cu√°ntica**: Las matrices de Hamiltonianos a menudo se diagonalizan para encontrar estados estacionarios y energ√≠as.

**Visualizaci√≥n de la Diagonalizaci√≥n**

Aunque la diagonalizaci√≥n en s√≠ misma es un proceso algebraico, podemos visualizar los efectos de la transformaci√≥n mediante la representaci√≥n de los vectores propios y su orientaci√≥n en el espacio, o simular c√≥mo la matriz transforma un conjunto de vectores unitarios en el espacio, mostrando la alineaci√≥n con los vectores propios. Estas visualizaciones ser√≠an similares a las proporcionadas en ejemplos de vectores propios.

### **5.1.10 Valores propios y vectores propios**

Los **valores propios** y **vectores propios** son conceptos fundamentales en √°lgebra lineal que juegan un papel crucial en numerosas aplicaciones matem√°ticas y cient√≠ficas, incluyendo mec√°nica, √≥ptica, an√°lisis de datos, y m√°s. Estos conceptos son esenciales para entender c√≥mo una transformaci√≥n lineal act√∫a sobre un espacio vectorial.

**Definici√≥n de Valores Propios y Vectores Propios**

Un **valor propio** $(\lambda)$ de una matriz $(A)$ es un escalar que satisface la condici√≥n:

$A \mathbf{v} = \lambda \mathbf{v}$

donde $(\mathbf{v})$ es un vector no nulo conocido como **vector propio** asociado a $(\lambda)$. En otras palabras, un vector propio es un vector que, cuando es transformado por $(A)$, no cambia su direcci√≥n sino solo su magnitud (escalado por $(\lambda)$.

**C√≥mo Encontrar Valores Propios y Vectores Propios**

Para encontrar los valores propios de una matriz $(A)$, debemos resolver la ecuaci√≥n caracter√≠stica:

$\det(A - \lambda I) = 0$

donde $(I)$ es la matriz identidad del mismo tama√±o que $(A)$. Esta ecuaci√≥n produce un polinomio cuyas ra√≠ces son los valores propios de $(A)$. Los vectores propios se obtienen resolviendo el sistema de ecuaciones lineales resultante de la sustituci√≥n de cada valor propio encontrado en:

$(A - \lambda I) \mathbf{v} = 0$

**Ejemplo Pr√°ctico**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix}
4 & 1 \\
1 & 4
\end{pmatrix}$

**Paso 1: Determinar los Valores Propios**

Primero, calculamos el polinomio caracter√≠stico y resolvemos la ecuaci√≥n caracter√≠stica:

$\det(A - \lambda I) = \det\left(\begin{pmatrix}
4 - \lambda & 1 \\
1 & 4 - \lambda
\end{pmatrix}\right) = (4-\lambda)^2 - 1^2 = \lambda^2 - 8\lambda + 15$

Resolviendo $(\lambda^2 - 8\lambda + 15 = 0)$:

$\lambda_{1,2} = \frac{8 \pm \sqrt{64 - 60}}{2} = \frac{8 \pm 2}{2} = 5, 3$

**Paso 2: Encontrar los Vectores Propios**

Para $(\lambda = 5)$:

$(A - 5I) = \begin{pmatrix}
-1 & 1 \\
1 & -1
\end{pmatrix}, \quad (A - 5I) \mathbf{v} = 0 \rightarrow \mathbf{v} = \begin{pmatrix}1 \\ 1\end{pmatrix}$

Para $(\lambda = 3)$:

$(A - 3I) = \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix}, \quad (A - 3I) \mathbf{v} = 0 \rightarrow \mathbf{v} = \begin{pmatrix}-1 \\ 1\end{pmatrix}$

**C√≥digo en Python para C√°lculo y Verificaci√≥n**

```python
import numpy as np

A = np.array([[4, 1], [1, 4]])

# Calculando valores y vectores propios
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Valores propios:", eigenvalues)
print("Vectores propios:\n", eigenvectors)
```

Estos conceptos no solo ayudan en la soluci√≥n de problemas matem√°ticos, sino que tambi√©n tienen aplicaciones pr√°cticas en din√°mica de sistemas, an√°lisis de estabilidad, procesamiento de se√±ales, y mucho m√°s.

### **5.1.11 Visualizaci√≥n de matrices y Eigen**

La **visualizaci√≥n de matrices y sus eigenvectores** (vectores propios) y **eigenvalores** (valores propios) es una herramienta fundamental en an√°lisis num√©rico, ciencia de datos, f√≠sica y otros campos que involucran √°lgebra lineal. La capacidad de visualizar estos componentes puede proporcionar una comprensi√≥n intuitiva de c√≥mo una matriz transforma el espacio en el que opera.

**Visualizaci√≥n de Matrices**

Visualizar una matriz en s√≠ misma puede referirse a varias cosas, dependiendo del contexto y del objetivo del an√°lisis. Puede incluir la representaci√≥n de la matriz como una imagen de p√≠xeles, donde cada color o tonalidad de gris representa un valor num√©rico, o la visualizaci√≥n de las transformaciones que la matriz aplica a conjuntos de vectores en el espacio.

**Visualizaci√≥n de Eigenvectores y Eigenvalores**

Los **eigenvectores** y **eigenvalores** revelan informaci√≥n sobre la "direcci√≥n" en la que una matriz act√∫a m√°s fuertemente o m√°s d√©bilmente cuando se aplica a vectores en el espacio. Los eigenvectores indican direcciones en las cuales la acci√≥n de la matriz es "pura" (escala los vectores sin rotarlos), y los eigenvalores indican cu√°nto se estira o comprime el espacio en esas direcciones.

**Ejemplo Pr√°ctico con Visualizaci√≥n**

Consideremos una matriz $(A)$ en $(\mathbb{R}^2)$ para simplificar la visualizaci√≥n:

$A = \begin{pmatrix}
3 & 2 \\
2 & 3
\end{pmatrix}$

Esta matriz es sim√©trica y se espera que tenga eigenvectores reales y ortogonales.

**C√°lculo de Eigenvectores y Eigenvalores**

Primero calculamos los eigenvalores y eigenvectores de $(A)$ usando Python:

```python
import numpy as np

# Definir la matriz A
A = np.array([[3, 2], [2, 3]])

# Calcular eigenvalores y eigenvectores
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalores:", eigenvalues)
print("Eigenvectores:\n", eigenvectors)
```

**Visualizaci√≥n**

Para visualizar c√≥mo $(A)$ transforma el espacio, graficaremos los efectos de $(A)$ en una rejilla de vectores unitarios, junto con sus eigenvectores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para graficar vectores
def plot_vector(v, origin, color='blue', label=None):
    plt.quiver(*origin, *v, color=color, scale=1, scale_units='xy', angles='xy', label=label, width=0.012, headwidth=4, headlength=6)

# Crear un grid de vectores unitarios
x, y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
u = np.vstack((x.flatten(), y.flatten()))

# Transformar cada vector por la matriz A
v = A @ u

# Graficar
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Graficar vectores transformados
for i in range(u.shape[1]):
    plot_vector(v[:,i], origin=[0,0], color='grey', label='Transformados' if i == 0 else "")  # Usar etiqueta solo una vez para evitar repetici√≥n en leyenda

# Graficar vectores propios
colors = ['red', 'blue']  # Colores para diferenciar los vectores propios
for i in range(len(eigenvalues)):
    plot_vector(eigenvalues[i] * eigenvectors[:, i], origin=[0,0], color=colors[i], label=f'Œª={eigenvalues[i]:.2f}')

# Configuraci√≥n adicional para mejorar la visualizaci√≥n
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.legend(loc='upper right', frameon=True, framealpha=0.9, shadow=True)
plt.title('Transformaci√≥n y Eigenvectores de A', fontsize=14, fontweight='bold')

# Mostrar el gr√°fico
plt.show()
```

**Explicaci√≥n de la Visualizaci√≥n**

- Los **vectores grises** en el gr√°fico representan c√≥mo la matriz $(A)$ transforma los vectores unitarios en el plano.

- Los **vectores coloreados** representan los eigenvectores de $(A)$, escalados por sus respectivos eigenvalores. Estos vectores muestran las direcciones en las cuales $(A)$ act√∫a sin rotaci√≥n, solo escalamiento.
