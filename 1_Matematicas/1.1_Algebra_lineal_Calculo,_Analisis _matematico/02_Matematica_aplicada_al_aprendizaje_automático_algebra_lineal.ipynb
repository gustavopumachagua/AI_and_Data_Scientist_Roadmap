### **2.1.1 Introducci√≥n a los Vectores**

Un vector es un elemento matem√°tico fundamental que posee tanto magnitud como direcci√≥n. Es usado ampliamente en f√≠sica, ingenier√≠a, matem√°ticas y ciencias de la computaci√≥n, entre otros campos, para representar fuerzas, velocidades, desplazamientos y otras magnitudes f√≠sicas o conceptuales que necesitan direcci√≥n adem√°s de tama√±o.

**Concepto B√°sico de Vectores**

**Definici√≥n**

En matem√°ticas, un **vector** se define como un elemento de un espacio vectorial. Este espacio es esencialmente una colecci√≥n de vectores que pueden ser sumados entre s√≠ y multiplicados ("escalados") por n√∫meros, conocidos como escalares en este contexto.

**Representaci√≥n**

Un vector en un plano bidimensional se representa com√∫nmente como un punto en un sistema de coordenadas, especificado por coordenadas $(x)$ e $(y)$. Por ejemplo, el vector $(\mathbf{v})$ en 2D se representa como $\mathbf{v} = (x, y)$.

**Tipos de Vectores**

1. **Vectores en 2D**: Representados por dos componentes.

2. **Vectores en 3D**: Incluyen una tercera componente, permitiendo representar profundidad adicional.

3. **Vectores de n-dimensiones**: En espacios de dimensiones superiores, utilizados en matem√°ticas avanzadas y ciencia de datos.

**Operaciones con Vectores**

Los vectores pueden ser manipulados mediante varias operaciones b√°sicas:

**Suma de vectores**: La suma de dos vectores $(\mathbf{u})$ y $(\mathbf{v})$, $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$, es otro vector $ \mathbf{w} = (u_1 + v_1, u_2 + v_2)$.

2. **Producto por un escalar**: Multiplicar un vector $\mathbf{v} = (v_1, v_2)$ por un escalar $(k)$ resulta en $\mathbf{w} = (k \cdot v_1, k \cdot v_2)$.

3. **Producto punto**: El producto de dos vectores que resulta en un escalar, $\mathbf{u} \cdot \mathbf{v} = u_1 \cdot v_1 + u_2 \cdot v_2$.

4. **Producto cruz**: Exclusivo de vectores en 3D, resulta en un vector perpendicular a ambos vectores originales.

**Ejemplo en Python**

Vamos a definir algunos vectores en Python, realizar operaciones b√°sicas y visualizarlas.

**C√≥digo para Definir y Operar con Vectores**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de vectores en 2D
v = np.array([2, 3])
u = np.array([5, 1])

# Suma de vectores
w = u + v

# Producto por escalar
k = 2
p = k * v

# Producto punto
dot_product = np.dot(u, v)

# Crear la figura y configurar los ejes
fig, ax = plt.subplots(figsize=(10, 6))  # Ajustar el tama√±o de la figura para mejor visualizaci√≥n

# Graficar vectores
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.015, label='v (2,3)')
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.015, label='u (5,1)')
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.015, label='w = u + v')
ax.quiver(0, 0, p[0], p[1], angles='xy', scale_units='xy', scale=1, color='yellow', width=0.015, label='p = 2 * v')

# Configurar l√≠mites del gr√°fico
ax.set_xlim(-1, 11)
ax.set_ylim(-1, 7)

# Mejorar la est√©tica del gr√°fico
ax.grid(True, linestyle='--', alpha=0.7)  # Personalizar la grilla
ax.axhline(0, color='black', linewidth=1)  # Hacer la l√≠nea del eje x m√°s visible
ax.axvline(0, color='black', linewidth=1)  # Hacer la l√≠nea del eje y m√°s visible

# A√±adir t√≠tulo y leyenda
ax.set_title('Operaciones con Vectores', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)  # Ubicar la leyenda

# Mostrar el gr√°fico
plt.show()

# Imprimir el producto punto
print("Producto punto de u y v:", dot_product)
```

### **2.1.2 M√≥dulo y producto interior**

El m√≥dulo (o magnitud) de un vector y el producto interior (tambi√©n conocido como producto punto o escalar) son conceptos fundamentales en el √°lgebra lineal y tienen aplicaciones pr√°cticas en f√≠sica, ingenier√≠a, y ciencia de datos. Estos conceptos ayudan a entender las propiedades de los vectores en relaci√≥n con el espacio en el que existen y en relaci√≥n entre s√≠.

**M√≥dulo de un Vector**

El **m√≥dulo de un vector** se refiere a su "longitud" o "tama√±o" en el espacio n-dimensional. Se calcula tomando la ra√≠z cuadrada de la suma de los cuadrados de sus componentes.

**F√≥rmula Matem√°tica**

Para un vector $(\mathbf{v})$ en un espacio 2D con componentes $(v_1, v_2)$, el m√≥dulo se calcula como:

$|\mathbf{v}| = \sqrt{v_1^2 + v_2^2}$

Para un vector en 3D $\mathbf{v} = (v_1, v_2, v_3)$:

$|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + v_3^2}$

**Ejemplo:**

Aqu√≠, calculamos la magnitud de un vector en Python utilizando NumPy.

```python
import numpy as np

# Definir un vector en 2D
v = np.array([3, 4])

# Calcular el m√≥dulo del vector
modulo_v = np.linalg.norm(v)

print("M√≥dulo del vector v:", modulo_v)
```

**Producto Interior (Producto Punto)**

El **producto interior** de dos vectores es una medida de su interacci√≥n lineal, proporcionando una forma de calcular √°ngulos y proyecciones entre vectores. Tambi√©n es fundamental para definiciones de ortogonalidad y para operaciones como la descomposici√≥n de vectores.

**F√≥rmula Matem√°tica**

El producto punto de dos vectores $(\mathbf{u})$ y $(\mathbf{v})$, ambos en 2D, se define como:

$\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2$

Extendido a 3D:

$\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + u_3v_3$

Esta operaci√≥n resulta en un escalar y se puede interpretar como la proyecci√≥n de un vector sobre otro.

**Ejemplo:**

Calculamos el producto punto de dos vectores en Python.

```python
u = np.array([1, 2, 3])
v = np.array([4, -5, 6])

# Producto punto de u y v
producto_punto = np.dot(u, v)

print("Producto punto de u y v:", producto_punto)
```

**Visualizaci√≥n**

### **2.1.3 Coseno y producto punto**

El coseno y el producto punto (o producto escalar) de vectores son conceptos estrechamente relacionados que juegan un papel crucial en el √°lgebra lineal y tienen aplicaciones importantes en √°reas como la f√≠sica, ingenier√≠a, y an√°lisis de datos. Estos conceptos nos permiten entender la relaci√≥n angular entre vectores y realizar c√°lculos importantes para la proyecci√≥n de vectores y determinaci√≥n de ortogonalidad.

**Producto Punto (Producto Escalar)**

El producto punto de dos vectores es una operaci√≥n que resulta en un escalar (un n√∫mero), el cual es igual a la suma de los productos de las correspondientes componentes de los vectores.

**F√≥rmula Matem√°tica**

Para dos vectores $(\mathbf{u})$ y $(\mathbf{v})$ en un espacio bidimensional con componentes $( u_1, u_2)$ y $( v_1, v_2)$ respectivamente, el producto punto se define como:

$\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2$

Extendiendo esto a tres dimensiones:

$\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + u_3v_3$

**Propiedades del Producto Punto**

- **Magnitud y √Ångulo**: El producto punto est√° relacionado con el coseno del √°ngulo entre los dos vectores, permitiendo calcular dicho √°ngulo.

- **Ortogonalidad**: Dos vectores son ortogonales si su producto punto es 0, ya que el coseno de 90¬∞ es 0.

**Relaci√≥n entre el Producto Punto y el Coseno**

El producto punto tambi√©n se puede expresar en t√©rminos del √°ngulo entre los dos vectores utilizando la magnitud (norma) de cada vector.

**F√≥rmula Extendida del Producto Punto**

Para dos vectores $(\mathbf{u})$ y $(\mathbf{v})$, el producto punto tambi√©n puede ser expresado como:

$\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \|\mathbf{v}\| \cos(\theta)$

donde $(\|\mathbf{u}\|)$ y $(\|\mathbf{v}\|)$ son las magnitudes de $(\mathbf{u})$ y $(\mathbf{v})$ respectivamente, y $(\theta)$ es el √°ngulo entre ellos.

**Ejemplo:**

Calcularemos el producto punto de dos vectores y usaremos la f√≥rmula extendida para determinar el √°ngulo entre ellos.

```python
import numpy as np

# Definir dos vectores
u = np.array([3, 4])
v = np.array([4, 6])

# Producto punto
dot_product = np.dot(u, v)

# Magnitudes de u y v
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

# Calcular el √°ngulo entre u y v
cos_theta = dot_product / (norm_u * norm_v)
angle = np.arccos(cos_theta) * (180 / np.pi)  # Convertir radianes a grados

print("Producto punto:", dot_product)
print("√Ångulo entre u y v (grados):", angle)
```

**Visualizaci√≥n en Python**

Para visualizar los vectores y su √°ngulo, podemos usar un gr√°fico simple.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Funci√≥n para graficar vectores
def plot_vector(vec, origin, color, label=None):
    plt.quiver(*origin, *vec, scale=1, scale_units='xy', angles='xy', color=color, width=0.02, label=label)

# Definir vectores
u = np.array([3, 4])
v = np.array([4, 6])

# Graficar vectores
plt.figure(figsize=(8, 8))  # Ajustando el tama√±o de la figura
plot_vector(u, [0, 0], 'red', 'u (3,4)')
plot_vector(v, [0, 0], 'blue', 'v (4,6)')

# Calcular el √°ngulo entre los vectores
angle_rad = np.arccos(np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v)))
angle_deg = np.degrees(angle_rad)

# A√±adir un arco para representar el √°ngulo
arc_radius = 0.5 * min(np.linalg.norm(u), np.linalg.norm(v))
arc = patches.Arc((0, 0), 2*arc_radius, 2*arc_radius, angle=0, theta1=0, theta2=angle_deg, color='green', label=f'√Ångulo = {angle_deg:.2f}¬∞')
plt.gca().add_patch(arc)

# Configurar l√≠mites y aspecto del gr√°fico
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # L√≠nea horizontal
plt.axvline(0, color='black', linewidth=0.5)  # L√≠nea vertical
plt.legend(loc='upper left')

# A√±adir t√≠tulo
plt.title('Visualizaci√≥n de Vectores y √Ångulo', fontsize=14, fontweight='bold')

# Mostrar el gr√°fico
plt.show()
```

**Conclusi√≥n**

El producto punto y el coseno del √°ngulo entre vectores son herramientas esenciales para entender la relaci√≥n entre vectores en t√©rminos de su orientaci√≥n y magnitud relativa. Estos conceptos son utilizados en una amplia gama de aplicaciones, desde la resoluci√≥n de problemas en f√≠sica hasta algoritmos en ciencia de datos y machine learning. ### **2.1.4 Proyecci√≥n**

La proyecci√≥n de vectores es un concepto fundamental en √°lgebra lineal que permite descomponer un vector en componentes paralelas y perpendiculares respecto a otro vector o superficie. Es esencial en muchas aplicaciones, como en la resoluci√≥n de sistemas de ecuaciones, en la optimizaci√≥n, y en an√°lisis de datos, especialmente en m√©todos como el An√°lisis de Componentes Principales (PCA) y en gr√°ficos computacionales.

**Concepto de Proyecci√≥n Vectorial**

La **proyecci√≥n de un vector $(\mathbf{u})$ sobre otro vector $(\mathbf{v})$** se refiere a encontrar el componente de $(\mathbf{u})$ que es paralelo a $(\mathbf{v})$.

**F√≥rmula Matem√°tica**

La proyecci√≥n del vector $(\mathbf{u})$ sobre $(\mathbf{v})$ se calcula con la siguiente f√≥rmula:

$\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{\mathbf{u} \cdot \mathbf{v}}{\mathbf{v} \cdot \mathbf{v}} \mathbf{v}$

donde $(\mathbf{u} \cdot \mathbf{v})$ es el producto punto de $(\mathbf{u})$ y $(\mathbf{v})$, y $(\mathbf{v} \cdot \mathbf{v})$ es el cuadrado de la norma de $(\mathbf{v})$.

**Interpretaci√≥n F√≠sica**

Si piensas en $(\mathbf{u})$ como una fuerza aplicada en la direcci√≥n de $(\mathbf{v})$, la proyecci√≥n de $(\mathbf{u})$ sobre $(\mathbf{v})$ te dir√° cu√°nta fuerza est√° actuando en la direcci√≥n de $(\mathbf{v})$.

**Ejemplo Pr√°ctico en Python**

Supongamos que queremos proyectar un vector $\mathbf{u} = (1, 3)$ sobre otro vector $\mathbf{v} = (2, 2)$.

**C√≥digo para Calcular la Proyecci√≥n**

```python
import numpy as np

# Definir los vectores
u = np.array([1, 3])
v = np.array([2, 2])

# Calcular la proyecci√≥n de u sobre v
proj_v_u = (np.dot(u, v) / np.dot(v, v)) * v

print("Proyecci√≥n de u sobre v:", proj_v_u)
```

**Visualizaci√≥n de la Proyecci√≥n de Vectores**

Podemos visualizar esto para entender mejor c√≥mo $(\mathbf{u})$ se descompone en componentes paralelas y perpendiculares a $(\mathbf{v})$.

**C√≥digo para Visualizaci√≥n en Python**

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para graficar vectores
def plot_vector(vec, origin=[0, 0], color='k', label=None):
    plt.quiver(*origin, *vec, scale=1, scale_units='xy', angles='xy', color=color, label=label, width=0.015)

# Definir vectores
u = np.array([1, 3])
v = np.array([2, 2])

# Funci√≥n para calcular la proyecci√≥n de u sobre v
def projection(u, v):
    v_norm = np.sqrt(sum(v**2))
    proj_of_u_on_v = (np.dot(u, v) / np.dot(v, v)) * v
    return proj_of_u_on_v

# Calcular la proyecci√≥n de u sobre v
proj_v_u = projection(u, v)

# Graficar los vectores originales
plot_vector(u, color='red', label='u (1,3)')
plot_vector(v, color='blue', label='v (2,2)')

# Graficar la proyecci√≥n de u sobre v
plot_vector(proj_v_u, color='green', label='Proyecci√≥n de u sobre v')

# Configurar l√≠mites y aspecto del gr√°fico
plt.xlim(0, 3)
plt.ylim(0, 4)
plt.grid(True, linestyle='--', alpha=0.7)  # Personalizar la grilla
plt.axhline(0, color='black', linewidth=1)  # L√≠nea horizontal m√°s visible
plt.axvline(0, color='black', linewidth=1)  # L√≠nea vertical m√°s visible
plt.legend(loc='upper left', fontsize=10)  # Personalizar ubicaci√≥n y tama√±o de la leyenda

# A√±adir t√≠tulo
plt.title('Visualizaci√≥n de Vectores y Proyecci√≥n', fontsize=14, fontweight='bold')

# Mostrar el gr√°fico
plt.show()
```

### **2.1.5 Cambio de base**

El cambio de base de vectores es un concepto fundamental en √°lgebra lineal que implica representar un vector en diferentes sistemas de coordenadas. Cada sistema de coordenadas se define por una base, que es un conjunto de vectores que son linealmente independientes y que pueden abarcar el espacio vectorial en cuesti√≥n. Cambiar la base de un vector permite reexpresar ese vector en t√©rminos de diferentes vectores base, lo cual es √∫til en muchas aplicaciones matem√°ticas y de ingenier√≠a, como la rotaci√≥n de coordenadas, la simplificaci√≥n de problemas en f√≠sica, y la transformaci√≥n de im√°genes en computaci√≥n gr√°fica.

**Concepto de Cambio de Base**

Una **base** de un espacio vectorial es un conjunto de vectores que son linealmente independientes y que generan todo el espacio. Cambiar la base de un vector significa reescribir ese vector en t√©rminos de otra base.

**F√≥rmulas Matem√°ticas**

Supongamos que tenemos un vector $(\mathbf{v})$ expresado en una base est√°ndar $\mathcal{E} = \{\mathbf{e}_1, \mathbf{e}_2\}$ como

$\mathbf{v} = v_1 \mathbf{e}_1 + v_2 \mathbf{e}_2$

Si queremos cambiar este vector a una nueva base $\mathcal{B} = \{\mathbf{b}_1, \mathbf{b}_2\}$, necesitamos encontrar los coeficientes $(c_1)$ y $(c_2)$ tales que

$\mathbf{v} = c_1 \mathbf{b}_1 + c_2 \mathbf{b}_2$

La relaci√≥n entre las coordenadas en la base original y la nueva base se da a trav√©s de la matriz de cambio de base $(P)$ que transforma coordenadas de la base $(\mathcal{B})$ a la base $(\mathcal{E})$:

$[ \mathbf{v} ]_{\mathcal{E}} = P [ \mathbf{v} ]_{\mathcal{B}}$

donde $(P)$ es la matriz cuyas columnas son las coordenadas de los vectores de la base $(\mathcal{B})$ expresados en la base $(\mathcal{E})$.

**Ejemplo Pr√°ctico en Python**

Supongamos que queremos cambiar la base del vector $\mathbf{v} = \begin{pmatrix} 3 \\ 2 \end{pmatrix}$ de la base est√°ndar $(\mathcal{E})$ a una nueva base $\mathcal{B} = \{\begin{pmatrix} 1 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 \\ 1 \end{pmatrix}\}$.

**C√≥digo para Cambio de Base**

```python
import numpy as np

# Definir el vector en la base est√°ndar
v = np.array([3, 2])

# Definir la nueva base
B = np.array([[1, 1],
              [0, 1]])

# Calcular la matriz de cambio de base de B a la base est√°ndar (inversa de B)
P_inv = np.linalg.inv(B)

# Cambiar el vector v a la nueva base
v_B = np.dot(P_inv, v)

print("Coordenadas de v en la nueva base:", v_B)
```

**Visualizaci√≥n del Cambio de Base**

Podemos visualizar el vector en ambas bases para entender mejor c√≥mo cambia.

**C√≥digo para Visualizaci√≥n en Python**

### **2.1.6 Base, espacio vectorial e independencia lineal**

Vamos a profundizar en estos conceptos con definiciones claras, ejemplos y visualizaciones en Python.

**Espacio Vectorial**

Un **espacio vectorial** es una colecci√≥n de vectores, que se pueden sumar entre s√≠ y multiplicar por escalares (n√∫meros), cumpliendo diez axiomas espec√≠ficos que incluyen la conmutatividad, asociatividad de la suma y la distributividad de la multiplicaci√≥n escalar sobre la suma de vectores. Los espacios vectoriales son el marco en el que se desarrolla todo el √°lgebra lineal.

**Ejemplo de Espacio Vectorial**

El conjunto de todos los vectores bidimensionales $(\mathbb{R}^2)$ es un ejemplo de espacio vectorial bajo las operaciones est√°ndar de suma de vectores y multiplicaci√≥n por escalar.

**Independencia Lineal**

Un conjunto de vectores es **independiente linealmente** si ninguno de los vectores en el conjunto puede ser escrito como una combinaci√≥n lineal de los otros. Esto implica que el √∫nico conjunto de escalares que satisface la ecuaci√≥n $c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_n\mathbf{v}_n = 0$ es $c_1 = c_2 = \dots = c_n = 0$.

**Ejemplo de Independencia Lineal**

Considera los vectores $\mathbf{v}_1 = (1, 0)$ y $\mathbf{v}_2 = (0, 1)$ en $(\mathbb{R}^2)$. Estos vectores son independientes linealmente porque no hay manera de escribir uno como una combinaci√≥n lineal del otro.

**Base de un Espacio Vectorial**

Una **base** de un espacio vectorial es un conjunto de vectores que son linealmente independientes y que generan todo el espacio vectorial (cualquier vector en el espacio puede ser expresado como una combinaci√≥n lineal de vectores de la base).

**Ejemplo de Base**

Los vectores $\mathbf{v}_1 = (1, 0)$ y $\mathbf{v}_2 = (0, 1)$ forman una base para $(\mathbb{R}^2)$ porque cualquier vector en $(\mathbb{R}^2)$ puede ser expresado como una combinaci√≥n lineal de $(\mathbf{v}_1)$ y $(\mathbf{v}_2)$.

**Ejemplo Pr√°ctico en Python: Verificando Independencia Lineal y Calculando una Base**

A continuaci√≥n, presentamos un c√≥digo en Python para verificar si un conjunto de vectores es linealmente independiente y para calcular la base de un espacio generado por esos vectores usando descomposici√≥n SVD (Singular Value Decomposition), que es √∫til para entender la estructura de un espacio vectorial.

**C√≥digo Python**

```python
import numpy as np

# Definir vectores
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])

# Crear una matriz con v1, v2 y v3 como columnas
A = np.column_stack((v1, v2, v3))

# Usar SVD para analizar la matriz
U, S, Vt = np.linalg.svd(A)

# Verificar la independencia lineal
print("Valores singulares:", S)
independientes = np.all(S > 1e-10)  # Asumiendo un peque√±o umbral para la precisi√≥n num√©rica
print("¬øSon independientes linealmente?", "S√≠" if independientes else "No")

# Determinar la base
print("Base del espacio generado por v1, v2, v3:", U[:, :2])  # Asumiendo que v1 y v2 son independientes
```

### **2.1.7 Aplicaciones del cambio de base**

El cambio de base es una operaci√≥n fundamental en √°lgebra lineal que tiene m√∫ltiples aplicaciones en ciencias, ingenier√≠a, inform√°tica, y otras disciplinas t√©cnicas. Cambiar la base de un conjunto de vectores implica representar esos vectores en un nuevo sistema de coordenadas, lo cual puede simplificar problemas, optimizar c√°lculos y proporcionar nuevas perspectivas sobre los datos. **Aplicaciones del Cambio de Base de Vectores**

**Simplificaci√≥n de Ecuaciones y Sistemas**

El cambio de base se usa para simplificar sistemas de ecuaciones lineales transformando la matriz de coeficientes en una forma m√°s manejable, como la forma diagonal o la forma de Jordan. Esto facilita encontrar soluciones o realizar otras operaciones matem√°ticas.

**Diagonalizaci√≥n de Matrices**

Una aplicaci√≥n importante del cambio de base es la diagonalizaci√≥n de matrices. Una matriz es diagonalizable si existe una base compuesta por sus vectores propios. En estad√≠stica y aprendizaje autom√°tico, PCA es una t√©cnica que utiliza el cambio de base para transformar un conjunto de datos posiblemente correlacionados en un conjunto de valores de variables linealmente no correlacionadas llamadas componentes principales. Este cambio de base se realiza hacia la base de vectores propios de la matriz de covarianza del conjunto de datos, lo que ayuda en la reducci√≥n de la dimensionalidad y la visualizaci√≥n de datos.

**Ejemplo Pr√°ctico: Diagonalizaci√≥n de una Matriz**

Supongamos que queremos diagonalizar una matriz, que es un caso espec√≠fico de cambio de base. Vamos a ver c√≥mo hacerlo en Python.

**Matriz a Diagonalizar**

Consideremos la matriz $(A)$:

$A = \begin{pmatrix} 4 & 1 \\ 2 & 3 \end{pmatrix}$

**C√≥digo en Python para Diagonalizaci√≥n**

```python
import numpy as np

# Definir la matriz A
A = np.array([[4, 1], [2, 3]])

# Calcular los vectores propios y valores propios
eigenvalues, eigenvectors = np.linalg.eig(A)

# Matriz P de vectores propios y su inversa
P = eigenvectors
P_inv = np.linalg.inv(P)

# Diagonalizar A
D = np.dot(np.dot(P_inv, A), P)

print("Matriz original A:\n", A)
print("Matriz diagonal D:\n", D)
print("Matriz de vectores propios P:\n", P)
```

### **2.1.9 Producto punto de vectores**

El producto punto, tambi√©n conocido como producto escalar, es una operaci√≥n fundamental en √°lgebra lineal que toma dos vectores y devuelve un escalar. Esta operaci√≥n es esencial en muchos campos, incluyendo f√≠sica, ingenier√≠a, y ciencias de la computaci√≥n, por sus aplicaciones en el c√°lculo de √°ngulos entre vectores, proyecciones, y en la determinaci√≥n de la ortogonalidad.

**Concepto de Producto Punto**

El producto punto de dos vectores se define como la suma de los productos de sus correspondientes componentes. Este resultado es un n√∫mero (escalar) y no un vector.

**F√≥rmula Matem√°tica**

Para dos vectores $\mathbf{u} = (u_1, u_2, \dots, u_n)$ y $\mathbf{v} = (v_1, v_2, \dots, v_n)$ en un espacio $(\mathbb{R}^n)$, el producto punto se calcula como:

$\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + \cdots + u_nv_n = \sum_{i=1}^n u_i v_i$

Adem√°s, el producto punto se relaciona con el √°ngulo $(\theta)$ entre $(\mathbf{u})$ y $(\mathbf{v})$ mediante la ley del coseno:

$\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \|\mathbf{v}\| \cos(\theta)$

donde $(\|\mathbf{u}\|)$ y $(\|\mathbf{v}\|)$ son las magnitudes (o normas) de $(\mathbf{u})$ y $(\mathbf{v})$, respectivamente.

**Ejemplo Pr√°ctico**

Supongamos que queremos calcular el producto punto de los vectores $\mathbf{u} = (1, 2, 3)$ y $\mathbf{v} = (4, -5, 6)$ y determinar el √°ngulo entre ellos.

**C√≥digo en Python para Producto Punto y √Ångulo**

```python
import numpy as np

# Definir vectores
u = np.array([1, 2, 3])
v = np.array([4, -5, 6])

# Calcular el producto punto
producto_punto = np.dot(u, v)

# Calcular magnitudes de u y v
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

# Calcular el √°ngulo entre u y v
cos_theta = producto_punto / (norm_u * norm_v)
angle = np.arccos(cos_theta) * (180 / np.pi)  # Convertir de radianes a grados

print("Producto punto:", producto_punto)
print("√Ångulo entre u y v (grados):", angle)
```

**Visualizaci√≥n del Producto Punto**

Visualizar el concepto del producto punto y el √°ngulo entre dos vectores puede ser muy √∫til para entender mejor su relaci√≥n geom√©trica.

**C√≥digo para Visualizaci√≥n en Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector(vec, origin, color, label):
    plt.quiver(*origin, vec[0], vec[1], color=color, scale=1, scale_units='xy', angles='xy', label=label)

# Definir vectores 3D
u = np.array([1, 2, 3])
v = np.array([4, -5, 6])

# Calcular el producto punto de u y v
dot_product = np.dot(u, v)

# Graficar proyecciones 2D de los vectores
plt.figure(figsize=(8, 8))
plot_vector(u[:2], [0, 0], 'red', f'u (1,2) proyecci√≥n de (1,2,3)')
plot_vector(v[:2], [0, 0], 'blue', f'v (4,-5) proyecci√≥n de (4,-5,6)')

# Configuraci√≥n del gr√°fico
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.title('Visualizaci√≥n de Vectores y Producto Punto', fontsize=14, fontweight='bold')

# Mostrar el producto punto en el gr√°fico
plt.figtext(0.5, 0.01, f'Producto Punto: {dot_product}', ha='center', fontsize=12, bbox={"facecolor":"white", "alpha":0.5, "pad":5})

# Mostrar el gr√°fico
plt.show()
```

Adem√°s, la propiedad de que dos vectores son ortogonales si su producto punto es cero es fundamental en la definici√≥n de bases ortogonales, que son muy √∫tiles en transformaciones lineales y m√©todos num√©ricos.

### **2.1.10 Cambio de base**

El cambio de base en √°lgebra lineal es un concepto crucial que se refiere a la transformaci√≥n de vectores de una base a otra dentro del mismo espacio vectorial. Este proceso es fundamental para simplificar problemas matem√°ticos, optimizar c√°lculos, y adaptar datos a contextos m√°s apropiados para an√°lisis espec√≠ficos.

**Concepto de Cambio de Base**

Una **base** de un espacio vectorial es un conjunto de vectores que son linealmente independientes y que abarcan todo el espacio. Cambiar la base de un vector implica reexpresar ese vector en t√©rminos de un nuevo conjunto de vectores base.

**Definici√≥n Matem√°tica**

Supongamos que tenemos un vector $(\mathbf{v})$ en un espacio vectorial, y este vector est√° expresado en t√©rminos de una base $\mathcal{B} = \{\mathbf{b}_1, \mathbf{b}_2, \ldots, \mathbf{b}_n\}$. Si deseamos expresar $(\mathbf{v})$ en t√©rminos de otra base $\mathcal{C} = \{\mathbf{c}_1, \mathbf{c}_2, \ldots, \mathbf{c}_n\}$, necesitamos encontrar los coeficientes $(c_1, c_2, \ldots, c_n)$ tales que:

$\mathbf{v} = c_1 \mathbf{c}_1 + c_2 \mathbf{c}_2 + \ldots + c_n \mathbf{c}_n$

La **matriz de cambio de base** $(P)$ de $(\mathcal{B})$ a $(\mathcal{C})$ transforma las coordenadas de $(\mathbf{v})$ en $(\mathcal{B})$ a sus coordenadas en $(\mathcal{C})$:

$[\mathbf{v} ]_{\mathcal{C}} = P [ \mathbf{v} ]_{\mathcal{B}}$

donde $(P)$ es la matriz cuyas columnas son las coordenadas de los vectores de $(\mathcal{C})$ expresados en la base $(\mathcal{B})$.

**Ejemplo Pr√°ctico: Cambio de Base**

Supongamos que queremos cambiar la base de un vector $\mathbf{v} = (3, 2)$ desde la base est√°ndar $(\mathcal{E})$ a una nueva base $\mathcal{B} = \{(1, 0), (0, 1)\}$ y luego a $\mathcal{C} = \{(2, 1), (1, 2)\}$.

**C√°lculo de la Matriz de Cambio de Base**

Primero, vamos a calcular la matriz de cambio de base $(P)$ de $(\mathcal{E})$ a $(\mathcal{C})$:

```python
import numpy as np

# Definir la nueva base C
C = np.array([[2, 1], [1, 2]])

# Calcular la matriz de cambio de base P (desde la base est√°ndar E a C)
P = np.linalg.inv(C)  # P es la inversa de C porque C est√° definida respecto a E

# Vector en la base est√°ndar
v_E = np.array([3, 2])

# Cambiar el vector v a la nueva base C
v_C = np.dot(P, v_E)

print("Coordenadas de v en la nueva base C:", v_C)
```

**Visualizaci√≥n del Cambio de Base**

Podemos visualizar c√≥mo el vector $(\mathbf{v})$ se representa en diferentes bases para entender mejor el efecto del cambio de base.

**C√≥digo para Visualizaci√≥n en Python**

```python
import matplotlib.pyplot as plt

def plot_vector(vec, origin, color, label):
    # Dibuja un vector desde el punto "origin" con una determinada configuraci√≥n de color y etiqueta
    plt.quiver(*origin, *vec, color=color, scale=1, scale_units='xy', angles='xy', width=0.02, headwidth=3, headlength=5, label=label)

# Configurar la figura
plt.figure(figsize=(6, 6))
origin = [0, 0]

# Vector en la base est√°ndar
plot_vector((3, 2), origin, 'red', 'v in E (3,2)')

# Vectores de la nueva base
plot_vector((2, 1), origin, 'green', 'c1 (2,1)')
plot_vector((1, 2), origin, 'blue', 'c2 (1,2)')

# Configuraciones del gr√°fico
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(True, linestyle='--', alpha=0.7)  # Grilla personalizada para mejor visualizaci√≥n
plt.axhline(0, color='black', linewidth=1)  # Hacer las l√≠neas de los ejes m√°s visibles
plt.axvline(0, color='black', linewidth=1)
plt.legend(loc='upper right', fontsize=10, frameon=True, framealpha=0.9, shadow=True)  # Leyenda con detalles mejorados
plt.title('Cambio de Base de Vectores', fontsize=14, fontweight='bold')  # T√≠tulo m√°s prominente y claro

# Mostrar el gr√°fico
plt.show()
```

**Conclusi√≥n**

El cambio de base es una herramienta poderosa en √°lgebra lineal que permite adaptar problemas y datos a contextos donde pueden ser manejados de manera m√°s eficiente. Se utiliza ampliamente en transformaciones geom√©tricas, resoluci√≥n de sistemas lineales, y en la simplificaci√≥n de modelos matem√°ticos en ciencias e ingenier√≠a. Entender y aplicar correctamente este concepto es esencial para la manipulaci√≥n efectiva de espacios vectoriales en muchas aplicaciones t√©cnicas.

### **2.1.11 Dependencia lineal de un conjunto de vectores**

La dependencia lineal de un conjunto de vectores es un concepto central en √°lgebra lineal que ayuda a determinar si un conjunto de vectores puede ser expresado en t√©rminos de una combinaci√≥n lineal de otros vectores en el conjunto. Es fundamental para entender estructuras como bases y dimensiones de espacios vectoriales.

**Concepto de Dependencia Lineal**

Un conjunto de vectores se considera **linealmente dependiente** si al menos uno de los vectores en el conjunto puede ser escrito como una combinaci√≥n lineal de los otros. Si ning√∫n vector en el conjunto puede ser expresado de esta manera, entonces los vectores son **linealmente independientes**.

**Definici√≥n Matem√°tica**

Un conjunto de vectores $(\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k)$ en un espacio vectorial es linealmente dependiente si existen escalares $(c_1, c_2, \ldots, c_k)$, no todos cero, tales que:

$c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \cdots + c_k \mathbf{v}_k = 0$

Si la √∫nica soluci√≥n a esta ecuaci√≥n es $(c_1 = c_2 = \cdots = c_k = 0)$, entonces los vectores son linealmente independientes.

**Ejemplo Pr√°ctico: Evaluando la Dependencia Lineal**

Supongamos que queremos determinar si el conjunto de vectores ${(2, 3), (4, 6), (1, -1)}$ es linealmente dependiente.

**Pasos para la Evaluaci√≥n**

1. Formar una matriz con los vectores como columnas.

2. Usar operaciones de fila (reducci√≥n de filas) para determinar si la matriz tiene rango completo.

**C√≥digo en Python**

```python
import numpy as np

# Definir los vectores como columnas de una matriz
A = np.array([[2, 4, 1],
              [3, 6, -1]])

# Reducir la matriz a su forma escalonada de fila reducida
U = np.linalg.matrix_rank(A)

# Determinar si los vectores son linealmente dependientes
num_vectors = A.shape[1]
is_dependent = U < num_vectors

print("Rango de la matriz:", U)
print("¬øEs el conjunto linealmente dependiente?", "S√≠" if is_dependent else "No")
```

**Visualizaci√≥n de la Dependencia Lineal**

Podemos visualizar los vectores para ver c√≥mo se relacionan entre s√≠ en el plano.

**C√≥digo para Visualizaci√≥n en Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector(vec, origin, color, label):
    # Ajusta el ancho del vector, la cabeza y la longitud de la cabeza para una mejor visualizaci√≥n
    plt.quiver(*origin, vec[0], vec[1], scale=1, scale_units='xy', angles='xy', color=color, width=0.02, headwidth=4, headlength=6, label=label)

# Crear la figura
plt.figure(figsize=(6, 6))  # Aumentar el tama√±o para una mejor visualizaci√≥n
origin = [0, 0]

# Definir los vectores
v1 = np.array([2, 3])
v2 = np.array([4, 6])
v3 = np.array([1, -1])

# Graficar los vectores
plot_vector(v1, origin, 'red', 'v1 (2,3)')
plot_vector(v2, origin, 'green', 'v2 (4,6) [2*v1]')  # Aclarar que v2 es el doble de v1
plot_vector(v3, origin, 'blue', 'v3 (1,-1)')

# Configurar el gr√°fico
plt.xlim(-2, 5)
plt.ylim(-2, 7)
plt.grid(True, linestyle='--', alpha=0.7)  # Mejorar la visibilidad de la grilla
plt.axhline(0, color='black', linewidth=1)  # Hacer las l√≠neas de los ejes m√°s visibles
plt.axvline(0, color='black', linewidth=1)
plt.legend(loc='upper right', frameon=True, framealpha=0.9, shadow=True)  # Mejorar la leyenda con un marco y sombra
plt.title('Visualizaci√≥n de Vectores', fontsize=14, fontweight='bold')  # T√≠tulo m√°s prominente

# Mostrar el gr√°fico
plt.show()
```

**Conclusi√≥n**

La dependencia lineal es una propiedad crucial de conjuntos de vectores que nos informa sobre su capacidad para generar espacios vectoriales y formar bases. La evaluaci√≥n de la dependencia lineal es esencial en la simplificaci√≥n de sistemas, optimizaci√≥n de problemas y comprensi√≥n de la estructura geom√©trica subyacente de conjuntos de datos en ciencias e ingenier√≠a. Conocer si un conjunto de vectores es dependiente o independiente ayuda en la toma de decisiones cr√≠ticas en el dise√±o de algoritmos y la implementaci√≥n de soluciones t√©cnicas.
