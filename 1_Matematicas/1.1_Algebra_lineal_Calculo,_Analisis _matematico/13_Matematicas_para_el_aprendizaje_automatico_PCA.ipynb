### **13.1.1 Producto de puntos**

El **producto punto** (tambi√©n conocido como producto escalar o dot product) es una operaci√≥n fundamental en √°lgebra lineal que se realiza entre dos vectores. Esta operaci√≥n resulta en un escalar (un n√∫mero √∫nico) que es una medida de la magnitud de los vectores y del coseno del √°ngulo entre ellos. El producto punto tiene importantes aplicaciones en f√≠sica, ingenier√≠a, ciencias de la computaci√≥n y m√°s.

**Definici√≥n del Producto Punto**

El producto punto de dos vectores $(\mathbf{a})$ y $(\mathbf{b})$, ambos en $(\mathbb{R}^n)$, se define como:

$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i = a_1b_1 + a_2b_2 + \cdots + a_nb_n$

Donde $(a_i)$ y $(b_i)$ son las componentes de los vectores $(\mathbf{a})$ y $(\mathbf{b})$ respectivamente.

**Interpretaci√≥n Geom√©trica**

Geom√©tricamente, el producto punto puede ser interpretado como:

$\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos(\theta)$

donde $(\|\mathbf{a}\|)$ y $(\|\mathbf{b}\|)$ son las magnitudes (o normas) de los vectores $(\mathbf{a})$ y $(\mathbf{b})$, y $(\theta)$ es el √°ngulo entre ellos. Esta interpretaci√≥n es especialmente √∫til porque muestra c√≥mo el producto punto es una medida de la tendencia de dos vectores a apuntar en la misma direcci√≥n.

**Ejemplo con F√≥rmulas**

**Ejemplo**: Consideremos dos vectores en $(\mathbb{R}^3)$:

$\mathbf{a} = [1, 2, 3]$

$\mathbf{b} = [4, -5, 6]$

El producto punto de $(\mathbf{a})$ y $(\mathbf{b})$ es:

$\mathbf{a} \cdot \mathbf{b} = 1\cdot4 + 2\cdot(-5) + 3\cdot6 = 4 - 10 + 18 = 12$

**C√≥digo en Python y Visualizaci√≥n**

Podemos calcular y visualizar el producto punto utilizando Python. Aqu√≠ tambi√©n mostraremos c√≥mo este producto cambia con el √°ngulo entre los dos vectores.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir los vectores
a = np.array([1, 2, 3])
b = np.array([4, -5, 6])

# Calcular el producto punto
dot_product = np.dot(a, b)

# Imprimir el resultado
print("Producto punto de a y b:", dot_product)

# Visualizaci√≥n de los vectores en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# A√±adir los vectores al gr√°fico con estilo mejorado
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='blue', label='Vector a [1, 2, 3]', arrow_length_ratio=0.1, linewidth=2)
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='red', label='Vector b [4, -5, 6]', arrow_length_ratio=0.1, linewidth=2)

# Establecer l√≠mites y etiquetas
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X Axis', fontsize=12)
ax.set_ylabel('Y Axis', fontsize=12)
ax.set_zlabel('Z Axis', fontsize=12)
ax.set_title("3D Vector Representation", fontsize=14, fontweight='bold')
ax.view_init(elev=20., azim=30)  # Ajustar la perspectiva de visualizaci√≥n

# A√±adir cuadr√≠cula, leyenda y mostrar el gr√°fico
ax.grid(True)
ax.legend(fontsize='medium')
plt.show()
```

**Interpretaci√≥n**

- El c√≥digo calcula el producto punto de los vectores $(\mathbf{a})$ y $(\mathbf{b})$, resultando en 12, lo que indica una cierta medida de alineaci√≥n entre ellos, dado que el resultado no es cero.

- La visualizaci√≥n muestra los vectores en el espacio tridimensional, proporcionando una representaci√≥n gr√°fica de su orientaci√≥n relativa.

El producto punto es fundamental no solo en matem√°ticas, sino tambi√©n en muchas aplicaciones pr√°cticas, incluyendo la proyecci√≥n de vectores, la determinaci√≥n de la ortogonalidad, y en el c√°lculo de √°ngulos entre vectores en algoritmos de machine learning y gr√°ficos por computadora.

### **13.1.2 Producto interior: definici√≥n**

El **producto interior** (o producto escalar, producto punto) es una operaci√≥n fundamental en matem√°ticas, especialmente en √°lgebra lineal y en espacios vectoriales. Esta operaci√≥n toma dos vectores y devuelve un escalar, que es esencial para muchas aplicaciones en matem√°ticas y f√≠sica, como calcular √°ngulos entre vectores, longitudes de vectores, y proyecciones de vectores.

**Definici√≥n del Producto Interior**

El producto interior entre dos vectores $(\mathbf{a})$ y $(\mathbf{b})$ en un espacio $(\mathbb{R}^n)$ se define como:

$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i = a_1b_1 + a_2b_2 + \cdots + a_nb_n$

donde $(a_i)$ y $(b_i)$ son las componentes del vector $(\mathbf{a})$ y $(\mathbf{b})$ respectivamente.

**Propiedades del Producto Interior**

El producto interior tiene varias propiedades importantes:

1. **Conmutativo**: $(\mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a})$
2. **Lineal en el primer argumento**:

   - $(c\mathbf{a}) \cdot \mathbf{b} = c(\mathbf{a} \cdot \mathbf{b})$ para cualquier escalar $(c)$

   - $(\mathbf{a} + \mathbf{b}) \cdot \mathbf{c} = \mathbf{a} \cdot \mathbf{c} + \mathbf{b} \cdot \mathbf{c}$

3. **Positivamente definido**: $(\mathbf{a} \cdot \mathbf{a} \geq 0)$ y $(\mathbf{a} \cdot \mathbf{a} = 0)$ si y solo si $(\mathbf{a} = 0)$

**Interpretaci√≥n Geom√©trica**

Geom√©tricamente, el producto punto se puede interpretar como:

$\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos(\theta)$

donde $(\|\mathbf{a}\|)$ y $(\|\mathbf{b}\|)$ son las magnitudes de los vectores $(\mathbf{a})$ y $(\mathbf{b})$, respectivamente, y $(\theta)$ es el √°ngulo entre ellos. Esto muestra que el producto punto es una medida de c√≥mo los dos vectores est√°n orientados el uno respecto al otro.

**Ejemplo con F√≥rmulas**

**Ejemplo**: Calculemos el producto punto de dos vectores en $(\mathbb{R}^3)$:

$\mathbf{a} = [3, -5, 4]$

$\mathbf{b} = [2, 6, 5]$

$\mathbf{a} \cdot \mathbf{b} = 3 \times 2 + (-5) \times 6 + 4 \times 5 = 6 - 30 + 20 = -4$

**C√≥digo en Python y Visualizaci√≥n**

Vamos a calcular este producto en Python y visualizar los vectores en un gr√°fico.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir vectores
a = np.array([3, -5, 4])
b = np.array([2, 6, 5])

# Calcular producto punto
dot_product = np.dot(a, b)

# Imprimir el resultado
print("Producto interior de a y b:", dot_product)

# Visualizaci√≥n en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# A√±adir los vectores al gr√°fico con estilo mejorado
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='blue', label='Vector a [3, -5, 4]', arrow_length_ratio=0.1, linewidth=2)
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='red', label='Vector b [2, 6, 5]', arrow_length_ratio=0.1, linewidth=2)

# Establecer l√≠mites y etiquetas
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X Axis', fontsize=12)
ax.set_ylabel('Y Axis', fontsize=12)
ax.set_zlabel('Z Axis', fontsize=12)
ax.set_title("3D Vector Representation", fontsize=14, fontweight='bold')
ax.view_init(elev=20., azim=30)  # Ajustar la perspectiva de visualizaci√≥n

# A√±adir cuadr√≠cula, leyenda y mostrar el gr√°fico
ax.grid(True)
ax.legend(fontsize='medium')
plt.show()
```

**Interpretaci√≥n**

- El producto punto resultante, $(-4)$, sugiere que los vectores $(\mathbf{a})$ y $(\mathbf{b})$ est√°n orientados de tal manera que forman un √°ngulo obtuso, ya que el coseno de un √°ngulo obtuso es negativo.

- La visualizaci√≥n proporciona una representaci√≥n gr√°fica de la orientaci√≥n de los vectores en el espacio tridimensional, ayudando a visualizar c√≥mo se relacionan geom√©tricamente.

El producto interior es una herramienta poderosa para entender la relaci√≥n entre vectores y tiene numerosas aplicaciones en la teor√≠a y la pr√°ctica. Espero que esta explicaci√≥n te haya ayudado a entender mejor el concepto.

### **13.1.3 Producto interior: longitud de los vectores**

El **producto interior** no solo permite determinar el √°ngulo entre dos vectores y c√≥mo est√°n orientados entre s√≠, sino que tambi√©n se utiliza para calcular la longitud o norma de un vector. Este concepto es clave en la geometr√≠a del espacio vectorial y tiene aplicaciones directas en f√≠sica, ingenier√≠a, inform√°tica y m√°s.

**Longitud de un Vector Usando el Producto Interior**

La longitud, o norma, de un vector se calcula usando el producto interior del vector consigo mismo. Esta medida es esencialmente la magnitud del vector.

**F√≥rmula de la Norma**

Para un vector $\mathbf{v} = [v_1, v_2, \dots, v_n]$ en un espacio $(\mathbb{R}^n)$, la norma (o magnitud) de $(\mathbf{v})$ se define como:

$\|\mathbf{v}\| = \sqrt{\mathbf{v} \cdot \mathbf{v}} = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$

Esta f√≥rmula se deriva del producto interior del vector consigo mismo:

$\mathbf{v} \cdot \mathbf{v} = \sum_{i=1}^n v_i^2$

El resultado de esta suma es un escalar que, al ser introducido bajo la ra√≠z cuadrada, da la longitud del vector.

**Ejemplo Pr√°ctico**

**Ejemplo**: Calculemos la longitud del vector $\mathbf{v} = [3, 4]$ en $(\mathbb{R}^2)$.

Usando la f√≥rmula de la norma:

$\|\mathbf{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$

Este resultado tambi√©n ilustra el famoso Teorema de Pit√°goras en un tri√°ngulo rect√°ngulo con lados 3 y 4, y la hipotenusa siendo la longitud del vector.

**C√≥digo en Python y Visualizaci√≥n**

A continuaci√≥n, veremos c√≥mo calcular esta norma en Python y visualizaremos el vector.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo utilizando seaborn
sns.set_theme(style="darkgrid")

# Definir el vector
v = np.array([3, 4])

# Calcular la norma del vector
norm_v = np.linalg.norm(v)

# Imprimir la norma
print("Longitud del vector v:", norm_v)

# Visualizaci√≥n en 2D
fig, ax = plt.subplots(figsize=(6, 6))
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='deepskyblue',
           label=f'Vector v [3, 4] | Norma: {norm_v:.2f}', width=0.02)

# Dibujar un c√≠rculo con radio igual a la norma del vector
circle = plt.Circle((0, 0), norm_v, color='gray', fill=False, linestyle='dotted', linewidth=1)
ax.add_patch(circle)

# Ajustar l√≠mites para mejor visualizaci√≥n
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)

# Mejorar etiquetas y t√≠tulo
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('Visualizaci√≥n del Vector y su Longitud', fontsize=14)

# A√±adir cuadr√≠cula y leyendas
ax.legend()

# Mostrar el gr√°fico
plt.show()
```

**Interpretaci√≥n**

- **C√≥digo**: Calculamos la norma del vector $(\mathbf{v})$ usando la funci√≥n `np.linalg.norm()`, que internamente realiza el c√°lculo del producto interior de $(\mathbf{v})$ consigo mismo y toma la ra√≠z cuadrada del resultado.

- **Gr√°fica**: La visualizaci√≥n muestra el vector en un plano 2D junto con su longitud. La flecha azul representa el vector, y su longitud en el gr√°fico es consistente con el valor calculado de la norma.

Esta metodolog√≠a para calcular la longitud del vector usando el producto interior es extremadamente √∫til en muchos contextos, especialmente en aquellos que requieren la normalizaci√≥n de vectores (hacer que la longitud del vector sea igual a 1), que es com√∫n en procesamiento de se√±ales, gr√°ficos por computadora y aprendizaje autom√°tico.

### **13.1.4 Producto interior: distancias entre vectores**

El **producto interior** es una herramienta esencial en √°lgebra lineal para muchas aplicaciones, incluida la determinaci√≥n de distancias entre vectores. Vamos a explorar c√≥mo se puede usar el producto interior para calcular la distancia entre dos vectores, utilizando el concepto de norma y la ley del coseno, que son fundamentales para entender la geometr√≠a en espacios vectoriales.

**Definici√≥n y F√≥rmula**

La distancia entre dos vectores $(\mathbf{a})$ y $(\mathbf{b})$ en un espacio vectorial se define t√≠picamente como la norma del vector diferencia $(\mathbf{a} - \mathbf{b})$. Esta distancia es una medida de la "cercan√≠a" entre los dos puntos definidos por los vectores en el espacio vectorial.

**F√≥rmula de la Distancia Euclidiana**

La distancia Euclidiana entre dos vectores $(\mathbf{a})$ y $(\mathbf{b})$ se calcula como:

$d(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\| = \sqrt{(\mathbf{a} - \mathbf{b}) \cdot (\mathbf{a} - \mathbf{b})}$

Esta f√≥rmula utiliza el producto interior para calcular el cuadrado de cada componente de la diferencia $(\mathbf{a} - \mathbf{b})$, sumar estos cuadrados y luego tomar la ra√≠z cuadrada del resultado.

**Ejemplo con C√°lculo**

**Ejemplo**: Supongamos que queremos calcular la distancia entre los vectores $\mathbf{a} = [3, 5]$ y $\mathbf{b} = [1, 1]$ en $(\mathbb{R}^2)$.

1. Calculamos la diferencia $(\mathbf{a} - \mathbf{b})$:

$\mathbf{a} - \mathbf{b} = [3 - 1, 5 - 1] = [2, 4]$

2. Aplicamos la f√≥rmula de la distancia:

$d(\mathbf{a}, \mathbf{b}) = \sqrt{[2, 4] \cdot [2, 4]} = \sqrt{2^2 + 4^2} = \sqrt{4 + 16} = \sqrt{20} \approx 4.47$

**C√≥digo en Python y Visualizaci√≥n**

Podemos calcular y visualizar esta distancia utilizando Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir vectores
a = np.array([3, 5])
b = np.array([1, 1])

# Calcular la diferencia y la distancia
difference = a - b
distance = np.linalg.norm(difference)

# Imprimir la distancia
print("Distancia entre a y b:", distance)

# Visualizaci√≥n en 2D
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a [3, 5]', width=0.012)
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector b [1, 1]', width=0.012)
plt.plot([a[0], b[0]], [a[1], b[1]], 'g--', linewidth=2, label=f'Distancia = {distance:.2f}')

# Configuraci√≥n del gr√°fico
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.xlabel('X Axis', fontsize=12)
plt.ylabel('Y Axis', fontsize=12)
plt.title('Distancia entre Vectores', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right', fontsize=10)
plt.show()
```

**Interpretaci√≥n**

- **C√≥digo**: Este script calcula la distancia entre dos vectores utilizando la funci√≥n `np.linalg.norm()`, que implementa la norma Euclidiana.

- **Gr√°fica**: La visualizaci√≥n muestra los dos vectores y la l√≠nea verde punteada que representa la distancia entre ellos en el plano 2D. Esta representaci√≥n ayuda a visualizar c√≥mo la distancia es la longitud del vector que conecta los dos puntos.

Esta metodolog√≠a es fundamental en muchas √°reas, incluyendo el an√°lisis de datos para medir similitudes entre observaciones, en f√≠sica para determinar distancias en campos de fuerza, y en gr√°ficos por computadora para manejar colisiones y proximidades. Esta comprensi√≥n de c√≥mo calcular y visualizar distancias entre vectores es esencial para una variedad de aplicaciones pr√°cticas y te√≥ricas en ciencia e ingenier√≠a.

### **13.1.5 Producto interior: √°ngulos y ortogonalidad**

El **producto interior** es fundamental en √°lgebra lineal para analizar √°ngulos y ortogonalidad entre vectores. A trav√©s del producto punto, podemos medir el √°ngulo entre dos vectores y determinar si son ortogonales, es decir, si el √°ngulo entre ellos es de 90 grados.

**√Ångulos Entre Vectores Usando el Producto Interior**

El producto interior entre dos vectores no solo proporciona informaci√≥n sobre su magnitud combinada, sino que tambi√©n sobre el coseno del √°ngulo entre ellos. Esto hace posible calcular el √°ngulo directamente.

**F√≥rmula del √Ångulo**

Si tenemos dos vectores $(\mathbf{a})$ y $(\mathbf{b})$, el √°ngulo $(\theta)$ entre ellos puede determinarse con la f√≥rmula:

$\cos(\theta) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}$

donde $(|\mathbf{a}|)$ y $(|\mathbf{b}\|)$ son las normas de $(\mathbf{a})$ y $(\mathbf{b})$, respectivamente, y $(\mathbf{a} \cdot \mathbf{b})$ es el producto punto de $(\mathbf{a})$ y $(\mathbf{b})$.

**Ejemplo de C√°lculo de √Ångulo**

**Ejemplo**: Determinemos el √°ngulo entre los vectores $\mathbf{a} = [1, 2, 3]$ y $\mathbf{b} = [4, -5, 6]$.

1. Calculamos el producto punto:

$\mathbf{a} \cdot \mathbf{b} = 1\cdot4 + 2\cdot(-5) + 3\cdot6 = 4 - 10 + 18 = 12$

2. Calculamos las normas de $(\mathbf{a})$ y $(\mathbf{b})$:

$\|\mathbf{a}\| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{14}, \quad \|\mathbf{b}\| = \sqrt{4^2 + (-5)^2 + 6^2} = \sqrt{77}$

3. Calculamos el coseno del √°ngulo:

$\cos(\theta) = \frac{12}{\sqrt{14} \times \sqrt{77}}$

4. Finalmente, calculamos $(\theta)$:

$\theta = \cos^{-1}\left(\frac{12}{\sqrt{14} \times \sqrt{77}}\right)$

**Ortogonalidad**

Dos vectores son ortogonales si el √°ngulo entre ellos es de 90 grados, lo cual implica que su producto punto es cero.

$\mathbf{a} \cdot \mathbf{b} = 0 \Rightarrow \text{Los vectores son ortogonales}$

**C√≥digo en Python y Visualizaci√≥n**

Veamos c√≥mo implementar estos c√°lculos en Python y visualicemos los vectores.

```python
import numpy as np

# Definir los vectores
a = np.array([1, 2, 3])
b = np.array([4, -5, 6])

# Calcular el producto punto
dot_product = np.dot(a, b)

# Calcular las normas
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Calcular el coseno del √°ngulo
cos_theta = dot_product / (norm_a * norm_b)

# Calcular el √°ngulo en grados
angle = np.arccos(cos_theta) * (180 / np.pi)

# Imprimir el √°ngulo
print("√Ångulo entre a y b:", angle, "grados")

# Verificaci√≥n de ortogonalidad
is_orthogonal = "S√≠" if np.isclose(dot_product, 0) else "No"
print("¬øSon ortogonales a y b?:", is_orthogonal)
```

Este script calcula el √°ngulo entre dos vectores y verifica si son ortogonales. No se incluye una visualizaci√≥n aqu√≠, pero se podr√≠a utilizar matplotlib para crear una representaci√≥n gr√°fica si se deseara.

**Conclusi√≥n**

El producto interior es una herramienta poderosa para entender relaciones angulares entre vectores y para verificar la ortogonalidad, que tiene importantes aplicaciones en proyecciones, optimizaci√≥n, y en el dise√±o de algoritmos en √°reas como la computaci√≥n gr√°fica y el aprendizaje autom√°tico.

Este enfoque matem√°tico y computacional proporciona una base s√≥lida para trabajar con conceptos geom√©tricos en espacios multidimensionales.

### **13.1.6 Productos internos de funciones y variables aleatorias**

El **producto interno de funciones** y el **producto interno de variables aleatorias** son conceptos esenciales en el an√°lisis funcional y en la teor√≠a de probabilidades, respectivamente. Estas ideas generalizan el concepto de producto punto en espacios de funciones y espacios de variables aleatorias.

**Producto Interno de Funciones**

En el contexto de funciones, el producto interno es una generalizaci√≥n del producto punto para funciones continuas. Se utiliza t√≠picamente en espacios de funciones como $(L^2)$, el espacio de funciones cuadrado-integrables.

**Definici√≥n y F√≥rmula**

Para dos funciones $(f)$ y $(g)$ en el espacio $(L^2)$ sobre un intervalo $[a, b]$, el producto interno se define como:

$\langle f, g \rangle = \int_a^b f(x) g(x) \, dx$

**Ejemplo**

**Ejemplo**: Consideremos las funciones $f(x) = x$ y $g(x) = x^2$ sobre el intervalo $[0, 1]$.

El producto interno de $(f)$ y $(g)$ ser√≠a:

$\langle f, g \rangle = \int_0^1 x \cdot x^2 \, dx = \int_0^1 x^3 \, dx = \left[\frac{x^4}{4}\right]_0^1 = \frac{1}{4}$

**Producto Interno de Variables Aleatorias**

El producto interno de variables aleatorias es similar al concepto de covarianza y es fundamental en la teor√≠a de procesos estoc√°sticos.

**Definici√≥n y F√≥rmula**

Para dos variables aleatorias $(X)$ y $(Y)$, el producto interno se define como la esperanza del producto de estas variables:

$\langle X, Y \rangle = E[XY]$

Esto es equivalente a la covarianza si las variables tienen media cero.

**Ejemplo**

**Ejemplo**: Sean $(X)$ y $(Y)$ dos variables aleatorias donde $E[X] = 1$, $E[Y] = 2$, y $E[XY] = 5$.

El producto interno de $(X)$ y $(Y)$ ser√≠a:

$\langle X, Y \rangle = E[XY] = 5$

Para calcular estos ejemplos en Python, podemos usar NumPy para el caso de las variables aleatorias y SciPy para integrar funciones.

```python
import numpy as np
from scipy.integrate import quad

# Definir las funciones
def f(x):
    return x

def g(x):
    return x**2

# Calcular el producto interno de las funciones
result, _ = quad(lambda x: f(x)*g(x), 0, 1)
print("Producto interno de las funciones f y g:", result)

# Variables aleatorias ejemplo (supongamos valores esperados conocidos)
E_XY = 5  # Valor esperado del producto XY
print("Producto interno de las variables aleatorias X y Y:", E_XY)
```

### **13.1.8 Propiedades de los productos internos**

El producto interno, tambi√©n conocido como producto escalar, es una operaci√≥n clave en √°lgebra lineal con muchas propiedades importantes que facilitan su aplicaci√≥n en diversos campos, como la geometr√≠a, la f√≠sica, y la estad√≠stica. Vamos a detallar estas propiedades, ofrecer ejemplos claros y demostrar c√≥mo se pueden implementar en Python.

**Propiedades Fundamentales del Producto Interno**

El producto interno entre dos vectores $(\mathbf{a})$ y $(\mathbf{b})$ en un espacio vectorial tiene las siguientes propiedades:

1. **Conmutatividad**:

   $\mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a}$

2. **Linealidad en el primer argumento** (y tambi√©n en el segundo, si el campo es conmutativo, como los n√∫meros reales):

   - **Homogeneidad**:

     $(c\mathbf{a}) \cdot \mathbf{b} = c(\mathbf{a} \cdot \mathbf{b}) \quad \text{para cualquier escalar } c$

   - **Aditividad**:

     $(\mathbf{a} + \mathbf{b}) \cdot \mathbf{c} = \mathbf{a} \cdot \mathbf{c} + \mathbf{b} \cdot \mathbf{c}$

3. **No negatividad**:

   $\mathbf{a} \cdot \mathbf{a} \geq 0$

   Con igualdad si y solo si $\mathbf{a} = \mathbf{0}$.

4. **Definici√≥n de la norma a partir del producto interno**:

   $\|\mathbf{a}\| = \sqrt{\mathbf{a} \cdot \mathbf{a}}$

5. **Ortogonalidad**:

   Si $\mathbf{a} \cdot \mathbf{b} = 0$, entonces $(\mathbf{a})$ y $(\mathbf{b})$ son ortogonales.

**Ejemplo Ilustrativo**

Consideremos dos vectores en $(\mathbb{R}^3)$: $\mathbf{a} = [3, 4, 0]$ y $\mathbf{b} = [4, -3, 0]$.

1. **Calculamos el Producto Interno**:

   $\mathbf{a} \cdot \mathbf{b} = 3 \times 4 + 4 \times (-3) + 0 \times 0 = 12 - 12 + 0 = 0$

   Esto indica que $(\mathbf{a})$ y $(\mathbf{b})$ son ortogonales, demostrando la propiedad de ortogonalidad.

2. **Norma de $(\mathbf{a})$**:

   $\|\mathbf{a}\| = \sqrt{3^2 + 4^2 + 0^2} = \sqrt{9 + 16} = 5$


**Implementaci√≥n en Python y Visualizaci√≥n**

Vamos a implementar este ejemplo en Python y visualizar los vectores para entender mejor la relaci√≥n entre ellos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir vectores
a = np.array([3, 4, 0])
b = np.array([4, -3, 0])

# Calcular producto interno
dot_product = np.dot(a, b)

# Calcular normas
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Imprimir resultados
print("Producto interno de a y b:", dot_product)
print("Norma de a:", norm_a)
print("Norma de b:", norm_b)
print("¬øSon ortogonales a y b?:", "S√≠" if dot_product == 0 else "No")

# Visualizaci√≥n en 2D
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.015, label='Vector a [3, 4]')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.015, label='Vector b [4, -3]')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('Visualizaci√≥n de Vectores y su Ortogonalidad', fontsize=14, fontweight='bold')
plt.axhline(0, color='black', linewidth=0.5)  # Eje Y
plt.axvline(0, color='black', linewidth=0.5)  # Eje X
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

**Interpretaci√≥n**

- **C√≥digo**: Calculamos el producto interno y las normas, verificando que los vectores son ortogonales porque su producto interno es 0.

### **13.1.9 √Ångulos entre vectores utilizando un producto interior no est√°ndar**

El concepto de **√°ngulos entre vectores utilizando un producto interior no est√°ndar** implica el uso de un producto interno definido de manera diferente al producto punto euclidiano est√°ndar. En espacios vectoriales m√°s generales, podemos definir productos internos que incluyan una matriz de pesos o que operen en contextos m√°s abstractos. Este enfoque es esencial en √°reas como la geometr√≠a diferencial y el √°lgebra lineal avanzada.

**Producto Interior Generalizado**

En un espacio vectorial $(V)$, un producto interior puede ser definido de manera general usando una matriz sim√©trica positiva definida $(A)$. Para dos vectores $(\mathbf{u})$ y $(\mathbf{v})$ en $(\mathbb{R}^n)$, el producto interno $(\langle \mathbf{u}, \mathbf{v} \rangle_A)$ es definido como:

$\langle \mathbf{u}, \mathbf{v} \rangle_A = \mathbf{u}^T A \mathbf{v}$

donde $(A)$ es una matriz $(n \times n)$ que es sim√©trica y positiva definida.

**√Ångulos entre Vectores con un Producto Interior no Est√°ndar**

El √°ngulo $(\theta)$ entre dos vectores $(\mathbf{u})$ y $(\mathbf{v})$ usando este producto interior modificado se puede calcular como:

$\cos(\theta) = \frac{\langle \mathbf{u}, \mathbf{v} \rangle_A}{\sqrt{\langle \mathbf{u}, \mathbf{u} \rangle_A \langle \mathbf{v}, \mathbf{v} \rangle_A}}$

**Ejemplo**

Supongamos que queremos calcular el √°ngulo entre $\mathbf{u} = [1, 2]$ y $\mathbf{v} = [2, -1]$ utilizando la matriz $A = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}$.

1. **Calcular los productos internos**:

   - $\langle \mathbf{u}, \mathbf{v} \rangle_A = [1, 2] \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix} [2, -1]^T = [1, 2] [4, -3] = 4 - 6 = -2$

   - $\langle \mathbf{u}, \mathbf{u} \rangle_A = [1, 2] [2, 6] = 14$

   - $\langle \mathbf{v}, \mathbf{v} \rangle_A = [2, -1] [4, -3] = 11$

2. **Calcular el √°ngulo**:

   - $\cos(\theta) = \frac{-2}{\sqrt{14 \times 11}}$

   - $\theta = \cos^{-1}\left(\frac{-2}{\sqrt{154}}\right)$

**C√≥digo en Python y Visualizaci√≥n**

Implementemos este c√°lculo en Python y visualicemos los vectores y el √°ngulo entre ellos.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Definir vectores y matriz A
u = np.array([1, 2])
v = np.array([2, -1])
A = np.array([[2, 0], [0, 3]])

# Funci√≥n para calcular el producto interior generalizado
def generalized_inner_product(x, y, A):
    return np.dot(x, np.dot(A, y))

# Calcular productos internos
uv_A = generalized_inner_product(u, v, A)
uu_A = generalized_inner_product(u, u, A)
vv_A = generalized_inner_product(v, v, A)

# Calcular el coseno del √°ngulo
cos_theta = uv_A / np.sqrt(uu_A * vv_A)
angle_degrees = np.arccos(cos_theta) * (180 / np.pi)

# Imprimir resultados
print("Coseno del √°ngulo:", cos_theta)
print("√Ångulo entre u y v:", angle_degrees, "grados")

# Visualizaci√≥n en 2D
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector u [1, 2]', width=0.012)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v [2, -1]', width=0.012)

# A√±adir un arco para visualizar el √°ngulo
arc = patches.Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=angle_degrees, color='green', label='√Ångulo')
plt.gca().add_patch(arc)

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('√Ångulo entre Vectores con Producto Interior no Est√°ndar')
plt.grid(True)
plt.legend()
plt.show()
```

**Interpretaci√≥n**

Este ejemplo muestra c√≥mo el producto interno modificado afecta la medida del √°ngulo entre dos vectores, ajust√°ndose a las ponderaciones espec√≠ficas dadas por la matriz $(A)$. La visualizaci√≥n ayuda a ilustrar estas relaciones en un contexto geom√©trico m√°s amplio.

Estos conceptos son cruciales cuando trabajamos en espacios vectoriales con m√©tricas no est√°ndar, com√∫n en aplicaciones avanzadas de matem√°ticas y f√≠sica.

### **13.1.10 Productos internos generales: longitudes y distancias**

El concepto de **productos internos generales** es una extensi√≥n del producto interior est√°ndar que se usa para definir medidas de longitud y distancia en espacios vectoriales equipados con una m√©trica m√°s general. Este enfoque permite la adaptaci√≥n del concepto de √°ngulo, longitud y distancia a contextos donde las propiedades geom√©tricas del espacio se describen de manera diferente a la euclidiana est√°ndar.

**Definici√≥n de Producto Interno General**

Un producto interno general en un espacio vectorial $(V)$ sobre $(\mathbb{R})$ puede definirse usando una matriz $(A)$ sim√©trica y positiva definida. Para dos vectores $(\mathbf{u})$ y $(\mathbf{v})$ en $(V)$, el producto interno se define como:

$\langle \mathbf{u}, \mathbf{v} \rangle_A = \mathbf{u}^T A \mathbf{v}$

**Longitudes y Distancias**

La longitud (o norma) de un vector $(\mathbf{v})$ utilizando este producto interno generalizado se calcula como:

$\|\mathbf{v}\|_A = \sqrt{\langle \mathbf{v}, \mathbf{v} \rangle_A} = \sqrt{\mathbf{v}^T A \mathbf{v}}$

La distancia entre dos vectores $(\mathbf{u})$ y $(\mathbf{v})$ se define usando la norma del vector diferencia $\mathbf{u} - \mathbf{v}$:

$d_A(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|_A = \sqrt{(\mathbf{u} - \mathbf{v})^T A (\mathbf{u} - \mathbf{v})}$

**Ejemplo con C√°lculo**

Supongamos que queremos calcular la longitud y la distancia entre los vectores $\mathbf{u} = [1, 2]$ y $\mathbf{v} = [3, 4]$ usando la matriz $A = \begin{bmatrix} 1 & 0.5 \\ 0.5 & 2 \end{bmatrix}$.

1. **Longitud de $(\mathbf{u})$**:

   $\|\mathbf{u}\|_A = \sqrt{\mathbf{u}^T A \mathbf{u}} = \sqrt{[1, 2] \begin{bmatrix} 1 & 0.5 \\ 0.5 & 2 \end{bmatrix} [1, 2]^T}$

   $= \sqrt{1 \cdot 1 + 2 \cdot 0.5 + 0.5 \cdot 2 + 2 \cdot 2 \cdot 2} = \sqrt{1 + 1 + 1 + 8} = \sqrt{11}$

2. **Distancia entre $(\mathbf{u})$ y $(\mathbf{v})$**:

   $d_A(\mathbf{u}, \mathbf{v}) = \sqrt{(\mathbf{u} - \mathbf{v})^T A (\mathbf{u} - \mathbf{v})}$

   $= \sqrt{[-2, -2] \begin{bmatrix} 1 & 0.5 \\ 0.5 & 2 \end{bmatrix} [-2, -2]^T} = \sqrt{4 + 2 + 2 + 8} = \sqrt{16} = 4$

**C√≥digo en Python y Visualizaci√≥n**

Aqu√≠ est√° c√≥mo podr√≠amos calcular y visualizar esto en Python:

```python
import numpy as np

# Definir vectores y matriz A
u = np.array([1, 2])
v = np.array([3, 4])
A = np.array([[1, 0.5], [0.5, 2]])

# Funci√≥n para calcular el producto interno generalizado
def generalized_inner_product(x, y, A):
    return np.dot(x, np.dot(A, y))

# Calcular longitud de u
length_u = np.sqrt(generalized_inner_product(u, u, A))

# Calcular distancia entre u y v
difference = u - v
distance_uv = np.sqrt(generalized_inner_product(difference, difference, A))

# Imprimir resultados
print("Longitud de u:", length_u)
print("Distancia entre u y v:", distance_uv)
```

### **13.1.11 Productos interiores y √°ngulos**

El **producto interior** (tambi√©n conocido como producto escalar) es una operaci√≥n matem√°tica que no solo proporciona una forma de calcular la proyecci√≥n de un vector sobre otro, sino que tambi√©n permite medir √°ngulos entre dos vectores en espacios vectoriales. Esta herramienta es esencial para entender la geometr√≠a del espacio en el que operan los vectores, especialmente en el contexto de √°lgebra lineal y geometr√≠a anal√≠tica.

**Definici√≥n de Producto Interior**

El producto interior entre dos vectores $(\mathbf{a})$ y $(\mathbf{b})$ en un espacio $(\mathbb{R}^n)$ se define como:

$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i = a_1b_1 + a_2b_2 + \cdots + a_nb_n$

**√Ångulos entre Vectores**

El √°ngulo $(\theta)$ entre los vectores $(\mathbf{a})$ y $(\mathbf{b})$ se puede determinar utilizando la definici√≥n del coseno del √°ngulo que forman, que se deriva del producto interior:

$\cos(\theta) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\|\|\mathbf{b}\|}$

donde $\|\mathbf{a}\|$ y $\|\mathbf{b}\|$ son las normas (magnitudes) de $(\mathbf{a})$ y $(\mathbf{b})$, respectivamente.

**Ejemplo Pr√°ctico**

**Ejemplo**: Calculemos el √°ngulo entre los vectores $\mathbf{a} = [1, 2]$ y $\mathbf{b} = [2, 3]$.

1. **Producto Punto**:

   $\mathbf{a} \cdot \mathbf{b} = 1 \times 2 + 2 \times 3 = 2 + 6 = 8$

2. **Normas de los Vectores**:

   $\|\mathbf{a}\| = \sqrt{1^2 + 2^2} = \sqrt{1 + 4} = \sqrt{5}$

   $\|\mathbf{b}\| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13}$

3. **Coseno del √Ångulo**:

   $\cos(\theta) = \frac{8}{\sqrt{5} \times \sqrt{13}} = \frac{8}{\sqrt{65}}$

4. **√Ångulo $(\theta)$**:

   $\theta = \cos^{-1}\left(\frac{8}{\sqrt{65}}\right)$

**C√≥digo en Python y Visualizaci√≥n**

Podemos calcular este √°ngulo en Python y visualizar los vectores para entender mejor su orientaci√≥n relativa.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Definir vectores
a = np.array([1, 2])
b = np.array([2, 3])

# Calcular producto punto
dot_product = np.dot(a, b)

# Calcular normas
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Calcular coseno del √°ngulo
cos_theta = dot_product / (norm_a * norm_b)

# Calcular √°ngulo en radianes y luego convertir a grados
angle = np.arccos(cos_theta)
angle_degrees = np.degrees(angle)

# Imprimir el √°ngulo
print("√Ångulo entre a y b:", angle_degrees, "grados")

# Visualizaci√≥n en 2D
plt.figure(figsize=(5, 5))
ax = plt.gca()
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'Vector a [1, 2]')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'Vector b [2, 3]')

# Dibujar un arco para visualizar el √°ngulo
arc = patches.Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=angle_degrees, color='green', label='√Ångulo')
ax.add_patch(arc)

# Configurar gr√°fico
ax.set_xlim(0, 3)
ax.set_ylim(0, 4)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Visualizaci√≥n de Vectores y √Ångulo entre Ellos')
ax.grid(True)
ax.legend()

plt.show()
```

**Interpretaci√≥n**

- **C√°lculos**: El script calcula el producto interno, las normas de cada vector, y utiliza estos valores para calcular el √°ngulo entre los dos vectores.
- **Visualizaci√≥n**: La gr√°fica muestra c√≥mo los dos vectores se posicionan uno con respecto al otro en el plano 2D, ayudando a visualizar el √°ngulo calculado.

### **13.1.12 Algoritmo del vecino m√°s pr√≥ximo K**

El **algoritmo del vecino m√°s pr√≥ximo $(K)$**, m√°s conocido como **K-nearest neighbors (K-NN)**, es un m√©todo simple pero poderoso utilizado tanto para clasificaci√≥n como para regresi√≥n en aprendizaje autom√°tico y miner√≠a de datos. Este algoritmo se basa en la premisa de que elementos similares se encuentran cerca unos de otros en un espacio de caracter√≠sticas.

**Descripci√≥n del Algoritmo K-NN**

**Funcionamiento B√°sico**

1. **Elecci√≥n de $(K)$**: El n√∫mero $(K)$ representa la cantidad de vecinos m√°s cercanos que se considerar√°n para determinar la clase o valor de un nuevo punto. $(K)$ es un par√°metro definido por el usuario.

2. **C√°lculo de Distancia**: Calcula la distancia entre el punto de consulta y todos los puntos en el conjunto de datos de entrenamiento. Usualmente se utiliza la distancia euclidiana, aunque se pueden emplear otras m√©tricas dependiendo del contexto.

3. **Identificaci√≥n de Vecinos**: Selecciona los $(K)$ puntos m√°s cercanos (vecinos) al punto de consulta.

4. **Decisi√≥n**:

   - **Para clasificaci√≥n**: La clase del nuevo punto se determina por mayor√≠a de votos entre sus $(K)$ vecinos m√°s cercanos.
   - **Para regresi√≥n**: El valor del nuevo punto se estima como el promedio (o mediana) de los valores de sus $(K)$ vecinos m√°s cercanos.

**Ejemplo de Clasificaci√≥n con K-NN**

Supongamos que tenemos un conjunto de datos con dos caracter√≠sticas (Feature1, Feature2) y dos clases (Clase A y Clase B). Queremos clasificar un nuevo punto $(P)$ usando $(K = 3)$.

**Paso a Paso**:

1. **Calculamos la distancia** de $(P)$ a cada punto en nuestro conjunto de entrenamiento.
2. **Seleccionamos los 3 vecinos m√°s cercanos** (supongamos que dos pertenecen a la Clase A y uno a la Clase B).
3. **Clasificaci√≥n de $(P)$**: Como la mayor√≠a de los vecinos pertenecen a la Clase A, $(P)$ se clasificar√° como Clase A.

**C√≥digo en Python y Visualizaci√≥n**

Implementemos un ejemplo b√°sico del algoritmo K-NN en Python utilizando la librer√≠a `scikit-learn`, que proporciona herramientas eficientes para el aprendizaje autom√°tico.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification

# Generar un conjunto de datos de clasificaci√≥n
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1, n_samples=100)
colors = np.array(['red', 'blue'])[y]

# Crear el clasificador K-NN
knn = KNeighborsClassifier(n_neighbors=3)

# Entrenar el clasificador
knn.fit(X, y)

# Nuevo punto P
P = np.array([[0, -1]])

# Clasificaci√≥n del nuevo punto P
prediction = knn.predict(P)

# Crear malla para la regi√≥n de decisi√≥n
h = .02  # paso de la malla
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predecir la clase para cada punto en la malla
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Visualizaci√≥n
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdBu)
plt.scatter(X[:, 0], X[:, 1], color=colors, edgecolor='k', s=20, label='Datos de entrenamiento')
plt.scatter(P[:, 0], P[:, 1], color='green', marker='x', s=100, label='Nuevo punto P')
plt.title(f'Clasificaci√≥n K-NN de P, K=3, Clase Predicha: {"A" if prediction == 0 else "B"}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

**Interpretaci√≥n**

En el c√≥digo anterior:

- Generamos datos sint√©ticos para clasificaci√≥n con dos caracter√≠sticas.
- Utilizamos la clase `KNeighborsClassifier` de `scikit-learn` para configurar el algoritmo K-NN con $(K = 3)$.
- Entrenamos el modelo con el conjunto de datos y luego clasificamos un nuevo punto $(P)$, visualizando la disposici√≥n de los puntos y la clasificaci√≥n resultante.

Este ejemplo ilustra c√≥mo K-NN puede ser implementado f√°cilmente para resolver problemas de clasificaci√≥n. El algoritmo es intuitivo y efectivo, particularmente en sistemas donde las relaciones de proximidad entre puntos son indicativas de su similitud en t√©rminos de salida (clase o valor).

El K-NN es una herramienta b√°sica pero potente en ciencia de datos, con la ventaja de ser f√°cil de entender y aplicar, aunque puede volverse computacionalmente intenso a medida que el tama√±o del conjunto de datos crece.
