**Estad√≠stica Descriptiva**

La estad√≠stica descriptiva ayuda a describir, mostrar o resumir datos de manera que puedan ser entendidos f√°cilmente. Esto es especialmente √∫til al comenzar cualquier an√°lisis de datos para obtener una vista general de la informaci√≥n.

* **Ejemplo:**

Vamos a generar algunos datos y calcular estad√≠sticas descriptivas b√°sicas como la media, mediana y desviaci√≥n est√°ndar.

```python
import numpy as np
import matplotlib.pyplot as plt

# Usando un estilo predefinido de matplotlib para el fondo y la cuadr√≠cula
plt.style.use('ggplot')

# Generando datos aleatorios
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculando estad√≠sticas descriptivas
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

# Creando el histograma
plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='white')

# Lineas para la media y la mediana
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Media: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashdot', linewidth=2, label=f'Mediana: {median:.2f}')

# A√±adiendo t√≠tulo y etiquetas con personalizaci√≥n
plt.title("Histograma de Datos con Media y Mediana", fontsize=16, fontweight='bold', color='navy')
plt.xlabel("Valor", fontsize=14, fontstyle='italic')
plt.ylabel("Frecuencia", fontsize=14, fontstyle='italic')

# Modificando la leyenda
plt.legend(loc='upper right', frameon=True, facecolor='lightgrey', framealpha=0.9, edgecolor='black')

# Personalizando los ticks
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# Mostrando la cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gr√°fico
plt.show()
```

**Probabilidad**

La probabilidad permite modelar y estudiar incertidumbres. Es fundamental para los m√©todos de inferencia estad√≠stica y aprendizaje autom√°tico.

* **Ejemplo:**

Simulaci√≥n de lanzamientos de una moneda para entender la probabilidad de obtener cara o cruz.

```python
import numpy as np

# Funci√≥n para simular lanzamientos de moneda
def lanzar_moneda(n):
    lanzamientos = np.random.randint(0, 2, size=n)
    caras = np.sum(lanzamientos)
    return caras, n - caras

n_lanzamientos = 1000
caras, cruces = lanzar_moneda(n_lanzamientos)

print(f"En {n_lanzamientos} lanzamientos, sali√≥ cara {caras} veces y cruz {cruces} veces.")
```

**Regresi√≥n Lineal**

La regresi√≥n lineal es una t√©cnica estad√≠stica usada para modelar la relaci√≥n entre una variable dependiente y una o m√°s variables independientes.

* **Ejemplo:**

Vamos a simular algunos datos y ajustar un modelo de regresi√≥n lineal.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos simulados
x = np.random.rand(100, 1) * 100  # Caracter√≠sticas (independientes)
y = 3 * x + np.random.randn(100, 1) * 30 + 50  # Variable dependiente

# Ajuste del modelo de regresi√≥n
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

# Gr√°fica de los datos y la l√≠nea de ajuste
plt.figure(figsize=(10, 6))  # Ajustando el tama√±o de la figura
plt.scatter(x, y, alpha=0.6, color='blue', edgecolor='black', label='Datos')  # Puntos de datos
plt.plot(x, y_pred, color='darkred', linewidth=2, label='L√≠nea de Regresi√≥n')  # L√≠nea de regresi√≥n
plt.title("An√°lisis de Regresi√≥n Lineal", fontsize=14, fontweight='bold', color='darkblue')  # T√≠tulo
plt.xlabel("Variable Independiente", fontsize=12)  # Etiqueta del eje X
plt.ylabel("Variable Dependiente", fontsize=12)  # Etiqueta del eje Y
plt.legend(frameon=True, loc='best', fontsize=10)  # Leyenda
plt.grid(True, linestyle='--', alpha=0.5)  # Grilla personalizada
plt.show()  # Mostrar gr√°fico
```

La relaci√≥n entre una variable dependiente y una o m√°s variables independientes es un concepto central en muchos tipos de an√°lisis estad√≠stico y modelado predictivo, particularmente en la regresi√≥n. Aqu√≠ te explico detalladamente:

**Variable Dependiente**

La **variable dependiente** es la que se quiere predecir o explicar en un modelo estad√≠stico. Esta variable se considera "dependiente" porque su valor se presume que depende de los valores de otras variables. Por ejemplo, en un estudio sobre el rendimiento acad√©mico, la variable dependiente podr√≠a ser la calificaci√≥n final de un estudiante, que se intenta predecir a partir de otros factores.

**Variables Independientes**

Las **variables independientes** son aquellas que se utilizan para predecir el valor de la variable dependiente. En teor√≠a, se consideran independientes porque se presume que son la causa de los cambios en la variable dependiente. En el ejemplo del rendimiento acad√©mico, las variables independientes podr√≠an incluir horas de estudio, asistencia a clase, y otros factores socioecon√≥micos que podr√≠an influir en las calificaciones de los estudiantes.

**Relaci√≥n entre Variables**

El an√°lisis de la relaci√≥n entre la variable dependiente y las variables independientes implica examinar c√≥mo cambia la primera en respuesta a variaciones en las segundas. Este an√°lisis se puede hacer a trav√©s de varios m√©todos estad√≠sticos, siendo la regresi√≥n lineal uno de los m√°s comunes.

**Ejemplo de Regresi√≥n Lineal:**

En un modelo de regresi√≥n lineal simple, donde solo hay una variable independiente, la relaci√≥n se modela como una l√≠nea recta:

$Y = \beta_0 + \beta_1X + \epsilon$

donde:

- $(Y)$ es la variable dependiente.

- $(X)$ es la variable independiente.

- $(\beta_0)$ es el intercepto, el valor esperado de $(Y)$ cuando $(X)$ es 0.

- $(\beta_1)$ es la pendiente, que indica el cambio en $(Y)$ por una unidad de cambio en $(X)$.

- $(\epsilon)$ es el t√©rmino de error, que captura la variaci√≥n en $(Y)$ no explicada por $(X)$.

**Importancia en la Ciencia de Datos**

Entender y modelar correctamente estas relaciones es crucial en la ciencia de datos, ya que permite:

- **Predecir** valores futuros de la variable dependiente bas√°ndose en los valores de las variables independientes.

- **Interpretar** la influencia de cada variable independiente sobre la dependiente, lo cual es vital para comprender los mecanismos subyacentes en muchos campos, desde la econom√≠a hasta la biolog√≠a.

Mediante el estudio de estas relaciones, los cient√≠ficos y analistas pueden formular recomendaciones, tomar decisiones informadas, y desarrollar estrategias basadas en un entendimiento s√≥lido y cuantitativo de los datos.

**Conclusi√≥n**

Las matem√°ticas son fundamentales en la ciencia de datos para dise√±ar modelos que interpretan correctamente los datos y para asegurar que las conclusiones basadas en esos datos sean v√°lidas y √∫tiles. Los ejemplos proporcionados ilustran c√≥mo se pueden aplicar t√©cnicas estad√≠sticas b√°sicas y modelos matem√°ticos para comenzar a resolver problemas en este campo.

El √°lgebra lineal es una rama fundamental de las matem√°ticas aplicadas, que se enfoca en el estudio de vectores, espacios vectoriales (tambi√©n conocidos como espacios lineales), transformaciones lineales, y sistemas de ecuaciones lineales. **Sistemas de Ecuaciones Lineales**

Uno de los usos m√°s inmediatos del √°lgebra lineal es resolver sistemas de ecuaciones lineales. Estos sistemas aparecen en contextos como la econom√≠a (para modelar equilibrios de mercado), ingenier√≠a (balance de fuerzas), y en la computaci√≥n (algoritmos y estructuras de datos).

**Ejemplo Matem√°tico:**

Consideremos el sistema de ecuaciones lineales:

\begin{align*}
x + 2y &= 5 \\
3x + 4y &= 11
\end{align*}

Podemos resolver este sistema usando matrices.

```python
import numpy as np

# Definir la matriz de coeficientes y el vector de t√©rminos independientes
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 11])

# Resolver el sistema de ecuaciones
solucion = np.linalg.solve(A, b)

print(f"La soluci√≥n del sistema es x = {solucion[0]}, y = {solucion[1]}")
```

**Transformaciones Lineales y Cambio de Base**

Las transformaciones lineales son funciones entre espacios vectoriales que preservan las operaciones de suma de vectores y multiplicaci√≥n por escalares. Estas son cruciales para entender cambios de coordenadas y perspectivas en f√≠sica y gr√°ficos por computadora.

**Ejemplo de Rotaci√≥n en el Plano:**

Una rotaci√≥n en el plano puede ser descrita por una matriz de transformaci√≥n. Si queremos rotar puntos en el plano por un √°ngulo $(\theta)$, usamos la matriz:

$
R(\theta) = \begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix}
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para rotar un punto
def rotar_punto(punto, theta):
    R = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    return np.dot(R, punto)

# Punto original y punto rotado
punto_original = np.array([1, 0])
punto_rotado = rotar_punto(punto_original, np.pi / 4)  # Rotar 45 grados

# Graficar
plt.figure(figsize=(6, 6))  # Ajustando el tama√±o de la figura
plt.plot([0, punto_original[0]], [0, punto_original[1]], 'ro-', label='Original', linewidth=2, markersize=8)
plt.plot([0, punto_rotado[0]], [0, punto_rotado[1]], 'bo-', label='Rotado', linewidth=2, markersize=8)
plt.axis('equal')
plt.legend(frameon=True, loc='upper right', fontsize=10)  # Leyenda
plt.title('Rotaci√≥n de un Punto en el Plano', fontsize=14, fontweight='bold', color='navy')  # T√≠tulo
plt.xlabel('X', fontsize=12)  # Etiqueta del eje X
plt.ylabel('Y', fontsize=12)  # Etiqueta del eje Y
plt.grid(True, linestyle='--', alpha=0.7)  # Grilla personalizada

# A√±adir anotaciones para los puntos
plt.annotate('Original (1,0)', xy=punto_original, xytext=(10, -10), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5'))
plt.annotate('Rotado', xy=punto_rotado, xytext=(-90, 0), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5'))

plt.show()  # Mostrar gr√°fico
```

**An√°lisis de Datos y Machine Learning**

### **1.1.3 Dominar los vectores**

Los vectores son una parte esencial del √°lgebra lineal y tienen numerosas aplicaciones en f√≠sica, ingenier√≠a, ciencias de la computaci√≥n y an√°lisis de datos. Un vector se puede considerar como una representaci√≥n de una magnitud que tiene direcci√≥n y tama√±o. En este contexto, vamos a explorar c√≥mo trabajar con vectores en Python, incluyendo c√≥mo realizar operaciones b√°sicas, visualizaciones y aplicaciones pr√°cticas.

**Conceptos B√°sicos de Vectores**

Un vector en matem√°ticas es una entidad que tiene tanto magnitud como direcci√≥n. Puede representarse en varias dimensiones, siendo los vectores en 2D y 3D los m√°s comunes en visualizaci√≥n.

**Representaci√≥n de un Vector**

En 2 dimensiones, un vector $(\mathbf{v})$ puede ser representado como:

$\mathbf{v} = \begin{pmatrix} x \\ y \end{pmatrix}$

donde $(x)$ y $(y)$ son las componentes horizontal y vertical del vector, respectivamente.

```python
import numpy as np

v = np.array([2, 3])
print("Vector v:", v)
```

**Operaciones B√°sicas con Vectores**

Las operaciones fundamentales con vectores incluyen la suma, la resta, el producto escalar y el producto vectorial (este √∫ltimo solo en 3D).

**Suma de Vectores**

La suma de dos vectores se realiza sumando sus correspondientes componentes:

$\mathbf{v} + \mathbf{w} = \begin{pmatrix} v_x + w_x \\ v_y + w_y \end{pmatrix}$

**Ejemplo:**

```python
w = np.array([1, 2])
suma = v + w
print("Suma de v y w:", suma)
```

**Producto Escalar**

El producto escalar (o producto punto) de dos vectores es una operaci√≥n que resulta en un escalar. Se calcula como:

$\mathbf{v} \cdot \mathbf{w} = v_x \times w_x + v_y \times w_y$

**Ejemplo:**

```python
producto_punto = np.dot(v, w)
print("Producto punto de v y w:", producto_punto)
```

**Visualizaci√≥n de Vectores**

La visualizaci√≥n de vectores es crucial para entender su direcci√≥n y magnitud de manera intuitiva. Podemos usar Matplotlib para graficar vectores en Python.

**Ejemplo de Gr√°fico de Vectores:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los vectores
v = np.array([3, 4])
w = np.array([1, 2])

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))  # Ajustando el tama√±o de la figura

# Graficar el vector v
ax.quiver(0, 0, *v, scale=1, scale_units='xy', angles='xy', color='darkred', width=0.015, label='Vector v (3,4)')
ax.quiver(0, 0, *w, scale=1, scale_units='xy', angles='xy', color='navy', width=0.015, label='Vector w (1,2)')

# Establecer l√≠mites para los ejes
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

# A√±adir cuadr√≠cula
ax.grid(True, linestyle='--', alpha=0.5)  # Grilla personalizada

# Hacer que los ejes tengan la misma proporci√≥n para que los vectores se muestren correctamente
ax.set_aspect('equal')

# A√±adir t√≠tulo y leyenda
ax.set_title('Visualizaci√≥n de Vectores', fontsize=14, fontweight='bold')  # T√≠tulo personalizado
ax.legend(loc='upper right', fontsize=10)  # Leyenda personalizada

# Anotaciones para cada vector
ax.annotate('Vector v (3,4)', xy=(1.5, 2), xytext=(3, 4), arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='darkred'))
ax.annotate('Vector w (1,2)', xy=(0.5, 1), xytext=(1, 2), arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='navy'))

# Mostrar el gr√°fico
plt.show()
```

**Aplicaci√≥n Pr√°ctica: Movimiento en F√≠sica**

Los vectores son muy utilizados para describir movimientos en f√≠sica, donde cada vector puede representar la posici√≥n, velocidad o aceleraci√≥n de un objeto en el espacio.

**Ejemplo:**

Supongamos que un objeto se mueve con una velocidad inicial representada por el vector $\mathbf{v} = \begin{pmatrix} 2 \\ 3 \end{pmatrix}$ metros/segundo. Queremos calcular su posici√≥n despu√©s de 5 segundos si sigue movi√©ndose en la misma direcci√≥n y velocidad.

```python
tiempo = 5  # segundos
posicion_final = v * tiempo  # Operaci√≥n elemento a elemento
print("Posici√≥n final despu√©s de 5 segundos:", posicion_final)
```

### **1.1.4 Operaciones con vectores**

Operar con vectores es fundamental en muchas disciplinas cient√≠ficas y t√©cnicas, incluyendo f√≠sica, ingenier√≠a y ciencias de la computaci√≥n. Vamos a detallar algunas operaciones b√°sicas con vectores, c√≥mo se representan matem√°ticamente, y c√≥mo podemos implementar y visualizar estas operaciones en Python.

**Tipos de Operaciones Vectoriales**

1. **Suma de Vectores**

2. **Resta de Vectores**

3. **Multiplicaci√≥n por Escalar**

4. **Producto Punto (o Escalar)**

5. **Producto Cruz (solo para vectores en 3D)**

**Suma de Vectores**

La **suma de dos vectores** se realiza a√±adiendo las componentes correspondientes de cada vector.

**F√≥rmula Matem√°tica**

Si tenemos dos vectores $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$, la suma es:

$\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2)$

**Ejemplo:**

```python
import numpy as np

# Definir dos vectores
u = np.array([2, 5])
v = np.array([3, 1])

# Sumar vectores
suma = u + v
print("Suma de vectores:", suma)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para graficar vectores
def plot_vector(vec, origin, color, label):
    plt.quiver(*origin, *vec, scale=1, scale_units='xy', angles='xy', color=color, width=0.02, label=label)  # Ajuste en el ancho

# Definir vectores
u = np.array([2, 5])
v = np.array([3, 1])
suma = u + v

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))  # Ajustar tama√±o de la figura

# Graficar vectores
plot_vector(u, [0, 0], 'red', 'Vector u (2,5)')
plot_vector(v, [0, 0], 'blue', 'Vector v (3,1)')
plot_vector(suma, [0, 0], 'green', 'Suma u+v')

# Establecer l√≠mites para los ejes
ax.set_xlim(0, 6)
ax.set_ylim(0, 6)

# A√±adir cuadr√≠cula
ax.grid(True, linestyle='--', alpha=0.7)  # Grilla personalizada

# Hacer que los ejes tengan la misma proporci√≥n para que los vectores se muestren correctamente
ax.set_aspect('equal')

# A√±adir t√≠tulo y leyenda
ax.set_title('Suma de Vectores', fontsize=14, fontweight='bold')  # T√≠tulo personalizado
ax.legend(loc='upper right', fontsize=10)  # Leyenda personalizada

# A√±adir anotaciones para cada vector
ax.annotate('u (2,5)', xy=(1, 2.5), xytext=(2, 5), textcoords="offset points", arrowprops=dict(arrowstyle='->', color='red'))
ax.annotate('v (3,1)', xy=(1.5, 0.5), xytext=(3, 1), textcoords="offset points", arrowprops=dict(arrowstyle='->', color='blue'))
ax.annotate('u+v (5,6)', xy=(2.5, 3), xytext=(5, 6), textcoords="offset points", arrowprops=dict(arrowstyle='->', color='green'))

# Mostrar el gr√°fico
plt.show()
```

**Resta de Vectores**

La **resta de vectores** se realiza restando las componentes del segundo vector de las del primero.

**F√≥rmula Matem√°tica**

$\mathbf{u} - \mathbf{v} = (u_1 - v_1, u_2 - v_2)$

**Ejemplo:**

```python
# Restar vectores
resta = u - v
print("Resta de vectores:", resta)
```

**Multiplicaci√≥n por Escalar**

La **multiplicaci√≥n de un vector por un escalar** ampl√≠a o reduce la magnitud del vector sin cambiar su direcci√≥n.

**F√≥rmula Matem√°tica**

$k \mathbf{v} = (k v_1, k v_2)$

**Ejemplo:**

```python
# Multiplicar vector por un escalar
k = 3
escalar_multiplicacion = k * v
print("Multiplicaci√≥n por escalar:", escalar_multiplicacion)
```

**Producto Punto**

El **producto punto** entre dos vectores es una operaci√≥n fundamental que resulta en un escalar, el cual es el producto de las magnitudes de los vectores y el coseno del √°ngulo entre ellos.

**F√≥rmula Matem√°tica**

$\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2$

**Ejemplo:**

```python
# Producto punto
producto_punto = np.dot(u, v)
print("Producto punto:", producto_punto)
```

**Producto Cruz**

El **producto cruz** es espec√≠fico para vectores en tres dimensiones y resulta en un vector perpendicular a ambos vectores originales.

**F√≥rmula Matem√°tica**

$\mathbf{u} \times \mathbf{v} = (u_2 v_3 - u_3 v_2, u_3 v_1 - u_1 v_3, u_1 v_2 - u_2 v_1)$

**Ejemplo:**

```python
# Definir vectores en 3D
u = np.array([2, 5, 0])
v = np.array([3, 1, 0])

# Producto cruz
producto_cruz = np.cross(u, v)
print("Producto cruz:", producto_cruz)
```

Cada una de estas operaciones tiene aplicaciones espec√≠ficas en diversos campos, desde la f√≠sica, donde vectores pueden representar fuerzas o velocidades, hasta la ciencia de datos y el aprendizaje autom√°tico, donde los vectores se utilizan para modelar caracter√≠sticas y realizar c√°lculos en algoritmos. Estos fundamentos son esenciales para una profunda comprensi√≥n y habilidad pr√°ctica en ciencias aplicadas y tecnolog√≠a.

### **1.1.5 Exploraci√≥n del espacio de par√°metros**

La exploraci√≥n del espacio de par√°metros es un concepto esencial en muchas √°reas de la ciencia y la ingenier√≠a, especialmente en estad√≠sticas, machine learning, y f√≠sica. Se refiere al proceso de analizar c√≥mo las variaciones en los par√°metros de un modelo afectan sus predicciones o comportamientos. Esta exploraci√≥n es crucial para optimizar modelos, entender su sensibilidad y mejorar su interpretaci√≥n.

**Conceptos B√°sicos**

**Espacio de par√°metros**: Conjunto de todos los posibles valores que pueden tomar los par√°metros de un modelo.

**Par√°metro**: Variable que se ajusta durante el proceso de calibraci√≥n de un modelo para mejorar su ajuste a un conjunto de datos.

**Aplicaciones**

1. **Optimizaci√≥n de Modelos**: Buscar los par√°metros que minimizan o maximizan una funci√≥n objetivo.

2. **An√°lisis de Sensibilidad**: Estudiar c√≥mo la variaci√≥n en los par√°metros afecta los resultados del modelo.

3. **Validaci√≥n de Modelos**: Asegurar que el modelo funcione bien bajo una amplia gama de condiciones de par√°metros.

**Ejemplo en Machine Learning: Optimizaci√≥n de un Modelo de Regresi√≥n**

Supongamos que tenemos un modelo de regresi√≥n simple y queremos entender c√≥mo var√≠an los errores del modelo con diferentes valores de los par√°metros. Vamos a explorar esto con un ejemplo concreto en Python.

**Definici√≥n del Problema**

Consideramos un modelo de regresi√≥n lineal:

$y = \beta_0 + \beta_1 x + \epsilon$

donde $(\beta_0)$ es el intercepto, $(\beta_1)$ es la pendiente, y $(\epsilon)$ es el t√©rmino de error.

**Simulaci√≥n de Datos**

Generamos datos sint√©ticos donde $(x)$ es una variable independiente y $(y)$ es la variable dependiente afectada por $(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sint√©ticos
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Visualizar los datos
plt.figure(figsize=(8, 6))  # Ajustando el tama√±o de la figura
plt.scatter(x, y, alpha=0.75, color='magenta', edgecolor='black', s=50)  # Cambio de color, tama√±o y bordes de los puntos
plt.title('Distribuci√≥n de Datos Generados', fontsize=14, fontweight='bold', color='navy')  # T√≠tulo personalizado
plt.xlabel('x - Variable Independiente', fontsize=12)  # Etiqueta del eje X personalizada
plt.ylabel('y - Variable Dependiente', fontsize=12)  # Etiqueta del eje Y personalizada
plt.grid(True, linestyle='--', alpha=0.5)  # Grilla personalizada
plt.show()  # Mostrar gr√°fico
```

**Exploraci√≥n del Espacio de Par√°metros**

Vamos a variar el par√°metro $(\beta_1)$ (pendiente) mientras mantenemos $(\beta_0)$ (intercepto) fijo, para ver c√≥mo afecta al error del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Generar datos sint√©ticos
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Rango de valores para beta_1
betas = np.linspace(2.5, 3.5, 200)
errors = []

for beta in betas:
    y_pred = 4 + beta * x  # Usando beta_0 = 4 fijo
    error = mean_squared_error(y, y_pred)
    errors.append(error)

# Graficar beta_1 vs. error
plt.figure(figsize=(8, 6))  # Ajustar tama√±o de la figura
plt.plot(betas, errors, color='darkblue', linewidth=2)  # Cambio de color y grosor de la l√≠nea
plt.title("Error Cuadr√°tico Medio vs. Pendiente del Modelo", fontsize=14, fontweight='bold')  # T√≠tulo personalizado
plt.xlabel("Pendiente (\u03B2_1)", fontsize=12)  # Etiqueta del eje X personalizada
plt.ylabel("Error Cuadr√°tico Medio", fontsize=12)  # Etiqueta del eje Y personalizada
plt.grid(True, linestyle='--', alpha=0.5)  # Grilla personalizada
plt.show()  # Mostrar gr√°fico
```

Este gr√°fico nos muestra c√≥mo var√≠a el error cuadr√°tico medio (MSE) a medida que cambiamos la pendiente del modelo. Podemos observar que hay un valor m√≠nimo de error en alg√∫n punto del rango de $(\beta_1)$ explorado, indicando la pendiente √≥ptima para nuestro conjunto de datos sint√©ticos.

**Conclusi√≥n**

Explorar el espacio de par√°metros es una t√©cnica poderosa para entender y mejorar modelos en ciencias y ingenier√≠a. Nos permite identificar los mejores par√°metros, evaluar la robustez del modelo, y predecir c√≥mo cambios en los par√°metros pueden influir en el desempe√±o del modelo en diferentes condiciones. Esta t√©cnica es fundamental en optimizaci√≥n, an√°lisis de sensibilidad y en el proceso de validaci√≥n de modelos predictivos.

### **1.1.6 Resolver algunas ecuaciones simult√°neas**

Resolver ecuaciones simult√°neas, tambi√©n conocidas como sistemas de ecuaciones lineales, es un problema cl√°sico en √°lgebra lineal que tiene aplicaciones en numerosos campos como la econom√≠a, ingenier√≠a y ciencia de datos. Las ecuaciones simult√°neas pueden resolverse mediante varios m√©todos, incluyendo eliminaci√≥n gaussiana, regla de Cramer, y m√©todos matriciales como la inversi√≥n de matrices o el uso de la descomposici√≥n LU. Aqu√≠ exploraremos c√≥mo resolver un sistema de ecuaciones usando matrices en Python, acompa√±ado de explicaciones te√≥ricas y visualizaciones.

**Sistema de Ecuaciones Lineales**

Supongamos que tenemos el siguiente sistema de ecuaciones lineales:


\begin{align*}
x + 2y + 3z &= 9, \\
2x + 3y + 7z &= 8, \\
4x + y + 2z &= 3.
\end{align*}


Queremos encontrar los valores de $(x)$, $(y)$, y $(z)$ que satisfacen todas estas ecuaciones simult√°neamente.

**Representaci√≥n Matricial**

Este sistema puede representarse en forma matricial como $(A\mathbf{x} = \mathbf{b})$, donde $(A)$ es la matriz de coeficientes, $(\mathbf{x})$ es el vector de inc√≥gnitas, y $(\mathbf{b})$ es el vector de t√©rminos constantes:

$
A = \begin{pmatrix}
1 & 2 & 3 \\
2 & 3 & 7 \\
4 & 1 & 2
\end{pmatrix}, \quad
\mathbf{x} = \begin{pmatrix}
x \\
y \\
z
\end{pmatrix}, \quad
\mathbf{b} = \begin{pmatrix}
9 \\
8 \\
3
\end{pmatrix}
$

**M√©todo de Soluci√≥n en Python**

Vamos a utilizar NumPy, una biblioteca de Python para c√°lculos num√©ricos, para resolver este sistema de ecuaciones. Usaremos la funci√≥n `numpy.linalg.solve()`, que es eficiente y maneja internamente los posibles problemas de precisi√≥n num√©rica asociados con la inversi√≥n de matrices.

**C√≥digo para Resolver el Sistema**

```python
import numpy as np

# Definir la matriz A y el vector b
A = np.array([[1, 2, 3],
              [2, 3, 7],
              [4, 1, 2]])

b = np.array([9, 8, 3])

# Resolver el sistema de ecuaciones
x = np.linalg.solve(A, b)

print("Soluci√≥n del sistema:")
print("x =", x[0])
print("y =", x[1])
print("z =", x[2])
```

**Verificaci√≥n de la Soluci√≥n**

Es importante verificar que la soluci√≥n obtenida realmente satisface el sistema de ecuaciones original. Podemos hacerlo multiplicando la matriz de coeficientes por el vector soluci√≥n y compar√°ndolo con el vector de t√©rminos constantes.

**C√≥digo para Verificar la Soluci√≥n**

```python
# Verificar la soluci√≥n
b_check = np.dot(A, x)
print("Comprobaci√≥n del resultado:")
print("b calculado =", b_check)
```

**Visualizaci√≥n**

Aunque visualizar un sistema de tres variables en un espacio 3D puede ser complejo, se puede intentar visualizar sistemas de dos variables para entender mejor c√≥mo se intersectan las l√≠neas o planos. Aqu√≠, nos enfocaremos en explicar y resolver el sistema.

**Conclusi√≥n**

Resolver sistemas de ecuaciones lineales es fundamental para entender c√≥mo diferentes variables se relacionan entre s√≠ en un modelo. El uso de Python y NumPy facilita la implementaci√≥n de m√©todos num√©ricos complejos con pocas l√≠neas de c√≥digo, lo que permite a los investigadores y profesionales concentrarse m√°s en el problema y menos en los detalles computacionales. Esta capacidad es esencial para la modelizaci√≥n en ciencia y tecnolog√≠a modernas.

### **1.1.7 Realizar algunas operaciones vectoriales**

Operar con vectores es una habilidad esencial en matem√°ticas, f√≠sica, ingenier√≠a y ciencia de datos. Los vectores se utilizan para representar magnitudes que tienen direcci√≥n y sentido, como fuerzas, velocidades, y posiciones. A continuaci√≥n, exploraremos algunas operaciones vectoriales b√°sicas, con ejemplos pr√°cticos y visualizaciones usando Python.

**Tipos de Operaciones Vectoriales**

1. **Suma de Vectores**

2. **Resta de Vectores**

3. **Multiplicaci√≥n por un Escalar**

4. **Producto Escalar (o Producto Punto)**

Cada una de estas operaciones ayuda a manipular vectores para diferentes prop√≥sitos anal√≠ticos y pr√°cticos.

**Suma de Vectores**

La **suma de vectores** implica sumar correspondientes componentes de dos o m√°s vectores.

**F√≥rmula Matem√°tica**

Si tenemos dos vectores en 2D, $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$, la suma es:

$\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2)$

**Ejemplo:**

```python
import numpy as np

u = np.array([2, 3])
v = np.array([5, -2])

suma = u + v
print("Suma de vectores:", suma)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para graficar vectores
def plot_vector(vec, origin, color, name):
    plt.quiver(*origin, *vec, scale=1, scale_units='xy', angles='xy', color=color, label=name, width=0.02)

# Definir vectores
u = np.array([2, 3])
v = np.array([5, -2])
suma = u + v

# Graficar vectores
plot_vector(u, [0, 0], 'red', 'u (2,3)')
plot_vector(v, u, 'blue', 'v (5,-2)')  # origen en la punta de u
plot_vector(suma, [0, 0], 'green', 'u+v')

# Establecer l√≠mites para los ejes
plt.xlim(-1, 8)
plt.ylim(-3, 6)

# Mejorar la visualizaci√≥n del gr√°fico
plt.grid(True, linestyle='--', alpha=0.7)  # Grilla personalizada
plt.axhline(0, color='black', linewidth=1)  # L√≠nea horizontal m√°s visible
plt.axvline(0, color='black', linewidth=1)  # L√≠nea vertical m√°s visible

# A√±adir t√≠tulo y leyenda
plt.title('Suma de Vectores', fontsize=14, fontweight='bold')  # T√≠tulo personalizado
plt.legend(loc='upper right', fontsize=10)  # Leyenda personalizada

# A√±adir anotaciones para cada vector
plt.annotate('u (2,3)', xy=(1, 1.5), xytext=(3, 4), textcoords="offset points", arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='red'))
plt.annotate('v (5,-2)', xy=(u[0] + v[0]/2, u[1] + v[1]/2), xytext=(u[0] + 2.5, u[1] - 1), textcoords="offset points", arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='blue'))
plt.annotate('u+v (7,1)', xy=(suma[0]/2, suma[1]/2), xytext=(suma[0] - 1, suma[1] + 1), textcoords="offset points", arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5', color='green'))

# Mostrar el gr√°fico
plt.show()
```

**Producto Escalar**

El **producto escalar** (o producto punto) es una operaci√≥n fundamental que resulta en un escalar, calculado como el producto de las componentes correspondientes de los vectores y sumando los resultados.

**F√≥rmula Matem√°tica**

$\mathbf{u} \cdot \mathbf{v} = u_1 \times v_1 + u_2 \times v_2$

**Ejemplo:**

```python
producto_punto = np.dot(u, v)
print("Producto escalar de u y v:", producto_punto)
```

**Producto Cruz**

El **producto cruz** es espec√≠fico para vectores en tres dimensiones y resulta en un vector que es perpendicular a ambos vectores de entrada.

**F√≥rmula Matem√°tica**

$\mathbf{u} \times \mathbf{v} = (u_2 v_3 - u_3 v_2, u_3 v_1 - u_1 v_3, u_1 v_2 - u_2 v_1)$

**Ejemplo:**

```python
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])

producto_cruz = np.cross(u, v)
print("Producto cruz de u y v:", producto_cruz)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))  # Ajustando el tama√±o de la figura
ax = fig.add_subplot(111, projection='3d')

# Funci√≥n para graficar vectores en 3D
def plot_vector3d(ax, vec, origin, color, label):
    ax.quiver(*origin, *vec, color=color, label=label, length=1, arrow_length_ratio=0.1, linewidth=2)

# Definir vectores
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
producto_cruz = np.cross(u, v)

# Graficar vectores
plot_vector3d(ax, u, [0, 0, 0], 'red', 'u (1,0,0)')
plot_vector3d(ax, v, [0, 0, 0], 'blue', 'v (0,1,0)')
plot_vector3d(ax, producto_cruz, [0, 0, 0], 'green', 'u x v')

# Ajustar los l√≠mites del eje para visualizaci√≥n completa
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

# Etiquetas de los ejes
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# A√±adir t√≠tulo y leyenda
ax.set_title('Visualizaci√≥n 3D de Vectores', fontsize=14, fontweight='bold')
plt.legend(loc='upper right', fontsize=10)  # Leyenda personalizada

# Mostrar el gr√°fico
plt.show()
```
