**Diferenciaci贸n en Diversos Contextos**

**Diferenciaci贸n Matricial**

**Derivada Matricial:**

### **7.1.3 El jacobiano**

El **jacobiano** es un concepto fundamental en c谩lculo multivariante que representa una matriz de todas las primeras derivadas parciales de un vector de funciones. El determinante de esta matriz se conoce tambi茅n como el determinante jacobiano, y es crucial en la transformaci贸n y el cambio de variables en integrales m煤ltiples, as铆 como en la descripci贸n de la din谩mica de sistemas en varias dimensiones.

**Definici贸n del Jacobiano**

Supongamos que $\mathbf{F} : \mathbb{R}^n \rightarrow \mathbb{R}^m$ es una funci贸n vectorial definida por $\mathbf{F}(\mathbf{x}) = \left(f_1(\mathbf{x}), f_2(\mathbf{x}), \ldots, f_m(\mathbf{x})\right)$, donde $\mathbf{x} = (x_1, x_2, \ldots, x_n)$. La matriz jacobiana $(J)$ de $(\mathbf{F})$ se define como:

$J = \begin{pmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \frac{\partial f_m}{\partial x_2} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{pmatrix}$

**Ejemplo Pr谩ctico de Jacobiano**

Consideremos la funci贸n vectorial $\mathbf{F} : \mathbb{R}^2 \rightarrow \mathbb{R}^2$ dada por:

$\mathbf{F}(x, y) = \left( x^2 y, 5x + \sin(y) \right)$

Las derivadas parciales necesarias para el jacobiano son:

$\frac{\partial (x^2 y)}{\partial x} = 2xy, \quad \frac{\partial (x^2 y)}{\partial y} = x^2, \quad \frac{\partial (5x + \sin(y))}{\partial x} = 5, \quad \frac{\partial (5x + \sin(y))}{\partial y} = \cos(y)$

Entonces, la matriz jacobiana $(J)$ es:

$J = \begin{pmatrix}
2xy & x^2 \\
5 & \cos(y)
\end{pmatrix}$

**Visualizaci贸n en Python del Campo Vectorial y su Jacobiano**

Vamos a graficar el campo vectorial y calcular el jacobiano en un punto espec铆fico usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def F(x, y):
    return np.array([x**2 * y, 5*x + np.sin(y)])

x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
u, v = F(x, y)

plt.figure(figsize=(8, 6))
plt.quiver(x, y, u, v, color='r')
plt.title('Campo Vectorial de F(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# C谩lculo del Jacobiano en el punto (1, )
x_point, y_point = 1, np.pi
J_at_point = np.array([[2 * x_point * y_point, x_point**2], [5, np.cos(y_point)]])
print("Jacobian at (1, ):")
print(J_at_point)
```

**Mec谩nica y Din谩mica de Fluidos**: Utilizado para estudiar c贸mo peque帽as variaciones en las condiciones iniciales afectan a los estados futuros del sistema.

**Ejemplo 1: Transformaci贸n de Coordenadas**

**Transformaci贸n de Cartesiano a Polar**

Supongamos que queremos cambiar de coordenadas cartesianas $(x, y)$ a coordenadas polares $(r, \theta)$, donde:

$x = r \cos(\theta), \quad y = r \sin(\theta)$

La matriz jacobiana para esta transformaci贸n es:

$J = \begin{pmatrix}
\frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} \\
\frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta}

\cos(\theta) & -r \sin(\theta) \\
\sin(\theta) & r \cos(\theta)
\end{pmatrix}$

El determinante del jacobiano $\text{det}(J)$, que es necesario para el cambio de variables en integrales, es:

$\text{det}(J) = r (\cos^2(\theta) + \sin^2(\theta)) = r$

Aqu铆, mostraremos c贸mo las l铆neas de cuadr铆cula cartesianas se transforman en polares.

### **7.1.5 El Hessian**

El **Hessiano** o la **matriz hessiana** es una herramienta fundamental en c谩lculo multivariante, que se utiliza para analizar la curvatura de funciones de varias variables. Esta matriz cuadrada de segundas derivadas parciales es crucial en la optimizaci贸n, particularmente en la determinaci贸n de m谩ximos, m铆nimos y puntos de silla de funciones multivariables.

**Definici贸n de la Matriz Hessiana**

Para una funci贸n $f: \mathbb{R}^n \rightarrow \mathbb{R}$ que es dos veces diferenciable, el Hessiano de $(f)$ en un punto $\mathbf{x} = (x_1, x_2, ..., x_n)$ est谩 definido como:

$H_f(\mathbf{x}) = \begin{pmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{pmatrix}$

**Ejemplo Pr谩ctico: Funci贸n de Dos Variables**

Consideremos la funci贸n $f(x, y) = x^2 y + 3xy + y^2$.

**C谩lculo del Hessiano**

Las segundas derivadas parciales de $(f)$ son:

- $\frac{\partial^2 f}{\partial x^2} = 2y$

- $\frac{\partial^2 f}{\partial y^2} = 2x + 2$

- $\frac{\partial^2 f}{\partial x \partial y} = \frac{\partial^2 f}{\partial y \partial x} = 2x + 3$

As铆, el Hessiano de $(f)$ es:

$H_f = \begin{pmatrix}
2y & 2x + 3 \\
2x + 3 & 2
\end{pmatrix}$

**Visualizaci贸n en Python del Hessiano**

A continuaci贸n, visualizaremos la funci贸n y su Hessiano en Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Define the function and calculate its Hessian
def f(x, y):
    return x**2 * y + 3 * x * y + y**2

def hessian(x, y):
    return np.array([
        [2*y, 2*x + 3],
        [2*x + 3, 2]
    ])

# Setup the meshgrid for X, Y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(12, 8))

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap=cm.viridis, edgecolor='none', alpha=0.8)
cbar = fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('f(X, Y)', rotation=270, labelpad=15)

# Customize the plot with labels and title
ax.set_title('Superficie de $f(x, y) = x^2y + 3xy + y^2$', fontsize=16, pad=20)
ax.set_xlabel('X', fontsize=14, labelpad=10)
ax.set_ylabel('Y', fontsize=14, labelpad=10)
ax.set_zlabel('f(X, Y)', fontsize=14, labelpad=10)

# Enhance visual appearance with lighting
ax.view_init(elev=30, azim=-45)  # Adjust the viewing angle
ax.grid(True, which='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

# Optional: Add contours for better depth perception
ax.contour(X, Y, Z, zdir='z', offset=Z.min(), cmap=cm.viridis)
ax.contour(X, Y, Z, zdir='x', offset=-4, cmap=cm.viridis)
ax.contour(X, Y, Z, zdir='y', offset=4, cmap=cm.viridis)

# Display the plot
plt.show()
```

**Aplicaciones del Hessiano**

**Optimizaci贸n**: El Hessiano es utilizado para determinar la naturaleza de los puntos cr铆ticos en funciones de varias variables. Un Hessiano positivo definido en un punto cr铆tico indica un m铆nimo local, un Hessiano negativo definido indica un m谩ximo local, y un Hessiano indefinido indica un punto de silla.

**Econom铆a**: En teor铆a econ贸mica, el Hessiano se utiliza para analizar la curvatura de funciones de utilidad y funciones de producci贸n, lo que ayuda a entender el comportamiento de los consumidores y productores.

**Ejemplo 1: Funci贸n de Dos Variables**

En este caso, $(x^2)$ se multiplica por la derivada de $(y)$, que es 1, y $(e^y)$ se deriva normalmente como si $(y)$ fuera la 煤nica variable.

### **7.1.7 C谩lculo de la jacobiana**

El **c谩lculo de la jacobiana** implica determinar la matriz jacobiana para una funci贸n vectorial de varias variables. Esta matriz es fundamental en muchos campos de la ciencia y la ingenier铆a, especialmente en la transformaci贸n de coordenadas, optimizaci贸n multivariable y an谩lisis de sistemas din谩micos.

**Definici贸n de la Matriz Jacobiana**

Para una funci贸n vectorial $\mathbf{F} : \mathbb{R}^n \rightarrow \mathbb{R}^m$ definida por $\mathbf{F}(\mathbf{x}) = \left(f_1(\mathbf{x}), f_2(\mathbf{x}), \ldots, f_m(\mathbf{x})\right)$, donde $\mathbf{x} = (x_1, x_2, ..., x_n)$, la matriz jacobiana $(J)$ de $(\mathbf{F})$ en un punto es la matriz de todas sus primeras derivadas parciales, es decir:

$J = \begin{pmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \frac{\partial f_m}{\partial x_2} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{pmatrix}$

**Ejemplo Pr谩ctico: C谩lculo de la Jacobiana**

Las derivadas parciales necesarias para calcular la jacobiana son:

- $\frac{\partial (x^2 y)}{\partial x} = 2xy$

- $\frac{\partial (x^2 y)}{\partial y} = x^2$

- $\frac{\partial (5x + \sin(y))}{\partial x} = 5$

- $\frac{\partial (5x + \sin(y))}{\partial y} = \cos(y)$

Por lo tanto, la matriz jacobiana de $(\mathbf{F})$ es:

$J = \begin{pmatrix}
2xy & x^2 \\
5 & \cos(y)
\end{pmatrix}$

**C贸digo Python para C谩lculo y Visualizaci贸n**

### **7.1.9 C谩lculo de hessianos**

El **c谩lculo de hessianos** se refiere al proceso de formar la matriz hessiana para una funci贸n de m煤ltiples variables. Esta matriz es crucial en el an谩lisis de curvatura de superficies y en la optimizaci贸n multivariable, especialmente para determinar la naturaleza de los puntos cr铆ticos (m铆nimos, m谩ximos y puntos de silla).

**Definici贸n de la Matriz Hessiana**

La matriz hessiana de una funci贸n $f: \mathbb{R}^n \rightarrow \mathbb{R}$ es una matriz cuadrada que contiene todas las segundas derivadas parciales de $(f)$. Si $(f)$ es una funci贸n de las variables $(x_1, x_2, \ldots, x_n)$, la matriz hessiana se define como:

$H = \begin{pmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{pmatrix}$

**Ejemplo Pr谩ctico: C谩lculo de la Hessiana**

Consideremos una funci贸n de dos variables, $f(x, y) = x^4 + y^4 - 4xy + 1$.

**Derivadas Parciales Segundas**

Calculamos las segundas derivadas parciales de $(f)$:

\begin{align*}
\frac{\partial^2 f}{\partial x^2} &= \frac{\partial}{\partial x}(4x^3 - 4y) = 12x^2, \\
\frac{\partial^2 f}{\partial y^2} &= \frac{\partial}{\partial y}(4y^3 - 4x) = 12y^2, \\
\frac{\partial^2 f}{\partial x \partial y} &= \frac{\partial}{\partial y}(-4x) = -4, \\
\frac{\partial^2 f}{\partial y \partial x} &= \frac{\partial}{\partial x}(-4y) = -4.
\end{align*}

**Matriz Hessiana**

Por lo tanto, la matriz hessiana para $(f)$ es:

$H_f = \begin{pmatrix}
12x^2 & -4 \\
-4 & 12y^2
\end{pmatrix}$

**C贸digo Python para C谩lculo y Visualizaci贸n**

Podemos usar Python para visualizar la funci贸n y calcular su matriz hessiana en diferentes puntos.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and its Hessian
def f(x, y):
    return x**4 + y**4 - 4*x*y + 1

def hess_f(x, y):
    return np.array([
        [12*x**2, -4],
        [-4, 12*y**2]
    ])

# Generate data for the plots
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10, label='f(X, Y)')

# Enhance the plot with labels and title
ax.set_title('Surface of $f(x, y) = x^4 + y^4 - 4xy + 1$', fontsize=14, pad=20)
ax.set_xlabel('X', fontsize=12, labelpad=10)
ax.set_ylabel('Y', fontsize=12, labelpad=10)
ax.set_zlabel('f(X, Y)', fontsize=12, labelpad=10)

# Adjust view angle for better visualization
ax.view_init(elev=30, azim=225)

# Calculate the Hessian at a specific point
x_point, y_point = 1, 1
hessian_matrix = hess_f(x_point, y_point)
print("Hessian at (1, 1):")
print(hessian_matrix)

plt.show()
```
