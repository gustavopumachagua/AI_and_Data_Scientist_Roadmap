### **8.1.1 Regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** es una extensi贸n de la regla de la cadena b谩sica de c谩lculo diferencial, adaptada para funciones de m煤ltiples variables. Esta regla es esencial cuando se trata de composiciones de funciones donde cada funci贸n involucra varias variables. Es fundamental en el an谩lisis de gradientes, derivadas direccionales, y en la diferenciaci贸n de funciones compuestas m谩s complejas.

**Definici贸n y Principios**

La regla de la cadena multivariante permite calcular la derivada de una funci贸n compuesta $f(g(x, y), h(x, y))$ donde $(g)$ y $(h)$ son funciones de las variables $(x)$ y $(y)$, y $(f)$ es una funci贸n de dos variables, que son los resultados de $(g)$ y $(h)$, respectivamente.

**Formulaci贸n Matem谩tica**

Si tenemos:

- Una funci贸n $z = f(u, v)$.

- Donde $u = g(x, y)$ y $v = h(x, y)$.

La derivada de $(z)$ con respecto a $(x)$ y $(y)$ utilizando la regla de la cadena ser铆a:

$\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$

$\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$

**Ejemplo Pr谩ctico**

Supongamos que tenemos las siguientes funciones:

- $f(u, v) = u^2 + v^2$.

- $u = g(x, y) = x^2 + y^2$.

- $v = h(x, y) = e^{x+y}$.

**Derivadas Parciales de $(g)$ y $(h)$**

$\frac{\partial g}{\partial x} = 2x, \quad \frac{\partial g}{\partial y} = 2y$

$\frac{\partial h}{\partial x} = e^{x+y}, \quad \frac{\partial h}{\partial y} = e^{x+y}$

**Aplicaci贸n de la Regla de la Cadena Multivariante**

$\frac{\partial z}{\partial x} = 2u \cdot 2x + 2v \cdot e^{x+y} = 4x(x^2 + y^2) + 2e^{x+y}e^{x+y}$

$\frac{\partial z}{\partial y} = 2u \cdot 2y + 2v \cdot e^{x+y} = 4y(x^2 + y^2) + 2e^{x+y}e^{x+y}$

**C贸digo Python y Visualizaci贸n**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the functions
def f(u, v):
    return u**2 + v**2

def g(x, y):
    return x**2 + y**2

def h(x, y):
    return np.exp(x + y)

# Calculate partial derivatives using the chain rule
def partial_derivatives(x, y):
    u = g(x, y)
    v = h(x, y)
    dz_dx = 2 * u * 2 * x + 2 * v * np.exp(x + y)
    dz_dy = 2 * u * 2 * y + 2 * v * np.exp(x + y)
    return dz_dx, dz_dy

# Range of values for x and y
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
DZ_DX, DZ_DY = partial_derivatives(X, Y)

# Visualization of partial derivatives
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
surface1 = ax1.plot_surface(X, Y, DZ_DX, cmap='viridis', edgecolor='none')
ax1.set_title('Partial Derivative $\partial z / \partial x$')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('DZ_DX')
fig.colorbar(surface1, ax=ax1, shrink=0.5, aspect=10)

ax2 = fig.add_subplot(122, projection='3d')
surface2 = ax2.plot_surface(X, Y, DZ_DY, cmap='viridis', edgecolor='none')
ax2.set_title('Partial Derivative $\partial z / \partial y$')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('DZ_DY')
fig.colorbar(surface2, ax=ax2, shrink=0.5, aspect=10)

plt.show()
```

**Conclusi贸n**

La regla de la cadena multivariante es una herramienta esencial en el c谩lculo diferencial para funciones de varias variables. Permite descomponer c贸mo las variaciones en variables subyacentes afectan a una funci贸n compuesta, facilitando an谩lisis m谩s profundos en campos como la f铆sica, la ingenier铆a, y la econom铆a. Su comprensi贸n y aplicaci贸n correcta son fundamentales para avanzar en el estudio y aplicaci贸n de matem谩ticas avanzadas en problemas reales.

### **8.1.2 M谩s regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** en c谩lculo es una extensi贸n crucial de la regla de la cadena est谩ndar, aplicable cuando se trabaja con funciones de varias variables. Esta t茅cnica permite calcular la derivada de funciones compuestas que son el resultado de aplicar m煤ltiples funciones, cada una dependiendo de varias variables. Es fundamental en 谩reas como la din谩mica de sistemas, donde las variables de estado evolucionan con el tiempo y dependen de m煤ltiples factores.

**Formulaci贸n General**

Supongamos que queremos calcular la derivada de una funci贸n compuesta $z = f(g(x, y), h(x, y))$ respecto a m煤ltiples variables. Aqu铆, $(g)$ y $(h)$ son funciones de las variables $(x)$ y $(y)$, y $(f)$ es una funci贸n que depende de $u = g(x, y)$ y $v = h(x, y)$.

La regla de la cadena multivariante nos dice que:

$\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$

$\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$

**Interpretaci贸n**

- $\frac{\partial f}{\partial u}$ y $\frac{\partial f}{\partial v}$ son las derivadas parciales de $(f)$ respecto a sus argumentos $(u)$ y $(v)$, evaluadas en $g(x, y)$ y $h(x, y)$ respectivamente.

- $\frac{\partial g}{\partial x}$ y $\frac{\partial h}{\partial x}$, as铆 como $\frac{\partial g}{\partial y}$ y $\frac{\partial h}{\partial y}$, son las derivadas parciales de $(g)$ y $(h)$ respecto a $(x)$ y $(y)$.

**Ejemplo Extendido con Aplicaci贸n**

Consideremos una situaci贸n concreta donde $f(u, v) = u^2 + v^2$, $g(x, y) = x^2 - y^2$, y $h(x, y) = 2xy$.

**C谩lculo de Derivadas Parciales**

Para $(g)$ y $(h)$:

$\frac{\partial g}{\partial x} = 2x, \quad \frac{\partial g}{\partial y} = -2y$

$\frac{\partial h}{\partial x} = 2y, \quad \frac{\partial h}{\partial y} = 2x$

Para $(f)$:

$\frac{\partial f}{\partial u} = 2u, \quad \frac{\partial f}{\partial v} = 2v$

**Aplicaci贸n de la Regla de la Cadena**

$\frac{\partial z}{\partial x} = 2u \cdot 2x + 2v \cdot 2y = 4(x^2 - y^2)x + 4(2xy)y$

$\frac{\partial z}{\partial y} = 2u \cdot -2y + 2v \cdot 2x = -4(x^2 - y^2)y + 4(2xy)x$

**C贸digo Python para Visualizaci贸n**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(u, v):
    return u**2 + v**2

def g(x, y):
    return x**2 - y**2

def h(x, y):
    return 2*x*y

def partials(x, y):
    u = g(x, y)
    v = h(x, y)
    df_dx = 2 * u * 2 * x + 2 * v * 2 * y
    df_dy = 2 * u * (-2 * y) + 2 * v * 2 * x
    return df_dx, df_dy

# Generate data
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
DF_DX, DF_DY = partials(X, Y)

# Plotting the partial derivatives
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Contour plot for z/x
cp = ax1.contourf(X, Y, DF_DX, levels=np.linspace(DF_DX.min(), DF_DX.max(), 100), cmap='viridis')
fig.colorbar(cp, ax=ax1, label='Value of $\partial z / \partial x$')
ax1.set_title('$\partial z / \partial x$', fontsize=14)
ax1.set_xlabel('X', fontsize=12)
ax1.set_ylabel('Y', fontsize=12)
ax1.set_aspect('equal')  # Set equal scaling by changing aspect ratio of the x and y axes

# Contour plot for z/y
cp2 = ax2.contourf(X, Y, DF_DY, levels=np.linspace(DF_DY.min(), DF_DY.max(), 100), cmap='viridis')
fig.colorbar(cp2, ax=ax2, label='Value of $\partial z / \partial y$')
ax2.set_title('$\partial z / \partial y$', fontsize=14)
ax2.set_xlabel('X', fontsize=12)
ax2.set_ylabel('Y', fontsize=12)
ax2.set_aspect('equal')  # Consistent aspect ratio

# Enhancements
plt.subplots_adjust(wspace=0.3)  # Adjust the spacing between subplots
plt.suptitle('Visualization of Partial Derivatives', fontsize=16)  # Overall title for all subplots
plt.show()
```

**Conclusi贸n**

La regla de la cadena multivariante es una herramienta poderosa en c谩lculo diferencial, permitiendo descomponer c贸mo las variaciones en m煤ltiples variables subyacentes afectan a una funci贸n compuesta. Su comprensi贸n y aplicaci贸n son esenciales para el estudio avanzado en muchas disciplinas, incluyendo f铆sica, ingenier铆a, econom铆a y biolog铆a.

### **8.1.3 Redes neuronales simples**

Las **redes neuronales simples** son modelos computacionales inspirados en el cerebro humano que se utilizan para aprender y tomar decisiones a partir de datos. A menudo, una red neuronal simple se refiere a un modelo de perceptr贸n o a redes neuronales de una sola capa, que son los bloques de construcci贸n b谩sicos para redes m谩s complejas. Estos modelos son capaces de realizar tareas de clasificaci贸n y regresi贸n bas谩ndose en la entrada que reciben y los pesos que aprenden durante el entrenamiento.

**Conceptos B谩sicos de Redes Neuronales**

**Neurona Artificial**

La unidad b谩sica de una red neuronal es la **neurona artificial**, que recibe varias entradas, las combina linealmente usando pesos, a帽ade un sesgo, y luego aplica una funci贸n de activaci贸n para producir una salida. La estructura b谩sica de una neurona se puede describir con la siguiente f贸rmula matem谩tica:

$y = f\left(\sum_{i=1}^n w_i x_i + b\right)$

Donde:

- $(x_i)$ son las entradas.

- $(w_i)$ son los pesos.

- $(b)$ es el sesgo.

- $(f)$ es la funci贸n de activaci贸n, como la funci贸n sigmoide, ReLU, etc.,

- $(y)$ es la salida.

**Funciones de Activaci贸n**

Las funciones de activaci贸n m谩s comunes incluyen:

- **Sigmoide**: $f(x) = \frac{1}{1 + e^{-x}}$, 煤til para clasificaci贸n binaria.

- **ReLU (Rectified Linear Unit)**: $f(x) = \max(0, x)$, com煤n en redes profundas por su eficiencia.

- **Tanh**: $f(x) = \tanh(x)$, que es una versi贸n escalada de la funci贸n sigmoide.

**Ejemplo de una Red Neuronal Simple: El Perceptr贸n**

El perceptr贸n es la forma m谩s simple de una red neuronal, que consiste en una sola neurona con aprendizaje supervisado. Es capaz de clasificar dos clases linealmente separables.

**Configuraci贸n del Problema**

Supongamos que queremos clasificar puntos de datos que pertenecen a dos clases usando un perceptr贸n con una funci贸n de activaci贸n sigmoide. Los puntos de datos son $(x_1, y_1)$ con etiquetas de clase.

**C贸digo Python para un Perceptr贸n Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Perceptron model
class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 0 else 0 for i in y])

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = sigmoid(linear_output)

                update = self.lr * (y_[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_output)
        return np.array([1 if i > 0.5 else 0 for i in y_predicted])

# Example data
X = np.array([[2, 3], [1, 1], [5, 6], [2, 1]])
y = np.array([0, 0, 1, 1])

# Train perceptron
p = Perceptron(learning_rate=0.1, n_iters=1000)
p.fit(X, y)

# Predict new data
predictions = p.predict(X)

# Visualization of data and decision boundary
fig, ax = plt.subplots()
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', edgecolor='k', s=100)
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

# Calculate decision boundary (x2 = -(b + w1*x1)/w2)
x1 = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, 100)
x2 = -(p.bias + p.weights[0] * x1) / p.weights[1]

ax.plot(x1, x2, color='red', linestyle='--', label='Decision Boundary')
ax.set_title('Perceptron Classification')
ax.set_xlabel('Feature 1 (X1)')
ax.set_ylabel('Feature 2 (X2)')
ax.legend()

plt.show()
```

**Conclusi贸n**

Las redes neuronales simples, como el perceptr贸n, proporcionan una introducci贸n fundamental a los conceptos m谩s profundos de las redes neuronales y el aprendizaje autom谩tico. Aunque los perceptrones tienen limitaciones, como su incapacidad para resolver problemas no linealmente separables, son la base sobre la que se construyen arquitecturas m谩s complejas como las redes neuronales profundas (deep learning). Estas herramientas b谩sicas son esenciales para entender c贸mo las m谩quinas pueden aprender a partir de datos y tomar decisiones.

### **8.1.4 Redes neuronales m谩s sencillas**

Las **redes neuronales m谩s sencillas**, generalmente conocidas como redes de una sola capa o perceptrones, son modelos b谩sicos de aprendizaje autom谩tico dise帽ados para la clasificaci贸n y regresi贸n lineal. Estos modelos son fundamentales para entender c贸mo las redes neuronales m谩s complejas, como las redes neuronales profundas, funcionan y se construyen.

**Conceptos B谩sicos**

Una red neuronal simple generalmente consiste en una 煤nica "neurona" que procesa m煤ltiples entradas para producir una salida. La estructura b谩sica de una neurona incluye:

- **Entradas (Inputs)**: Valores num茅ricos que representan caracter铆sticas del dato.

- **Pesos (Weights)**: Coeficientes que se aprenden durante el entrenamiento, que multiplican a las entradas.

- **Sesgo (Bias)**: Un t茅rmino adicional que permite ajustar la salida de la neurona adem谩s de los pesos aplicados a las entradas.

- **Funci贸n de Activaci贸n**: Una funci贸n matem谩tica que transforma la suma ponderada de las entradas y el sesgo en la salida que la red proporciona.

**F贸rmula de la Neurona**

La salida de una neurona, $(y)$, se calcula como:

$y = f\left(\sum_{i=1}^n w_i x_i + b\right)$

Donde:

- $(x_i)$ son las entradas.

- $(w_i)$ son los pesos.

- $(b)$ es el sesgo.

- $(f)$ es la funci贸n de activaci贸n, como sigmoide, tanh, o ReLU,

**Ejemplo Pr谩ctico: Perceptr贸n para Clasificaci贸n Binaria**

Vamos a considerar un ejemplo donde utilizamos un perceptr贸n para clasificar datos que pertenecen a dos categor铆as distintas basadas en dos caracter铆sticas.

**Datos de Ejemplo**

Imagina que tenemos un conjunto de datos con dos caracter铆sticas, donde cada punto pertenece a una de dos clases (0 o 1). El objetivo es aprender una l铆nea que pueda separar estas dos clases.

**Funci贸n de Activaci贸n**

Usaremos la funci贸n sigmoide, que es com煤n en las neuronas de clasificaci贸n binaria:

$\sigma(z) = \frac{1}{1 + e^{-z}}$

**C贸digo Python para un Perceptr贸n Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Perceptron class
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.lr = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Train the model
        for _ in range(self.n_iterations):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = sigmoid(linear_output)
                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_output)
        return [1 if i > 0.5 else 0 for i in y_predicted]

# Sample data
X = np.array([[0.1, 1.1], [2.2, 2.9], [1.6, 1.8], [3.0, 3.1], [1.2, 0.9], [3.1, 3.3]])
y = np.array([0, 1, 0, 1, 0, 1])

# Train the model
p = Perceptron(learning_rate=0.1)
p.fit(X, y)

# Visualization of data and decision boundary
fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', edgecolor='k', s=50)
legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)  # Correct use of add_artist
points = np.linspace(min(X[:, 0]) - 0.1, max(X[:, 0]) + 0.1, 100)
decision_boundary = -(p.weights[0] * points + p.bias) / p.weights[1]
ax.plot(points, decision_boundary, color='red', label='Decision Boundary')
ax.set_title('Perceptron Classification')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.legend()
plt.show()
```

**Conclusi贸n**

Las redes neuronales simples como el perceptr贸n son herramientas poderosas para aprender relaciones lineales entre caracter铆sticas y clasificaciones. Aunque limitadas a problemas linealmente separables y a clasificaci贸n binaria, estas redes forman la base para entender y construir modelos de redes neuronales m谩s complejos. A medida que se a帽aden capas y se utilizan diferentes funciones de activaci贸n, estos modelos pueden aprender patrones mucho m谩s complejos y resolver problemas de clasificaci贸n y regresi贸n de alta dimensionalidad.

### **8.1.5 Ejercicio de regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** en c谩lculo diferencial es una herramienta esencial para tratar con funciones compuestas que dependen de varias variables. Esta regla permite calcular las derivadas de estas funciones compuestas de manera sistem谩tica y eficaz. A continuaci贸n, proporcionar茅 una explicaci贸n detallada con ejemplos, f贸rmulas matem谩ticas y un c贸digo Python para ilustrar c贸mo aplicar esta regla en un ejercicio pr谩ctico.

**Ejemplo de Regla de la Cadena Multivariante**

Consideremos el siguiente problema: queremos calcular las derivadas parciales de la funci贸n $(z)$ compuesta por $z = f(u, v)$ donde $u = g(x, y)$ y $v = h(x, y)$. Supongamos que:

- $f(u, v) = u^2 + 3uv + v^2$.

- $g(x, y) = x^2 + y^2$.

- $h(x, y) = e^{x+y}$.

**Objetivo**

Calcular $\frac{\partial z}{\partial x}$ y $\frac{\partial z}{\partial y}$.

**Paso a Paso de la Regla de la Cadena**

**Determinar las Derivadas Parciales de $(g)$ y $(h)$:**

   - $\frac{\partial g}{\partial x} = 2x$.

   - $\frac{\partial g}{\partial y} = 2y$.

   - $\frac{\partial h}{\partial x} = e^{x+y}$.

   - $\frac{\partial h}{\partial y} = e^{x+y}$.

**Determinar las Derivadas Parciales de $(f)$ respecto a $(u)$ y $(v)$:**

   - $\frac{\partial f}{\partial u} = 2u + 3v$.

   - $\frac{\partial f}{\partial v} = 3u + 2v$.

**Aplicar la Regla de la Cadena:**

   - $\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$.

   - $\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$.

**C贸digo Python para Visualizaci贸n y C谩lculo**

```python
import numpy as np

# Definir las funciones g, h, f y sus derivadas parciales
def g(x, y):
    return x**2 + y**2

def h(x, y):
    return np.exp(x + y)

def f(u, v):
    return u**2 + 3*u*v + v**2

def df_du(u, v):
    return 2*u + 3*v

def df_dv(u, v):
    return 3*u + 2*v

# Calcular derivadas parciales usando la regla de la cadena
def calculate_partials(x, y):
    u = g(x, y)
    v = h(x, y)
    partial_x = df_du(u, v) * 2*x + df_dv(u, v) * np.exp(x + y)
    partial_y = df_du(u, v) * 2*y + df_dv(u, v) * np.exp(x + y)
    return partial_x, partial_y

# Ejemplo de punto para c谩lculo
x, y = 1, 1
partial_x, partial_y = calculate_partials(x, y)
print(f"z/x at (x={x}, y={y}) is {partial_x}")
print(f"z/y at (x={x}, y={y}) is {partial_y}")
```

Este script define las funciones $g(x, y)$, $h(x, y)$ y $f(u, v)$, junto con sus derivadas parciales necesarias. Luego calcula las derivadas parciales de $(z)$ utilizando la regla de la cadena para un punto espec铆fico $(x, y) = (1, 1)$.

### **8.1.6 Redes neuronales artificiales simples**

Las **Redes Neuronales Artificiales Simples** (ANN, por sus siglas en ingl茅s) son modelos computacionales que se inspiran en las redes neuronales biol贸gicas del cerebro humano. Estas redes son estructuras compuestas por nodos, llamados "neuronas artificiales", que est谩n conectadas entre s铆 y transmiten se帽ales de una a otra. Las redes neuronales simples generalmente se refieren a modelos con pocas capas, como el perceptr贸n o redes con una sola capa oculta, que pueden realizar tareas de clasificaci贸n y regresi贸n.

**Componentes B谩sicos de una Red Neuronal Simple**

Una red neuronal t铆pica incluye:

- **Neuronas de Entrada**: Donde cada neurona representa una caracter铆stica del conjunto de datos.

- **Capas Ocultas**: Capas intermedias entre las entradas y salidas, que pueden tener una o m谩s neuronas. En redes simples, frecuentemente hay solo una capa oculta.

- **Neuronas de Salida**: Representan las predicciones o clasificaciones del modelo.

- **Pesos y Sesgos**: Cada conexi贸n entre neuronas tiene un peso y cada neurona tiene un sesgo, que son ajustados durante el entrenamiento.

- **Funciones de Activaci贸n**: Como sigmoide, tanh, o ReLU, que ayudan a introducir no linealidades en el modelo.

**Ejemplo Pr谩ctico: Red Neuronal con una Capa Oculta**

Consideremos un ejemplo sencillo donde construimos una red neuronal con una capa oculta para clasificar datos que son linealmente no separables.

**Configuraci贸n del Problema**

Imagina que tenemos un conjunto de datos de puntos que pertenecen a dos clases, y queremos usar una red neuronal para clasificar estos puntos.

**Funci贸n de Activaci贸n**

Utilizaremos la funci贸n de activaci贸n ReLU para las neuronas de la capa oculta y la funci贸n sigmoide para la salida, adecuada para la clasificaci贸n binaria.

**C贸digo Python para una Red Neuronal Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

# Visualizar la funci贸n de activaci贸n
x_vals = np.linspace(-10, 10, 100)
plt.figure(figsize=(8, 6))  # Ajusta el tama帽o de la figura
plt.plot(x_vals, sigmoid(x_vals), label='Sigmoide', color='blue', linewidth=2, linestyle='--')
plt.plot(x_vals, relu(x_vals), label='ReLU', color='red', linewidth=2, linestyle='-')

plt.title('Funciones de Activaci贸n', fontsize=14, fontweight='bold')
plt.xlabel('X', fontsize=12)
plt.ylabel('f(X)', fontsize=12)
plt.grid(True)  # A帽adir cuadr铆cula
plt.legend()

# Puedes tambi茅n marcar puntos espec铆ficos, como por ejemplo el punto donde ReLU cambia de 0 a x
plt.scatter([0], [0], color='black', zorder=5)  # Marca el origen para ReLU
plt.annotate('Origen (0,0)', (0, 1), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

**Explicaci贸n del C贸digo**

- **Inicializaci贸n**: Se inicializan aleatoriamente los pesos y se establecen los sesgos a cero.

- **Feedforward**: Se calcula la salida de la red mediante la propagaci贸n de las entradas a trav茅s de las capas usando las funciones de activaci贸n.

**Conclusi贸n**

Las redes neuronales artificiales simples son herramientas poderosas para modelar relaciones no lineales y patrones complejos en datos, aunque su simplicidad puede limitar su capacidad para manejar tareas m谩s complejas que requieren una representaci贸n m谩s rica, como puede ofrecer una red profunda. ### **8.1.7 Entrenamiento de redes neuronales**

El **entrenamiento de redes neuronales** es el proceso mediante el cual una red neuronal aprende a realizar tareas espec铆ficas, como la clasificaci贸n o la regresi贸n, a partir de datos de entrenamiento. Este proceso involucra ajustar los pesos y sesgos de la red de tal manera que el error entre las salidas predichas y las salidas reales deseadas se minimice. Vamos a explorar este proceso detalladamente con ejemplos, f贸rmulas y un script de Python para ilustrar el entrenamiento de una red neuronal simple.

**Componentes Clave del Entrenamiento**

**Datos de Entrenamiento**: Conjunto de entradas y salidas esperadas que la red usar谩 para aprender.

**Modelo de Red Neuronal**: Estructura de la red, incluyendo n煤mero y tipo de capas, n煤mero de neuronas por capa, funciones de activaci贸n, etc.

**Funci贸n de P茅rdida (Loss Function)**: Mide el error entre las salidas predichas y las reales. Ejemplos comunes incluyen el Error Cuadr谩tico Medio (MSE) para regresi贸n y la Entrop铆a Cruzada para clasificaci贸n.

**Optimizador**: M茅todo usado para ajustar los pesos y sesgos en funci贸n del gradiente de la funci贸n de p茅rdida. Ejemplos comunes son el Descenso del Gradiente Estoc谩stico (SGD), Adam y RMSprop.

**Backpropagation**: Algoritmo utilizado para calcular los gradientes de la funci贸n de p茅rdida respecto a cada peso y sesgo en la red.

**Ejemplo de Entrenamiento: Red Neuronal para XOR**

Supongamos que queremos entrenar una red neuronal para resolver el problema XOR, que no es linealmente separable y requiere al menos una capa oculta.

**Configuraci贸n del Problema**

- **Datos XOR**: Entradas $[0, 0], [0, 1], [1, 0], [1, 1]$ y Salidas $[0], [1], [1], [0]$.

**Red Neuronal**

- **Arquitectura**: 2 neuronas en la capa de entrada, 2 neuronas en una capa oculta, 1 neurona en la capa de salida.

- **Funci贸n de Activaci贸n**: Sigmoide para todas las neuronas.

**C贸digo Python para el Entrenamiento**

```python
import numpy as np

# Funciones de activaci贸n y derivadas
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Entrenamiento de la red neuronal
class NeuralNetwork:
    def __init__(self, x, y):
        self.input      = x
        self.weights1   = np.random.rand(self.input.shape[1], 2)  # pesos capa oculta
        self.weights2   = np.random.rand(2, 1)  # pesos capa salida
        self.y          = y
        self.output     = np.zeros(self.y.shape)

    def feedforward(self):
        self.layer1 = sigmoid(np.dot(self.input, self.weights1))
        self.output = sigmoid(np.dot(self.layer1, self.weights2))

    def backprop(self):
        # aplicaci贸n de la regla de la cadena para encontrar la derivada de la p茅rdida con respecto a weights2 y weights1
        d_weights2 = np.dot(self.layer1.T, (2 * (self.y - self.output) * sigmoid_derivative(self.output)))
        d_weights1 = np.dot(self.input.T,  (np.dot(2 * (self.y - self.output) * sigmoid_derivative(self.output), self.weights2.T) * sigmoid_derivative(self.layer1)))

        # actualizar los pesos con las derivadas
        self.weights1 += d_weights1
        self.weights2 += d_weights2

    def train(self, epochs=10000):
        for epoch in range(epochs):
            self.feedforward()
            self.backprop()

# Datos XOR
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

nn = NeuralNetwork(X, y)
nn.train()

# Predicciones despu茅s del entrenamiento
print("Salidas despu茅s del entrenamiento:")
print(nn.output)
```

**Visualizaci贸n de la Funci贸n de P茅rdida**

Una parte esencial del entrenamiento es visualizar c贸mo cambia la funci贸n de p茅rdida a medida que la red aprende.

```python
import matplotlib.pyplot as plt

losses = []
for epoch in range(10000):
    nn.feedforward()
    loss = np.mean(np.square(y - nn.output))  # Calculo de la p茅rdida MSE
    losses.append(loss)
    nn.backprop()

plt.figure(figsize=(10, 6))  # Ajusta el tama帽o de la figura para una mejor visualizaci贸n
plt.plot(losses, label='P茅rdida MSE', color='darkred', linewidth=2, linestyle='-')
plt.title('Funci贸n de P茅rdida durante el Entrenamiento', fontsize=14, fontweight='bold')
plt.xlabel('poca', fontsize=12)
plt.ylabel('P茅rdida', fontsize=12)
plt.yscale('log')  # Escala logar铆tmica en el eje Y, opcional dependiendo de tus preferencias
plt.grid(True)  # A帽ade una cuadr铆cula para mejorar la legibilidad
plt.legend()  # A帽ade una leyenda si hay m谩s de una serie de datos
plt.show()
```

**Conclusi贸n**

El entrenamiento de redes neuronales involucra ajustar sistem谩ticamente los pesos y sesgos de la red para minimizar la diferencia entre las salidas predichas y las reales. Este proceso se realiza t铆picamente utilizando algoritmos de optimizaci贸n como el descenso del gradiente, apoyado por el algoritmo de backpropagation para calcular los gradientes necesarios. A medida que la red se entrena, deber铆amos ver una disminuci贸n en la funci贸n de p茅rdida, indicando que la red est谩 aprendiendo a clasificar correctamente las entradas proporcionadas.
