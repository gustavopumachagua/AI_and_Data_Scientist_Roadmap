## **8.1 Regla de la cadena multivariante y sus aplicaciones**

Una vez visto que el cálculo multivariante no es en realidad más complicado que el caso univariante, nos centraremos ahora en las aplicaciones de la regla de la cadena. Las redes neuronales son una de las estructuras conceptuales más populares y exitosas en el aprendizaje automático. Se construyen a partir de una red conectada de neuronas y se inspiran en la estructura de los cerebros biológicos. El comportamiento de cada neurona está influido por un conjunto de parámetros de control, cada uno de los cuales debe optimizarse para ajustarse lo mejor posible a los datos. La regla de la cadena multivariante puede utilizarse para calcular la influencia de cada parámetro de las redes y permitir su actualización durante el entrenamiento.

### **8.1.1 Regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** es una extensión de la regla de la cadena básica de cálculo diferencial, adaptada para funciones de múltiples variables. Esta regla es esencial cuando se trata de composiciones de funciones donde cada función involucra varias variables. Es fundamental en el análisis de gradientes, derivadas direccionales, y en la diferenciación de funciones compuestas más complejas.

**Definición y Principios**

La regla de la cadena multivariante permite calcular la derivada de una función compuesta $f(g(x, y), h(x, y))$ donde $(g)$ y $(h)$ son funciones de las variables $(x)$ y $(y)$, y $(f)$ es una función de dos variables, que son los resultados de $(g)$ y $(h)$, respectivamente.

**Formulación Matemática**

Si tenemos:

- Una función $z = f(u, v)$.

- Donde $u = g(x, y)$ y $v = h(x, y)$.

La derivada de $(z)$ con respecto a $(x)$ y $(y)$ utilizando la regla de la cadena sería:

$\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$

$\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$

**Ejemplo Práctico**

Supongamos que tenemos las siguientes funciones:

- $f(u, v) = u^2 + v^2$.

- $u = g(x, y) = x^2 + y^2$.

- $v = h(x, y) = e^{x+y}$.

**Derivadas Parciales de $(g)$ y $(h)$**

$\frac{\partial g}{\partial x} = 2x, \quad \frac{\partial g}{\partial y} = 2y$

$\frac{\partial h}{\partial x} = e^{x+y}, \quad \frac{\partial h}{\partial y} = e^{x+y}$

**Aplicación de la Regla de la Cadena Multivariante**

$\frac{\partial z}{\partial x} = 2u \cdot 2x + 2v \cdot e^{x+y} = 4x(x^2 + y^2) + 2e^{x+y}e^{x+y}$

$\frac{\partial z}{\partial y} = 2u \cdot 2y + 2v \cdot e^{x+y} = 4y(x^2 + y^2) + 2e^{x+y}e^{x+y}$

**Código Python y Visualización**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the functions
def f(u, v):
    return u**2 + v**2

def g(x, y):
    return x**2 + y**2

def h(x, y):
    return np.exp(x + y)

# Calculate partial derivatives using the chain rule
def partial_derivatives(x, y):
    u = g(x, y)
    v = h(x, y)
    dz_dx = 2 * u * 2 * x + 2 * v * np.exp(x + y)
    dz_dy = 2 * u * 2 * y + 2 * v * np.exp(x + y)
    return dz_dx, dz_dy

# Range of values for x and y
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
DZ_DX, DZ_DY = partial_derivatives(X, Y)

# Visualization of partial derivatives
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
surface1 = ax1.plot_surface(X, Y, DZ_DX, cmap='viridis', edgecolor='none')
ax1.set_title('Partial Derivative $\partial z / \partial x$')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('DZ_DX')
fig.colorbar(surface1, ax=ax1, shrink=0.5, aspect=10)

ax2 = fig.add_subplot(122, projection='3d')
surface2 = ax2.plot_surface(X, Y, DZ_DY, cmap='viridis', edgecolor='none')
ax2.set_title('Partial Derivative $\partial z / \partial y$')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('DZ_DY')
fig.colorbar(surface2, ax=ax2, shrink=0.5, aspect=10)

plt.show()
```

**Conclusión**

La regla de la cadena multivariante es una herramienta esencial en el cálculo diferencial para funciones de varias variables. Permite descomponer cómo las variaciones en variables subyacentes afectan a una función compuesta, facilitando análisis más profundos en campos como la física, la ingeniería, y la economía. Su comprensión y aplicación correcta son fundamentales para avanzar en el estudio y aplicación de matemáticas avanzadas en problemas reales.

### **8.1.2 Más regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** en cálculo es una extensión crucial de la regla de la cadena estándar, aplicable cuando se trabaja con funciones de varias variables. Esta técnica permite calcular la derivada de funciones compuestas que son el resultado de aplicar múltiples funciones, cada una dependiendo de varias variables. Es fundamental en áreas como la dinámica de sistemas, donde las variables de estado evolucionan con el tiempo y dependen de múltiples factores.

**Formulación General**

Supongamos que queremos calcular la derivada de una función compuesta $z = f(g(x, y), h(x, y))$ respecto a múltiples variables. Aquí, $(g)$ y $(h)$ son funciones de las variables $(x)$ y $(y)$, y $(f)$ es una función que depende de $u = g(x, y)$ y $v = h(x, y)$.

La regla de la cadena multivariante nos dice que:

$\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$

$\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$

**Interpretación**

- $\frac{\partial f}{\partial u}$ y $\frac{\partial f}{\partial v}$ son las derivadas parciales de $(f)$ respecto a sus argumentos $(u)$ y $(v)$, evaluadas en $g(x, y)$ y $h(x, y)$ respectivamente.

- $\frac{\partial g}{\partial x}$ y $\frac{\partial h}{\partial x}$, así como $\frac{\partial g}{\partial y}$ y $\frac{\partial h}{\partial y}$, son las derivadas parciales de $(g)$ y $(h)$ respecto a $(x)$ y $(y)$.

**Ejemplo Extendido con Aplicación**

Consideremos una situación concreta donde $f(u, v) = u^2 + v^2$, $g(x, y) = x^2 - y^2$, y $h(x, y) = 2xy$.

**Cálculo de Derivadas Parciales**

Para $(g)$ y $(h)$:

$\frac{\partial g}{\partial x} = 2x, \quad \frac{\partial g}{\partial y} = -2y$

$\frac{\partial h}{\partial x} = 2y, \quad \frac{\partial h}{\partial y} = 2x$

Para $(f)$:

$\frac{\partial f}{\partial u} = 2u, \quad \frac{\partial f}{\partial v} = 2v$

**Aplicación de la Regla de la Cadena**

$\frac{\partial z}{\partial x} = 2u \cdot 2x + 2v \cdot 2y = 4(x^2 - y^2)x + 4(2xy)y$

$\frac{\partial z}{\partial y} = 2u \cdot -2y + 2v \cdot 2x = -4(x^2 - y^2)y + 4(2xy)x$

**Código Python para Visualización**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(u, v):
    return u**2 + v**2

def g(x, y):
    return x**2 - y**2

def h(x, y):
    return 2*x*y

def partials(x, y):
    u = g(x, y)
    v = h(x, y)
    df_dx = 2 * u * 2 * x + 2 * v * 2 * y
    df_dy = 2 * u * (-2 * y) + 2 * v * 2 * x
    return df_dx, df_dy

# Generate data
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
DF_DX, DF_DY = partials(X, Y)

# Plotting the partial derivatives
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Contour plot for ∂z/∂x
cp = ax1.contourf(X, Y, DF_DX, levels=np.linspace(DF_DX.min(), DF_DX.max(), 100), cmap='viridis')
fig.colorbar(cp, ax=ax1, label='Value of $\partial z / \partial x$')
ax1.set_title('$\partial z / \partial x$', fontsize=14)
ax1.set_xlabel('X', fontsize=12)
ax1.set_ylabel('Y', fontsize=12)
ax1.set_aspect('equal')  # Set equal scaling by changing aspect ratio of the x and y axes

# Contour plot for ∂z/∂y
cp2 = ax2.contourf(X, Y, DF_DY, levels=np.linspace(DF_DY.min(), DF_DY.max(), 100), cmap='viridis')
fig.colorbar(cp2, ax=ax2, label='Value of $\partial z / \partial y$')
ax2.set_title('$\partial z / \partial y$', fontsize=14)
ax2.set_xlabel('X', fontsize=12)
ax2.set_ylabel('Y', fontsize=12)
ax2.set_aspect('equal')  # Consistent aspect ratio

# Enhancements
plt.subplots_adjust(wspace=0.3)  # Adjust the spacing between subplots
plt.suptitle('Visualization of Partial Derivatives', fontsize=16)  # Overall title for all subplots
plt.show()
```

**Conclusión**

La regla de la cadena multivariante es una herramienta poderosa en cálculo diferencial, permitiendo descomponer cómo las variaciones en múltiples variables subyacentes afectan a una función compuesta. Su comprensión y aplicación son esenciales para el estudio avanzado en muchas disciplinas, incluyendo física, ingeniería, economía y biología.

### **8.1.3 Redes neuronales simples**

Las **redes neuronales simples** son modelos computacionales inspirados en el cerebro humano que se utilizan para aprender y tomar decisiones a partir de datos. A menudo, una red neuronal simple se refiere a un modelo de perceptrón o a redes neuronales de una sola capa, que son los bloques de construcción básicos para redes más complejas. Estos modelos son capaces de realizar tareas de clasificación y regresión basándose en la entrada que reciben y los pesos que aprenden durante el entrenamiento.

**Conceptos Básicos de Redes Neuronales**

**Neurona Artificial**

La unidad básica de una red neuronal es la **neurona artificial**, que recibe varias entradas, las combina linealmente usando pesos, añade un sesgo, y luego aplica una función de activación para producir una salida. La estructura básica de una neurona se puede describir con la siguiente fórmula matemática:

$y = f\left(\sum_{i=1}^n w_i x_i + b\right)$

Donde:

- $(x_i)$ son las entradas.

- $(w_i)$ son los pesos.

- $(b)$ es el sesgo.

- $(f)$ es la función de activación, como la función sigmoide, ReLU, etc.,

- $(y)$ es la salida.

**Funciones de Activación**

Las funciones de activación más comunes incluyen:

- **Sigmoide**: $f(x) = \frac{1}{1 + e^{-x}}$, útil para clasificación binaria.

- **ReLU (Rectified Linear Unit)**: $f(x) = \max(0, x)$, común en redes profundas por su eficiencia.

- **Tanh**: $f(x) = \tanh(x)$, que es una versión escalada de la función sigmoide.

**Ejemplo de una Red Neuronal Simple: El Perceptrón**

El perceptrón es la forma más simple de una red neuronal, que consiste en una sola neurona con aprendizaje supervisado. Es capaz de clasificar dos clases linealmente separables.

**Configuración del Problema**

Supongamos que queremos clasificar puntos de datos que pertenecen a dos clases usando un perceptrón con una función de activación sigmoide. Los puntos de datos son $(x_1, y_1)$ con etiquetas de clase.

**Código Python para un Perceptrón Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Perceptron model
class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 0 else 0 for i in y])

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = sigmoid(linear_output)

                update = self.lr * (y_[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_output)
        return np.array([1 if i > 0.5 else 0 for i in y_predicted])

# Example data
X = np.array([[2, 3], [1, 1], [5, 6], [2, 1]])
y = np.array([0, 0, 1, 1])

# Train perceptron
p = Perceptron(learning_rate=0.1, n_iters=1000)
p.fit(X, y)

# Predict new data
predictions = p.predict(X)

# Visualization of data and decision boundary
fig, ax = plt.subplots()
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', edgecolor='k', s=100)
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

# Calculate decision boundary (x2 = -(b + w1*x1)/w2)
x1 = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, 100)
x2 = -(p.bias + p.weights[0] * x1) / p.weights[1]

ax.plot(x1, x2, color='red', linestyle='--', label='Decision Boundary')
ax.set_title('Perceptron Classification')
ax.set_xlabel('Feature 1 (X1)')
ax.set_ylabel('Feature 2 (X2)')
ax.legend()

plt.show()
```

**Conclusión**

Las redes neuronales simples, como el perceptrón, proporcionan una introducción fundamental a los conceptos más profundos de las redes neuronales y el aprendizaje automático. Aunque los perceptrones tienen limitaciones, como su incapacidad para resolver problemas no linealmente separables, son la base sobre la que se construyen arquitecturas más complejas como las redes neuronales profundas (deep learning). Estas herramientas básicas son esenciales para entender cómo las máquinas pueden aprender a partir de datos y tomar decisiones.

### **8.1.4 Redes neuronales más sencillas**

Las **redes neuronales más sencillas**, generalmente conocidas como redes de una sola capa o perceptrones, son modelos básicos de aprendizaje automático diseñados para la clasificación y regresión lineal. Estos modelos son fundamentales para entender cómo las redes neuronales más complejas, como las redes neuronales profundas, funcionan y se construyen.

**Conceptos Básicos**

Una red neuronal simple generalmente consiste en una única "neurona" que procesa múltiples entradas para producir una salida. La estructura básica de una neurona incluye:

- **Entradas (Inputs)**: Valores numéricos que representan características del dato.

- **Pesos (Weights)**: Coeficientes que se aprenden durante el entrenamiento, que multiplican a las entradas.

- **Sesgo (Bias)**: Un término adicional que permite ajustar la salida de la neurona además de los pesos aplicados a las entradas.

- **Función de Activación**: Una función matemática que transforma la suma ponderada de las entradas y el sesgo en la salida que la red proporciona.

**Fórmula de la Neurona**

La salida de una neurona, $(y)$, se calcula como:

$y = f\left(\sum_{i=1}^n w_i x_i + b\right)$

Donde:

- $(x_i)$ son las entradas.

- $(w_i)$ son los pesos.

- $(b)$ es el sesgo.

- $(f)$ es la función de activación, como sigmoide, tanh, o ReLU,
- $(n)$ es el número de entradas.

**Ejemplo Práctico: Perceptrón para Clasificación Binaria**

Vamos a considerar un ejemplo donde utilizamos un perceptrón para clasificar datos que pertenecen a dos categorías distintas basadas en dos características.

**Datos de Ejemplo**

Imagina que tenemos un conjunto de datos con dos características, donde cada punto pertenece a una de dos clases (0 o 1). El objetivo es aprender una línea que pueda separar estas dos clases.

**Función de Activación**

Usaremos la función sigmoide, que es común en las neuronas de clasificación binaria:

$\sigma(z) = \frac{1}{1 + e^{-z}}$

**Código Python para un Perceptrón Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Perceptron class
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.lr = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Train the model
        for _ in range(self.n_iterations):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = sigmoid(linear_output)
                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_output)
        return [1 if i > 0.5 else 0 for i in y_predicted]

# Sample data
X = np.array([[0.1, 1.1], [2.2, 2.9], [1.6, 1.8], [3.0, 3.1], [1.2, 0.9], [3.1, 3.3]])
y = np.array([0, 1, 0, 1, 0, 1])

# Train the model
p = Perceptron(learning_rate=0.1)
p.fit(X, y)

# Visualization of data and decision boundary
fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', edgecolor='k', s=50)
legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)  # Correct use of add_artist
points = np.linspace(min(X[:, 0]) - 0.1, max(X[:, 0]) + 0.1, 100)
decision_boundary = -(p.weights[0] * points + p.bias) / p.weights[1]
ax.plot(points, decision_boundary, color='red', label='Decision Boundary')
ax.set_title('Perceptron Classification')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.legend()
plt.show()
```

**Conclusión**

Las redes neuronales simples como el perceptrón son herramientas poderosas para aprender relaciones lineales entre características y clasificaciones. Aunque limitadas a problemas linealmente separables y a clasificación binaria, estas redes forman la base para entender y construir modelos de redes neuronales más complejos. A medida que se añaden capas y se utilizan diferentes funciones de activación, estos modelos pueden aprender patrones mucho más complejos y resolver problemas de clasificación y regresión de alta dimensionalidad.

### **8.1.5 Ejercicio de regla de la cadena multivariante**

La **Regla de la Cadena Multivariante** en cálculo diferencial es una herramienta esencial para tratar con funciones compuestas que dependen de varias variables. Esta regla permite calcular las derivadas de estas funciones compuestas de manera sistemática y eficaz. A continuación, proporcionaré una explicación detallada con ejemplos, fórmulas matemáticas y un código Python para ilustrar cómo aplicar esta regla en un ejercicio práctico.

**Ejemplo de Regla de la Cadena Multivariante**

Consideremos el siguiente problema: queremos calcular las derivadas parciales de la función $(z)$ compuesta por $z = f(u, v)$ donde $u = g(x, y)$ y $v = h(x, y)$. Supongamos que:

- $f(u, v) = u^2 + 3uv + v^2$.

- $g(x, y) = x^2 + y^2$.

- $h(x, y) = e^{x+y}$.

**Objetivo**

Calcular $\frac{\partial z}{\partial x}$ y $\frac{\partial z}{\partial y}$.

**Paso a Paso de la Regla de la Cadena**

**Determinar las Derivadas Parciales de $(g)$ y $(h)$:**

   - $\frac{\partial g}{\partial x} = 2x$.

   - $\frac{\partial g}{\partial y} = 2y$.

   - $\frac{\partial h}{\partial x} = e^{x+y}$.

   - $\frac{\partial h}{\partial y} = e^{x+y}$.

**Determinar las Derivadas Parciales de $(f)$ respecto a $(u)$ y $(v)$:**

   - $\frac{\partial f}{\partial u} = 2u + 3v$.

   - $\frac{\partial f}{\partial v} = 3u + 2v$.

**Aplicar la Regla de la Cadena:**

   - $\frac{\partial z}{\partial x} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial x} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial x}$.

   - $\frac{\partial z}{\partial y} = \frac{\partial f}{\partial u} \frac{\partial g}{\partial y} + \frac{\partial f}{\partial v} \frac{\partial h}{\partial y}$.

**Código Python para Visualización y Cálculo**

```python
import numpy as np

# Definir las funciones g, h, f y sus derivadas parciales
def g(x, y):
    return x**2 + y**2

def h(x, y):
    return np.exp(x + y)

def f(u, v):
    return u**2 + 3*u*v + v**2

def df_du(u, v):
    return 2*u + 3*v

def df_dv(u, v):
    return 3*u + 2*v

# Calcular derivadas parciales usando la regla de la cadena
def calculate_partials(x, y):
    u = g(x, y)
    v = h(x, y)
    partial_x = df_du(u, v) * 2*x + df_dv(u, v) * np.exp(x + y)
    partial_y = df_du(u, v) * 2*y + df_dv(u, v) * np.exp(x + y)
    return partial_x, partial_y

# Ejemplo de punto para cálculo
x, y = 1, 1
partial_x, partial_y = calculate_partials(x, y)
print(f"∂z/∂x at (x={x}, y={y}) is {partial_x}")
print(f"∂z/∂y at (x={x}, y={y}) is {partial_y}")
```

Este script define las funciones $g(x, y)$, $h(x, y)$ y $f(u, v)$, junto con sus derivadas parciales necesarias. Luego calcula las derivadas parciales de $(z)$ utilizando la regla de la cadena para un punto específico $(x, y) = (1, 1)$.

**Conclusión**

La regla de la cadena multivariante es un principio fundamental en cálculo diferencial que permite el manejo de funciones compuestas en múltiples dimensiones. A través de su aplicación, podemos analizar cómo las variaciones en las variables independientes afectan a una función compuesta en términos de sus derivadas parciales. Este enfoque es invaluable en áreas tales como la física, ingeniería, economía y ciencias de la computación, donde las funciones complejas a menudo modelan fenómenos interdependientes.

### **8.1.6 Redes neuronales artificiales simples**

Las **Redes Neuronales Artificiales Simples** (ANN, por sus siglas en inglés) son modelos computacionales que se inspiran en las redes neuronales biológicas del cerebro humano. Estas redes son estructuras compuestas por nodos, llamados "neuronas artificiales", que están conectadas entre sí y transmiten señales de una a otra. Las redes neuronales simples generalmente se refieren a modelos con pocas capas, como el perceptrón o redes con una sola capa oculta, que pueden realizar tareas de clasificación y regresión.

**Componentes Básicos de una Red Neuronal Simple**

Una red neuronal típica incluye:

- **Neuronas de Entrada**: Donde cada neurona representa una característica del conjunto de datos.

- **Capas Ocultas**: Capas intermedias entre las entradas y salidas, que pueden tener una o más neuronas. En redes simples, frecuentemente hay solo una capa oculta.

- **Neuronas de Salida**: Representan las predicciones o clasificaciones del modelo.

- **Pesos y Sesgos**: Cada conexión entre neuronas tiene un peso y cada neurona tiene un sesgo, que son ajustados durante el entrenamiento.

- **Funciones de Activación**: Como sigmoide, tanh, o ReLU, que ayudan a introducir no linealidades en el modelo.

**Ejemplo Práctico: Red Neuronal con una Capa Oculta**

Consideremos un ejemplo sencillo donde construimos una red neuronal con una capa oculta para clasificar datos que son linealmente no separables.

**Configuración del Problema**

Imagina que tenemos un conjunto de datos de puntos que pertenecen a dos clases, y queremos usar una red neuronal para clasificar estos puntos.

**Función de Activación**

Utilizaremos la función de activación ReLU para las neuronas de la capa oculta y la función sigmoide para la salida, adecuada para la clasificación binaria.

**Código Python para una Red Neuronal Simple**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

# Visualizar la función de activación
x_vals = np.linspace(-10, 10, 100)
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura
plt.plot(x_vals, sigmoid(x_vals), label='Sigmoide', color='blue', linewidth=2, linestyle='--')
plt.plot(x_vals, relu(x_vals), label='ReLU', color='red', linewidth=2, linestyle='-')

plt.title('Funciones de Activación', fontsize=14, fontweight='bold')
plt.xlabel('X', fontsize=12)
plt.ylabel('f(X)', fontsize=12)
plt.grid(True)  # Añadir cuadrícula
plt.legend()

# Puedes también marcar puntos específicos, como por ejemplo el punto donde ReLU cambia de 0 a x
plt.scatter([0], [0], color='black', zorder=5)  # Marca el origen para ReLU
plt.annotate('Origen (0,0)', (0, 1), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

**Explicación del Código**

- **Inicialización**: Se inicializan aleatoriamente los pesos y se establecen los sesgos a cero.

- **Feedforward**: Se calcula la salida de la red mediante la propagación de las entradas a través de las capas usando las funciones de activación.

**Conclusión**

Las redes neuronales artificiales simples son herramientas poderosas para modelar relaciones no lineales y patrones complejos en datos, aunque su simplicidad puede limitar su capacidad para manejar tareas más complejas que requieren una representación más rica, como puede ofrecer una red profunda. Estos modelos son un excelente punto de partida para entender los fundamentos del aprendizaje profundo y sus aplicaciones en la clasificación automática y otras tareas de predicción.

### **8.1.7 Entrenamiento de redes neuronales**

El **entrenamiento de redes neuronales** es el proceso mediante el cual una red neuronal aprende a realizar tareas específicas, como la clasificación o la regresión, a partir de datos de entrenamiento. Este proceso involucra ajustar los pesos y sesgos de la red de tal manera que el error entre las salidas predichas y las salidas reales deseadas se minimice. Vamos a explorar este proceso detalladamente con ejemplos, fórmulas y un script de Python para ilustrar el entrenamiento de una red neuronal simple.

**Componentes Clave del Entrenamiento**

**Datos de Entrenamiento**: Conjunto de entradas y salidas esperadas que la red usará para aprender.

**Modelo de Red Neuronal**: Estructura de la red, incluyendo número y tipo de capas, número de neuronas por capa, funciones de activación, etc.

**Función de Pérdida (Loss Function)**: Mide el error entre las salidas predichas y las reales. Ejemplos comunes incluyen el Error Cuadrático Medio (MSE) para regresión y la Entropía Cruzada para clasificación.

**Optimizador**: Método usado para ajustar los pesos y sesgos en función del gradiente de la función de pérdida. Ejemplos comunes son el Descenso del Gradiente Estocástico (SGD), Adam y RMSprop.

**Backpropagation**: Algoritmo utilizado para calcular los gradientes de la función de pérdida respecto a cada peso y sesgo en la red.

**Ejemplo de Entrenamiento: Red Neuronal para XOR**

Supongamos que queremos entrenar una red neuronal para resolver el problema XOR, que no es linealmente separable y requiere al menos una capa oculta.

**Configuración del Problema**

- **Datos XOR**: Entradas $[0, 0], [0, 1], [1, 0], [1, 1]$ y Salidas $[0], [1], [1], [0]$.

**Red Neuronal**

- **Arquitectura**: 2 neuronas en la capa de entrada, 2 neuronas en una capa oculta, 1 neurona en la capa de salida.

- **Función de Activación**: Sigmoide para todas las neuronas.

**Código Python para el Entrenamiento**

```python
import numpy as np

# Funciones de activación y derivadas
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Entrenamiento de la red neuronal
class NeuralNetwork:
    def __init__(self, x, y):
        self.input      = x
        self.weights1   = np.random.rand(self.input.shape[1], 2)  # pesos capa oculta
        self.weights2   = np.random.rand(2, 1)  # pesos capa salida
        self.y          = y
        self.output     = np.zeros(self.y.shape)

    def feedforward(self):
        self.layer1 = sigmoid(np.dot(self.input, self.weights1))
        self.output = sigmoid(np.dot(self.layer1, self.weights2))

    def backprop(self):
        # aplicación de la regla de la cadena para encontrar la derivada de la pérdida con respecto a weights2 y weights1
        d_weights2 = np.dot(self.layer1.T, (2 * (self.y - self.output) * sigmoid_derivative(self.output)))
        d_weights1 = np.dot(self.input.T,  (np.dot(2 * (self.y - self.output) * sigmoid_derivative(self.output), self.weights2.T) * sigmoid_derivative(self.layer1)))

        # actualizar los pesos con las derivadas
        self.weights1 += d_weights1
        self.weights2 += d_weights2

    def train(self, epochs=10000):
        for epoch in range(epochs):
            self.feedforward()
            self.backprop()

# Datos XOR
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

nn = NeuralNetwork(X, y)
nn.train()

# Predicciones después del entrenamiento
print("Salidas después del entrenamiento:")
print(nn.output)
```

**Visualización de la Función de Pérdida**

Una parte esencial del entrenamiento es visualizar cómo cambia la función de pérdida a medida que la red aprende.

```python
import matplotlib.pyplot as plt

losses = []
for epoch in range(10000):
    nn.feedforward()
    loss = np.mean(np.square(y - nn.output))  # Calculo de la pérdida MSE
    losses.append(loss)
    nn.backprop()

plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura para una mejor visualización
plt.plot(losses, label='Pérdida MSE', color='darkred', linewidth=2, linestyle='-')
plt.title('Función de Pérdida durante el Entrenamiento', fontsize=14, fontweight='bold')
plt.xlabel('Época', fontsize=12)
plt.ylabel('Pérdida', fontsize=12)
plt.yscale('log')  # Escala logarítmica en el eje Y, opcional dependiendo de tus preferencias
plt.grid(True)  # Añade una cuadrícula para mejorar la legibilidad
plt.legend()  # Añade una leyenda si hay más de una serie de datos
plt.show()
```

**Conclusión**

El entrenamiento de redes neuronales involucra ajustar sistemáticamente los pesos y sesgos de la red para minimizar la diferencia entre las salidas predichas y las reales. Este proceso se realiza típicamente utilizando algoritmos de optimización como el descenso del gradiente, apoyado por el algoritmo de backpropagation para calcular los gradientes necesarios. A medida que la red se entrena, deberíamos ver una disminución en la función de pérdida, indicando que la red está aprendiendo a clasificar correctamente las entradas proporcionadas.
