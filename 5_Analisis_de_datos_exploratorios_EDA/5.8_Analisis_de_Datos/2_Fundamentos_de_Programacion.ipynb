# **2. Fundamentos de Programación y Manipulación de Datos**

##  **Programación en Python para Análisis de Datos**

### **Sintaxis básica**

Python es un lenguaje de programación versátil y potente que se utiliza ampliamente en análisis de datos. Las bibliotecas como Pandas, NumPy y Matplotlib proporcionan herramientas esenciales para manipular datos y crear visualizaciones. A continuación, se presenta una guía paso a paso con ejemplos y gráficos personalizados en Python.

1. **Instalación de Bibliotecas:**

Antes de comenzar, asegúrate de tener instaladas las bibliotecas necesarias. Puedes instalarlas utilizando el siguiente comando:

```bash
pip install pandas numpy matplotlib
```

2. **Importación de Bibliotecas:**

Importa las bibliotecas necesarias en tu script de Python:

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

3. **Carga de Datos:**

Lee tus datos en un DataFrame de Pandas. Para este ejemplo, utilizaremos un conjunto de datos ficticio:

````python
# Crear un DataFrame de ejemplo
datos = {
    'Fecha': pd.date_range('2022-01-01', '2022-12-31'),
    'Ventas': np.random.randint(100, 1000, size=(365,))
}

df = pd.DataFrame(datos)
````

4. **Exploración de Datos:**

Realiza una exploración inicial de tus datos:

````python
# Mostrar las primeras filas del DataFrame
print(df.head())

# Obtener estadísticas descriptivas
print(df.describe())
````

5. **Visualización de Datos:**

Crea gráficos personalizados para visualizar tus datos:

````python
# Gráfico de línea para las ventas a lo largo del tiempo
plt.figure(figsize=(10, 5))
plt.plot(df['Fecha'], df['Ventas'], label='Ventas')
plt.title('Ventas a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.show()
````

6. **Análisis Exploratorio de Datos (EDA):**

Realiza un análisis más profundo utilizando herramientas como gráficos de dispersión y correlación:

````python
# Crear una columna 'OtraVariable' con datos de ejemplo
df['OtraVariable'] = np.random.randint(1, 100, size=len(df))

# Gráfico de dispersión para analizar la relación entre dos variables
plt.scatter(df['Ventas'], df['OtraVariable'])
plt.title('Relación entre Ventas y OtraVariable')
plt.xlabel('Ventas')
plt.ylabel('OtraVariable')
plt.show()

# Matriz de correlación
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
````

7. **Agregación y Agrupación:**

Agrega y agrupa datos según necesites:

````python
ventas_por_mes = df.groupby(df['Fecha'].dt.month)['Ventas'].sum()

# Gráfico de barras para las ventas por mes
ventas_por_mes.plot(kind='bar')
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
````

8. **Transformación de Datos:**

Realiza transformaciones en tus datos según sea necesario:

````python
# Crear una nueva columna calculada
df['NuevaColumna'] = df['Ventas'] * 2

# Filtrar datos
datos_filtrados = df[df['Ventas'] > 500]
````

````python
datos_filtrados
````

9. **Exportación de Datos:**

Guarda tus datos transformados si es necesario:

````python
df.to_csv('../../Data/datos_transformados.csv', index=False)
````

### **Estructuras de datos en Python**

Las estructuras de datos en Python son formas de organizar y almacenar datos para que puedan ser utilizados y manipulados de manera eficiente. Algunas de las estructuras de datos más comunes en Python son listas, tuplas, conjuntos, diccionarios y arrays de NumPy. A continuación, se proporcionará una explicación detallada de cada una, junto con ejemplos y gráficos personalizados en código Python.

**1. Listas:**

Las listas son colecciones ordenadas y modificables de elementos. Pueden contener cualquier tipo de dato, incluso otras listas.

````python
# Ejemplo de lista
mi_lista = [1, 2, 3, 4, 5]

# Acceder a elementos de la lista
print(mi_lista[0])  # Salida: 1

# Modificar elementos de la lista
mi_lista[1] = 10

# Agregar elementos al final de la lista
mi_lista.append(6)

# Gráfico personalizado de una lista
plt.plot(mi_lista, marker='o')
plt.title('Gráfico de una Lista')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
````

**2. Tuplas:**

Las tuplas son colecciones ordenadas e inmutables de elementos. A diferencia de las listas, no se pueden modificar después de la creación.

````python
# Ejemplo de tupla
mi_tupla = (1, 2, 3, 4, 5)

# Acceder a elementos de la tupla
print(mi_tupla[0])  # Salida: 1

# No se pueden modificar
# mi_tupla[1] = 10  # Esto generará un error

# Gráfico personalizado de una tupla
plt.plot(mi_tupla, marker='o')
plt.title('Gráfico de una Tupla')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
````

**3. Conjuntos:**

Los conjuntos son colecciones desordenadas de elementos únicos.

````python
# Ejemplo de conjunto
mi_conjunto = {1, 2, 3, 4, 5}

# No hay índices, no se puede acceder a elementos por posición

# Agregar elementos al conjunto
mi_conjunto.add(6)

# Gráfico personalizado de un conjunto (no es común, ya que no tiene orden)
# Se puede hacer un gráfico de Venn si se tienen varios conjuntos
````

````python
mi_conjunto
````

**4. Diccionarios:**

Los diccionarios son colecciones de pares clave-valor.

````python
# Ejemplo de diccionario
mi_diccionario = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4, 'cinco': 5}

# Acceder a elementos por clave
print(mi_diccionario['dos'])  # Salida: 2

# Modificar valores por clave
mi_diccionario['dos'] = 10

# Agregar nuevos pares clave-valor
mi_diccionario['seis'] = 6

# Gráfico personalizado de un diccionario (pueden no tener un orden específico)
plt.bar(mi_diccionario.keys(), mi_diccionario.values())
plt.title('Gráfico de un Diccionario')
plt.xlabel('Clave')
plt.ylabel('Valor')
plt.show()
````

**5. Arrays de NumPy:**

Los arrays de NumPy son estructuras eficientes para manipular datos numéricos.

````python
import numpy as np

# Ejemplo de array de NumPy
mi_array = np.array([1, 2, 3, 4, 5])

# Operaciones vectorizadas son posibles
mi_array = mi_array * 2

# Gráfico personalizado de un array de NumPy
plt.plot(mi_array, marker='o')
plt.title('Gráfico de un Array de NumPy')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
````

Estos son solo ejemplos básicos para cada estructura de datos. Puedes adaptarlos según tus necesidades específicas. Recuerda que la elección de la estructura de datos depende del contexto y del tipo de operaciones que planeas realizar con tus datos.

### **Introducción a la biblioteca Pandas para manipulación y análisis de datos**

Pandas es una biblioteca de Python que proporciona estructuras de datos de alto rendimiento y herramientas de análisis de datos. Sus dos estructuras de datos principales son `Series` (para datos unidimensionales) y `DataFrame` (para datos bidimensionales).
````python
import pandas as pd
````

3. **Creación de Series y DataFrames:**

Crea una Serie y un DataFrame básicos:

````python
# Crear una Serie
serie = pd.Series([1, 3, 5, np.nan, 6, 8])

# Crear un DataFrame con fechas y valores aleatorios
fechas = pd.date_range('2022-01-01', periods=6)
df = pd.DataFrame(np.random.randn(6, 4), index=fechas, columns=list('ABCD'))
````

Realiza operaciones básicas de manipulación de datos:

````python
# Seleccionar una columna por nombre
columna_A = df['A']

# Seleccionar filas por índice
filas_2_4 = df[1:4]

# Filtrar datos basado en condiciones
df_filtrado = df[df['A'] > 0]
````

Pandas proporciona métodos de visualización integrados:

Realiza operaciones de agrupación y agregación:

### **Trabajo con DataFrames y Series**

En Pandas, las `Series` y los `DataFrames` son las estructuras de datos fundamentales. Una `Serie` es un objeto unidimensional similar a un array, mientras que un `DataFrame` es una estructura bidimensional que se puede considerar como una tabla de datos. A continuación, se proporciona una explicación detallada con ejemplos y gráficos personalizados en código Python.

1. **Creación de Series y DataFrames:**

**Creación de Series:**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear una Serie desde una lista
serie = pd.Series([1, 3, 5, np.nan, 6, 8], name='MiSerie')
print(serie)
````

**Creación de DataFrames:**

````python
# Crear un DataFrame desde un array numpy y fechas
fechas = pd.date_range('2022-01-01', periods=6)
data = np.random.randn(6, 4)
columnas = list('ABCD')

df = pd.DataFrame(data, index=fechas, columns=columnas)
print(df)
````

2. **Acceso y Manipulación de Datos:**

**Acceso a Datos:**

````python
# Acceder a una columna por nombre
columna_A = df['A']

# Acceder a varias columnas por nombre
columnas_AB = df[['A', 'B']]

# Acceder a una fila por índice
fila_2 = df.loc['2022-01-02']

# Acceder a un elemento específico
elemento = df.at['2022-01-02', 'A']
````

````python
columna_A
````

````python
columnas_AB
````

````python
fila_2
````

````python
elemento
````

**Manipulación de Datos:**


````python
# Agregar una nueva columna
df['E'] = np.random.randn(6)

# Operaciones aritméticas entre columnas
df['F'] = df['A'] + df['B']

# Aplicar una función a una columna
df['G'] = df['C'].apply(lambda x: 'Mayor' if x > 0 else 'Menor')
````

Pandas proporciona métodos de visualización integrados que utilizan Matplotlib.

````python
# Gráfico de líneas
df.plot(kind='line', title='Gráfico de Líneas')
plt.show()

# Gráfico de barras
df.plot(kind='bar', title='Gráfico de Barras')
plt.show()

# Gráfico de dispersión
df.plot(kind='scatter', x='A', y='B', title='Gráfico de Dispersión')
plt.show()
````
