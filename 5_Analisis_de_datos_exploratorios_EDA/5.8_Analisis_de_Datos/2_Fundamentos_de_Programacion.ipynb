6. **An谩lisis Exploratorio de Datos (EDA):**

### **Estructuras de datos en Python**

Las estructuras de datos en Python son formas de organizar y almacenar datos para que puedan ser utilizados y manipulados de manera eficiente. Algunas de las estructuras de datos m谩s comunes en Python son listas, tuplas, conjuntos, diccionarios y arrays de NumPy. A continuaci贸n, se proporcionar谩 una explicaci贸n detallada de cada una, junto con ejemplos y gr谩ficos personalizados en c贸digo Python.

**1. Listas:**

Las listas son colecciones ordenadas y modificables de elementos. Pueden contener cualquier tipo de dato, incluso otras listas.

````python
# Ejemplo de lista
mi_lista = [1, 2, 3, 4, 5]

# Acceder a elementos de la lista
print(mi_lista[0])  # Salida: 1

# Modificar elementos de la lista
mi_lista[1] = 10

# Agregar elementos al final de la lista
mi_lista.append(6)

# Gr谩fico personalizado de una lista
plt.plot(mi_lista, marker='o')
plt.title('Gr谩fico de una Lista')
plt.xlabel('ndice')
plt.ylabel('Valor')
plt.show()
````

**2. Tuplas:**

Las tuplas son colecciones ordenadas e inmutables de elementos. A diferencia de las listas, no se pueden modificar despu茅s de la creaci贸n.

````python
# Ejemplo de tupla
mi_tupla = (1, 2, 3, 4, 5)

# Acceder a elementos de la tupla
print(mi_tupla[0])  # Salida: 1

# No se pueden modificar
# mi_tupla[1] = 10  # Esto generar谩 un error

# Gr谩fico personalizado de una tupla
plt.plot(mi_tupla, marker='o')
plt.title('Gr谩fico de una Tupla')
plt.xlabel('ndice')
plt.ylabel('Valor')
plt.show()
````

**3. Conjuntos:**

Los conjuntos son colecciones desordenadas de elementos 煤nicos.

````python
# Ejemplo de conjunto
mi_conjunto = {1, 2, 3, 4, 5}

# No hay 铆ndices, no se puede acceder a elementos por posici贸n

# Agregar elementos al conjunto
mi_conjunto.add(6)

# Gr谩fico personalizado de un conjunto (no es com煤n, ya que no tiene orden)
# Se puede hacer un gr谩fico de Venn si se tienen varios conjuntos
````

````python
mi_conjunto
````

**4. Diccionarios:**

Los diccionarios son colecciones de pares clave-valor.

````python
# Ejemplo de diccionario
mi_diccionario = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4, 'cinco': 5}

# Acceder a elementos por clave
print(mi_diccionario['dos'])  # Salida: 2

# Modificar valores por clave
mi_diccionario['dos'] = 10

# Agregar nuevos pares clave-valor
mi_diccionario['seis'] = 6

# Gr谩fico personalizado de un diccionario (pueden no tener un orden espec铆fico)
plt.bar(mi_diccionario.keys(), mi_diccionario.values())
plt.title('Gr谩fico de un Diccionario')
plt.xlabel('Clave')
plt.ylabel('Valor')
plt.show()
````

**5. Arrays de NumPy:**

Los arrays de NumPy son estructuras eficientes para manipular datos num茅ricos.

````python
import numpy as np

# Ejemplo de array de NumPy
mi_array = np.array([1, 2, 3, 4, 5])

# Operaciones vectorizadas son posibles
mi_array = mi_array * 2

# Gr谩fico personalizado de un array de NumPy
plt.plot(mi_array, marker='o')
plt.title('Gr谩fico de un Array de NumPy')
plt.xlabel('ndice')
plt.ylabel('Valor')
plt.show()
````

Estos son solo ejemplos b谩sicos para cada estructura de datos. Puedes adaptarlos seg煤n tus necesidades espec铆ficas. Recuerda que la elecci贸n de la estructura de datos depende del contexto y del tipo de operaciones que planeas realizar con tus datos.
