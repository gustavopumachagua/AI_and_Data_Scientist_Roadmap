# **3. INTRODUCCI√ìN A LA ESTAD√çSTICA DESCRIPTIVA Y MACHINE LEARNING**

La introducci√≥n a la estad√≠stica descriptiva y el aprendizaje autom√°tico (machine learning) es un tema amplio, pero puedo darte una explicaci√≥n general y mostrarte un ejemplo b√°sico en Python para ilustrar algunos conceptos clave.

* **Estad√≠stica descriptiva:**

La estad√≠stica descriptiva es una rama de la estad√≠stica que se centra en la recopilaci√≥n, organizaci√≥n, resumen y presentaci√≥n de datos. Su objetivo principal es describir y resumir las caracter√≠sticas importantes de un conjunto de datos sin hacer inferencias o conclusiones m√°s all√° de los datos observados.

**Ejemplo en Python:**

Supongamos que tenemos una lista de alturas de personas y queremos calcular algunas medidas descriptivas b√°sicas:

```python
import numpy as np

# Datos de ejemplo
alturas = [165, 170, 155, 180, 160, 175, 150, 185, 168, 172]

# Calcular medidas descriptivas
media = np.mean(alturas)
mediana = np.median(alturas)
desviacion_estandar = np.std(alturas)

print("Media:", media)
print("Mediana:", mediana)
print("Desviaci√≥n est√°ndar:", desviacion_estandar)
```

En este ejemplo, calculamos la media (promedio), la mediana y la desviaci√≥n est√°ndar de las alturas de las personas. Estas medidas descriptivas nos brindan informaci√≥n sobre el valor t√≠pico, la dispersi√≥n y la variabilidad de los datos.

* **Aprendizaje autom√°tico (Machine Learning):**

El aprendizaje autom√°tico es un subcampo de la inteligencia artificial que se enfoca en desarrollar algoritmos y modelos que permiten a las m√°quinas aprender y tomar decisiones basadas en datos sin ser programadas expl√≠citamente. El objetivo principal es construir modelos predictivos o descriptivos a partir de datos de entrenamiento.

**Ejemplo en Python:**

Consideremos un problema de clasificaci√≥n en el que queremos construir un modelo para predecir si una flor es de tipo `Iris-setosa` o `Iris-versicolor` en funci√≥n de sus longitudes y anchos de p√©talos. Utilizaremos el conjunto de datos `Iris`, ampliamente utilizado en el aprendizaje autom√°tico.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data[:, 2:]  # Longitud y ancho de los p√©talos
y = iris.target

# Entrenar un modelo de regresi√≥n log√≠stica
model = LogisticRegression()
model.fit(X, y)

# Graficar los puntos de datos y las regiones de decisi√≥n del modelo
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Personalizar la gr√°fica
plt.figure(figsize=(10, 8))

# Graficar las regiones de decisi√≥n
plt.contourf(xx, yy, Z, alpha=0.8, cmap='viridis')

# Graficar los puntos de datos
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='viridis', s=100)

# Agregar leyenda
legend_labels = ['Setosa', 'Versicolor', 'Virginica']
plt.legend(handles=scatter.legend_elements()[0], labels=legend_labels, title='Clases')

# Ajustar etiquetas y t√≠tulo
plt.xlabel('Longitud del p√©talo')
plt.ylabel('Ancho del p√©talo')
plt.title('Clasificaci√≥n de flores Iris')

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, utilizamos el algoritmo de regresi√≥n log√≠stica para construir un modelo de clasificaci√≥n que puede predecir la especie de una flor `Iris` en funci√≥n de sus caracter√≠sticas de p√©talos. Graficamos los puntos de datos y las regiones de decisi√≥n del modelo para visualizar c√≥mo se realiza la clasificaci√≥n.

Esto es solo una introducci√≥n b√°sica a la estad√≠stica descriptiva y el aprendizaje autom√°tico. Ambos campos tienen muchos conceptos y t√©cnicas m√°s avanzados que se pueden explorar, pero espero que este ejemplo te brinde una idea inicial de c√≥mo se pueden aplicar en Python.

## **La estad√≠stica descriptiva**

La estad√≠stica descriptiva es una rama de la estad√≠stica que se encarga de recopilar, organizar, resumir y presentar datos de forma num√©rica o gr√°fica. Su objetivo principal es describir las caracter√≠sticas principales de un conjunto de datos, sin realizar inferencias o conclusiones m√°s all√° de los datos observados. A trav√©s de la estad√≠stica descriptiva, se busca resumir y visualizar la informaci√≥n contenida en los datos, proporcionando una comprensi√≥n m√°s clara y concisa de los mismos.

A continuaci√≥n, te mostrar√© un ejemplo en c√≥digo Python para ilustrar algunos conceptos de la estad√≠stica descriptiva, as√≠ como su representaci√≥n gr√°fica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: edades de una muestra de personas
edades = [22, 25, 31, 35, 40, 42, 45, 50, 55, 60]

# C√°lculo de medidas descriptivas b√°sicas
media = np.mean(edades)
mediana = np.median(edades)
desviacion_estandar = np.std(edades)

# Creaci√≥n de un histograma de frecuencias
plt.hist(edades, bins=5, edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')

# Agregar l√≠neas para mostrar la media y la mediana
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label='Mediana')

# Mostrar la leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()

# Impresi√≥n de las medidas descriptivas
print("Media:", media)
print("Mediana:", mediana)
print("Desviaci√≥n est√°ndar:", desviacion_estandar)
```

En este ejemplo, tenemos una lista de edades de una muestra de personas. Aplicaremos algunas medidas descriptivas b√°sicas y crearemos un histograma de frecuencias para visualizar la distribuci√≥n de las edades.

El histograma muestra c√≥mo se distribuyen las edades en diferentes rangos. Los valores de las edades se agrupan en "`bins`" o intervalos y se muestra la frecuencia de observaciones en cada `bin`. Esto nos da una idea de la forma de la distribuci√≥n y nos permite identificar patrones o caracter√≠sticas importantes de los datos.

Adem√°s, calculamos la media (promedio), la mediana y la desviaci√≥n est√°ndar de las edades. Estas medidas descriptivas nos proporcionan informaci√≥n sobre el valor central, la dispersi√≥n y la variabilidad de los datos.

En resumen, la estad√≠stica descriptiva nos ayuda a resumir y comprender los datos mediante medidas num√©ricas y gr√°ficas. Esto nos permite obtener una visi√≥n general de los datos y extraer informaci√≥n importante sin hacer inferencias o generalizaciones m√°s all√° de los datos observados.

## **Gr√°ficos estad√≠sticos**

Los gr√°ficos estad√≠sticos son herramientas visuales que nos permiten representar los datos de manera clara y comprensible. Estas representaciones gr√°ficas ayudan a visualizar la distribuci√≥n, la relaci√≥n o la variaci√≥n de los datos, lo que facilita la interpretaci√≥n y el an√°lisis de la informaci√≥n estad√≠stica. A continuaci√≥n, te mostrar√© algunos ejemplos de gr√°ficos estad√≠sticos comunes junto con su implementaci√≥n en c√≥digo Python.

* **Histograma:**

El histograma es un gr√°fico de barras que muestra la frecuencia de los valores en diferentes intervalos o `bins`. Es √∫til para visualizar la distribuci√≥n de una variable continua.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(0, 1, 1000)

# Crear histograma con personalizaci√≥n
plt.hist(datos, bins=20, edgecolor='black', color='skyblue', alpha=0.7)  # A√±adir transparencia y cambiar color
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma')

# Agregar l√≠nea vertical para la media
media = np.mean(datos)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media')

# Agregar leyenda
plt.legend()

# A√±adir cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr√°fica
plt.show()
```

* **Diagrama de dispersi√≥n:**

El diagrama de dispersi√≥n muestra la relaci√≥n entre dos variables num√©ricas. Cada punto en el gr√°fico representa una observaci√≥n y su ubicaci√≥n en el plano cartesiano est√° determinada por los valores de ambas variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Crear diagrama de dispersi√≥n con personalizaci√≥n
plt.scatter(x, y, color='skyblue', marker='o', label='Datos')  # Cambiar color y marcador
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Diagrama de Dispersi√≥n')

# Agregar l√≠nea de regresi√≥n
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
plt.plot(x, polynomial(x), color='red', linestyle='--', label='L√≠nea de Regresi√≥n')

# Agregar leyenda
plt.legend()

# A√±adir cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr√°fica
plt.show()
```

* **Gr√°fico de barras:**

El gr√°fico de barras muestra la comparaci√≥n de diferentes categor√≠as o grupos mediante barras rectangulares. Es √∫til para representar datos categ√≥ricos o variables discretas.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [20, 35, 30, 15]

# Crear gr√°fico de barras con personalizaci√≥n
plt.bar(categorias, valores, color='skyblue', edgecolor='black')  # Cambiar color de las barras y del borde
plt.xlabel('Categor√≠as', fontsize=12)  # Cambiar tama√±o de la fuente en el eje x
plt.ylabel('Valores', fontsize=12)  # Cambiar tama√±o de la fuente en el eje y
plt.title('Gr√°fico de Barras', fontsize=14)  # Cambiar tama√±o de la fuente del t√≠tulo
plt.xticks(fontsize=10)  # Cambiar tama√±o de la fuente en el eje x
plt.yticks(fontsize=10)  # Cambiar tama√±o de la fuente en el eje y

# Agregar etiquetas en las barras
for i, valor in enumerate(valores):
    plt.text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=10)

# Mostrar la gr√°fica
plt.show()
```

* **Gr√°fico de pastel:**

El gr√°fico de pastel o de torta muestra la proporci√≥n de cada categor√≠a en relaci√≥n con el total. Es √∫til para representar datos porcentuales o proporcionales.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [30, 20, 25, 15]

# Colores personalizados para cada categor√≠a
colores = ['lightblue', 'lightcoral', 'lightgreen', 'lightsalmon']

# Explode para resaltar una porci√≥n (opcional)
explode = (0, 0.1, 0, 0)

# Crear gr√°fico de pastel con personalizaci√≥n
plt.pie(valores, labels=categorias, autopct='%1.1f%%', colors=colores, explode=explode, startangle=90)
plt.title('Gr√°fico de Pastel', fontsize=14)

# Mostrar la gr√°fica
plt.show()
```

Estos son solo algunos ejemplos de gr√°ficos estad√≠sticos. Existen muchos otros tipos de gr√°ficos, como diagramas de caja, gr√°ficos de l√≠nea, gr√°ficos de √°rea, entre otros, que se utilizan en diferentes situaciones para representar datos de manera efectiva. La elecci√≥n del gr√°fico adecuado depende del tipo de datos y el objetivo de la visualizaci√≥n.

## **Gr√°ficos de dispersi√≥n y burbujas**

Los gr√°ficos de dispersi√≥n y burbujas son tipos de gr√°ficos que se utilizan para visualizar la relaci√≥n entre dos o tres variables num√©ricas. Estos gr√°ficos son √∫tiles para identificar patrones, tendencias o correlaciones entre las variables. A continuaci√≥n, te mostrar√© ejemplos de implementaci√≥n en c√≥digo Python y sus gr√°ficas correspondientes.

* **Gr√°fico de Dispersi√≥n:**

Un gr√°fico de dispersi√≥n muestra la relaci√≥n entre dos variables num√©ricas. Cada punto en el gr√°fico representa una observaci√≥n y su ubicaci√≥n en el plano cartesiano est√° determinada por los valores de ambas variables. Este gr√°fico nos permite identificar la distribuci√≥n de los datos y detectar posibles patrones o relaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Personalizar el gr√°fico de dispersi√≥n
plt.scatter(x, y, color='blue', marker='o', edgecolors='black', alpha=0.7, label='Datos')

# Agregar l√≠neas de referencia
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Etiquetas y t√≠tulo
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.title('Gr√°fico de Dispersi√≥n Personalizado', fontsize=14)

# A√±adir leyenda
plt.legend()

# Ajustar dise√±o
plt.grid(True)
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

En el ejemplo anterior, se generan datos aleatorios para las variables `x` e `y`. Luego, se crea un gr√°fico de dispersi√≥n utilizando la funci√≥n `scatter` de `Matplotlib`. Cada punto en el gr√°fico representa una observaci√≥n con sus valores correspondientes en el eje `x` e `y`.

* **Gr√°fico de Burbujas:**

Un gr√°fico de burbujas es similar a un gr√°fico de dispersi√≥n, pero agrega una tercera dimensi√≥n al tama√±o de los puntos. Esto se logra asignando un tama√±o proporcional a los valores de una tercera variable. El tama√±o de las burbujas puede representar una medida adicional o una variable adicional en el conjunto de datos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
sizes = np.random.randint(10, 100, 100)  # Tama√±o de las burbujas
colors = np.random.rand(100)  # Colores aleatorios

# Personalizar el gr√°fico de burbujas
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, edgecolors='black', cmap='viridis')

# Barra de color
cbar = plt.colorbar()
cbar.set_label('Color')

# Etiquetas y t√≠tulo
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.title('Gr√°fico de Burbujas Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(True)
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se generan datos aleatorios para las variables `x` e `y`, as√≠ como tama√±os aleatorios para las burbujas en la variable sizes. Luego, se crea un gr√°fico de burbujas utilizando la funci√≥n `scatter`, donde el par√°metro `s` se establece como `sizes` para asignar el tama√±o de las burbujas.

El gr√°fico de burbujas permite visualizar tres variables de manera simult√°nea: dos variables en los ejes `x` e `y`, y una tercera variable representada por el tama√±o de las burbujas.

Estos son solo ejemplos b√°sicos de gr√°ficos de dispersi√≥n y burbujas. Es posible personalizar a√∫n m√°s estos gr√°ficos ajustando los colores, etiquetas, leyendas y otros elementos visuales seg√∫n tus necesidades y preferencias.

## **Gr√°ficos de barras, barras apiladas y agrupadas**

Los gr√°ficos de barras, barras apiladas y barras agrupadas son tipos de gr√°ficos que se utilizan para representar datos categ√≥ricos o discretos en forma de barras rectangulares. Estos gr√°ficos son √∫tiles para comparar diferentes categor√≠as y observar las diferencias o similitudes entre ellas. A continuaci√≥n, te mostrar√© ejemplos de implementaci√≥n en c√≥digo Python y sus gr√°ficas correspondientes.

* **Gr√°fico de Barras:**

Un gr√°fico de barras muestra la frecuencia o el valor de una variable en diferentes categor√≠as. Cada categor√≠a se representa por una barra individual cuya altura representa la frecuencia o valor correspondiente. Este tipo de gr√°fico es ideal para comparar categor√≠as y visualizar la magnitud de las diferencias entre ellas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [25, 40, 30, 15]

# Crear gr√°fico de barras con personalizaciones
plt.bar(categorias, valores, color=['blue', 'green', 'red', 'purple'], edgecolor='black', linewidth=2)

# Etiquetas y t√≠tulo
plt.xlabel('Categor√≠as', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.title('Gr√°fico de Barras Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar la gr√°fica
plt.show()
```

En el ejemplo anterior, se definen las categor√≠as y los valores correspondientes en las variables categorias y valores. Luego, se crea un gr√°fico de barras utilizando la funci√≥n `bar` de `Matplotlib`. Cada barra representa una categor√≠a y su altura indica el valor asociado.

* **Gr√°fico de Barras Apiladas:**

Un gr√°fico de barras apiladas muestra m√∫ltiples variables o subcategor√≠as dentro de cada categor√≠a. Las barras se apilan una encima de la otra, y la altura total de cada barra representa la suma de los valores de las subcategor√≠as. Este tipo de gr√°fico es √∫til para comparar la composici√≥n o la contribuci√≥n de diferentes subcategor√≠as dentro de cada categor√≠a.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores1 = [25, 40, 30, 15]
valores2 = [15, 30, 20, 25]

# Crear gr√°fico de barras apiladas con personalizaciones
plt.bar(categorias, valores1, label='Grupo 1', color='blue', edgecolor='black', linewidth=1)
plt.bar(categorias, valores2, bottom=valores1, label='Grupo 2', color='orange', edgecolor='black', linewidth=1)

# Etiquetas y t√≠tulo
plt.xlabel('Categor√≠as', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.title('Gr√°fico de Barras Apiladas Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se tienen dos grupos de valores representados por las variables `valores1` y `valores2`. Mediante el uso de la funci√≥n `bar`, se crean barras apiladas para cada categor√≠a. El par√°metro `bottom` se utiliza para indicar el valor base de cada barra apilada.

* **Gr√°fico de Barras Agrupadas:**

Un gr√°fico de barras agrupadas muestra m√∫ltiples variables o subcategor√≠as dentro de cada categor√≠a, pero en lugar de apilar las barras, se agrupan una al lado de la otra. Esto permite una comparaci√≥n directa entre las subcategor√≠as de diferentes categor√≠as.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores1 = [25, 40, 30, 15]
valores2 = [15, 30, 20, 25]

# Configuraci√≥n de posici√≥n de las barras
pos = np.arange(len(categorias))
ancho = 0.35

# Crear gr√°fico de barras agrupadas con personalizaciones
plt.bar(pos, valores1, ancho, label='Grupo 1', color='blue', edgecolor='black', linewidth=1)
plt.bar(pos + ancho, valores2, ancho, label='Grupo 2', color='orange', edgecolor='black', linewidth=1)

# Etiquetas y t√≠tulo
plt.xlabel('Categor√≠as', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.title('Gr√°fico de Barras Agrupadas Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(pos + ancho / 2, categorias, fontsize=10)
plt.yticks(fontsize=10)

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se tienen dos grupos de valores representados por las variables `valores1` y `valores2`. El arreglo pos se utiliza para establecer la posici√≥n de cada grupo de barras agrupadas. Mediante el uso de la funci√≥n `bar` y ajustando la posici√≥n de las barras, se crean barras agrupadas para cada categor√≠a.

Estos son solo ejemplos b√°sicos de gr√°ficos de barras, barras apiladas y barras agrupadas. Es posible personalizar a√∫n m√°s estos gr√°ficos ajustando los colores, etiquetas, leyendas y otros elementos visuales seg√∫n tus necesidades y preferencias.

## **Gr√°ficos de l√≠neas, columnas, combinados**

Los gr√°ficos de l√≠neas, columnas y combinados son tipos de gr√°ficos utilizados para visualizar datos en forma de l√≠neas o columnas. Cada tipo de gr√°fico tiene sus propias caracter√≠sticas y se utiliza en diferentes situaciones. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada y ejemplos de implementaci√≥n en c√≥digo Python para cada uno de ellos, junto con las respectivas gr√°ficas.

* **Gr√°fico de L√≠neas:**

Un gr√°fico de l√≠neas muestra la relaci√≥n y tendencia de un conjunto de datos a lo largo de un eje. Es √∫til para representar datos continuos y resaltar patrones, cambios o tendencias en los datos a lo largo del tiempo o de otra variable continua.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear gr√°fico de l√≠neas con personalizaciones
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, label='y = sin(x)')

# Etiquetas y t√≠tulo
plt.xlabel('Eje x', fontsize=12)
plt.ylabel('Eje y', fontsize=12)
plt.title('Gr√°fico de L√≠neas Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se genera una serie de datos `x` utilizando la funci√≥n `linspace` para crear una secuencia de valores desde 0 hasta 10. Luego, se calcula la funci√≥n `seno` de cada valor de `x` y se guarda en la variable `y`. Mediante la funci√≥n `plot`, se traza el gr√°fico de l√≠neas, donde los valores de `x` representan el eje horizontal y los valores de `y` representan el eje vertical.

* **Gr√°fico de Columnas:**

Un gr√°fico de columnas muestra la comparaci√≥n de datos discretos o categ√≥ricos utilizando columnas verticales. Cada columna representa una categor√≠a y su altura indica la magnitud de la variable asociada. Este tipo de gr√°fico es √∫til para comparar diferentes categor√≠as y resaltar las diferencias entre ellas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [25, 40, 30, 15]

# Crear gr√°fico de columnas con personalizaciones
plt.bar(categorias, valores, color='green', edgecolor='black', linewidth=2, alpha=0.7)

# Etiquetas y t√≠tulo
plt.xlabel('Categor√≠as', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.title('Gr√°fico de Columnas Personalizado', fontsize=14)

# Ajustar dise√±o
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se definen las categor√≠as y los valores correspondientes en las variables categorias y valores. Luego, se crea un gr√°fico de columnas utilizando la funci√≥n `bar`, donde cada columna representa una categor√≠a y su altura indica el valor asociado.

* **Gr√°fico Combinado:**

Un gr√°fico combinado permite mostrar m√∫ltiples series de datos utilizando diferentes tipos de gr√°ficos en el mismo gr√°fico. Esto es √∫til cuando se desea comparar diferentes variables o tendencias en un solo gr√°fico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear gr√°fico combinado
fig, ax1 = plt.subplots()

# Gr√°fico de l√≠neas en el primer eje y
ax1.plot(x, y1, 'b-', label='Seno')
ax1.set_xlabel('Eje x')
ax1.set_ylabel('Seno', color='b')
ax1.tick_params('y', colors='b')
ax1.set_title('Gr√°fico Combinado')

# Segundo eje y
ax2 = ax1.twinx()

# Gr√°fico de columnas en el segundo eje y
ax2.bar(x, y2, alpha=0.3, label='Coseno', color='g')
ax2.set_ylabel('Coseno', color='g')
ax2.tick_params('y', colors='g')

# Leyenda
fig.legend(loc='upper right')

plt.show()
```

En este ejemplo, se generan dos series de datos `y1` y `y2` utilizando funciones trigonom√©tricas. Se crea un gr√°fico combinado utilizando `subplots` para generar dos ejes y superponer los gr√°ficos. En el eje principal (`ax1`), se traza el gr√°fico de l√≠neas para `y1`, mientras que en el segundo eje `y` (`ax2`), se traza un gr√°fico de columnas para `y2`. Cada eje `y` tiene su propia escala y etiquetas correspondientes.

Estos ejemplos proporcionan una idea de c√≥mo crear gr√°ficos de l√≠neas, columnas y combinados utilizando Python y la biblioteca `Matplotlib`. Puedes personalizar a√∫n m√°s estos gr√°ficos seg√∫n tus necesidades, como ajustar colores, estilos, leyendas, etiquetas de ejes, entre otros.

## **La estad√≠stica y el Machine Learning**

La estad√≠stica y el aprendizaje autom√°tico (Machine Learning) son dos disciplinas interrelacionadas que se utilizan para analizar y comprender los datos. Ambas tienen como objetivo extraer informaci√≥n √∫til y tomar decisiones basadas en patrones y relaciones presentes en los datos. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada de cada una, junto con ejemplos de implementaci√≥n en c√≥digo Python y algunas visualizaciones comunes.

* **Estad√≠stica:**

La estad√≠stica es una rama de las matem√°ticas que se centra en recopilar, analizar e interpretar datos para obtener informaci√≥n significativa. Se utiliza para describir y resumir los datos, as√≠ como para inferir conclusiones sobre una poblaci√≥n m√°s amplia a partir de una muestra de datos.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos una lista de alturas de personas y queremos calcular la media y la desviaci√≥n est√°ndar.

```python
import numpy as np

# Datos de ejemplo (alturas en cent√≠metros)
alturas = [165, 170, 155, 180, 160, 175, 150, 185, 165, 170]

# C√°lculo de la media y desviaci√≥n est√°ndar
media = np.mean(alturas)
desviacion_estandar = np.std(alturas)

print("Media:", media)
print("Desviaci√≥n est√°ndar:", desviacion_estandar)
```

En este ejemplo, utilizamos la biblioteca `NumPy` para calcular la media y la desviaci√≥n est√°ndar de la lista de alturas. La media representa el valor promedio de las alturas, mientras que la desviaci√≥n est√°ndar indica qu√© tan dispersos est√°n los datos con respecto a la media.

* **Gr√°ficas asociadas:**

Algunas gr√°ficas comunes utilizadas en estad√≠stica incluyen el histograma y el diagrama de caja (`boxplot`).

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo (aseg√∫rate de tener la variable 'alturas' definida)
alturas = np.random.normal(170, 10, 100)

# Gr√°fico de histograma
plt.hist(alturas, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Alturas')
plt.ylabel('Frecuencia')
plt.title('Histograma de Alturas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Gr√°fico de diagrama de caja
plt.boxplot(alturas, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.xlabel('Alturas')
plt.title('Diagrama de Caja de Alturas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

En el primer gr√°fico, se muestra un histograma que representa la distribuci√≥n de las alturas. Los datos se agrupan en intervalos (`bins`) y se muestra la frecuencia con la que aparecen en cada intervalo.

* **Machine Learning:**

El aprendizaje autom√°tico (Machine Learning) es una rama de la inteligencia artificial que se enfoca en el desarrollo de algoritmos y modelos que permiten a las computadoras aprender autom√°ticamente a partir de los datos y tomar decisiones o hacer predicciones sin ser programadas expl√≠citamente. Se basa en la construcci√≥n de modelos estad√≠sticos y utiliza t√©cnicas para identificar patrones y relaciones en los datos.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos con informaci√≥n sobre el tama√±o de las casas (en metros cuadrados) y sus precios de venta. Queremos construir un modelo de regresi√≥n lineal para predecir el precio de una casa en funci√≥n de su tama√±o.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Datos de ejemplo (tama√±o de las casas y precios)
tamanios = np.array([80, 100, 120, 150, 200, 250, 300]).reshape(-1, 1)
precios = np.array([150000, 200000, 250000, 300000, 400000, 500000, 600000])

# Construcci√≥n del modelo de regresi√≥n lineal
modelo = LinearRegression()
modelo.fit(tamanios, precios)

# Predicci√≥n del precio de una casa de tama√±o 180 metros cuadrados
tamanio_nuevo = np.array([[180]])
precio_predicho = modelo.predict(tamanio_nuevo)

print("Precio predicho:", precio_predicho)
```

En este ejemplo, utilizamos la biblioteca `scikit-learn` para construir un modelo de regresi√≥n lineal. El modelo se ajusta a los datos de tama√±o de las casas y precios para aprender la relaci√≥n entre ambas variables. Luego, podemos usar el modelo entrenado para hacer predicciones, como predecir el precio de una casa con un tama√±o de 180 metros cuadrados.

* **Gr√°ficas asociadas:**

En el caso de la regresi√≥n lineal, una gr√°fica com√∫nmente utilizada es el gr√°fico de dispersi√≥n con la l√≠nea de regresi√≥n.

```python
import matplotlib.pyplot as plt

# Supongamos que 'tamanios' y 'precios' est√°n definidos

# Crear el gr√°fico de dispersi√≥n
plt.scatter(tamanios, precios, label='Datos reales', color='blue')

# A√±adir la l√≠nea de regresi√≥n
plt.plot(tamanios, modelo.predict(tamanios), color='red', label='L√≠nea de Regresi√≥n')

# Personalizar el gr√°fico
plt.xlabel('Tama√±o (metros cuadrados)')
plt.ylabel('Precio')
plt.title('Regresi√≥n Lineal')
plt.legend()  # Mostrar la leyenda
plt.grid(True, linestyle='--', alpha=0.7)  # A√±adir cuadr√≠cula punteada
plt.show()
```

En esta gr√°fica, se muestran los puntos de datos representados por el tama√±o de las casas y los precios en un gr√°fico de dispersi√≥n. Adem√°s, se traza la l√≠nea de regresi√≥n que representa la relaci√≥n aprendida por el modelo de regresi√≥n lineal.

Estos ejemplos ilustran c√≥mo se aplican la estad√≠stica y el aprendizaje autom√°tico utilizando Python y las bibliotecas `NumPy`, `scikit-learn` y `Matplotlib`. ## **Paradigmas de Aprendizaje en Machine Learning**

Los paradigmas de aprendizaje en Machine Learning se refieren a las diferentes aproximaciones o enfoques utilizados para resolver problemas de aprendizaje autom√°tico. Cada paradigma tiene sus propias caracter√≠sticas, suposiciones y t√©cnicas asociadas. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada de algunos de los paradigmas de aprendizaje m√°s comunes, junto con ejemplos de implementaci√≥n en c√≥digo Python y visualizaciones asociadas.

* **Aprendizaje Supervisado:**

En el aprendizaje supervisado, se utiliza un conjunto de datos etiquetados (datos de entrenamiento) que consisten en pares de entradas y salidas esperadas. El objetivo es aprender una funci√≥n que pueda mapear nuevas entradas a las salidas correctas. Los algoritmos de aprendizaje supervisado se dividen en clasificaci√≥n (cuando la salida es categ√≥rica) y regresi√≥n (cuando la salida es continua).

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos que representa el tama√±o de las flores y su correspondiente especie (rosa, tulip√°n o lirio). Queremos construir un modelo de clasificaci√≥n utilizando el algoritmo de `Support Vector Machine` (`SVM`) para predecir la especie de una flor desconocida basada en su tama√±o.

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo (tama√±o de las flores y especie)
tamanios = np.array([[3, 5], [4, 6], [2, 4], [6, 8], [5, 7], [4.5, 5.5]])
especies = np.array(['rosa', 'rosa', 'tulip√°n', 'lirio', 'lirio', 'tulip√°n'])

# Construcci√≥n del modelo de clasificaci√≥n SVM
modelo = SVC()
modelo.fit(tamanios, especies)

# Predicci√≥n de la especie para una flor de tama√±o (4, 5)
flor_nueva = np.array([[4, 5]])
especie_predicha = modelo.predict(flor_nueva)

print("Especie predicha:", especie_predicha)
```

En este ejemplo, utilizamos la biblioteca `scikit-learn` para construir un modelo de clasificaci√≥n `SVM`. El modelo se ajusta a los datos de tama√±o de las flores y especies para aprender la relaci√≥n entre ambas variables. Luego, podemos usar el modelo entrenado para hacer predicciones, como predecir la especie de una flor con un tama√±o espec√≠fico.

* **Gr√°fica asociada:**

Una visualizaci√≥n com√∫n en clasificaci√≥n supervisada es el gr√°fico de dispersi√≥n con diferentes colores para cada clase.

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo (tama√±o de las flores y especie)
tamanios = np.array([[3, 5], [4, 6], [2, 4], [6, 8], [5, 7], [4.5, 5.5]])
especies = np.array(['rosa', 'rosa', 'tulip√°n', 'lirio', 'lirio', 'tulip√°n'])

# Mapeo de etiquetas de especie a valores num√©ricos
especies_numeros = np.unique(especies, return_inverse=True)[1]

# Construcci√≥n del modelo de clasificaci√≥n SVM
modelo = SVC()
modelo.fit(tamanios, especies_numeros)

# Predicci√≥n de la especie para una flor de tama√±o (4, 5)
flor_nueva = np.array([[4, 5]])
especie_predicha_numero = modelo.predict(flor_nueva)

especie_predicha = np.unique(especies)[especie_predicha_numero]

print("Especie predicha:", especie_predicha)

# Colores y marcadores para cada especie
colores = ['red', 'blue', 'green']
marcadores = ['o', 's', '^']

# Gr√°fico de dispersi√≥n con colores seg√∫n la especie
plt.figure(figsize=(8, 6))
for i, especie in enumerate(np.unique(especies_numeros)):
    indices = especies_numeros == especie
    plt.scatter(tamanios[indices, 0], tamanios[indices, 1], c=colores[i], marker=marcadores[i], label=f'Especie {i}')

# Marcar la flor nueva
plt.scatter(flor_nueva[:, 0], flor_nueva[:, 1], c='black', marker='*', s=200, label='Flor Nueva')

plt.xlabel('Tama√±o (caracter√≠stica 1)')
plt.ylabel('Tama√±o (caracter√≠stica 2)')
plt.title('Clasificaci√≥n de Flores')
plt.legend()
plt.show()
```

En esta gr√°fica, los puntos de datos representan el tama√±o de las flores en un gr√°fico de dispersi√≥n. Los puntos se colorean de acuerdo a la especie correspondiente, lo que permite visualizar la separaci√≥n de las clases y c√≥mo se distribuyen los datos.

* **Aprendizaje No Supervisado:**

En el aprendizaje no supervisado, no se utilizan etiquetas en los datos de entrenamiento. El objetivo principal es descubrir patrones, estructuras o agrupamientos inherentes en los datos. Los algoritmos de aprendizaje no supervisado se dividen en `clustering` (agrupamiento) y reducci√≥n de la dimensionalidad.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos que representa las caracter√≠sticas de diferentes flores, pero no tenemos informaci√≥n sobre sus especies. Queremos utilizar el algoritmo de clustering K-means para agrupar las flores en diferentes categor√≠as basadas en sus caracter√≠sticas.

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (caracter√≠sticas de flores)
X = np.array([[3, 5], [4, 6], [2, 4], [6, 8], [5, 7], [4.5, 5.5]])

# Entrenamiento del modelo de clustering K-means
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)

# Predicci√≥n del grupo al que pertenece cada muestra
y_pred = kmeans.predict(X)

# Colores y marcadores distintivos para cada grupo
colores = ['blue', 'green']
marcadores = ['o', 's']

# Gr√°fica de los datos y los grupos encontrados
plt.figure(figsize=(8, 6))
for i, grupo in enumerate(np.unique(y_pred)):
    indices = y_pred == grupo
    plt.scatter(X[indices, 0], X[indices, 1], c=colores[i], marker=marcadores[i], label=f'Grupo {i}')

# Centroides marcados en rojo con 'x'
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroides')

# Etiquetas para los puntos y centroides
for i, txt in enumerate(np.arange(1, len(X) + 1)):
    plt.annotate(txt, (X[i, 0], X[i, 1]), textcoords="offset points", xytext=(0, 5), ha='center')

for i, txt in enumerate(np.arange(1, len(kmeans.cluster_centers_) + 1)):
    plt.annotate(f'Centroide {txt}', (kmeans.cluster_centers_[i, 0], kmeans.cluster_centers_[i, 1]),
                 textcoords="offset points", xytext=(0, 5), ha='center')

plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.title('Clustering con K-means')
plt.legend()
plt.grid(True)
plt.show()
```

En la gr√°fica resultante, los puntos se colorean seg√∫n el grupo al que pertenecen, determinado por el algoritmo K-means. Tambi√©n se muestran los centroides de cada grupo como cruces rojas.

* **Aprendizaje por refuerzo:**

En el aprendizaje por refuerzo, un agente aprende a tomar decisiones en un entorno para maximizar una recompensa acumulativa. El agente interact√∫a con el entorno a trav√©s de acciones y recibe retroalimentaci√≥n en forma de recompensas o castigos seg√∫n sus acciones. El objetivo es aprender una pol√≠tica √≥ptima que gu√≠e al agente para tomar las mejores acciones en cada estado. Un ejemplo popular de aprendizaje por refuerzo es el algoritmo Q-learning.
Aqu√≠ tienes un ejemplo simplificado de Q-learning en Python:

```python
import numpy as np

# Configuraci√≥n del entorno y recompensas
num_estados = 4
num_acciones = 2
recompensas = np.array([[0, -1], [-1, 1], [-1, 1], [-1, 0]])

# Matriz Q inicial
Q = np.zeros((num_estados, num_acciones))

# Hiperpar√°metros
factor_aprendizaje = 0.1
factor_descuento = 0.9
num_episodios = 100

# Algoritmo de Q-learning
for _ in range(num_episodios):
    estado_actual = 0
    while estado_actual != num_estados - 1:
        accion = np.argmax(Q[estado_actual])
        siguiente_estado = np.random.choice(range(num_estados))
        recompensa = recompensas[estado_actual, accion]
        Q[estado_actual, accion] += factor_aprendizaje * (recompensa + factor_descuento * np.max(Q[siguiente_estado]) - Q[estado_actual, accion])
        estado_actual = siguiente_estado

print("Matriz Q aprendida:")
print(Q)
```

En este ejemplo, se utiliza una matriz Q para almacenar los valores de utilidad (Q-values) para cada par estado-acci√≥n. El agente utiliza la matriz Q para seleccionar la mejor acci√≥n en cada estado y actualizar los valores de Q a medida que interact√∫a con el entorno. Despu√©s del entrenamiento, se imprime la matriz Q aprendida.

Ten en cuenta que estos ejemplos son solo ilustrativos y pueden simplificar algunos aspectos del problema real. Adem√°s, las gr√°ficas mostradas son solo ejemplos b√°sicos y podr√≠an ser m√°s complejas en situaciones reales.

## **Aprendizaje supervisado y no supervisado**

El aprendizaje supervisado y no supervisado son dos paradigmas fundamentales en el campo del aprendizaje autom√°tico (machine learning). A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada de cada uno, junto con ejemplos de c√≥digo en Python y visualizaciones gr√°ficas.

* **Aprendizaje Supervisado:**

El aprendizaje supervisado implica entrenar un modelo utilizando ejemplos de entrada y salida conocidos para que pueda realizar predicciones o clasificar nuevos ejemplos. En este paradigma, se proporciona un conjunto de datos de entrenamiento etiquetado, donde cada ejemplo tiene caracter√≠sticas de entrada y una etiqueta o valor de salida correspondiente.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos de flores con caracter√≠sticas como longitud y ancho de p√©talos, as√≠ como su especie (etiqueta). Podemos utilizar un algoritmo de aprendizaje supervisado, como la regresi√≥n log√≠stica, para predecir la especie de una flor desconocida en funci√≥n de sus caracter√≠sticas.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Datos de entrenamiento
X_train = np.array([[5.1, 3.5], [4.9, 3.0], [6.7, 3.1], [6.0, 3.0]])
y_train = np.array([0, 0, 1, 1])  # 0: Setosa, 1: Versicolor

# Crear y entrenar el modelo
model = LogisticRegression()
model.fit(X_train, y_train)

# Datos de prueba
X_test = np.array([[5.5, 2.5], [6.3, 2.9]])

# Realizar predicciones
y_pred = model.predict(X_test)

# Graficar los datos de entrenamiento
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', label='Entrenamiento')

# Graficar los datos de prueba y las predicciones
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, marker='x', s=100, label='Predicciones')

plt.xlabel('Longitud del p√©talo')
plt.ylabel('Ancho del p√©talo')
plt.title('Predicciones de especies de flores')
plt.legend()
plt.show()
```

En este ejemplo, utilizamos la regresi√≥n log√≠stica como algoritmo de aprendizaje supervisado para clasificar las flores en dos especies diferentes (Setosa y Versicolor). Entrenamos el modelo utilizando los datos de entrenamiento etiquetados y luego realizamos predicciones en nuevos datos de prueba. La gr√°fica muestra los puntos de datos de entrenamiento con diferentes colores seg√∫n su especie, as√≠ como las predicciones en los nuevos datos representadas por '`x`'.

* **Aprendizaje No Supervisado:**

El aprendizaje no supervisado implica descubrir patrones y estructuras ocultas en un conjunto de datos sin la necesidad de etiquetas o valores de salida conocidos. En este paradigma, el modelo aprende de forma aut√≥noma a partir de los datos y encuentra agrupaciones, relaciones o caracter√≠sticas significativas.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos de clientes con caracter√≠sticas demogr√°ficas y de comportamiento de compra. Podemos utilizar un algoritmo de aprendizaje no supervisado, como el clustering (agrupamiento), para agrupar a los clientes en segmentos similares basados en sus caracter√≠sticas.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Datos de entrada
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])

# Crear y entrenar el modelo de clustering
model = KMeans(n_clusters=2, n_init=10)  # Puedes ajustar el valor de n_init aqu√≠
model.fit(X)

# Obtener las etiquetas de los clusters y los centroides
labels = model.labels_
centroids = model.cluster_centers_

# Definir colores y marcadores
colors = ['r', 'g']
markers = ['o', 's']

# Graficar los datos y los centroides
for i in range(len(X)):
    plt.scatter(X[i, 0], X[i, 1], color=colors[labels[i]], marker=markers[labels[i]])

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='b', s=100, label='Centroides')
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.title('Clustering de clientes')

# Agregar leyenda
plt.legend()

plt.show()
```

En este ejemplo, utilizamos el algoritmo de `clustering K-means` como un ejemplo de aprendizaje no supervisado. Agrupamos a los clientes en dos segmentos (2 clusters) en funci√≥n de sus caracter√≠sticas demogr√°ficas y de comportamiento de compra. La gr√°fica muestra los puntos de datos de entrada coloreados seg√∫n su cluster asignado, y los centroides de los clusters se representan con '`x`'.

Espero que estos ejemplos te hayan proporcionado una comprensi√≥n clara de los conceptos de aprendizaje supervisado y no supervisado, as√≠ como su implementaci√≥n en Python y visualizaci√≥n gr√°fica.

## **Identificando el tipo de problema a resolver**

Para identificar el tipo de problema a resolver en Machine Learning, es importante comprender la naturaleza de los datos y los objetivos que se quieren lograr. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada de los diferentes tipos de problemas en Machine Learning, junto con ejemplos de c√≥digo en Python y visualizaciones gr√°ficas.

* **Problemas de clasificaci√≥n:**

En los problemas de clasificaci√≥n, el objetivo es asignar una etiqueta o categor√≠a a una instancia de datos basada en sus caracter√≠sticas. Es decir, se busca predecir a qu√© clase pertenece cada ejemplo. Algunos ejemplos de problemas de clasificaci√≥n incluyen la detecci√≥n de spam en correos electr√≥nicos, la clasificaci√≥n de im√°genes en diferentes categor√≠as o la predicci√≥n de la probabilidad de enfermedad basada en s√≠ntomas.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos que contiene informaci√≥n sobre frutas, como su peso y textura, y queremos clasificar si una fruta es una manzana o una naranja. Podemos utilizar un algoritmo de clasificaci√≥n, como el clasificador de vecinos m√°s cercanos (K-Nearest Neighbors), para resolver este problema.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Datos de entrenamiento
X_train = np.array([[100, 2], [120, 3], [130, 1], [140, 3], [150, 2], [160, 2]])
y_train = np.array(['naranja', 'naranja', 'manzana', 'manzana', 'manzana', 'naranja'])

# Asignar colores num√©ricos a las clases
color_map = {'naranja': 0, 'manzana': 1}
colors = [color_map[label] for label in y_train]

# Crear y entrenar el clasificador
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

# Datos de prueba
X_test = np.array([[110, 2]])

# Realizar la clasificaci√≥n
y_pred = classifier.predict(X_test)

# Graficar los datos y la clasificaci√≥n
plt.scatter(X_train[:, 0], X_train[:, 1], c=colors, cmap='viridis', label='Datos de entrenamiento', edgecolors='k')
plt.scatter(X_test[:, 0], X_test[:, 1], c=color_map[y_pred[0]], marker='x', s=200, label='Fruta clasificada')
plt.xlabel('Peso')
plt.ylabel('Textura')
plt.title('Clasificaci√≥n de frutas')

# Agregar leyenda
plt.legend()

# Mostrar barra de color
plt.colorbar(label='Clases')

plt.show()
```

En este ejemplo, utilizamos el clasificador de vecinos m√°s cercanos para clasificar las frutas en manzanas y naranjas seg√∫n su peso y textura. Entrenamos el clasificador con los datos de entrenamiento etiquetados y realizamos la clasificaci√≥n en nuevos datos de prueba. La gr√°fica muestra los datos de entrenamiento coloreados seg√∫n su clase, y la nueva fruta a clasificar se representa con '`x`'.

* **Problemas de regresi√≥n:**

En los problemas de regresi√≥n, el objetivo es predecir un valor num√©rico continuo en funci√≥n de las caracter√≠sticas de entrada. En otras palabras, se busca encontrar una relaci√≥n funcional entre las variables de entrada y salida. Algunos ejemplos de problemas de regresi√≥n incluyen la predicci√≥n de los precios de las viviendas seg√∫n sus caracter√≠sticas, la estimaci√≥n de la demanda de productos en funci√≥n de factores externos o la predicci√≥n del rendimiento de un estudiante en funci√≥n de sus horas de estudio.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos que contiene informaci√≥n sobre el tama√±o de las casas (en metros cuadrados) y sus precios de venta, y queremos predecir el precio de una casa nueva basada en su tama√±o. Podemos utilizar un algoritmo de regresi√≥n, como la regresi√≥n lineal, para resolver este problema.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de entrenamiento
X_train = np.array([[80], [100], [120], [150], [180]])
y_train = np.array([150000, 200000, 230000, 280000, 320000])

# Crear y entrenar el modelo de regresi√≥n lineal
regression = LinearRegression()
regression.fit(X_train, y_train)

# Datos de prueba
X_test = np.array([[110]])

# Realizar la predicci√≥n
y_pred = regression.predict(X_test)

# Graficar los datos y la predicci√≥n
plt.scatter(X_train, y_train, label='Datos de entrenamiento')
plt.plot(X_test, y_pred, color='red', linestyle='--', label='Predicci√≥n')
plt.xlabel('Tama√±o')
plt.ylabel('Precio')
plt.title('Predicci√≥n de precios de casas')

# Agregar leyenda
plt.legend()

# Mostrar la grilla
plt.grid(True)

plt.show()
```

En este ejemplo, utilizamos la regresi√≥n lineal para predecir los precios de las casas seg√∫n su tama√±o. Entrenamos el modelo con los datos de entrenamiento y realizamos la predicci√≥n en un nuevo dato de prueba. La gr√°fica muestra los datos de entrenamiento y la l√≠nea de regresi√≥n que representa la predicci√≥n de los precios de las casas.

* **Problemas de agrupamiento (clustering):**

En los problemas de agrupamiento, el objetivo es agrupar instancias de datos similares en clusters o grupos, sin tener etiquetas previas. El algoritmo de agrupamiento busca encontrar patrones y estructuras ocultas en los datos. Algunos ejemplos de problemas de agrupamiento incluyen la segmentaci√≥n de clientes en grupos de comportamiento similar, la clasificaci√≥n de noticias en categor√≠as tem√°ticas o la identificaci√≥n de grupos de genes con perfiles de expresi√≥n similares.

**Ejemplo en c√≥digo Python:**

Supongamos que tenemos un conjunto de datos que contiene informaci√≥n sobre las caracter√≠sticas de diferentes flores, como su longitud y ancho de p√©talos, y queremos agrupar las flores en diferentes tipos seg√∫n su similitud. Podemos utilizar un algoritmo de clustering, como K-means, para resolver este problema.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Datos de entrada
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])

# Crear y entrenar el modelo de clustering
model = KMeans(n_clusters=2, n_init=10)  # Establecer expl√≠citamente n_init
model.fit(X)

# Obtener las etiquetas de los clusters y los centroides
labels = model.labels_
centroids = model.cluster_centers_

# Mapear etiquetas a colores
colors = np.array(['r', 'g'])[labels]

# Graficar los datos y los centroides
plt.scatter(X[:, 0], X[:, 1], color=colors)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='b', s=100, label='Centroides')

plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.title('Clustering de flores')

# Agregar leyenda
plt.legend()

plt.show()
```

En este ejemplo, utilizamos el algoritmo de clustering K-means para agrupar las flores en dos segmentos basados en las caracter√≠sticas de longitud y ancho de p√©talos. La gr√°fica muestra los datos de entrada coloreados seg√∫n su cluster asignado, y los centroides de los clusters se representan con una '`x`'.
