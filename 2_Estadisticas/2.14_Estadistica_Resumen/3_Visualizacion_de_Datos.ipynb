**Histogramas:**

Un histograma es una representaci贸n gr谩fica de la distribuci贸n de un conjunto de datos.
```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear un histograma
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

**Diagramas de Dispersi贸n:**

Los diagramas de dispersi贸n muestran la relaci贸n entre dos variables. Cada punto en el gr谩fico representa una observaci贸n con respecto a dos variables. Este tipo de gr谩fico es 煤til para identificar patrones, tendencias y correlaciones.

```python
# Generar datos de ejemplo para un diagrama de dispersi贸n
x = np.random.normal(loc=50, scale=10, size=100)
y = 2 * x + np.random.normal(scale=20, size=100)

# Crear un diagrama de dispersi贸n
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', alpha=0.7)
plt.title('Diagrama de Dispersi贸n')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

En este ejemplo, el diagrama de dispersi贸n muestra una relaci贸n lineal entre las variables X e Y.

**Boxplots (Diagrama de Caja y Bigotes):**

El boxplot es un gr谩fico que proporciona informaci贸n sobre la distribuci贸n de los datos, incluyendo la mediana, los cuartiles y los valores at铆picos.

```python
# Generar datos de ejemplo para un boxplot
datos_boxplot = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(5, 2, 100)]

# Crear un boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(datos_boxplot, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Boxplot de Datos')
plt.ylabel('Valor')
plt.show()
```

Este boxplot muestra la comparaci贸n de tres conjuntos de datos. La caja representa el rango intercuartil (IQR), la l铆nea en el centro de la caja es la mediana y los "bigotes" indican la dispersi贸n de los datos.
