# **3. Visualización de Datos**

## **Gráficos Estadísticos**

### **Histogramas, diagramas de dispersión, boxplots, etc.**

**Histogramas:**

Un histograma es una representación gráfica de la distribución de un conjunto de datos. Se divide en "bins" o intervalos, y cada barra representa la frecuencia o la cantidad de datos en cada intervalo. Es útil para visualizar la forma de la distribución, la tendencia central y la dispersión de los datos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear un histograma
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

Este ejemplo utiliza un histograma para visualizar la distribución de datos normales. Cambiar el número de bins ajusta la cantidad de detalle en el histograma.

**Diagramas de Dispersión:**

Los diagramas de dispersión muestran la relación entre dos variables. Cada punto en el gráfico representa una observación con respecto a dos variables. Este tipo de gráfico es útil para identificar patrones, tendencias y correlaciones.

```python
# Generar datos de ejemplo para un diagrama de dispersión
x = np.random.normal(loc=50, scale=10, size=100)
y = 2 * x + np.random.normal(scale=20, size=100)

# Crear un diagrama de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', alpha=0.7)
plt.title('Diagrama de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

En este ejemplo, el diagrama de dispersión muestra una relación lineal entre las variables X e Y.

**Boxplots (Diagrama de Caja y Bigotes):**

El boxplot es un gráfico que proporciona información sobre la distribución de los datos, incluyendo la mediana, los cuartiles y los valores atípicos.

```python
# Generar datos de ejemplo para un boxplot
datos_boxplot = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(5, 2, 100)]

# Crear un boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(datos_boxplot, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Boxplot de Datos')
plt.ylabel('Valor')
plt.show()
```

Este boxplot muestra la comparación de tres conjuntos de datos. La caja representa el rango intercuartil (IQR), la línea en el centro de la caja es la mediana y los "bigotes" indican la dispersión de los datos.

Estos son ejemplos básicos, y se pueden personalizar más según tus necesidades. Los gráficos proporcionan una visualización efectiva de los datos y facilitan la interpretación de las características clave de la distribución.

### **Selección adecuada de gráficos según el tipo de datos**

La selección adecuada de gráficos depende del tipo de datos y de la información que deseas comunicar. Aquí hay algunas pautas generales para la elección de gráficos según el tipo de datos:

1. **Datos Unidimensionales (Univariados):**

   - **Histograma:** Para visualizar la distribución y frecuencia de datos numéricos.
   - **Gráfico de Barras:** Para datos categóricos o discretos, donde cada barra representa la frecuencia o proporción de cada categoría.

2. **Datos Bidimensionales (Bivariados):**

   - **Diagrama de Dispersión:** Para visualizar la relación entre dos variables numéricas.
   - **Gráfico de Barras Agrupadas o Apiladas:** Para comparar frecuencias o proporciones de diferentes categorías en dos o más grupos.

3. **Datos Multidimensionales:**

   - **Diagrama de Caja y Bigotes (Boxplot):** Para comparar la distribución de datos entre múltiples grupos.
   - **Gráfico de Dispersión 3D:** Si tienes tres variables numéricas y deseas visualizar la relación tridimensional.

4. **Datos Temporales o de Series Temporales:**

   - **Gráfico de Líneas:** Para mostrar cambios a lo largo del tiempo.
   - **Gráfico de Área:** Puede ser útil para resaltar la acumulación de datos a lo largo del tiempo.

5. **Comparaciones y Proporciones:**

   - **Gráfico de Barras o Gráfico de Columnas:** Para comparar magnitudes entre diferentes categorías.
   - **Gráfico Circular (Pie Chart):** Para representar proporciones en un conjunto completo.

6. **Distribuciones y Tendencias:**

   - **Histograma o Kernel Density Plot:** Para visualizar la forma de la distribución.
   - **Gráfico de Puntos o Gráfico de Líneas:** Para mostrar tendencias o patrones a lo largo de un eje.

7. **Datos Geoespaciales:**

   - **Mapa de Calor (Heatmap):** Para visualizar la densidad de datos en un mapa.
   - **Mapa de Coropletas:** Para mostrar variaciones espaciales de datos categóricos.

Recuerda que estas son pautas generales y hay muchas otras opciones de gráficos. La elección final debe basarse en la claridad, la efectividad de la comunicación y la adecuación a tus objetivos específicos. Además, ten en cuenta la audiencia a la que te diriges, ya que diferentes gráficos pueden ser más efectivos para diferentes públicos.
