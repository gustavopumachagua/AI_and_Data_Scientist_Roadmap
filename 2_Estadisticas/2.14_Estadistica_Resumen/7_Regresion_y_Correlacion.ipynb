## **Regresi贸n Lineal Simple**

### **Modelo de regresi贸n, coeficiente de determinaci贸n**

**Modelo de Regresi贸n:**

Un modelo de regresi贸n es una herramienta estad铆stica que se utiliza para examinar la relaci贸n entre una variable dependiente (respuesta) y una o m谩s variables independientes (predictores). El objetivo es desarrollar una ecuaci贸n matem谩tica que describa la relaci贸n entre estas variables.

**Ecuaci贸n del Modelo de Regresi贸n Simple:**

$Y = \beta_0 + \beta_1 \cdot X + \epsilon$

- $(Y)$ es la variable dependiente.
- $(X)$ es la variable independiente.
- $(\beta_0)$ es la ordenada al origen (intercepto).
- $(\beta_1)$ es la pendiente.
- $(\epsilon)$ es el error.

**Ecuaci贸n del Modelo de Regresi贸n M煤ltiple:**

$Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \ldots + \beta_n \cdot X_n + \epsilon$

**Coeficiente de Determinaci贸n $(R^2)$:**

El coeficiente de determinaci贸n $(R^2)$ es una medida que indica la proporci贸n de la variabilidad en la variable dependiente que puede explicarse por el modelo de regresi贸n. $(R^2)$ var铆a entre 0 y 1, donde 0 indica que el modelo no explica ninguna variabilidad y 1 indica que el modelo explica toda la variabilidad.

$R^2 = 1 - \frac{\text{Suma de los cuadrados de los residuos}}{\text{Suma total de los cuadrados}}$

**Ejemplo y C贸digo Python:**

Supongamos que queremos ajustar un modelo de regresi贸n simple y calcular $(R^2)$ utilizando datos ficticios.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Crear datos ficticios
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Ajustar un modelo de regresi贸n lineal
model = LinearRegression()
model.fit(X, y)

# Predicciones del modelo
y_pred = model.predict(X)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Visualizaci贸n en gr谩fico
plt.scatter(X, y, alpha=0.7, label='Datos Observados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo de Regresi贸n')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Modelo de Regresi贸n Lineal y $R^2$')
plt.legend()
plt.show()

# Imprimir resultados
print(f"Coeficientes del modelo: {model.coef_[0][0]} (pendiente), {model.intercept_[0]} (intercepto)")
print(f"Coeficiente de Determinaci贸n (R^2): {r2}")
```

En este c贸digo, hemos creado datos ficticios con una relaci贸n lineal y ajustado un modelo de regresi贸n lineal. La visualizaci贸n incluye una dispersi贸n de puntos para los datos observados y la l铆nea roja para el modelo de regresi贸n. Adem谩s, imprimimos los coeficientes del modelo y el coeficiente de determinaci贸n $(R^2)$. Puedes ajustar este c贸digo para incluir m煤ltiples variables independientes para un modelo de regresi贸n m煤ltiple.

### **Interpretaci贸n de resultados**

La interpretaci贸n de los resultados de un modelo de regresi贸n y el coeficiente de determinaci贸n $(R^2)$ es fundamental para entender la relaci贸n entre las variables y la calidad del ajuste del modelo. **Interpretaci贸n de Coeficientes del Modelo de Regresi贸n:**

En un modelo de regresi贸n lineal simple, la ecuaci贸n del modelo es:

$Y = \beta_0 + \beta_1 \cdot X + \epsilon$

Donde:

- $(\beta_0)$ es la ordenada al origen (intercepto).
- $(\beta_1)$ es la pendiente.
- $(X)$ es la variable independiente.
- $(Y)$ es la variable dependiente.

**Ejemplo en Python:**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Crear datos ficticios
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Ajustar un modelo de regresi贸n lineal
model = LinearRegression()
model.fit(X, y)

# Imprimir coeficientes del modelo
print(f"Intercepto (ordenada al origen): {model.intercept_[0]}")
print(f"Pendiente: {model.coef_[0][0]}")
```

En este ejemplo, el intercepto $(\beta_0)$ representa el valor esperado de $(Y)$ cuando $(X)$ es cero, y la pendiente $(\beta_1)$ indica c贸mo cambia $(Y)$ por unidad de cambio en $(X)$.

**Interpretaci贸n del Coeficiente de Determinaci贸n $(R^2)$:**

El coeficiente de determinaci贸n $(R^2)$ mide la proporci贸n de la variabilidad en la variable dependiente que es explicada por el modelo. Su interpretaci贸n var铆a entre 0 (ninguna explicaci贸n) y 1 (explicaci贸n completa).

$R^2 = 1 - \frac{\text{Suma de los cuadrados de los residuos}}{\text{Suma total de los cuadrados}}$

**Ejemplo en Python:**

```python
from sklearn.metrics import r2_score

# Predicciones del modelo
y_pred = model.predict(X)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Imprimir coeficiente de determinaci贸n (R^2)
print(f"Coeficiente de Determinaci贸n (R^2): {r2}")
```

**Visualizaci贸n con Gr谩ficos Personalizados:**

```python
import matplotlib.pyplot as plt

# Visualizaci贸n en gr谩fico
plt.scatter(X, y, alpha=0.7, label='Datos Observados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo de Regresi贸n')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Modelo de Regresi贸n Lineal y $R^2$')
plt.legend()
plt.show()
```

En este gr谩fico, los puntos azules representan los datos observados, y la l铆nea roja es la predicci贸n del modelo de regresi贸n. Este gr谩fico ayuda a visualizar c贸mo el modelo se ajusta a los datos observados.

**Interpretaci贸n Conjunta:**

- **Coeficientes del Modelo:**

  - El intercepto $(\beta_0)$ puede interpretarse como el valor esperado de la variable dependiente cuando la variable independiente es cero.
  - La pendiente $(\beta_1)$ indica el cambio promedio en la variable dependiente por unidad de cambio en la variable independiente.

- **Coeficiente de Determinaci贸n $(R^2)$:**

  - Un $(R^2)$ cercano a 1 indica que el modelo explica una gran proporci贸n de la variabilidad en la variable dependiente.
  - Un $(R^2)$ cercano a 0 sugiere que el modelo no explica mucha variabilidad.

## **Regresi贸n M煤ltiple**

### **Extensi贸n del modelo de regresi贸n a m煤ltiples variables independientes**

Cuando trabajamos con modelos de regresi贸n, es com煤n extender el enfoque a m煤ltiples variables independientes, creando as铆 un modelo de regresi贸n m煤ltiple. La ecuaci贸n del modelo de regresi贸n m煤ltiple se ve as铆:

$Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \ldots + \beta_n \cdot X_n + \epsilon$

Donde:

- $( Y )$ es la variable dependiente.
- $( X_1, X_2, \ldots, X_n )$ son las variables independientes.
- $( \beta_0 )$ es el t茅rmino de intersecci贸n (ordenada al origen).
- $( \beta_1, \beta_2, \ldots, \beta_n )$ son los coeficientes asociados a las variables independientes.
- $( \epsilon )$ es el t茅rmino de error.

**Ejemplo con C贸digo Python:**

Supongamos que tenemos un conjunto de datos con dos variables independientes $( X_1 )$ y $( X_2 )$, y queremos ajustar un modelo de regresi贸n m煤ltiple.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Crear datos ficticios con dos variables independientes
np.random.seed(42)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 5 * X2 + np.random.randn(100, 1)

# Ajustar un modelo de regresi贸n m煤ltiple
model = LinearRegression()
X_multi = np.column_stack((X1, X2))
model.fit(X_multi, y)

# Obtener coeficientes y realizar predicciones
intercepto = model.intercept_[0]
coeficientes = model.coef_[0]
y_pred = model.predict(X_multi)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Imprimir coeficientes y R^2
print(f"Intercepto (ordenada al origen): {intercepto}")
print(f"Coeficientes: {coeficientes}")
print(f"Coeficiente de Determinaci贸n (R^2): {r2}")

# Visualizaci贸n en gr谩fico 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, y, color='blue', label='Datos Observados')
ax.scatter(X1, X2, y_pred, color='red', label='Predicciones del Modelo', marker='^')

ax.set_xlabel('Variable Independiente 1')
ax.set_ylabel('Variable Independiente 2')
ax.set_zlabel('Variable Dependiente')
ax.set_title('Modelo de Regresi贸n M煤ltiple')

plt.legend()
plt.show()
```


En este ejemplo, hemos creado un conjunto de datos con dos variables independientes $(X_1)$ y $(X_2)$ y ajustado un modelo de regresi贸n m煤ltiple. La visualizaci贸n incluye un gr谩fico 3D para mostrar c贸mo el modelo se ajusta a los datos observados y predice valores en funci贸n de dos variables independientes.

**Interpretaci贸n de Resultados:**

- **Coeficientes del Modelo:**

  - Cada coeficiente $(\beta_1, \beta_2)$ representa el cambio promedio en la variable dependiente por unidad de cambio en la variable independiente correspondiente, manteniendo las otras variables constantes.

- **Coeficiente de Determinaci贸n $(R^2)$:**

  - Indica la proporci贸n de la variabilidad en la variable dependiente que es explicada por el modelo de regresi贸n m煤ltiple.

- **Visualizaci贸n:**

  - El gr谩fico 3D proporciona una representaci贸n visual del modelo y c贸mo se ajusta a los datos.

La extensi贸n a m煤ltiples variables independientes permite modelar relaciones m谩s complejas y capturar la variabilidad asociada a m煤ltiples factores. La interpretaci贸n de los coeficientes sigue siendo similar, pero ahora consideramos el impacto de cada variable independiente controlando las dem谩s.
