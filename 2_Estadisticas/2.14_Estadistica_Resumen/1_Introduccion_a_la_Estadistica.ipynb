# **1. Introducci贸n a la Estad铆stica**

## **Conceptos B谩sicos de Estad铆stica**

### **Definiciones clave (poblaci贸n, muestra, variable, etc.)**

Vamos a explorar las definiciones clave en estad铆stica: poblaci贸n, muestra, y variable, y c贸mo puedes representarlas visualmente en Python utilizando gr谩ficos personalizados. Utilizaremos las bibliotecas `matplotlib` y `numpy` para crear visualizaciones simples.

Ahora, aqu铆 tienes un ejemplo con explicaciones detalladas:

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir una poblaci贸n
np.random.seed(42)  # Para reproducibilidad
poblacion = np.random.normal(loc=50, scale=10, size=10000)  # Distribuci贸n normal, media=50, desviaci贸n est谩ndar=10

# Graficar la poblaci贸n con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(poblacion, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# L铆nea vertical para la media
media_poblacion = np.mean(poblacion)
plt.axvline(x=media_poblacion, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media_poblacion:.2f}')

# L铆nea vertical para la desviaci贸n est谩ndar
std_dev_poblacion = np.std(poblacion)
plt.axvline(x=media_poblacion + std_dev_poblacion, color='orange', linestyle='dashed', linewidth=2, label=f'SD = {std_dev_poblacion:.2f}')
plt.axvline(x=media_poblacion - std_dev_poblacion, color='orange', linestyle='dashed', linewidth=2)

# T铆tulo y etiquetas del eje
plt.title('Distribuci贸n de la Poblaci贸n - Media y Desviaci贸n Est谩ndar', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
````

En este ejemplo:

- **Poblaci贸n:** Hemos creado una poblaci贸n de 10,000 elementos que sigue una distribuci贸n normal con una media de 50 y una desviaci贸n est谩ndar de 10. La poblaci贸n es el conjunto completo de datos que queremos estudiar.

- **Muestra:** Ahora, tomaremos una muestra de esta poblaci贸n. Utilizaremos solo 500 elementos de la poblaci贸n para formar nuestra muestra.

````python
# Tomar una muestra de la poblaci贸n
muestra = np.random.choice(poblacion, size=500)

# Personalizar la gr谩fica
plt.figure(figsize=(12, 6))

# Histograma de la poblaci贸n
plt.hist(poblacion, bins=30, color='skyblue', alpha=0.5, label='Poblaci贸n', edgecolor='black')

# Histograma de la muestra
plt.hist(muestra, bins=30, color='orange', alpha=0.7, label='Muestra', edgecolor='black')

# T铆tulo y etiquetas del eje
plt.title('Comparaci贸n entre Poblaci贸n y Muestra', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr谩fica
plt.show()
````

En esta visualizaci贸n, puedes ver c贸mo la muestra (en naranja) es un subconjunto m谩s peque帽o pero a煤n representa la forma general de la poblaci贸n.

- **Variable:** Ahora, introduzcamos una variable. Supongamos que estamos interesados en la altura de las personas en nuestra poblaci贸n y muestra.

````python
# Generar datos de altura para la poblaci贸n y la muestra
altura_poblacion = np.random.normal(loc=170, scale=10, size=10000)
altura_muestra = np.random.choice(altura_poblacion, size=500)

# Personalizar la gr谩fica
plt.figure(figsize=(12, 6))

# Histograma de la poblaci贸n
plt.hist(altura_poblacion, bins=30, color='skyblue', alpha=0.5, label='Altura en Poblaci贸n', edgecolor='black')

# Histograma de la muestra
plt.hist(altura_muestra, bins=30, color='orange', alpha=0.7, label='Altura en Muestra', edgecolor='black')

# L铆nea vertical para la media de la poblaci贸n
media_altura_poblacion = np.mean(altura_poblacion)
plt.axvline(x=media_altura_poblacion, color='blue', linestyle='dashed', linewidth=2, label=f'Media Poblaci贸n = {media_altura_poblacion:.2f}')

# L铆nea vertical para la media de la muestra
media_altura_muestra = np.mean(altura_muestra)
plt.axvline(x=media_altura_muestra, color='orange', linestyle='dashed', linewidth=2, label=f'Media Muestra = {media_altura_muestra:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Distribuci贸n de Altura en Poblaci贸n y Muestra', fontsize=16)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr谩fica
plt.show()
````

En este caso, hemos introducido la variable "altura", y graficamos c贸mo se distribuye en la poblaci贸n y la muestra.

Estos son ejemplos simples para comprender las definiciones clave en estad铆stica. La visualizaci贸n de datos es esencial para entender las caracter铆sticas de la poblaci贸n y la muestra, as铆 como la distribuci贸n de las variables de inter茅s.

### **Tipos de datos: cualitativos y cuantitativos**

En estad铆stica, los datos se pueden clasificar en dos tipos principales: cualitativos y cuantitativos. Vamos a explorar cada uno de ellos con ejemplos y gr谩ficos personalizados utilizando Python.

**1. Datos Cualitativos:**

Los datos cualitativos representan atributos no num茅ricos y se dividen en dos subtipos: nominales y ordinales.

**a. Datos Nominales:**

Los datos nominales representan categor铆as con nombres 煤nicos, pero no hay un orden espec铆fico entre ellas.

**Ejemplo:** Colores de los autom贸viles.

````python
import matplotlib.pyplot as plt

# Datos nominales
colores = ['red', 'blue', 'green', 'yellow', 'black']
conteo_colores = [25, 20, 15, 10, 30]

# Colores personalizados para las barras
colores_barras = ['lightcoral', 'royalblue', 'lightgreen', 'gold', 'gray']

# Graficar un gr谩fico de barras para datos nominales con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(colores, conteo_colores, color=colores_barras, edgecolor='black', linewidth=1.2)

# T铆tulo y etiquetas del eje
plt.title('Distribuci贸n de Colores de Autom贸viles', fontsize=16)
plt.xlabel('Colores', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar la gr谩fica
plt.show()
````

**b. Datos Ordinales:**

Los datos ordinales tienen categor铆as con un orden espec铆fico, pero la distancia entre ellas no es cuantificable.

**Ejemplo:** Clasificaci贸n de productos por calidad (Baja, Media, Alta).

````python
import matplotlib.pyplot as plt

# Datos ordinales
productos = ['Baja', 'Media', 'Alta']
conteo_calidad = [15, 30, 20]

# Colores personalizados para las barras
colores_barras = ['lightcoral', 'gold', 'lightgreen']

# Graficar un gr谩fico de barras para datos ordinales con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(productos, conteo_calidad, color=colores_barras, edgecolor='black', linewidth=1.2)

# T铆tulo y etiquetas del eje
plt.title('Clasificaci贸n de Productos por Calidad', fontsize=16)
plt.xlabel('Calidad', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar la gr谩fica
plt.show()
````

**2. Datos Cuantitativos:**

Los datos cuantitativos representan cantidades num茅ricas y se dividen en dos subtipos: discretos y continuos.

**a. Datos Discretos:**

Los datos discretos son valores num茅ricos contables y espec铆ficos.

**Ejemplo:** N煤mero de estudiantes en una clase.

````python
import matplotlib.pyplot as plt

# Datos discretos
num_estudiantes = [20, 25, 30, 35, 40]
grados = ['Primero', 'Segundo', 'Tercero', 'Cuarto', 'Quinto']

# Colores personalizados para las barras
colores_barras = ['lightblue', 'royalblue', 'cornflowerblue', 'steelblue', 'navy']

# Graficar un gr谩fico de barras para datos discretos con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(grados, num_estudiantes, color=colores_barras, edgecolor='black', linewidth=1.2)

# T铆tulo y etiquetas del eje
plt.title('N煤mero de Estudiantes por Grado', fontsize=16)
plt.xlabel('Grados', fontsize=12)
plt.ylabel('N煤mero de Estudiantes', fontsize=12)

# Mostrar la gr谩fica
plt.show()
````

**b. Datos Continuos:**

Los datos continuos pueden tomar cualquier valor dentro de un rango y son medibles.

**Ejemplo:** Altura de los estudiantes.

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos continuos
altura_estudiantes = np.random.normal(loc=170, scale=5, size=100)

# Graficar un histograma para datos continuos con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(altura_estudiantes, bins=20, color='mediumorchid', edgecolor='black', alpha=0.7)

# L铆nea vertical para la media
media_altura_estudiantes = np.mean(altura_estudiantes)
plt.axvline(x=media_altura_estudiantes, color='darkorange', linestyle='dashed', linewidth=2, label=f'Media = {media_altura_estudiantes:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Distribuci贸n de Altura de Estudiantes', fontsize=16)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
````

Estos ejemplos y gr谩ficos proporcionan una introducci贸n a los tipos de datos cualitativos y cuantitativos, y c贸mo visualizarlos en Python. La elecci贸n entre gr谩ficos de barras y histogramas depende del tipo de datos que est茅s representando.

## **Descriptiva vs. Inferencial**

### **Diferencias entre estad铆sticas descriptivas e inferenciales**

Las estad铆sticas descriptivas e inferenciales son dos ramas principales de la estad铆stica que se utilizan para analizar y comprender datos. A continuaci贸n, explicar茅 las diferencias entre ambas y proporcionar茅 ejemplos con gr谩ficos en c贸digo Python.

**Estad铆sticas Descriptivas:**

Las estad铆sticas descriptivas se centran en describir y resumir las caracter铆sticas fundamentales de un conjunto de datos. Estas medidas proporcionan informaci贸n sobre la tendencia central, la dispersi贸n y la forma de la distribuci贸n de los datos.

Ejemplo en Python:

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos aleatorio
datos = np.random.normal(loc=50, scale=10, size=1000)

# Estad铆sticas descriptivas
media = np.mean(datos)
mediana = np.median(datos)
desviacion_estandar = np.std(datos)

# Graficar un histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# L铆nea vertical para la media
plt.axvline(x=media, color='orange', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')

# L铆nea vertical para la mediana
plt.axvline(x=mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Histograma de Datos Aleatorios', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()

# Imprimir estad铆sticas descriptivas
print(f'Media: {media}')
print(f'Mediana: {mediana}')
print(f'Desviaci贸n Est谩ndar: {desviacion_estandar}')
````

En este ejemplo, se genera un conjunto de datos aleatorio y se calculan la media, la mediana y la desviaci贸n est谩ndar. El histograma proporciona una visualizaci贸n de la distribuci贸n de los datos.

**Estad铆sticas Inferenciales:**

Las estad铆sticas inferenciales se utilizan para hacer inferencias o generalizaciones sobre una poblaci贸n bas谩ndose en una muestra de datos. Esto implica utilizar m茅todos como pruebas de hip贸tesis y estimaci贸n de intervalos de confianza.

Ejemplo en Python:

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Crear dos muestras para comparar
muestra1 = np.random.normal(loc=50, scale=10, size=100)
muestra2 = np.random.normal(loc=55, scale=10, size=100)

# Prueba t de dos muestras
t_stat, p_valor = stats.ttest_ind(muestra1, muestra2)

# Boxplot para visualizar las muestras con personalizaciones
plt.figure(figsize=(10, 6))
plt.boxplot([muestra1, muestra2], labels=['Muestra 1', 'Muestra 2'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='orange'))

# L铆nea horizontal para el valor medio de la muestra 1
plt.axhline(y=np.mean(muestra1), color='blue', linestyle='dashed', linewidth=2, label=f'Media Muestra 1 = {np.mean(muestra1):.2f}')

# L铆nea horizontal para el valor medio de la muestra 2
plt.axhline(y=np.mean(muestra2), color='red', linestyle='dashed', linewidth=2, label=f'Media Muestra 2 = {np.mean(muestra2):.2f}')

# T铆tulo y etiquetas del eje
plt.title('Comparaci贸n de Muestras', fontsize=16)
plt.ylabel('Valor', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()

# Imprimir resultados de la prueba t
print(f'Estad铆stico t: {t_stat}')
print(f'Valor p: {p_valor}')
````

En este ejemplo, se comparan dos muestras utilizando una prueba t de dos muestras. El boxplot ayuda a visualizar las diferencias entre las muestras.

En resumen, las estad铆sticas descriptivas se utilizan para resumir y describir datos, mientras que las estad铆sticas inferenciales se utilizan para hacer inferencias sobre poblaciones bas谩ndose en muestras. Ambas son fundamentales para comprender y analizar conjuntos de datos.

### **Aplicaciones pr谩cticas**

Vamos a explorar aplicaciones pr谩cticas de estad铆sticas descriptivas e inferenciales utilizando ejemplos con gr谩ficos en Python.

**1. Estad铆sticas Descriptivas: An谩lisis de Ventas Mensuales**

Supongamos que tenemos datos mensuales de ventas de una tienda. Queremos utilizar estad铆sticas descriptivas para entender la tendencia y la variabilidad de las ventas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ventas mensuales
ventas_mensuales = np.random.normal(loc=50000, scale=10000, size=12)

# Estad铆sticas descriptivas
media_ventas = np.mean(ventas_mensuales)
desviacion_ventas = np.std(ventas_mensuales)

# Definir los meses
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Gr谩fico de barras con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(meses, ventas_mensuales, color='skyblue', edgecolor='black')

# L铆nea horizontal para la media de las ventas
plt.axhline(y=media_ventas, color='orange', linestyle='dashed', linewidth=2, label=f'Media de Ventas = {media_ventas:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Ventas Mensuales', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()

# Imprimir estad铆sticas descriptivas
print(f'Media de Ventas: {media_ventas}')
print(f'Desviaci贸n Est谩ndar de Ventas: {desviacion_ventas}')
````

En este ejemplo, utilizamos estad铆sticas descriptivas para calcular la media y la desviaci贸n est谩ndar de las ventas mensuales. El gr谩fico de barras muestra la variaci贸n en las ventas a lo largo de los meses.

**2. Estad铆sticas Inferenciales: Prueba A/B en una P谩gina Web**

Supongamos que una empresa realiza cambios en su p谩gina web y quiere evaluar si el nuevo dise帽o aumenta la tasa de clics en un bot贸n espec铆fico en comparaci贸n con el dise帽o anterior.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de la tasa de clics para el dise帽o anterior y nuevo
dise帽o_anterior = np.random.normal(loc=0.1, scale=0.02, size=1000)
dise帽o_nuevo = np.random.normal(loc=0.12, scale=0.02, size=1000)

# Prueba t para comparar las tasas de clics
t_stat, p_valor = stats.ttest_ind(dise帽o_anterior, dise帽o_nuevo)

# Boxplot con personalizaciones
plt.figure(figsize=(10, 6))
plt.boxplot([dise帽o_anterior, dise帽o_nuevo], labels=['Dise帽o Anterior', 'Nuevo Dise帽o'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='orange'))

# L铆nea horizontal para la media del dise帽o anterior
plt.axhline(y=np.mean(dise帽o_anterior), color='blue', linestyle='dashed', linewidth=2, label=f'Media Dise帽o Anterior = {np.mean(dise帽o_anterior):.2f}')

# L铆nea horizontal para la media del nuevo dise帽o
plt.axhline(y=np.mean(dise帽o_nuevo), color='red', linestyle='dashed', linewidth=2, label=f'Media Nuevo Dise帽o = {np.mean(dise帽o_nuevo):.2f}')

# T铆tulo y etiquetas del eje
plt.title('Comparaci贸n de Tasas de Clics', fontsize=16)
plt.ylabel('Tasa de Clics', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()

# Imprimir resultados de la prueba t
print(f'Estad铆stico t: {t_stat}')
print(f'Valor p: {p_valor}')
````

En este ejemplo, utilizamos una prueba t para comparar las tasas de clics entre el dise帽o anterior y el nuevo. El boxplot muestra las distribuciones y las diferencias entre los dos dise帽os.

Estas aplicaciones representan solo dos ejemplos de c贸mo se pueden utilizar las estad铆sticas descriptivas e inferenciales en situaciones pr谩cticas para comprender datos y tomar decisiones informadas.
