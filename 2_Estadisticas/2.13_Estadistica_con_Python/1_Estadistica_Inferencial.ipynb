**Ejemplo 1: Estimaci贸n de un par谩metro**

Supongamos que tenemos una poblaci贸n de 1000 estudiantes y queremos estimar la altura promedio de todos los estudiantes en la universidad. En lugar de medir a todos los estudiantes, tomamos una muestra aleatoria de 100 estudiantes y medimos sus alturas.

**Ejemplo 2: Intervalo de confianza**

Continuando con el ejemplo anterior, adem谩s de la estimaci贸n puntual, queremos calcular un intervalo de confianza del 95% para la altura promedio de todos los estudiantes en la universidad.

```python
from scipy import stats

# Calculamos el intervalo de confianza al 95% para la altura promedio
intervalo_confianza = stats.norm.interval(0.95, loc=altura_promedio_estimada, scale=altura_muestra.std()/np.sqrt(len(altura_muestra)))

print("Intervalo de confianza al 95%:", intervalo_confianza)
```


**Ejemplo 3: Prueba de hip贸tesis**

Supongamos que tenemos dos grupos de estudiantes, uno que tom贸 un curso de preparaci贸n para un examen y otro que no. Queremos determinar si el curso de preparaci贸n tuvo un impacto significativo en las calificaciones del examen.

```python
# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

print("Resultado de la prueba de hip贸tesis:", resultado_prueba)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

# Crear histograma para visualizar las calificaciones en ambos grupos
plt.figure(figsize=(10, 6))
plt.hist(grupo_control, bins=15, alpha=0.7, color='blue', label='Grupo Control', edgecolor='black')
plt.hist(grupo_preparacion, bins=15, alpha=0.7, color='orange', label='Grupo Preparaci贸n', edgecolor='black')
plt.xlabel('Calificaciones', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Distribuci贸n de Calificaciones en Ambos Grupos', fontsize=14)
plt.legend()

# Mostrar el resultado de la prueba de hip贸tesis en la gr谩fica
test_result_text = f'Diferencia significativa (p={resultado_prueba.pvalue:.3f})' if resultado_prueba.pvalue < 0.05 else f'No hay diferencia significativa (p={resultado_prueba.pvalue:.3f})'
plt.annotate(test_result_text, xy=(0.5, 0.9), xycoords='axes fraction', ha='center', va='center',
             fontsize=12, color='red' if resultado_prueba.pvalue < 0.05 else 'green', bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'))

plt.show()
```

En este ejemplo, la funci贸n `ttest_ind` realiza una prueba t de Student para determinar si hay una diferencia significativa entre las medias de los dos grupos.

Estos son solo ejemplos b谩sicos para ilustrar los fundamentos de la Estad铆stica Inferencial con Python. La Estad铆stica Inferencial es una herramienta poderosa para tomar decisiones basadas en datos y extraer informaci贸n valiosa de muestras para hacer inferencias sobre poblaciones m谩s grandes.

## **Estad铆stica descriptiva e inferencial**

**Estad铆stica Descriptiva:**

La Estad铆stica Descriptiva es una rama de la estad铆stica que se centra en resumir y describir datos de una manera concisa y significativa. Su objetivo principal es organizar y presentar la informaci贸n de manera que sea m谩s f谩cil de entender y analizar, sin hacer inferencias sobre una poblaci贸n m谩s amplia. Algunas de las t茅cnicas comunes utilizadas en la Estad铆stica Descriptiva incluyen medidas de tendencia central, medidas de dispersi贸n y gr谩ficos.

**Ejemplo:** Puntuaciones de un examen

Supongamos que tenemos las siguientes puntuaciones (en escala de 100) de los estudiantes en un examen:

85, 90, 78, 92, 80, 88, 95, 87, 82, 89

Para realizar un an谩lisis descriptivo, podemos calcular medidas como la media, la mediana y la desviaci贸n est谩ndar:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de las puntuaciones de los estudiantes
puntuaciones = [85, 90, 78, 92, 80, 88, 95, 87, 82, 89]

# Calculamos la media y la mediana
media = np.mean(puntuaciones)
mediana = np.median(puntuaciones)

# Calculamos la desviaci贸n est谩ndar
desviacion_estandar = np.std(puntuaciones)

# Mostramos los resultados
print("Media:", media)
print("Mediana:", mediana)
print("Desviaci贸n Est谩ndar:", desviacion_estandar)

# Crear un histograma para visualizar las puntuaciones
plt.hist(puntuaciones, bins=5, edgecolor='black')
plt.xlabel('Puntuaciones')
plt.ylabel('Frecuencia')
plt.title('Distribuci贸n de Puntuaciones en el Examen')
plt.axvline(media, color='red', linestyle='--', label='Media')
plt.axvline(mediana, color='green', linestyle='--', label='Mediana')
plt.legend()
plt.show()
```

En este ejemplo, calculamos la media (86.6), la mediana (87), y la desviaci贸n est谩ndar (5.78) para resumir la distribuci贸n de las puntuaciones. Adem谩s, hemos creado un histograma que muestra la frecuencia de las puntuaciones, junto con las l铆neas verticales para representar la media y la mediana.

**Ejemplo:** Estimaci贸n de la proporci贸n de estudiantes que aprobar谩n el examen

Supongamos que queremos estimar la proporci贸n de estudiantes en toda la escuela que aprobar谩n el examen (puntuaci贸n igual o superior a 60). Para hacerlo, tomamos una muestra aleatoria de 50 estudiantes y registramos si aprobaron o no:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la muestra: 1 representa aprobado, 0 representa no aprobado
muestra = np.random.choice([0, 1], size=50, p=[0.4, 0.6])

# Calculamos la proporci贸n de estudiantes que aprobaron en la muestra
proporcion_aprobados_muestra = np.mean(muestra)

# Mostramos el resultado
print("Proporci贸n de estudiantes que aprobaron en la muestra:", proporcion_aprobados_muestra)

# Crear un gr谩fico de barras para visualizar la proporci贸n de aprobados en la muestra
plt.bar(["Aprobados", "No Aprobados"], [proporcion_aprobados_muestra, 1 - proporcion_aprobados_muestra])
plt.ylabel('Proporci贸n')
plt.title('Proporci贸n de Estudiantes que Aprobaron en la Muestra')
plt.show()
```

En este ejemplo, hemos estimado la proporci贸n de estudiantes que aprobaron en nuestra muestra (la cual puede variar en diferentes ejecuciones debido al muestreo aleatorio). Para hacer una inferencia sobre la poblaci贸n completa, necesitar铆amos realizar una estimaci贸n puntual o un intervalo de confianza.  **Prueba t de Student con SciPy**

Supongamos que tenemos dos grupos de estudiantes, uno que tom贸 un curso de preparaci贸n para un examen y otro que no. Queremos determinar si hay una diferencia significativa en las calificaciones del examen entre los dos grupos.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

# Crear histogramas para visualizar las calificaciones en ambos grupos
plt.figure(figsize=(10, 6))  # Ajustar el tama帽o del gr谩fico
plt.hist(grupo_control, bins=15, alpha=0.7, color='skyblue', label='Grupo Control', edgecolor='black')
plt.hist(grupo_preparacion, bins=15, alpha=0.7, color='salmon', label='Grupo Preparaci贸n', edgecolor='black')
plt.xlabel('Calificaciones', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Distribuci贸n de Calificaciones en Ambos Grupos', fontsize=14)
plt.axvline(np.mean(grupo_control), color='blue', linestyle='dashed', linewidth=1, label='Media Control')
plt.axvline(np.mean(grupo_preparacion), color='red', linestyle='dashed', linewidth=1, label='Media Preparaci贸n')
plt.legend()

# Mostrar el resultado de la prueba de hip贸tesis en la gr谩fica
resultado_texto = 'Diferencia significativa' if resultado_prueba.pvalue < 0.05 else 'No hay diferencia significativa'
color_texto = 'red' if resultado_prueba.pvalue < 0.05 else 'green'
plt.text(55, 10, resultado_texto, color=color_texto, fontsize=12)  # Ajustar la posici贸n del texto seg煤n necesidad

plt.show()
```

En este ejemplo, utilizamos la funci贸n `ttest_ind` de SciPy para realizar una prueba t de Student y determinar si hay una diferencia significativa entre las medias de las calificaciones de los dos grupos. Luego, utilizamos matplotlib para crear histogramas que representan las distribuciones de calificaciones en ambos grupos, y se muestra en la gr谩fica el resultado de la prueba de hip贸tesis.
