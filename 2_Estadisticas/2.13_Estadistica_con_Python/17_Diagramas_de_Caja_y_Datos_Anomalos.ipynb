## **Creaci贸n de diagramas de caja**

Un diagrama de caja, tambi茅n conocido como diagrama de caja y bigotes o boxplot en ingl茅s, es una herramienta gr谩fica utilizada en estad铆sticas para representar la distribuci贸n y la dispersi贸n de un conjunto de datos. A continuaci贸n, te proporcionar茅 una explicaci贸n detallada, ejemplos y gr谩ficos para ilustrar la creaci贸n y la interpretaci贸n de diagramas de caja.

**Explicaci贸n:**

Un diagrama de caja se compone de varios elementos que representan diferentes estad铆sticas resumen de un conjunto de datos:

1. **L铆nea central:** Representa la mediana, que es el valor que divide el conjunto de datos en dos partes iguales, con la mitad de los valores por encima y la mitad por debajo.

2. **Caja:** La caja abarca el rango intercuart铆lico (IQR), que se extiende desde el primer cuartil (Q1) hasta el tercer cuartil (Q3). El IQR contiene el 50% central de los datos.

3. **Bigotes:** Los bigotes se extienden desde los bordes de la caja hasta los valores extremos dentro de un rango definido. Los valores at铆picos se representan como puntos individuales fuera de los bigotes.

4. **Valores at铆picos:** Son valores que est谩n significativamente lejos del resto de los datos y que pueden indicar la presencia de datos an贸malos.

**Ejemplo: Diagrama de Caja para Datos de Calificaciones**

Supongamos que tienes un conjunto de datos de calificaciones de un examen. Vamos a crear un diagrama de caja para visualizar la distribuci贸n de estas calificaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
grades = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100, 120])

# Crear un diagrama de caja
plt.boxplot(grades)
plt.xlabel('Calificaciones')
plt.title('Diagrama de Caja: Calificaciones')
plt.show()
```

En este ejemplo, el diagrama de caja muestra una l铆nea central que representa la mediana, una caja que abarca el IQR y bigotes que se extienden hasta los valores extremos. El valor 120 se representa como un valor at铆pico fuera de los bigotes.

**Interpretaci贸n del Diagrama de Caja:**

- La mediana se encuentra alrededor de 92.
- La caja abarca desde aproximadamente 80 a 98 (Q1 a Q3).
- Los bigotes se extienden desde 65 hasta 100.

El diagrama de caja te proporciona una visualizaci贸n r谩pida de la distribuci贸n, la mediana, los cuartiles y la presencia de valores at铆picos en los datos de calificaciones.

Los diagramas de caja son 煤tiles para comparar distribuciones entre diferentes grupos de datos y para identificar la presencia de valores at铆picos. Son especialmente valiosos en an谩lisis exploratorios y en la identificaci贸n de caracter铆sticas clave de un conjunto de datos.

## **Regla emp铆rica 68-95-99.7**

La Regla Emp铆rica, tambi茅n conocida como la Regla 68-95-99.7, es una gu铆a aproximada que se aplica a distribuciones de datos que siguen una distribuci贸n normal (tambi茅n llamada distribuci贸n gaussiana o campana de Gauss). Esta regla establece que en una distribuci贸n normal:

- Aproximadamente el 68% de los datos caen dentro de una desviaci贸n est谩ndar de la media.
- Aproximadamente el 95% de los datos caen dentro de dos desviaciones est谩ndar de la media.
- Aproximadamente el 99.7% de los datos caen dentro de tres desviaciones est谩ndar de la media.

Esta regla es 煤til para entender c贸mo se distribuyen los datos alrededor de la media en una distribuci贸n normal y proporciona una forma r谩pida de estimar la proporci贸n de datos en diferentes rangos de desviaciones est谩ndar.

**Explicaci贸n:**

Supongamos que tienes un conjunto de datos que sigue una distribuci贸n normal. La distribuci贸n normal tiene una forma de campana con la media en el centro y simetr铆a en ambos lados. La Regla Emp铆rica se basa en la propiedad de la distribuci贸n normal de que ciertos porcentajes espec铆ficos de datos caen dentro de ciertos intervalos alrededor de la media.

**Ejemplo:**

Imagina que est谩s analizando las alturas de un grupo de personas y encuentras que la media de altura es 170 cm y la desviaci贸n est谩ndar es 10 cm. Vamos a aplicar la Regla Emp铆rica a este conjunto de datos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos de ejemplo que siguen una distribuci贸n normal
np.random.seed(0)
heights = np.random.normal(170, 10, 1000)

# Crear un histograma de los datos
plt.figure(figsize=(10, 6))
plt.hist(heights, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black', label='Alturas')

# Calcular las desviaciones est谩ndar y la media
media = np.mean(heights)
std_dev = np.std(heights)
std_dev_2 = 2 * std_dev
std_dev_3 = 3 * std_dev

# Dibujar l铆neas para mostrar los intervalos de desviaci贸n est谩ndar
colors = ['red', 'orange', 'green', 'purple']
labels = ['Media', '1 Desviaci贸n Est谩ndar', '2 Desviaciones Est谩ndar', '3 Desviaciones Est谩ndar']
for i, std in enumerate([0, std_dev, std_dev_2, std_dev_3]):
    plt.axvline(media - std, color=colors[i], linestyle='dashed', linewidth=2, label=f'-{labels[i]}')
    if std != 0:  # No repetir la l铆nea de la media
        plt.axvline(media + std, color=colors[i], linestyle='dashed', linewidth=2, label=f'+{labels[i]}')

# Superponer la curva de distribuci贸n normal
x = np.linspace(media - 3*std_dev, media + 3*std_dev, 1000)
plt.plot(x, norm.pdf(x, media, std_dev), 'm', linewidth=2, label='Curva Normal')

# Agregar leyendas, etiquetas y t铆tulo
plt.xlabel('Alturas (cm)', fontsize=12)
plt.ylabel('Frecuencia Relativa', fontsize=12)
plt.title('Distribuci贸n de Alturas con Regla Emp铆rica', fontsize=14)
plt.legend()

plt.show()
```

En el gr谩fico, ver谩s la distribuci贸n de alturas en forma de un histograma. Las l铆neas verticales punteadas marcan la media y diferentes intervalos de desviaciones est谩ndar alrededor de la media. Puedes observar c贸mo se cumplen aproximadamente los porcentajes indicados por la Regla Emp铆rica.

**Interpretaci贸n del Ejemplo:**

- Aproximadamente el 68% de las alturas caen dentro de una desviaci贸n est谩ndar (de 160 cm a 180 cm).
- Aproximadamente el 95% de las alturas caen dentro de dos desviaciones est谩ndar (de 150 cm a 190 cm).
- Aproximadamente el 99.7% de las alturas caen dentro de tres desviaciones est谩ndar (de 140 cm a 200 cm).

La Regla Emp铆rica es una herramienta 煤til para obtener una comprensi贸n r谩pida de c贸mo se distribuyen los datos en una distribuci贸n normal y para hacer estimaciones aproximadas de la proporci贸n de datos en diferentes intervalos.

## **Prueba de Tukey para detectar datos an贸malos**

La prueba de Tukey, tambi茅n conocida como el m茅todo de Tukey para detectar valores at铆picos, es una t茅cnica estad铆stica utilizada para identificar valores an贸malos en un conjunto de datos. Esta prueba se basa en el rango intercuart铆lico (IQR) y ayuda a identificar observaciones que se consideran at铆picas en comparaci贸n con el resto de los datos. La prueba de Tukey es especialmente 煤til cuando se trabaja con distribuciones que no son necesariamente normales.

**Explicaci贸n:**

La prueba de Tukey se basa en el rango intercuart铆lico (IQR), que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de un conjunto de datos. Los valores at铆picos se identifican como aquellos que est谩n por debajo de Q1 - 1.5 * IQR o por encima de Q3 + 1.5 * IQR.

**Pasos para Aplicar la Prueba de Tukey:**

1. Calcular el IQR: IQR = Q3 - Q1.
2. Calcular los l铆mites inferior y superior: L铆mite Inferior = Q1 - 1.5 * IQR, L铆mite Superior = Q3 + 1.5 * IQR.
3. Identificar valores que est茅n por debajo del l铆mite inferior o por encima del l铆mite superior como valores at铆picos.

**Ejemplo: Prueba de Tukey para Detectar Valores At铆picos**

Supongamos que tienes los siguientes datos de calificaciones de un examen:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
grades = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100, 120, 130])

# Calcular cuartiles
q1 = np.percentile(grades, 25)
q3 = np.percentile(grades, 75)

# Calcular IQR
iqr = q3 - q1

# Calcular l铆mites
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

# Identificar valores at铆picos
outliers = [x for x in grades if x < lower_limit or x > upper_limit]

# Crear un gr谩fico de dispersi贸n
plt.scatter(range(len(grades)), grades, color='blue', label='Calificaciones')
plt.scatter([grades.tolist().index(outlier) for outlier in outliers], outliers, color='red', label='Valores At铆picos')
plt.xlabel('Estudiantes')
plt.ylabel('Calificaciones')
plt.title('Prueba de Tukey para Valores At铆picos')
plt.legend()
plt.show()

print("Valores at铆picos:", outliers)
```

En este ejemplo, la prueba de Tukey se aplica a las calificaciones de los estudiantes. Los valores at铆picos se identifican y se resaltan en rojo en el gr谩fico de dispersi贸n.

**Interpretaci贸n del Ejemplo:**

- Los valores at铆picos identificados por la prueba de Tukey son 120 y 130, ya que est谩n por encima del l铆mite superior (118.75 + 1.5 * 15.5) y son considerados at铆picos en comparaci贸n con el resto de los datos.

La prueba de Tukey es una herramienta 煤til para identificar valores an贸malos en un conjunto de datos. Sin embargo, es importante recordar que esta prueba tiene sus limitaciones y debe usarse junto con otras t茅cnicas para una identificaci贸n m谩s precisa de valores at铆picos.
