# **17. Diagramas de Caja (BoxPlots) y Datos Anómalos (outliers) con la Prueba de Tukey en Python**

## **Qué son los datos anómalos**

Los datos anómalos, también conocidos como valores atípicos o outliers, son observaciones que difieren significativamente del resto de los datos en un conjunto. Estos valores anómalos son inusuales, inesperados o atípicos en relación con el patrón general de los datos y pueden tener un impacto desproporcionado en los resultados del análisis estadístico o en la construcción de modelos. Los datos anómalos pueden surgir debido a errores de medición, eventos inusuales o fenómenos raros.

Los datos anómalos pueden tener un efecto perjudicial en los análisis estadísticos y en la construcción de modelos porque pueden distorsionar la distribución de los datos y afectar las conclusiones que se derivan de ellos. Por lo tanto, es importante identificar y tratar adecuadamente los datos anómalos en el análisis de datos y en la construcción de modelos.

**Características de los Datos Anómalos:**

- **Valores Extremos:** Los datos anómalos tienden a ser valores que se encuentran en los extremos de la distribución de datos.

- **Desviación del Patrón:** Los valores anómalos se desvían significativamente del patrón general de los datos.

- **Impacto:** Los datos anómalos pueden tener un impacto desproporcionado en los resultados del análisis estadístico, como promedios, desviaciones estándar y correlaciones.

- **Causas:** Los datos anómalos pueden ser causados por errores de medición, eventos raros, errores de entrada de datos o fenómenos inusuales.

**Efectos de los Datos Anómalos:**

- **Distorsión Estadística:** Los datos anómalos pueden afectar las medidas de tendencia central (promedio) y de dispersión (desviación estándar), haciendo que no sean representativos de la mayoría de los datos.

- **Modelos Inexactos:** En la construcción de modelos, los datos anómalos pueden hacer que el modelo se ajuste de manera incorrecta, llevando a predicciones erróneas o poco fiables.

- **Interpretación Errónea:** Los datos anómalos pueden llevar a conclusiones incorrectas si no se tratan adecuadamente.

**Gestión de Datos Anómalos:**

- **Identificación:** La primera etapa es identificar los datos anómalos mediante técnicas estadísticas, gráficos y métodos visuales.

- **Tratamiento:** Dependiendo del contexto, los datos anómalos pueden ser eliminados, corregidos o tratados con métodos robustos.

- **Contextualización:** Es importante comprender la causa subyacente de los datos anómalos y considerar si deben ser tratados o si contienen información valiosa.

En resumen, los datos anómalos son valores que se desvían significativamente del patrón general de los datos y pueden tener un impacto negativo en análisis y modelos si no se manejan adecuadamente. Identificar y tratar los datos anómalos es una parte crucial del análisis de datos y la construcción de modelos precisos y confiables.

## **Creación de diagramas de caja**

Un diagrama de caja, también conocido como diagrama de caja y bigotes o boxplot en inglés, es una herramienta gráfica utilizada en estadísticas para representar la distribución y la dispersión de un conjunto de datos. Proporciona información sobre la mediana, los cuartiles y la presencia de valores atípicos en los datos. A continuación, te proporcionaré una explicación detallada, ejemplos y gráficos para ilustrar la creación y la interpretación de diagramas de caja.

**Explicación:**

Un diagrama de caja se compone de varios elementos que representan diferentes estadísticas resumen de un conjunto de datos:

1. **Línea central:** Representa la mediana, que es el valor que divide el conjunto de datos en dos partes iguales, con la mitad de los valores por encima y la mitad por debajo.

2. **Caja:** La caja abarca el rango intercuartílico (IQR), que se extiende desde el primer cuartil (Q1) hasta el tercer cuartil (Q3). El IQR contiene el 50% central de los datos.

3. **Bigotes:** Los bigotes se extienden desde los bordes de la caja hasta los valores extremos dentro de un rango definido. Los valores atípicos se representan como puntos individuales fuera de los bigotes.

4. **Valores atípicos:** Son valores que están significativamente lejos del resto de los datos y que pueden indicar la presencia de datos anómalos.

**Ejemplo: Diagrama de Caja para Datos de Calificaciones**

Supongamos que tienes un conjunto de datos de calificaciones de un examen. Vamos a crear un diagrama de caja para visualizar la distribución de estas calificaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
grades = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100, 120])

# Crear un diagrama de caja
plt.boxplot(grades)
plt.xlabel('Calificaciones')
plt.title('Diagrama de Caja: Calificaciones')
plt.show()
```

En este ejemplo, el diagrama de caja muestra una línea central que representa la mediana, una caja que abarca el IQR y bigotes que se extienden hasta los valores extremos. El valor 120 se representa como un valor atípico fuera de los bigotes.

**Interpretación del Diagrama de Caja:**

- La mediana se encuentra alrededor de 92.
- La caja abarca desde aproximadamente 80 a 98 (Q1 a Q3).
- Los bigotes se extienden desde 65 hasta 100.
- El valor 120 es un valor atípico.

El diagrama de caja te proporciona una visualización rápida de la distribución, la mediana, los cuartiles y la presencia de valores atípicos en los datos de calificaciones.

Los diagramas de caja son útiles para comparar distribuciones entre diferentes grupos de datos y para identificar la presencia de valores atípicos. Son especialmente valiosos en análisis exploratorios y en la identificación de características clave de un conjunto de datos.

## **Boxplot mostrando datos anómalos**

Por supuesto, aquí tienes una explicación detallada junto con un ejemplo y gráficos que muestran datos anómalos en un diagrama de caja.

**Explicación:**

Un diagrama de caja, también conocido como boxplot, es una herramienta gráfica que representa la distribución estadística de un conjunto de datos. Muestra la mediana, los cuartiles, los valores atípicos y la dispersión general de los datos. Los valores atípicos, también llamados outliers, son puntos que se encuentran lejos del resto de los datos y que pueden afectar la interpretación del conjunto.

**Ejemplo: Boxplot con Datos Anómalos**

Supongamos que tienes datos de calificaciones en un examen, y algunos estudiantes obtuvieron calificaciones extremadamente altas. Vamos a crear un boxplot que muestre estos datos anómalos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo con valores atípicos
grades = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100, 150, 160, 170])

# Crear un boxplot
plt.boxplot(grades)
plt.xlabel('Calificaciones')
plt.title('Boxplot con Datos Anómalos')
plt.show()
```

En este ejemplo, el boxplot muestra los siguientes elementos:

- La línea central representa la mediana, que se encuentra alrededor de 92.
- La caja abarca desde aproximadamente 80 a 98 (Q1 a Q3).
- Los bigotes se extienden desde alrededor de 65 hasta 100.
- Los valores 150, 160 y 170 se representan como valores atípicos fuera de los bigotes.

**Interpretación del Boxplot:**

- El boxplot indica la presencia de valores atípicos en las calificaciones más altas (150, 160 y 170), que se extienden más allá de los bigotes.
- Estos valores atípicos pueden haber influido en la mediana y en la longitud de los bigotes.

**Uso de Boxplot con Datos Anómalos:**

- Los boxplots con valores atípicos son útiles para visualizar cómo afectan esos valores extremos a la distribución y los estadísticos resumen.
- Pueden ayudar a tomar decisiones sobre cómo tratar los valores atípicos, ya sea eliminándolos, investigando sus causas o considerándolos en análisis específicos.

En resumen, el boxplot con datos anómalos te brinda una forma efectiva de visualizar cómo los valores atípicos influyen en la distribución y los resúmenes estadísticos de tus datos.

## **Regla empírica 68-95-99.7**

La Regla Empírica, también conocida como la Regla 68-95-99.7, es una guía aproximada que se aplica a distribuciones de datos que siguen una distribución normal (también llamada distribución gaussiana o campana de Gauss). Esta regla establece que en una distribución normal:

- Aproximadamente el 68% de los datos caen dentro de una desviación estándar de la media.
- Aproximadamente el 95% de los datos caen dentro de dos desviaciones estándar de la media.
- Aproximadamente el 99.7% de los datos caen dentro de tres desviaciones estándar de la media.

Esta regla es útil para entender cómo se distribuyen los datos alrededor de la media en una distribución normal y proporciona una forma rápida de estimar la proporción de datos en diferentes rangos de desviaciones estándar.

**Explicación:**

Supongamos que tienes un conjunto de datos que sigue una distribución normal. La distribución normal tiene una forma de campana con la media en el centro y simetría en ambos lados. La Regla Empírica se basa en la propiedad de la distribución normal de que ciertos porcentajes específicos de datos caen dentro de ciertos intervalos alrededor de la media.

**Ejemplo:**

Imagina que estás analizando las alturas de un grupo de personas y encuentras que la media de altura es 170 cm y la desviación estándar es 10 cm. Vamos a aplicar la Regla Empírica a este conjunto de datos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos de ejemplo que siguen una distribución normal
np.random.seed(0)
heights = np.random.normal(170, 10, 1000)

# Crear un histograma de los datos
plt.figure(figsize=(10, 6))
plt.hist(heights, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black', label='Alturas')

# Calcular las desviaciones estándar y la media
media = np.mean(heights)
std_dev = np.std(heights)
std_dev_2 = 2 * std_dev
std_dev_3 = 3 * std_dev

# Dibujar líneas para mostrar los intervalos de desviación estándar
colors = ['red', 'orange', 'green', 'purple']
labels = ['Media', '1 Desviación Estándar', '2 Desviaciones Estándar', '3 Desviaciones Estándar']
for i, std in enumerate([0, std_dev, std_dev_2, std_dev_3]):
    plt.axvline(media - std, color=colors[i], linestyle='dashed', linewidth=2, label=f'-{labels[i]}')
    if std != 0:  # No repetir la línea de la media
        plt.axvline(media + std, color=colors[i], linestyle='dashed', linewidth=2, label=f'+{labels[i]}')

# Superponer la curva de distribución normal
x = np.linspace(media - 3*std_dev, media + 3*std_dev, 1000)
plt.plot(x, norm.pdf(x, media, std_dev), 'm', linewidth=2, label='Curva Normal')

# Agregar leyendas, etiquetas y título
plt.xlabel('Alturas (cm)', fontsize=12)
plt.ylabel('Frecuencia Relativa', fontsize=12)
plt.title('Distribución de Alturas con Regla Empírica', fontsize=14)
plt.legend()

plt.show()
```

En el gráfico, verás la distribución de alturas en forma de un histograma. Las líneas verticales punteadas marcan la media y diferentes intervalos de desviaciones estándar alrededor de la media. Puedes observar cómo se cumplen aproximadamente los porcentajes indicados por la Regla Empírica.

**Interpretación del Ejemplo:**

- Aproximadamente el 68% de las alturas caen dentro de una desviación estándar (de 160 cm a 180 cm).
- Aproximadamente el 95% de las alturas caen dentro de dos desviaciones estándar (de 150 cm a 190 cm).
- Aproximadamente el 99.7% de las alturas caen dentro de tres desviaciones estándar (de 140 cm a 200 cm).

La Regla Empírica es una herramienta útil para obtener una comprensión rápida de cómo se distribuyen los datos en una distribución normal y para hacer estimaciones aproximadas de la proporción de datos en diferentes intervalos.

## **Prueba de Tukey para detectar datos anómalos**

La prueba de Tukey, también conocida como el método de Tukey para detectar valores atípicos, es una técnica estadística utilizada para identificar valores anómalos en un conjunto de datos. Esta prueba se basa en el rango intercuartílico (IQR) y ayuda a identificar observaciones que se consideran atípicas en comparación con el resto de los datos. La prueba de Tukey es especialmente útil cuando se trabaja con distribuciones que no son necesariamente normales.

**Explicación:**

La prueba de Tukey se basa en el rango intercuartílico (IQR), que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de un conjunto de datos. Los valores atípicos se identifican como aquellos que están por debajo de Q1 - 1.5 * IQR o por encima de Q3 + 1.5 * IQR.

**Pasos para Aplicar la Prueba de Tukey:**

1. Calcular el IQR: IQR = Q3 - Q1.
2. Calcular los límites inferior y superior: Límite Inferior = Q1 - 1.5 * IQR, Límite Superior = Q3 + 1.5 * IQR.
3. Identificar valores que estén por debajo del límite inferior o por encima del límite superior como valores atípicos.

**Ejemplo: Prueba de Tukey para Detectar Valores Atípicos**

Supongamos que tienes los siguientes datos de calificaciones de un examen:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
grades = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100, 120, 130])

# Calcular cuartiles
q1 = np.percentile(grades, 25)
q3 = np.percentile(grades, 75)

# Calcular IQR
iqr = q3 - q1

# Calcular límites
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

# Identificar valores atípicos
outliers = [x for x in grades if x < lower_limit or x > upper_limit]

# Crear un gráfico de dispersión
plt.scatter(range(len(grades)), grades, color='blue', label='Calificaciones')
plt.scatter([grades.tolist().index(outlier) for outlier in outliers], outliers, color='red', label='Valores Atípicos')
plt.xlabel('Estudiantes')
plt.ylabel('Calificaciones')
plt.title('Prueba de Tukey para Valores Atípicos')
plt.legend()
plt.show()

print("Valores atípicos:", outliers)
```

En este ejemplo, la prueba de Tukey se aplica a las calificaciones de los estudiantes. Los valores atípicos se identifican y se resaltan en rojo en el gráfico de dispersión.

**Interpretación del Ejemplo:**

- Los valores atípicos identificados por la prueba de Tukey son 120 y 130, ya que están por encima del límite superior (118.75 + 1.5 * 15.5) y son considerados atípicos en comparación con el resto de los datos.

La prueba de Tukey es una herramienta útil para identificar valores anómalos en un conjunto de datos. Sin embargo, es importante recordar que esta prueba tiene sus limitaciones y debe usarse junto con otras técnicas para una identificación más precisa de valores atípicos.
