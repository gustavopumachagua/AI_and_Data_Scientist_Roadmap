## **Distribuciones de probabilidad discretas**

Las distribuciones de probabilidad discretas son aquellas en las que la variable aleatoria toma valores aislados o discretos, es decir, valores espec铆ficos dentro de un conjunto finito o numerable. Cada valor tiene asociada una probabilidad de ocurrencia.

Aqu铆 hay algunas distribuciones de probabilidad discretas comunes y ejemplos de c贸mo trabajar con ellas en Python utilizando las bibliotecas NumPy y Matplotlib:

* **Distribuci贸n de probabilidad uniforme discreta:**

En esta distribuci贸n, todos los valores tienen la misma probabilidad de ocurrencia. Es 煤til cuando todos los resultados son igualmente probables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Distribuci贸n uniforme discreta
valores = np.array([1, 2, 3, 4, 5, 6])
probabilidades = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Graficar la distribuci贸n con personalizaci贸n
plt.bar(valores, probabilidades, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.title('Distribuci贸n uniforme discreta')

# A帽adir etiquetas a las barras
for valor, probabilidad in zip(valores, probabilidades):
    plt.text(valor, probabilidad + 0.01, f'{probabilidad:.2f}', ha='center', va='bottom')

# Personalizar ejes y cuadr铆cula
plt.xticks(valores)
plt.ylim(0, 0.2)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, definimos los posibles valores de la variable aleatoria y sus respectivas probabilidades. Graficamos la distribuci贸n utilizando el m茅todo `bar()` de Matplotlib, donde cada valor se representa en el eje `x` y su probabilidad en el eje `y`. En este caso, todos los valores tienen una probabilidad igual de `1/6`.

* **Distribuci贸n de probabilidad binomial:**

La distribuci贸n binomial se utiliza para modelar el n煤mero de 茅xitos en una serie de ensayos independientes con dos posibles resultados: 茅xito o fracaso. Se caracteriza por dos par谩metros: el n煤mero de ensayos (`n`) y la probabilidad de 茅xito en cada ensayo (`p`).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Distribuci贸n binomial
n = 10  # N煤mero de ensayos
p = 0.5  # Probabilidad de 茅xito

# Generar los posibles valores
valores = np.arange(0, n+1)
probabilidades = binom.pmf(valores, n, p)

# Graficar la distribuci贸n con personalizaci贸n
plt.bar(valores, probabilidades, color='lightblue', edgecolor='black', alpha=0.7)
plt.xlabel('N煤mero de 茅xitos')
plt.ylabel('Probabilidad')
plt.title('Distribuci贸n binomial')

# Personalizar ejes y cuadr铆cula
plt.xticks(valores)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, utilizamos la funci贸n `binom.pmf()` de la biblioteca SciPy para calcular las probabilidades de los posibles valores en una distribuci贸n binomial con 10 ensayos y una probabilidad de 茅xito de 0.5. ## **Distribuci贸n Uniforme Discreta**

La distribuci贸n uniforme discreta es una distribuci贸n de probabilidad en la que todos los valores posibles tienen la misma probabilidad de ocurrencia. Se utiliza cuando todos los resultados son igualmente probables.

En Python, puedes trabajar con la distribuci贸n uniforme discreta utilizando la biblioteca NumPy y la funci贸n `numpy.random.randint()`. A continuaci贸n se muestra un ejemplo de c贸mo generar una muestra aleatoria de una distribuci贸n uniforme discreta y graficar la distribuci贸n utilizando Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Par谩metros de la distribuci贸n uniforme discreta
valores = np.arange(1, 7)  # Valores posibles: 1, 2, 3, 4, 5, 6
probabilidad = 1 / len(valores)  # Probabilidad de cada valor

# Generar una muestra aleatoria de la distribuci贸n
muestra = np.random.choice(valores, size=1000, replace=True)

# Calcular las frecuencias de cada valor en la muestra
frecuencias = np.bincount(muestra)[1:]

# Graficar la distribuci贸n uniforme discreta con personalizaci贸n
plt.bar(valores, frecuencias / len(muestra), color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.title('Distribuci贸n uniforme discreta')

# Personalizar ejes y cuadr铆cula
plt.xticks(valores)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, definimos los valores posibles de la distribuci贸n uniforme discreta utilizando `np.arange()`, que genera una secuencia de n煤meros del 1 al 6. Calculamos la probabilidad de cada valor dividiendo 1 entre el n煤mero total de valores.

Luego, generamos una muestra aleatoria de la distribuci贸n utilizando `np.random.choice()`. Especificamos el tama帽o de la muestra (en este caso, 1000) y permitimos el reemplazo (replace=True), lo que significa que un mismo valor puede aparecer varias veces en la muestra.

A continuaci贸n, utilizamos `np.bincount()` para contar las frecuencias de cada valor en la muestra. La funci贸n `bincount()` devuelve un arreglo con las frecuencias, donde el 铆ndice del arreglo corresponde al valor.

Finalmente, graficamos la distribuci贸n utilizando `plt.bar()`, donde los valores se representan en el eje `x` y las probabilidades se representan en el eje `y`. Dividimos las frecuencias entre el tama帽o de la muestra para obtener las probabilidades relativas.

El resultado es un gr谩fico de barras que muestra la distribuci贸n uniforme discreta, donde cada valor tiene la misma probabilidad de ocurrencia. En este caso, se espera que todas las barras tengan una altura similar debido a la igual probabilidad asignada a cada valor.

## **Distribuci贸n de Bernoulli**

La distribuci贸n de Bernoulli es una distribuci贸n de probabilidad discreta que modela un experimento aleatorio con dos posibles resultados: 茅xito o fracaso. Se utiliza para describir situaciones en las que un evento ocurre con una probabilidad constante `p` de 茅xito y una probabilidad de (`1-p`) de fracaso.

En Python, puedes trabajar con la distribuci贸n de Bernoulli utilizando la biblioteca NumPy y la funci贸n `numpy.random.binomial()`. A continuaci贸n se muestra un ejemplo de c贸mo generar una muestra aleatoria de una distribuci贸n de Bernoulli y graficar la distribuci贸n utilizando Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Par谩metro de la distribuci贸n de Bernoulli
p = 0.3  # Probabilidad de 茅xito

# Generar una muestra aleatoria de la distribuci贸n
muestra = np.random.binomial(n=1, p=p, size=1000)

# Calcular las frecuencias de 茅xito y fracaso en la muestra
frecuencia_exito = np.sum(muestra == 1)
frecuencia_fracaso = np.sum(muestra == 0)

# Graficar la distribuci贸n de Bernoulli con personalizaci贸n
plt.bar(['xito', 'Fracaso'], [frecuencia_exito, frecuencia_fracaso], color=['green', 'red'], alpha=0.7, edgecolor='black')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title('Distribuci贸n de Bernoulli')

# Personalizar el texto en las barras
plt.text(0, frecuencia_exito + 10, f'{frecuencia_exito}', ha='center', va='bottom', color='black', fontweight='bold')
plt.text(1, frecuencia_fracaso + 10, f'{frecuencia_fracaso}', ha='center', va='bottom', color='black', fontweight='bold')

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, especificamos el par谩metro `p` que representa la probabilidad de 茅xito en la distribuci贸n de Bernoulli. Generamos una muestra aleatoria utilizando `np.random.binomial()`, donde especificamos el tama帽o de la muestra (`size=1000`), el n煤mero de ensayos (`n=1`) y la probabilidad de 茅xito (`p`).

Luego, calculamos las frecuencias de 茅xito y fracaso en la muestra contando el n煤mero de ocurrencias de cada resultado utilizando `np.sum()` y la comparaci贸n `muestra == 1` y `muestra == 0`.

Finalmente, graficamos la distribuci贸n utilizando `plt.bar()`, donde el eje `x` muestra los resultados posibles ('`xito`' y '`Fracaso`') y el eje `y` muestra las frecuencias correspondientes. El gr谩fico de barras muestra la distribuci贸n de Bernoulli, donde la altura de la barra de 茅xito representa la frecuencia de 茅xito y la altura de la barra de fracaso representa la frecuencia de fracaso.

La distribuci贸n de Bernoulli es un caso especial de la distribuci贸n binomial con un solo ensayo. Es 煤til para modelar eventos binarios y entender la probabilidad de ocurrencia de un resultado en particular.

En Python, puedes trabajar con la distribuci贸n binomial utilizando la biblioteca NumPy y la funci贸n `numpy.random.binomial()`. A continuaci贸n se muestra un ejemplo de c贸mo generar una muestra aleatoria de una distribuci贸n binomial y graficar la distribuci贸n utilizando Matplotlib:

En este ejemplo, especificamos los par谩metros de la distribuci贸n binomial: el n煤mero de ensayos (`n`) y la probabilidad de 茅xito en cada ensayo (`p`). Luego, generamos una muestra aleatoria utilizando `np.random.binomial()`, donde especificamos el tama帽o de la muestra (`size=1000`), el n煤mero de ensayos (`n`) y la probabilidad de 茅xito (`p`).

Despu茅s de obtener la muestra, utilizamos `np.bincount()` para contar las frecuencias de cada n煤mero de 茅xitos en la muestra. La funci贸n `bincount()` devuelve un arreglo con las frecuencias, donde el 铆ndice del arreglo corresponde al n煤mero de 茅xitos.

La distribuci贸n de Poisson es una distribuci贸n de probabilidad discreta que se utiliza para modelar eventos raros que ocurren de manera aleatoria en un intervalo de tiempo o espacio fijo.
```python
import numpy as np
import matplotlib.pyplot as plt

# Par谩metro de la distribuci贸n de Poisson
lambda_val = 3  # Tasa media de ocurrencia

# Generar una muestra aleatoria de la distribuci贸n
muestra = np.random.poisson(lam=lambda_val, size=1000)

# Calcular las frecuencias de cada n煤mero de ocurrencias en la muestra
frecuencias = np.bincount(muestra)

# Graficar la distribuci贸n de Poisson con personalizaci贸n
plt.bar(range(len(frecuencias)), frecuencias / len(muestra), color='green', alpha=0.7, edgecolor='black')
plt.xlabel('N煤mero de ocurrencias')
plt.ylabel('Probabilidad')
plt.title('Distribuci贸n de Poisson')

# Personalizar el eje x para mostrar solo los enteros
plt.xticks(range(len(frecuencias)))

# Mostrar el gr谩fico
plt.show()
```

En este ejemplo, especificamos el par谩metro lambda (`位`) de la distribuci贸n de Poisson, que representa la tasa media de ocurrencia de los eventos raros. Luego, generamos una muestra aleatoria utilizando `np.random.poisson()`, donde especificamos el tama帽o de la muestra (`size=1000`) y el valor de lambda (`lam=lambda_val`).

Despu茅s de obtener la muestra, utilizamos `np.bincount()` para contar las frecuencias de cada n煤mero de ocurrencias en la muestra. La funci贸n `bincount()` devuelve un arreglo con las frecuencias, donde el 铆ndice del arreglo corresponde al n煤mero de ocurrencias.

El resultado es un gr谩fico de barras que muestra la distribuci贸n de Poisson, donde cada barra representa la probabilidad de tener un determinado n煤mero de ocurrencias de eventos raros en el intervalo de tiempo o espacio.
