En este ejemplo, creamos dos distribuciones normales dependientes (`A` y `B`) utilizando la biblioteca SciPy. Luego, sumamos las distribuciones `A` y `B` para obtener la distribuci贸n resultante `C`. A continuaci贸n, se presenta un ejemplo de c贸mo generar y graficar un proceso de Poisson utilizando Python y la biblioteca NumPy:
