## **2.1 Modelado estadístico: tipos de variables**

**Tipos de Variables**

**1. Variables Cualitativas o Categóricas**

Estas variables representan categorías o etiquetas que generalmente no tienen un orden matemático natural. Se dividen en:

- **Nominales**: No hay un orden implícito entre las categorías. Ejemplo: color (rojo, verde, azul), género (masculino, femenino).

- **Ordinales**: Hay un orden o jerarquía entre las categorías. Ejemplo: nivel educativo (primaria, secundaria, universidad), rango militar.

**2. Variables Cuantitativas o Numéricas**

Estas variables son medibles y se expresan numéricamente. Se pueden sumar, restar o usar en cálculos matemáticos. Se clasifican en:

- **Discretas**: Valores que se cuentan y que tienen espacios fijos entre ellos. Ejemplo: número de hijos, número de coches en un hogar.

- **Continuas**: Valores que se miden y pueden tomar cualquier valor dentro de un rango. Ejemplo: peso, altura, temperatura.

## **2.2 Descripción general de los modelos**

**Tipos de Modelos Estadísticos**

1. **Modelos Lineales**

   - **Regresión Lineal Simple**: Modelo para predecir una variable dependiente (Y) usando una variable independiente (X).

   - **Árboles de Decisión**: Modelos que utilizan una estructura de árbol para tomar decisiones.

**Ejemplo en Python: Regresión Lineal**

Vamos a crear un ejemplo de una regresión lineal simple utilizando Python. Utilizaremos `matplotlib` para visualizar los datos y `scikit-learn` para crear el modelo.

**Paso 1: Instalación de Bibliotecas**

Si no tienes las bibliotecas necesarias, puedes instalarlas con:


**Paso 2: Generación de Datos de Ejemplo y Modelo**

Aquí está el código completo para generar datos sintéticos, ajustar un modelo de regresión lineal y visualizar tanto los datos como la línea de mejor ajuste.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos de ejemplo
np.random.seed(0)
X = 2.5 * np.random.randn(100) + 1.5   # Array de 100 valores con media = 1.5, desviación estándar = 2.5
res = 0.5 * np.random.randn(100)       # Generar 100 residuos
y = 2 + 0.3 * X + res                  # Ecuación lineal real y = 2 + 0.3X + ruido

# Cambio de forma de X para hacerlo un array 2D para sklearn
X = X.reshape(-1, 1)

# Crear el objeto de regresión lineal y entrenarlo
model = LinearRegression()
model.fit(X, y)

# Predecir y calcular la línea de mejor ajuste
predictions = model.predict(X)

# Visualización de datos y modelo
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='Datos originales', alpha=0.5)
plt.plot(X, predictions, color='red', linewidth=2, label='Línea de mejor ajuste')
plt.title('Regresión Lineal Simple')
plt.xlabel('Variable Independiente X')
plt.ylabel('Variable Dependiente Y')
plt.legend()
plt.show()
````

**Explicación del Código**

- **Datos de Ejemplo**: Generamos datos aleatorios que siguen aproximadamente una relación lineal.

## **2.3 Modelado explicativo versus predictivo**

El modelado estadístico se puede clasificar en dos grandes categorías dependiendo de su objetivo: modelado explicativo y modelado predictivo. Cada tipo tiene su propósito y metodología específica, así como sus propias herramientas y enfoques para evaluar la efectividad. A continuación, te explicaré cada uno y te mostraré cómo implementarlos con ejemplos en Python.

**Modelado Explicativo**

El **modelado explicativo** se centra en entender las relaciones entre las variables y explicar el impacto de unas variables sobre otras. Este tipo de modelo es común en la investigación científica y en la política, donde es importante entender las causas y los efectos.

**Objetivo**: Determinar si y cómo unas variables específicas afectan a otras.

**Ejemplo**: Investigar cómo diferentes factores socioeconómicos afectan el rendimiento educativo de los estudiantes.

**Ejemplo de Código en Python: Regresión Lineal Múltiple**

Vamos a utilizar `statsmodels`, una biblioteca de Python para realizar regresiones y otros análisis estadísticos, para explicar cómo diferentes variables independientes pueden afectar una variable dependiente.

**Instalación de Biblioteca**

Si necesitas instalar `statsmodels`, puedes hacerlo utilizando pip:

```bash
pip install statsmodels
```

**Código Python**

````python
import statsmodels.api as sm
import numpy as np

# Generar datos de ejemplo
np.random.seed(1)
x1 = np.random.randn(100)  # Variable independiente 1
x2 = np.random.randn(100)  # Variable independiente 2
y = 1 + 2*x1 + 3*x2 + np.random.randn(100) * 0.5  # Variable dependiente con ruido

X = sm.add_constant(np.column_stack((x1, x2)))  # Añadir constante para intercepto

# Crear el modelo de regresión lineal
model = sm.OLS(y, X)
results = model.fit()

# Mostrar los resultados
print(results.summary())
````

Este código ajusta un modelo de regresión lineal múltiple para entender cómo `x1` y `x2` afectan a `y`. El resumen del modelo incluirá coeficientes para cada variable independiente, mostrando su contribución estimada a la variable dependiente.

El **modelado predictivo** utiliza datos históricos para predecir datos futuros o desconocidos. Este enfoque es común en campos como el marketing, donde predecir el comportamiento futuro de los clientes puede ayudar a diseñar mejores estrategias.

**Objetivo**: Predecir resultados futuros basándose en datos pasados.

**Ejemplo**: Pronosticar las ventas futuras basándose en datos de ventas anteriores y variables de mercado.

## **2.4 Parámetros de población y estadísticas muestrales**

Exploraremos el concepto de parámetros de población frente a estadísticas muestrales, que son fundamentales en estadística para entender cómo inferir propiedades de una población grande basándose en una muestra. Además, veremos cómo implementar ejemplos de esto en Python con visualizaciones.

**Parámetros de Población vs. Estadísticas Muestrales**

1. **Parámetros de Población**

   - Son medidas que resumen características de toda una población. Una población es el conjunto completo de todas las posibles observaciones de interés.
   - Ejemplos comunes incluyen la media poblacional (μ), la varianza poblacional (σ²) y la proporción poblacional (p).

2. **Estadísticas Muestrales**

   - Son medidas que resumen características de una muestra. Una muestra es un subconjunto seleccionado de una población.
   - Ejemplos comunes son la media muestral $(\bar{x})$, la varianza muestral (s²) y la proporción muestral $(\hat{p})$.

El objetivo de usar estadísticas muestrales es estimar los parámetros de la población, ya que a menudo no es práctico o posible medir toda la población.

**Ejemplo en Python: Estimación de Parámetros Poblacionales**

Generaremos una población sintética, tomaremos muestras de ella y calcularemos estadísticas muestrales. Luego, visualizaremos cómo estas estadísticas pueden variar de una muestra a otra y cómo se comparan con los parámetros reales de la población.

**Instalación de Bibliotecas**

Necesitaremos `numpy` y `matplotlib` para generar datos y visualizarlos.

```bash
pip install numpy matplotlib
```

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una población grande
np.random.seed(0)
population = np.random.normal(loc=50, scale=10, size=10000)  # media=50, desviación estándar=10

# Tomar múltiples muestras de la población y calcular estadísticas
sample_means = []
sample_sizes = [10, 50, 100, 500, 1000, 5000]

for size in sample_sizes:
    sample = np.random.choice(population, size, replace=False)
    sample_means.append(np.mean(sample))

# Comparar medias muestrales con la media poblacional
population_mean = np.mean(population)

plt.figure(figsize=(10, 5))
plt.plot(sample_sizes, sample_means, marker='o', linestyle='-', color='b', label='Media Muestral')
plt.axhline(y=population_mean, color='r', linestyle='--', label='Media Poblacional')
plt.xlabel('Tamaño de la Muestra')
plt.ylabel('Media')
plt.title('Comparación de la Media Muestral con la Media Poblacional')
plt.legend()
plt.grid(True)
plt.show()
````

**Explicación del Código**

- **Generación de Población**: Creamos una población con una distribución normal, media de 50 y desviación estándar de 10.
- **Toma de Muestras y Cálculo**: Tomamos múltiples muestras de diferentes tamaños y calculamos la media para cada muestra.
- **Visualización**: Graficamos cómo la media muestral se compara con la media poblacional a medida que el tamaño de la muestra aumenta.

En este gráfico, verás cómo la media muestral tiende a acercarse a la media poblacional a medida que el tamaño de la muestra aumenta, ilustrando el concepto de la Ley de los Grandes Números.

**Conclusión**

Este ejemplo ilustra cómo las estadísticas muestrales se utilizan para estimar los parámetros de la población. A medida que aumenta el tamaño de la muestra, nuestras estimaciones suelen mejorar, acercándose a los verdaderos parámetros poblacionales. Este proceso es fundamental en muchos campos aplicados de la ciencia, donde tomar decisiones basadas en datos precisos y confiables es crucial.

## **2.5 Distribución normal (gaussiana)**

La **distribución normal**, también conocida como **distribución gaussiana**, es una de las distribuciones más importantes en estadística debido a su presencia frecuente en muchos fenómenos naturales y sociales. Aquí, te proporcionaré una explicación detallada de esta distribución, ejemplos de su aparición y cómo visualizarla utilizando Python.

**Características de la Distribución Normal**

1. **Forma de Campana**: La gráfica de la distribución normal es simétrica y tiene la forma de una campana, centrada en la media $(\mu)$.

2. **Desviación Estándar $(\sigma)$**: La dispersión de los datos alrededor de la media se mide con la desviación estándar. Un $(\sigma)$ más grande significa una curva más ancha.

3. **Media, Mediana y Moda Iguales**: En una distribución perfectamente normal, la media, la mediana y la moda son todas iguales y se encuentran en el centro de la distribución.

4. **Asintótica**: Las colas de la distribución se extienden infinitamente en ambos sentidos, aproximándose al eje horizontal pero nunca tocándolo.

5. **Regla Empírica (68-95-99.7)**: Aproximadamente el 68% de los datos caen dentro de $(\sigma)$, el 95% dentro de 2$(\sigma)$, y el 99.7% dentro de 3$(\sigma)$ de la media.

**Ejemplos de Fenómenos que Siguen una Distribución Normal**

- Alturas de personas.
- Presión arterial.
- Puntajes en ciertos tipos de pruebas estandarizadas.

**Ejemplo en Python: Visualización de la Distribución Normal**

Vamos a utilizar `numpy` para generar datos y `matplotlib` para visualizar una distribución normal.

**Código Python para Generar y Visualizar una Distribución Normal**

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
media = 0
desviacion_estandar = 1

# Generar datos de una distribución normal
datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Visualización de la distribución normal
plt.figure(figsize=(12, 8))  # Cambiando el tamaño para mejor visualización
plt.hist(datos, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Superponer la curva teórica de la distribución normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - media)**2 / (2 * desviacion_estandar**2)) / (np.sqrt(2 * np.pi) * desviacion_estandar)
plt.plot(x, p, 'r-', linewidth=3, label='Modelo Normal Teórico')  # Línea roja, más gruesa

# Añadir títulos y etiquetas
plt.title("Distribución Normal - Datos vs. Modelo Teórico", fontsize=16)
plt.xlabel('Datos', fontsize=14)
plt.ylabel('Densidad', fontsize=14)

# Añadir leyenda
plt.legend(fontsize=12)

# Añadir cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
````

## **2.6 Error estandar de la media**

El **error estándar de la media** (SEM, por sus siglas en inglés) es una medida estadística que cuantifica la dispersión o variabilidad de la media muestral respecto a la media poblacional. Es fundamental en estadística inferencial porque proporciona una estimación de cuán precisamente una muestra representa a toda la población.

**Conceptos Clave del Error Estándar de la Media**

1. **Definición**: El SEM se define como la desviación estándar de la distribución de las medias muestrales. Matemáticamente, se calcula como la desviación estándar de la población $(\sigma)$ dividida por la raíz cuadrada del tamaño de la muestra $(n)$:

   $SEM = \frac{\sigma}{\sqrt{n}}$

   Si la desviación estándar de la población no es conocida (lo cual es común), se puede utilizar la desviación estándar de la muestra como una estimación.

2. **Importancia**: El SEM es crucial para calcular intervalos de confianza y realizar pruebas de hipótesis sobre la media poblacional, especialmente en el teorema del límite central.

3. **Teorema del Límite Central**: Este teorema afirma que, con un tamaño de muestra suficientemente grande, la distribución de las medias muestrales tiende a ser normalmente distribuida, independientemente de la forma de la distribución de la población, siempre que la distribución no sea extremadamente sesgada.

**Ejemplo en Python: Cálculo y Visualización del SEM**

Vamos a generar datos de una distribución conocida, calcular la media muestral y el SEM, y visualizar esto para diferentes tamaños de muestra.

**Instalación de Bibliotecas**

Si necesitas instalar `numpy` y `matplotlib`, puedes hacerlo utilizando pip:

```bash
pip install numpy matplotlib
```

**Código Python para Generar Datos, Calcular SEM y Visualizar**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribución normal
np.random.seed(0)
media_poblacional = 50
desviacion_estandar_poblacional = 10
poblacion = np.random.normal(media_poblacional, desviacion_estandar_poblacional, 10000)

# Función para calcular la media muestral y el SEM
def calcular_sem(tamano_muestra):
    muestra = np.random.choice(poblacion, tamano_muestra)
    media_muestral = np.mean(muestra)
    sem = np.std(muestra, ddof=1) / np.sqrt(tamano_muestra)
    return media_muestral, sem

# Tamaños de muestra y cálculos
tamanos_de_muestra = [10, 30, 50, 100, 200, 500]
medias_muestrales = []
sems = []

for tamano in tamanos_de_muestra:
    media_muestral, sem = calcular_sem(tamano)
    medias_muestrales.append(media_muestral)
    sems.append(sem)

# Visualización
plt.figure(figsize=(10, 5))
plt.errorbar(tamanos_de_muestra, medias_muestrales, yerr=sems, fmt='o', color='b', ecolor='r', capsize=5)
plt.axhline(y=media_poblacional, color='g', linestyle='--')
plt.title('Media Muestral y Error Estándar de la Media para Diferentes Tamaños de Muestra')
plt.xlabel('Tamaño de la Muestra')
plt.ylabel('Media Muestral')
plt.grid(True)
plt.show()
````

**Descripción del Código**

- **Generación de Datos**: Creamos una población de 10,000 individuos con una distribución normal.
- **Función para Calcular SEM**: Esta función toma una muestra de la población, calcula la media muestral y el SEM utilizando la desviación estándar de la muestra.
- **Visualización**: Usamos un gráfico de error (`errorbar`) para mostrar cómo varía la media muestral y el SEM con diferentes tamaños de muestra. La línea verde punteada representa la media de la población.

**Conclusión**

El error estándar de la media es una herramienta estadística crucial que ayuda a entender la precisión de las medias muestrales como estimadores de la media poblacional. A medida que aumenta el tamaño de la muestra, el SEM disminuye, lo que indica una mayor precisión de la media muestral en representar a la población. Este concepto es fundamental en la teoría de la estimación y se utiliza ampliamente para inferir las características de poblaciones grandes a partir de muestras pequeñas.

## **2.7 Intervalos de confianza**

Los **intervalos de confianza** son una herramienta estadística crucial utilizada para estimar el rango dentro del cual podemos esperar encontrar un parámetro poblacional, como la media, con un cierto grado de confianza. Estos intervalos proporcionan una forma de cuantificar la incertidumbre asociada a las estimaciones muestrales.

**Conceptos Clave de los Intervalos de Confianza**

1. **Definición**: Un intervalo de confianza para un parámetro poblacional es un rango calculado a partir de los datos de una muestra que se espera que contenga el valor del parámetro con una cierta probabilidad, conocida como nivel de confianza (típicamente el 95% o 99%).

2. **Fórmula Matemática**: El intervalo de confianza para la media poblacional, suponiendo una distribución aproximadamente normal, se puede calcular con la fórmula:

   $\text{IC} = \bar{x} \pm z \times \text{SEM}$

   donde:

   - $(\bar{x})$ es la media muestral.

   - $(z)$ es el valor crítico del puntaje Z (por ejemplo, 1.96 para un IC del 95%).

   - SEM es el error estándar de la media, calculado como $\text{SEM} = \frac{s}{\sqrt{n}}$, con $(s)$ siendo la desviación estándar de la muestra y $(n)$ el tamaño de la muestra.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Configuración inicial
np.random.seed(0)
media_poblacional = 50
desviacion_estandar_poblacional = 10
tamaño_muestra = 100
datos = np.random.normal(media_poblacional, desviacion_estandar_poblacional, tamaño_muestra)
media_muestral = np.mean(datos)
sem = stats.sem(datos)

# Calcular el intervalo de confianza del 95%
confianza = 0.95
z = stats.norm.ppf((1 + confianza) / 2)
margen_error = z * sem
intervalo_confianza = (media_muestral - margen_error, media_muestral + margen_error)

# Visualización personalizada
plt.figure(figsize=(12, 6))
plt.hist(datos, bins=30, alpha=0.5, color='gray', label='Datos')
plt.axvline(x=media_muestral, color='blue', label='Media Muestral', linewidth=2, linestyle='dashed')
plt.axvline(x=intervalo_confianza[0], color='green', linestyle='dashed', label='Límite Inferior IC 95%', linewidth=2)
plt.axvline(x=intervalo_confianza[1], color='green', linestyle='dashed', label='Límite Superior IC 95%', linewidth=2)
plt.fill_betweenx([0, plt.ylim()[1]], intervalo_confianza[0], intervalo_confianza[1], color='green', alpha=0.3)
plt.title('Distribución de Datos y Intervalo de Confianza del 95%', fontsize=16)
plt.xlabel('Valor', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Resultados del intervalo de confianza
print("Intervalo de confianza del 95% para la media está entre {:.2f} y {:.2f}".format(intervalo_confianza[0], intervalo_confianza[1]))
````

## **2.8 Prueba de hipótesis estadística**

Las pruebas de hipótesis estadísticas son un componente fundamental de la inferencia estadística, utilizadas para tomar decisiones o probar afirmaciones sobre parámetros poblacionales basados en datos muestrales. A continuación, te proporciono una explicación detallada sobre cómo se realizan estas pruebas, incluyendo ejemplos y cómo implementarlas en Python.

**Conceptos Clave de las Pruebas de Hipótesis**

1. **Hipótesis Nula $(H_0)$**: Es la afirmación de que no hay diferencia o efecto, o que cualquier diferencia observada es producto del azar. Por ejemplo, $(H_0: \mu = \mu_0)$, donde $(\mu_0)$ es un valor específico de la media poblacional.

2. **Hipótesis Alternativa $(H_a)$**: Es la afirmación que se prueba contra la hipótesis nula, sugiriendo que hay una diferencia significativa. Por ejemplo, $(H_a: \mu \neq \mu_0)$, $(H_a: \mu > \mu_0)$, o $(H_a: \mu < \mu_0)$.

3. **Nivel de Significancia $(\alpha)$**: Es la probabilidad de rechazar la hipótesis nula cuando ésta es verdadera (error tipo I). Comúnmente se establece en 0.05.

4. **P-valor**: Es la probabilidad, bajo la suposición de que $(H_0)$ es cierta, de obtener un resultado al menos tan extremo como el que se observó. Un p-valor bajo $(< \alpha)$ indica que es improbable obtener tal resultado si $(H_0)$ es cierta, y por lo tanto, $(H_0)$ puede ser rechazada.

5. **Estadístico de Prueba**: Es un valor calculado a partir de los datos de la muestra que se usa para comparar con una distribución teórica bajo la suposición de que $(H_0)$ es cierta. Dependiendo del diseño de la prueba, este podría ser un valor z, t, F, $(\chi^2)$, etc.

**Ejemplo en Python: Prueba de Hipótesis para una Media Poblacional**

Realizaremos una prueba de hipótesis para determinar si la media de una población es diferente de un valor específico usando una prueba z.

**Instalación de Bibliotecas**

Si necesitas instalar `numpy` y `scipy`, puedes hacerlo utilizando pip:

```bash
pip install numpy scipy
```

**Código Python para la Prueba de Hipótesis**

````python
import numpy as np
from scipy import stats

# Parámetros y generación de datos de muestra
np.random.seed(0)
media_poblacional = 50
desviacion_estandar_poblacional = 10
tamaño_muestra = 30
datos = np.random.normal(media_poblacional, desviacion_estandar_poblacional, tamaño_muestra)

# Parámetros para la hipótesis nula
mu_0 = 52

# Cálculo del estadístico z
media_muestral = np.mean(datos)
sem = desviacion_estandar_poblacional / np.sqrt(tamaño_muestra)  # SEM teórico, si se conoce la desviación estándar poblacional
z = (media_muestral - mu_0) / sem

# Cálculo del p-valor para una prueba de dos colas
p_valor = 2 * (1 - stats.norm.cdf(np.abs(z)))

# Resultados
print(f"Estadístico Z: {z:.2f}")
print(f"P-valor: {p_valor:.4f}")

# Interpretación
if p_valor < 0.05:
    print("Rechazar la hipótesis nula, hay evidencia suficiente para decir que la media es diferente de {mu_0}.")
else:
    print("No rechazar la hipótesis nula, no hay evidencia suficiente para decir que la media es diferente de {mu_0}.")
````

**Explicación del Código**

- **Generación de Datos**: Creamos una muestra de datos normalmente distribuidos.

- **Estadístico Z**: Calculamos el estadístico z comparando la media muestral con la media bajo $(H_0)$.

- **P-valor**: Determinamos el p-valor asociado al valor z observado, considerando una prueba de dos colas (buscando diferencias en ambos sentidos).

**Conclusión**

Las pruebas de hipótesis son esenciales para evaluar afirmaciones sobre parámetros poblacionales basadas en muestras. Permiten a los investigadores tomar decisiones informadas sobre si los resultados observados pueden atribuirse al azar o a efectos reales. ## **2.9 Valor p: tamaño del efecto e influencia del tamaño de la muestra**

El **valor p** es una métrica estadística crucial utilizada para decidir si los resultados de un análisis estadístico son significativos. Este valor ayuda a determinar si los resultados observados en la muestra pueden atribuirse a una casualidad dada una hipótesis nula específica. Aquí, exploraremos detalladamente el concepto de valor p, cómo el tamaño del efecto y el tamaño de la muestra influyen en su cálculo, y proporcionaremos ejemplos prácticos con gráficos y código en Python.

**Conceptos Clave**

1. **Valor P**

   - Definición: Probabilidad de obtener un resultado al menos tan extremo como el observado, suponiendo que la hipótesis nula es cierta.
   - Fórmula: Dependiendo del test, pero comúnmente es $P(T \geq t)$ para un estadístico de prueba $(T)$ y un valor observado $(t)$.

2. **Tamaño del Efecto**

   - Indica la magnitud de la diferencia o relación observada en los datos, independientemente del tamaño de la muestra.
   - Ejemplos comunes incluyen la diferencia de medias, la correlación, o la odds ratio.
   - Fórmula para la diferencia de medias: $d = \frac{\bar{x}_1 - \bar{x}_2}{s}$, donde $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de dos grupos y $(s)$ es la desviación estándar combinada.

3. **Influencia del Tamaño de la Muestra**

   - Un tamaño de muestra más grande generalmente reduce el error estándar y hace que el valor p sea más sensible a pequeñas diferencias (aumentando la potencia estadística del test).

**Ejemplo en Python: Valor P y Tamaño del Efecto**

Realizaremos un ejemplo práctico donde compararemos dos grupos con tamaños de muestra variables para ilustrar cómo el tamaño del efecto y el tamaño de la muestra afectan el valor p.

**Instalación de Bibliotecas**

Si necesitas instalar `numpy`, `scipy`, y `matplotlib`, puedes hacerlo utilizando pip:

```bash
pip install numpy scipy matplotlib
```

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos para dos grupos
np.random.seed(0)
grupo1 = np.random.normal(50, 10, 100)  # Grupo 1: media 50, desviación estándar 10, n=100
grupo2 = np.random.normal(52, 10, 100)  # Grupo 2: media 52, desviación estándar 10, n=100

# Calcular el estadístico de prueba y el valor p
t_stat, p_valor = stats.ttest_ind(grupo1, grupo2)

# Calcular el tamaño del efecto (Cohen's d)
def cohens_d(x, y):
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    return (np.mean(x) - np.mean(y)) / np.sqrt(((nx-1)*np.std(x, ddof=1)**2 + (ny-1)*np.std(y, ddof=1)**2) / dof)

d = cohens_d(grupo1, grupo2)

# Visualización de los datos
plt.figure(figsize=(12, 6))
plt.hist(grupo1, color='blue', alpha=0.7, label='Grupo 1', bins=20, edgecolor='black')
plt.hist(grupo2, color='red', alpha=0.7, label='Grupo 2', bins=20, edgecolor='black')
plt.axvline(np.mean(grupo1), color='navy', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(grupo2), color='darkred', linestyle='dashed', linewidth=2)
plt.title(f'Test de Grupos Independientes, t={t_stat:.2f}, p={p_valor:.3f}, d={d:.2f}')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Resultados
print(f"Estadístico t: {t_stat:.2f}, Valor p: {p_valor:.3f}, Tamaño del efecto (Cohen's d): {d:.2f}")
````

**Explicación del Código**

- **Generación de Datos**: Creamos dos grupos con distribuciones normales ligeramente diferentes.
- **Test de t de Student**: Calculamos el estadístico t y el valor p entre los dos grupos para evaluar si las diferencias en las medias son estadísticamente significativas.
- **Tamaño del Efecto**: Usamos Cohen's d para medir la magnitud de la diferencia entre los grupos.
- **Visualización**: Mostramos los histogramas de ambos grupos con líneas que marcan las medias para visualizar la diferencia.

**Conclusión**

El valor p es una medida de significancia estadística que, junto con el tamaño del efecto, proporciona una comprensión profunda de la relevancia de los resultados observados. La influencia del tamaño de la muestra es crítica, ya que tamaños de muestra mayores pueden detectar diferencias menores como significativas. Este ejemplo muestra cómo evaluar estas métricas en Python y proporciona una base sólida para análisis estadístico avanzado.

## **2.10 Realizar una prueba t**

La **prueba t** (o test de t) es una herramienta estadística usada para determinar si existe una diferencia significativa entre las medias de dos grupos, que puede ser aplicada en diferentes contextos como pruebas independientes o apareadas. Es especialmente útil cuando el tamaño de la muestra es pequeño y la distribución de los datos se asume normal, pero no se conoce la varianza de la población.

**Tipos de Pruebas t**

1. **Prueba t para una muestra**: Compara la media de una muestra con un valor conocido (generalmente la media poblacional).
2. **Prueba t para muestras independientes**: Compara las medias de dos grupos independientes para ver si hay evidencia de que las medias poblacionales son diferentes.
3. **Prueba t pareada**: Compara las medias de dos grupos relacionados o emparejados; por ejemplo, medidas antes y después de un tratamiento en los mismos sujetos.

**Formula Matemática para la Prueba t**

Para dos muestras independientes, la fórmula de la prueba t es:

$t = \frac{\bar{x}_1 - \bar{x}_2}{s_p \cdot \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$

donde:

- $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de las muestras.
- $(n_1)$ y $(n_2)$ son los tamaños de las muestras.
- $(s_p)$ es la desviación estándar combinada para muestras independientes, calculada como:

$s_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}$

donde $(s_1^2)$ y $(s_2^2)$ son las varianzas de las muestras.

## **2.11 Realización de una prueba t de una muestra utilizando PROC TTEST**

La **prueba t de una muestra** es una técnica estadística usada para determinar si la media de una muestra difiere significativamente de una media poblacional conocida o un valor específico. Este tipo de prueba es útil cuando se quiere verificar suposiciones sobre características promedio de una población basadas en datos muestrales.

**Fundamentos de la Prueba t de una Muestra**

1. **Hipótesis Nula $(H_0)$**: La media de la muestra es igual a la media poblacional conocida o un valor específico $(H_0: \mu = \mu_0)$.

2. **Hipótesis Alternativa $(H_a)$**: La media de la muestra es diferente de la media poblacional $(H_a: \mu \neq \mu_0)$, dependiendo del contexto, también puede ser unilateral $(H_a: \mu > \mu_0)$ o $(H_a: \mu < \mu_0)$.

**Fórmula Matemática**

El estadístico de prueba t se calcula usando la fórmula:

$t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}$

donde:

- $(\bar{x})$ es la media de la muestra.

- $(\mu_0)$ es la media poblacional conocida o el valor específico a comparar.

- $(s)$ es la desviación estándar de la muestra.

- $(n)$ es el tamaño de la muestra.

El valor t resultante se compara con un valor crítico de la distribución t de Student con $(n-1)$ grados de libertad para determinar si se rechaza o no la hipótesis nula, basado en un nivel de significancia preestablecido $(\alpha)$, comúnmente 0.05.

## **2.12 Supuestos para la prueba t de dos muestras**

La **prueba t de dos muestras** (también conocida como prueba t independiente) es una técnica estadística usada para comparar las medias de dos grupos independientes para determinar si hay una diferencia estadísticamente significativa entre ellos. Como cualquier prueba estadística, tiene ciertos supuestos que deben cumplirse para que los resultados sean válidos.

**Supuestos de la Prueba t de Dos Muestras**

1. **Independencia de las Observaciones**: Las muestras de los dos grupos deben ser independientes entre sí, es decir, la selección de una observación en un grupo no debe afectar las observaciones en el otro grupo.

2. **Normalidad**: Las distribuciones de los datos en ambos grupos deben aproximarse a una distribución normal. Este supuesto puede ser menos crítico con tamaños de muestra grandes debido al Teorema del Límite Central.

3. **Varianzas Homogéneas (Homocedasticidad)**: Las varianzas de los dos grupos deben ser iguales. Si las varianzas son diferentes (heterocedasticidad), la estimación del error estándar que se utiliza para calcular el estadístico t podría estar sesgada.

**Fórmula Matemática**

La fórmula para el estadístico t de una prueba t de dos muestras, asumiendo varianzas iguales, es:

$t = \frac{\bar{x}_1 - \bar{x}_2}{s_p \cdot \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$

donde:

- $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de los dos grupos.

- $(n_1)$ y $(n_2)$ son los tamaños de las muestras de los dos grupos.

- $(s_p)$ es la desviación estándar combinada, que se calcula como:


$s_p = \sqrt{\frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}}$


y $(s_1^2)$ y $(s_2^2)$ son las varianzas de las dos muestras.

## **2.13 Pruebas de varianzas iguales y desiguales**

Al comparar las medias de dos grupos con una prueba t, es esencial verificar primero si las varianzas de los dos grupos son iguales (homocedasticidad) o diferentes (heterocedasticidad). Esta verificación es importante porque el tipo de prueba t que se debe aplicar depende de este supuesto.

**Pruebas de Varianzas Iguales vs. Desiguales**

**1. Prueba de Levene**: Comprueba si los grupos tienen varianzas iguales. Es menos sensible a las desviaciones de la normalidad que otras pruebas como la de Bartlett, que es más adecuada para datos que siguen una distribución normal.

- **Fórmula de la prueba de Levene**:

  La estadística W se calcula como:

  $W = \frac{(N - k) \sum_{i=1}^{k} n_i (\bar{Z}_{i.} - \bar{Z}_{..})^2}{(k - 1) \sum_{i=1}^{k} \sum_{j=1}^{n_i} (Z_{ij} - \bar{Z}_{i.})^2}$

  donde:

  - $(N)$ es el número total de observaciones,

  - $(k)$ es el número de grupos.

  - $(n_i)$ es el número de observaciones en el grupo $(i)$.

  - $(Z_{ij})$ es el valor absoluto de la desviación de cada observación del grupo $(i)$ de la media de su grupo.

  - $(\bar{Z}_{i.})$ y $(\bar{Z}_{..})$ son las medias de $(Z_{ij})$ para el grupo $(i)$ y para todos los grupos respectivamente.

**2. Prueba de Bartlett**: Evalúa la homogeneidad de varianzas entre grupos cuando los datos se distribuyen normalmente.

- **Fórmula de la prueba de Bartlett**:

  $X^2 = \frac{(N - k) \ln(S^2) - \sum_{i=1}^{k} (n_i - 1) \ln(s_i^2)}{1 + \frac{1}{3(k-1)} \left(\sum_{i=1}^{k} \frac{1}{n_i - 1} - \frac{1}{N - k}\right)}$

  donde:

  - $(N)$, $(k)$, y $(n_i)$ tienen el mismo significado que en Levene.

  - $(s_i^2)$ es la varianza muestral del grupo $(i)$.

  - $(S^2)$ es la varianza combinada.

**Ejemplo en Python: Pruebas de Varianzas Iguales y Desiguales**

A continuación, veremos cómo realizar estas pruebas en Python utilizando `scipy.stats`.

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos de muestra
np.random.seed(0)
grupo1 = np.random.normal(100, 10, 50)  # Grupo 1: media 100, desviación estándar 10, tamaño 50
grupo2 = np.random.normal(100, 20, 50)  # Grupo 2: media 100, desviación estándar 20, tamaño 50

# Prueba de Levene para igualdad de varianzas
levene_stat, levene_p = stats.levene(grupo1, grupo2)

# Prueba de Bartlett para igualdad de varianzas
bartlett_stat, bartlett_p = stats.bartlett(grupo1, grupo2)

# Visualización de los resultados
plt.figure(figsize=(12, 6))
plt.hist(grupo1, color='blue', alpha=0.7, label='Grupo 1', bins=15, edgecolor='black')
plt.hist(grupo2, color='red', alpha=0.7, label='Grupo 2', bins=15, edgecolor='black')
plt.title('Comparación de Distribuciones de Dos Grupos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Imprimir los resultados de las pruebas
print(f"Resultado de la Prueba de Levene: Estadístico = {levene_stat:.2f}, p-valor = {levene_p:.3f}")
print(f"Resultado de la Prueba de Bartlett: Estadístico = {bartlett_stat:.2f}, p-valor = {bartlett_p:.3f}")
````

**Explicación del Código**

 **Generación de Datos**: Se crean dos grupos con la misma media pero diferentes desviaciones estándar para ilustrar diferencias en la varianza.
- **Pruebas de Varianzas**: Se aplican la prueba de Levene y la prueba de Bartlett para evaluar la homogeneidad de las varianzas.
- **Visualización**: Los histogramas ayudan a visualizar la distribución de los datos y las diferencias en la dispersión entre los dos grupos.

**Conclusión**

Las pruebas de varianza son esenciales antes de realizar pruebas t de dos muestras para asegurarse de que se utiliza el enfoque correcto basado en si las varianzas son iguales o no. Las pruebas de Levene y Bartlett ofrecen métodos robustos para verificar la homogeneidad de varianzas y deben ser parte del análisis preliminar en cualquier estudio comparativo de grupos.

## **2.14 Realización de una prueba t de dos muestras utilizando PROC TTEST**

La **prueba t de dos muestras** es un método estadístico que se utiliza para determinar si las medias de dos grupos independientes son estadísticamente diferentes. Este tipo de prueba es común en investigaciones donde se comparan dos tratamientos, condiciones o grupos distintos bajo variables controladas.

**Supuestos para la Prueba t de Dos Muestras**

1. **Independencia**: Las muestras de los dos grupos deben ser independientes entre sí.
2. **Normalidad**: Los datos en ambos grupos deben seguir una distribución normal.
3. **Varianzas iguales (homocedasticidad)**: Las varianzas de los grupos deben ser iguales; sin embargo, existen versiones de la prueba t que ajustan por varianzas desiguales (como la prueba t de Welch).

**Fórmula de la Prueba t de Dos Muestras**

La fórmula para calcular el estadístico t en una prueba t de dos muestras, asumiendo varianzas iguales, es:

$t = \frac{\bar{x}_1 - \bar{x}_2}{s_p \cdot \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$

donde:

- $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de los dos grupos.

- $(s_p)$ es la desviación estándar combinada (pooled standard deviation), que se calcula como:

$s_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}$

- $(s_1)$ y $(s_2)$ son las desviaciones estándar de las muestras de los grupos 1 y 2, respectivamente.
- $(n_1)$ y $(n_2)$ son los tamaños de las muestras de los grupos 1 y 2.

**Ejemplo en Python: Prueba t de Dos Muestras**

A continuación, te muestro cómo realizar una prueba t de dos muestras en Python utilizando la biblioteca `scipy.stats`.

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos de muestra
np.random.seed(0)
grupo1 = np.random.normal(100, 10, 100)  # Grupo 1: media 100, desviación estándar 10, tamaño de muestra 100
grupo2 = np.random.normal(105, 10, 100)  # Grupo 2: media 105, desviación estándar 10, tamaño de muestra 100

# Realizar la prueba t de dos muestras
t_stat, p_valor = stats.ttest_ind(grupo1, grupo2)

# Visualización de los resultados
plt.figure(figsize=(10, 5))
plt.hist(grupo1, bins=20, alpha=0.7, label='Grupo 1', color='blue', edgecolor='black')
plt.hist(grupo2, bins=20, alpha=0.7, label='Grupo 2', color='red', edgecolor='black')
plt.axvline(np.mean(grupo1), color='blue', linestyle='dashed', label='Media Grupo 1')
plt.axvline(np.mean(grupo2), color='red', linestyle='dashed', label='Media Grupo 2')
plt.title('Resultados de la Prueba t de Dos Muestras')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Resultados de la prueba
print(f"Estadístico t: {t_stat:.2f}, P-valor: {p_valor:.3f}")
````

**Explicación del Código**

- **Generación de Datos**: Dos grupos con distribuciones normales y diferentes medias para probar la hipótesis de que son diferentes.
- **Prueba t de Dos Muestras**: Utilizando la función `ttest_ind` de `scipy.stats`, que calcula automáticamente el estadístico t y el p-valor para las muestras independientes.
- **Visualización**: Los histogramas superpuestos y las líneas de media ayudan a visualizar las diferencias entre los dos grupos.

**Conclusión**

La prueba t de dos muestras es una herramienta estadística esencial para comparar las medias de dos grupos independientes. En Python, `scipy.stats` proporciona funciones robustas para realizar estas pruebas fácilmente, permitiendo una evaluación rápida y eficaz de las hipótesis relacionadas con diferencias de medias. Este enfoque es fundamental en estudios comparativos en muchas disciplinas, desde la biología hasta las ciencias sociales.

## **2.15 Parámetros y estadísticas**

El estudio de la estadística involucra entender la diferencia entre **parámetros** y **estadísticas**, conceptos fundamentales en la teoría de la estimación. Aquí te proporciono una explicación detallada de estos términos, ejemplos prácticos, y cómo se aplican y visualizan usando Python.

**Parámetros vs. Estadísticas**

1. **Parámetros**

   - **Definición**: Un parámetro es una medida que describe una característica de una población entera. Es un valor fijo pero generalmente desconocido que queremos estimar.
   - **Ejemplos**: La media poblacional (μ), la varianza poblacional (σ²), y la proporción poblacional (p).

2. **Estadísticas**

   - **Definición**: Una estadística es una medida que describe una característica de una muestra, un subconjunto seleccionado de una población.
   - **Ejemplos**: La media muestral $(\bar{x})$, la varianza muestral (s²), y la proporción muestral $(\hat{p})$.

**Fórmulas Matemáticas**

- **Media Poblacional (μ)**

  $\mu = \frac{\sum_{i=1}^{N} x_i}{N}$

  donde $(x_i)$ son los valores individuales en la población y $(N)$ es el tamaño de la población.

- **Media Muestral $(\bar{x})$**

  $\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}$

  donde $(x_i)$ son los valores individuales en la muestra y $(n)$ es el tamaño de la muestra.

**Ejemplo en Python: Calculando y Visualizando Parámetros y Estadísticas**

Generaremos datos de una población simulada y tomaremos muestras de ella para calcular y comparar estadísticas y parámetros.

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una población grande y calcular el parámetro (media poblacional)
np.random.seed(0)
poblacion = np.random.normal(100, 15, 10000)  # media=100, desviación estándar=15
parametro_media_poblacional = np.mean(poblacion)

# Tomar una muestra y calcular la estadística (media muestral)
muestra = np.random.choice(poblacion, 100, replace=False)
estadistica_media_muestral = np.mean(muestra)

# Visualización de la población y la muestra
plt.figure(figsize=(12, 6))
plt.hist(poblacion, bins=50, alpha=0.6, color='gray', label='Población')
plt.hist(muestra, bins=50, alpha=0.8, color='blue', label='Muestra')
plt.axvline(parametro_media_poblacional, color='red', linestyle='dashed', linewidth=2, label=f'Media Poblacional: {parametro_media_poblacional:.2f}')
plt.axvline(estadistica_media_muestral, color='green', linestyle='dashed', linewidth=2, label=f'Media Muestral: {estadistica_media_muestral:.2f}')
plt.title('Distribución de la Población y la Muestra')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
````

**Explicación del Código**

- **Generación de Datos**: Creamos una población normalmente distribuida y extraemos una muestra de ella.
- **Cálculo de Medias**: Calculamos tanto la media poblacional (parámetro) como la media muestral (estadística).
- **Visualización**: Mostramos histogramas de la población y la muestra con líneas que indican las medias.

**Conclusión**

Entender la diferencia entre parámetros y estadísticas es crucial para la inferencia estadística. Los parámetros describen características totales de la población, mientras que las estadísticas se utilizan para estimar esos parámetros a partir de muestras seleccionadas. Este enfoque permite a los investigadores hacer inferencias sobre poblaciones grandes basadas en datos más manejables. El ejemplo en Python proporciona una base práctica para aplicar estos conceptos y visualizar diferencias entre parámetros y estadísticas.

## **2.16 Distribución normal**

La **distribución normal**, también conocida como **distribución gaussiana**, es una de las distribuciones más importantes en estadística debido a su presencia frecuente en muchos fenómenos naturales y sociales. Aquí, te proporcionaré una explicación detallada de esta distribución, ejemplos de su aparición y cómo visualizarla utilizando Python.

**Características de la Distribución Normal**

1. **Forma de Campana**: La gráfica de la distribución normal es simétrica y tiene la forma de una campana, centrada en la media $(\mu)$.

2. **Desviación Estándar $(\sigma)$**: La dispersión de los datos alrededor de la media se mide con la desviación estándar. Un $(\sigma)$ más grande significa una curva más ancha.

3. **Media, Mediana y Moda Iguales**: En una distribución perfectamente normal, la media, la mediana y la moda son todas iguales y se encuentran en el centro de la distribución.

4. **Asintótica**: Las colas de la distribución se extienden infinitamente en ambos sentidos, aproximándose al eje horizontal pero nunca tocándolo.

5. **Regla Empírica (68-95-99.7)**: Aproximadamente el 68% de los datos caen dentro de $(\sigma)$, el 95% dentro de 2$(\sigma)$, y el 99.7% dentro de 3$(\sigma)$ de la media.

**Ejemplos de Fenómenos que Siguen una Distribución Normal**

- Alturas de personas.
- Presión arterial.
- Puntajes en ciertos tipos de pruebas estandarizadas.

**Ejemplo en Python: Visualización de la Distribución Normal**

Vamos a utilizar `numpy` para generar datos y `matplotlib` para visualizar una distribución normal.

**Instalación de Bibliotecas**


**Código Python para Generar y Visualizar una Distribución Normal**

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
media = 0
desviacion_estandar = 1

# Generar datos de una distribución normal
datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Visualización de la distribución normal
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='b')

# Superponer la curva teórica de la distribución normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - media)**2 / (2 * desviacion_estandar**2)) / (np.sqrt(2 * np.pi) * desviacion_estandar)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  sigma = %.2f" % (media, desviacion_estandar)
plt.title(title)
plt.xlabel('Datos')
plt.ylabel('Densidad')
plt.show()
````

**Explicación del Código**

- **Generación de Datos**: Creamos una muestra de 1000 puntos de datos que siguen una distribución normal con media 0 y desviación estándar 1.
- **Histograma**: Visualizamos estos datos en un histograma para ver su distribución.
- **Curva Teórica**: Superponemos la curva teórica de la distribución normal basada en la fórmula de la función de densidad de probabilidad normal para verificar el ajuste de los datos generados al modelo teórico.

**Conclusión**

La distribución normal es esencial en estadística por su utilidad en el modelado de datos y en teoría de inferencia. Muchas técnicas estadísticas asumen que los datos siguen una distribución normal, lo que hace crucial entender y poder identificar esta distribución en sets de datos. El ejemplo en Python proporciona una manera práctica de visualizar y entender la distribución normal.

## **2.17 Uso de PROC TTEST para realizar una prueba t de una muestra**

El término "PROC TTEST" se refiere a un procedimiento en el software estadístico SAS, utilizado para realizar pruebas t, incluyendo la prueba t de una muestra. Aunque no se ejecuta "PROC TTEST" en Python, puedes realizar una prueba t de una muestra con funcionalidades similares usando bibliotecas como `scipy`. Aquí, te explicaré cómo realizar esta prueba en Python, proporcionando detalles sobre la fórmula matemática y un ejemplo práctico con visualización.

**Prueba t de Una Muestra**

Esta prueba se utiliza para determinar si la media de una muestra es significativamente diferente de una media poblacional conocida o un valor específico.

**Fórmula Matemática**

La fórmula para el estadístico t en una prueba t de una muestra es:

$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

donde:

- $(\bar{x})$ es la media muestral.

- $(\mu_0)$ es la media poblacional o un valor específico con el que se compara la muestra.

- $(s)$ es la desviación estándar de la muestra.

- $(n)$ es el número de observaciones en la muestra.

El valor t calculado se compara con un valor crítico de la distribución t (dependiendo del nivel de significancia y los grados de libertad $(n-1)$ para determinar si se rechaza o no la hipótesis nula.

**Ejemplo en Python: Realización de una Prueba t de Una Muestra**

A continuación, te muestro cómo realizar esta prueba en Python usando la biblioteca `scipy.stats`.

**Instalación de Bibliotecas**

Si necesitas instalar `numpy` y `scipy`, puedes hacerlo usando pip:
```bash
pip install numpy scipy
```

**Código Python**


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos de muestra
np.random.seed(0)
datos = np.random.normal(5.5, 1.5, 30)  # media=5.5, desviación estándar=1.5, tamaño de muestra=30

# Valor específico a comparar (media poblacional hipotética)
mu_0 = 5

# Realizar la prueba t de una muestra
t_stat, p_valor = stats.ttest_1samp(datos, popmean=mu_0)

# Visualización de los resultados
plt.figure(figsize=(10, 5))
plt.hist(datos, bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(x=np.mean(datos), color='red', linestyle='dashed', linewidth=2, label=f'Media Muestral: {np.mean(datos):.2f}')
plt.axvline(x=mu_0, color='green', linestyle='dashed', linewidth=2, label=f'Media Hipotética: {mu_0}')
plt.title('Prueba t de una Muestra')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Resultados de la prueba
print(f"Estadístico t: {t_stat:.2f}, P-valor: {p_valor:.3f}")
````

**Descripción del Código**

- **Generación de Datos**: Se crea una muestra de datos con distribución normal.
- **Prueba t**: Usamos `ttest_1samp` de `scipy.stats` para calcular el estadístico t y el p-valor, comparando la media muestral con un valor hipotético.
- **Visualización**: El histograma ayuda a visualizar la distribución de los datos y a comparar la media muestral con la media hipotética.

**Conclusión**

Realizar una prueba t de una muestra en Python es un proceso directo con `scipy.stats`, permitiendo a los usuarios evaluar si la media de una muestra difiere significativamente de un valor conocido. Esta técnica es fundamental en análisis estadístico para probar hipótesis sobre características poblacionales basadas en datos muestrales.

## **2.18 Uso de PROC TTEST para comparar grupos**

El "PROC TTEST" es una herramienta específica del software SAS utilizada para realizar pruebas t estadísticas, incluyendo la comparación de medias entre dos grupos. Aunque no se puede ejecutar directamente "PROC TTEST" en Python, podemos simular una funcionalidad similar usando la biblioteca `scipy.stats` para realizar pruebas t de dos muestras, que es el equivalente funcional para comparar grupos en Python.

**Prueba t de Dos Muestras**

Esta prueba se utiliza para comparar las medias de dos grupos independientes y determinar si hay una diferencia estadísticamente significativa entre ellas. Aquí te detallo cómo realizar esta prueba en Python.

**Fórmula Matemática**

La fórmula para el estadístico t en la prueba t de dos muestras, asumiendo varianzas iguales, es:

$t = \frac{\bar{x}_1 - \bar{x}_2}{s_p \cdot \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$

donde:

- $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de los dos grupos.

- $(s_p)$ es la desviación estándar combinada (pooled standard deviation), calculada como:

$s_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}$

- $(s_1)$ y $(s_2)$ son las desviaciones estándar de las muestras de los grupos 1 y 2, respectivamente.

- $(n_1)$ y $(n_2)$ son los tamaños de las muestras de los grupos 1 y 2.

**Ejemplo en Python: Prueba t de Dos Muestras**

Vamos a realizar una prueba t de dos muestras en Python para comparar dos grupos independientes.

**Código Python**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos de muestra
np.random.seed(0)
grupo1 = np.random.normal(100, 15, 50)  # Grupo 1: media 100, desviación estándar 15, tamaño de muestra 50
grupo2 = np.random.normal(105, 15, 50)  # Grupo 2: media 105, desviación estándar 15, tamaño de muestra 50

# Realizar la prueba t de dos muestras
t_stat, p_valor = stats.ttest_ind(grupo1, grupo2)

# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.hist(grupo1, bins=20, alpha=0.7, label='Grupo 1', color='blue', edgecolor='black')
plt.hist(grupo2, bins=20, alpha=0.7, label='Grupo 2', color='red', edgecolor='black')
plt.axvline(np.mean(grupo1), color='blue', linestyle='dashed', label='Media Grupo 1')
plt.axvline(np.mean(grupo2), color='red', linestyle='dashed', label='Media Grupo 2')
plt.title('Resultados de la Prueba t de Dos Muestras')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Resultados de la prueba
print(f"Estadístico t: {t_stat:.2f}, P-valor: {p_valor:.3f}")
````

**Descripción del Código**

- **Generación de Datos**: Se crean dos grupos con distribuciones normales similares pero con medias ligeramente diferentes para ilustrar una diferencia detectable por la prueba t.
- **Prueba t de Dos Muestras**: Usamos `scipy.stats.ttest_ind` para calcular el estadístico t y el p-valor, asumiendo que las varianzas son iguales.
- **Visualización**: Los histogramas superpuestos y las líneas de media ayudan a visualizar las diferencias entre los dos grupos.

**Conclusión**

Realizar una prueba t de dos muestras en Python es un proceso directo con `scipy.stats`, que permite a los investigadores evaluar si las medias de dos grupos independientes difieren significativamente. Este método es fundamental en estudios comparativos en muchas disciplinas científicas y es el equivalente en Python del "PROC TTEST" en SAS para comparar grupos.
