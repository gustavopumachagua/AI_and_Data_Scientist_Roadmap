# **27. Correlación de Pearson y cómo crear Mapas de Calor de la Matriz de Correlaciones**

## **1. Motivación para usar la correlación**

La correlación es una medida estadística importante que se utiliza para evaluar la relación entre dos variables. Al conocer la correlación entre dos variables, podemos entender cómo cambia una variable en respuesta a cambios en la otra variable.

La correlación es ampliamente utilizada en muchas áreas, como en ciencias sociales, psicología, finanzas, ciencia de datos, entre otras. Por ejemplo, en finanzas, la correlación se utiliza para analizar el rendimiento de dos o más activos financieros y para construir carteras de inversión diversificadas. En ciencia de datos, la correlación se utiliza para seleccionar las variables más relevantes para un modelo predictivo y para identificar posibles relaciones entre los atributos de los datos.

La correlación también es útil para validar hipótesis y para establecer relaciones causales. Sin embargo, es importante tener en cuenta que la correlación no implica necesariamente causalidad. La correlación solo indica la fuerza y dirección de la relación entre dos variables, pero no proporciona información sobre la causa y el efecto. Por lo tanto, es importante realizar un análisis cuidadoso y tener en cuenta otros factores antes de hacer conclusiones sobre la relación causal entre dos variables.

## **2. Contexto del ejemplo**

Para demostrar el uso de la correlación de Pearson y la creación de mapas de calor de la matriz de correlaciones con Python, utilizaré el conjunto de datos de iris, que contiene información sobre diferentes especies de iris y sus medidas de pétalos y sépalos.

Primero, cargaré los datos y calcularé la correlación de Pearson entre las diferentes características utilizando la biblioteca NumPy y Pandas:

```python
import pandas as pd
import numpy as np

# Cargar los datos de iris
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                   header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Seleccionar solo las columnas numéricas (todas menos 'species')
iris_numeric = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Calcular la matriz de correlaciones de Pearson solo para las columnas numéricas
corr_matrix = iris_numeric.corr(method='pearson')

# Mostrar la matriz de correlaciones
print(corr_matrix)
```

Luego, utilizaré la biblioteca Seaborn para visualizar la matriz de correlaciones como un mapa de calor:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un mapa de calor de la matriz de correlaciones
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaciones de Iris')
plt.show()
```

Este mapa de calor muestra la correlación de Pearson entre las diferentes características de iris. Los valores más oscuros indican una correlación positiva más fuerte, mientras que los valores más claros indican una correlación más débil o negativa.

En este ejemplo, se puede observar que hay una correlación positiva fuerte entre la longitud del pétalo y el ancho del pétalo, mientras que la correlación entre la longitud del sépalo y el ancho del sépalo es mucho más débil. También se puede ver que la longitud del pétalo está fuertemente correlacionada con todas las otras características, mientras que la longitud del sépalo está más débilmente correlacionada con las otras características.

En resumen, la correlación de Pearson y los mapas de calor de la matriz de correlaciones son herramientas útiles para explorar y visualizar relaciones entre diferentes características en un conjunto de datos.

## **3. Interpretación visual de la correlación**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de iris
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                   header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Crear un gráfico de dispersión de la longitud del pétalo vs el ancho del pétalo
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura
plt.scatter(iris['petal_length'], iris['petal_width'], c='blue', label='Iris Data')  # Personaliza el color y etiqueta
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.title('Correlación entre la longitud y el ancho del pétalo')
plt.grid(True, linestyle='--', alpha=0.7)  # Agrega una cuadrícula punteada
plt.legend()  # Muestra la leyenda

plt.show()
```

Este gráfico de dispersión muestra una correlación positiva fuerte entre la longitud del pétalo y el ancho del pétalo. Los puntos se agrupan en una línea diagonal que va desde la esquina inferior izquierda hasta la esquina superior derecha, lo que indica que a medida que la longitud del pétalo aumenta, también lo hace el ancho del pétalo.

En resumen, los gráficos de dispersión son una herramienta útil para interpretar visualmente la correlación entre dos variables. Si los puntos se agrupan en una línea diagonal, indica una correlación positiva, mientras que si los puntos se agrupan en una línea diagonal que va de la esquina superior izquierda a la esquina inferior derecha, indica una correlación negativa.

## **4. Obtener la matriz de correlaciones**

```python
import pandas as pd

# Crear un DataFrame con cuatro variables numéricas
df = pd.DataFrame({'Var1': [1, 2, 3, 4, 5],
                   'Var2': [2, 4, 6, 8, 10],
                   'Var3': [3, 6, 9, 12, 15],
                   'Var4': [4, 8, 12, 16, 20]})

# Calcular la matriz de correlaciones
corr_matrix = df.corr()

print(corr_matrix)
```

## **5. Graficar la matriz de correlaciones**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame con cuatro variables numéricas
df = pd.DataFrame({'Var1': [1, 2, 3, 4, 5],
                   'Var2': [2, 4, 6, 8, 10],
                   'Var3': [3, 6, 9, 12, 15],
                   'Var4': [4, 8, 12, 16, 20]})

# Calcular la matriz de correlaciones
corr_matrix = df.corr()

# Crear un mapa de calor de la matriz de correlaciones
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Matriz de Correlaciones')
plt.show()
```
