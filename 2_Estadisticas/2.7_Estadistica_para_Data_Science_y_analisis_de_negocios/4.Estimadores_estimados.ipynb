## **Intervalos de confianza una herramienta invaluable para la toma de decisiones**

La f贸rmula matem谩tica de la `t de Student (o t de Student)` se utiliza para calcular el estad铆stico `t` en pruebas de hip贸tesis o para construir intervalos de confianza cuando se trabaja con muestras peque帽as. La f贸rmula para el estad铆stico `t` es:

$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$

Donde:
- $(\bar{x})$ es la media muestral.
- $(\mu)$ es la media poblacional (hipot茅tica bajo la hip贸tesis nula).
- $(s)$ es la desviaci贸n est谩ndar muestral.
- $(n)$ es el tama帽o de la muestra.

En esta f贸rmula, $(\bar{x} - \mu)$ representa la diferencia entre la media muestral y la media poblacional hipot茅tica bajo la hip贸tesis nula. El t茅rmino $(\frac{s}{\sqrt{n}})$ es el error est谩ndar de la media, que indica cu谩nto puede variar la media muestral de una muestra a otra.

El estad铆stico `t` sigue una distribuci贸n t de Student con $(n-1)$ grados de libertad, lo que se denota como $(t(n-1))$. La f贸rmula general de la distribuci贸n `t` es m谩s compleja y se utiliza para calcular probabilidades asociadas con el estad铆stico  `t`.

$f(t; n-1) = \frac{\Gamma\left(\frac{n}{2}\right)}{\sqrt{\pi \cdot (n-1)} \cdot \Gamma\left(\frac{n-1}{2}\right)} \left(1 + \frac{t^2}{n-1}\right)^{-\frac{n}{2}}$

Donde:
- $( \Gamma )$ es la funci贸n gamma.

Esta f贸rmula se utiliza para calcular la probabilidad de obtener un valor de `t` espec铆fico en una distribuci贸n t de Student con $(n-1)$ grados de libertad. En la f贸rmula del estad铆stico `t`, se compara el valor calculado con la distribuci贸n `t` para determinar si hay evidencia suficiente para rechazar la hip贸tesis nula.

Los intervalos de confianza son una herramienta estad铆stica utilizada para estimar el valor desconocido de un par谩metro poblacional, como la media o la proporci贸n, a partir de una muestra aleatoria de datos. El intervalo de confianza proporciona un rango de valores plausibles para el verdadero valor del par谩metro, junto con un nivel de confianza asociado que refleja la probabilidad de que el intervalo contenga el verdadero valor del par谩metro.

Por ejemplo, supongamos que estamos interesados en estimar la media de altura de una poblaci贸n. Tomamos una muestra aleatoria de `50` personas y calculamos su media, que resulta ser de `170 cm`, con una desviaci贸n est谩ndar de `5 cm`. Usando un nivel de confianza del `95%`, podemos calcular el intervalo de confianza de la siguiente manera:

* Calculamos el error est谩ndar de la media utilizando la f贸rmula: $SE = \frac{s}{\sqrt{n}} = \frac{5}{\sqrt{50}} = 0.707$
* Calculamos la magnitud del intervalo de confianza utilizando la distribuci贸n `t de Student` para muestras peque帽as y un nivel de confianza del `95%` con `49` grados de libertad (50 - 1): $t_{0.025,49} = 2.009$
* Calculamos el intervalo de confianza: $170 \pm 2.009 \times 0.707 = [168.37, 171.63]$

Esto significa que podemos estar `95%` seguros de que el verdadero valor de la media de altura de la poblaci贸n se encuentra en el rango de `168.37 cm` a `171.63 cm`, basado en nuestra muestra.

En Python, podemos calcular el intervalo de confianza utilizando la biblioteca `scipy.stats` de la siguiente manera:

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Datos de la muestra
data = np.array([170, 165, 172, 176, 168, 171, 173, 169, 174, 167,
                 171, 169, 175, 172, 170, 169, 174, 168, 172, 171,
                 175, 170, 173, 169, 171, 172, 175, 168, 174, 172,
                 170, 169, 173, 171, 172, 170, 168, 173, 171, 170,
                 172, 175, 169, 173, 167, 170, 173, 172, 168, 171])

# Tama帽o de la muestra y nivel de confianza
n = len(data)
alpha = 0.05

# Calculamos la media y la desviaci贸n est谩ndar de la muestra
mean = np.mean(data)
std_dev = np.std(data, ddof=1)

# Calculamos el error est谩ndar
se = std_dev / np.sqrt(n)

# Calculamos el valor cr铆tico de t para el nivel de confianza y grados de libertad
t_crit = t.ppf(1 - alpha / 2, n - 1)

# Calculamos el intervalo de confianza
lower = mean - t_crit * se
upper = mean + t_crit * se

# Crear la gr谩fica
plt.figure(figsize=(10, 8))
plt.hist(data, bins=20, color="skyblue", alpha=0.7, label='Distribuci贸n de la Muestra')

# Agregar el intervalo de confianza
plt.axvline(mean, color='red', linestyle='solid', linewidth=2, label='Media')
plt.axvline(lower, color='green', linestyle='dashed', linewidth=2, label='L铆mites de Confianza')
plt.axvline(upper, color='green', linestyle='dashed', linewidth=2)

# Etiquetas y t铆tulo
plt.xlabel('Valores', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Distribuci贸n de la Muestra con Intervalo de Confianza', fontsize=16)
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

## **Calculando intervalos de confianza dentro de poblaci贸n con varianza conocida**

Cuando se conoce la varianza de la poblaci贸n, se puede calcular el intervalo de confianza para la media utilizando la distribuci贸n normal est谩ndar en lugar de la distribuci贸n `t de Student`. El intervalo de confianza se calcula utilizando la siguiente f贸rmula:

$Intervalo_{confianza} = media_{muestral} 卤 (valor_{cr铆tico} * error_{est谩ndar})$

donde el valor cr铆tico es el valor `z` de la tabla de distribuci贸n normal que corresponde al nivel de confianza deseado y el error est谩ndar se calcula como la desviaci贸n est谩ndar de la muestra dividida por la ra铆z cuadrada del tama帽o de la muestra.

A continuaci贸n se presenta un ejemplo de c贸mo calcular un intervalo de confianza dentro de la poblaci贸n con varianza conocida utilizando Python:

Suponga que se quiere calcular un intervalo de confianza del `95%` para la media de una poblaci贸n cuya desviaci贸n est谩ndar es conocida y es igual a `3.5`. Se toma una muestra de tama帽o `50` y se obtiene una media muestral de `24.5`. Primero se calcula el error est谩ndar:

```python
import numpy as np

# Datos de la muestra
n = 50
x_bar = 24.5
sigma = 3.5

# C谩lculo del error est谩ndar
se = sigma / np.sqrt(n)
```

A continuaci贸n, se calcula el valor cr铆tico utilizando la tabla de la distribuci贸n normal est谩ndar o la funci贸n `norm.ppf` de la biblioteca `SciPy`:

```python
from scipy.stats import norm

# Nivel de confianza del 95%
alpha = 0.05

# C谩lculo del valor cr铆tico
z_critico = norm.ppf(1 - alpha/2)
```

Luego, se calcula el intervalo de confianza:

```python
# C谩lculo del intervalo de confianza
lower = x_bar - z_critico * se
upper = x_bar + z_critico * se

print(f"Intervalo de confianza: ({lower}, {upper})")
```

Esto significa que se puede estar `95%` seguro de que el verdadero valor de la media poblacional se encuentra en el intervalo `(23.08, 25.92)` dado el tama帽o de muestra y la desviaci贸n est谩ndar conocida.

## **Intervalos de confianza. Varianza poblacional conocida**

Los intervalos de confianza son una herramienta estad铆stica que nos permite estimar un rango plausible de valores para un par谩metro de la poblaci贸n, como la media o proporci贸n, a partir de los datos de una muestra. Cuando la varianza poblacional es conocida, podemos utilizar la distribuci贸n normal para calcular los intervalos de confianza.

La f贸rmula para calcular el intervalo de confianza con varianza poblacional conocida es:

$Lower Bound = x - z*(/n)$

$Upper Bound = x + z*(/n)$

Donde `x` es la media muestral, `` es la desviaci贸n est谩ndar poblacional, `n` es el tama帽o de la muestra y `z` es el valor cr铆tico de la distribuci贸n normal est谩ndar correspondiente al nivel de confianza deseado.

A continuaci贸n, se presenta un ejemplo de c贸mo calcular un intervalo de confianza con varianza poblacional conocida en Python:

Supongamos que queremos estimar la altura media de los estudiantes universitarios en una poblaci贸n con una desviaci贸n est谩ndar poblacional conocida de `5 cm`. Tomamos una muestra aleatoria de `25` estudiantes y obtenemos una media muestral de `170 cm`. Queremos calcular un intervalo de confianza del `95%` para la altura media poblacional.

Podemos utilizar la funci贸n `scipy.stats.norm.ppf` para obtener el valor cr铆tico de la distribuci贸n normal est谩ndar correspondiente al nivel de confianza del `95%` `(伪=0.05)` y la f贸rmula anterior para calcular los l铆mites del intervalo de confianza.

```python
import numpy as np
from scipy.stats import norm

# datos de la muestra
data = np.array([168, 175, 172, 169, 174, 171, 170, 167, 173, 176,
                 170, 171, 169, 172, 175, 173, 171, 174, 169, 168,
                 172, 170, 173, 171, 172])

# tama帽o de la muestra, media muestral y desviaci贸n est谩ndar poblacional
n = len(data)
mean = np.mean(data)
sigma = 5

# valor cr铆tico de la distribuci贸n normal est谩ndar para un nivel de confianza del 95%
z = norm.ppf(1 - 0.05 / 2)

# l铆mites del intervalo de confianza
lower = mean - z * (sigma / np.sqrt(n))
upper = mean + z * (sigma / np.sqrt(n))

print("Intervalo de confianza del 95%: ({:.2f}, {:.2f})".format(lower, upper))
```

Esto nos dar谩 un intervalo de confianza del `95%` de `(168.69, 172.31)`, lo que significa que podemos estar seguros con un `95%` de confianza de que la altura media de los estudiantes universitarios en la poblaci贸n se encuentra en este rango.

Es importante tener en cuenta que el intervalo de confianza depende de la muestra aleatoria utilizada para estimar el par谩metro y del estad铆stico utilizado para calcular el intervalo. Por lo tanto, diferentes muestras aleatorias pueden dar lugar a diferentes intervalos de confianza.

En resumen, los intervalos de confianza son una herramienta invaluable en estad铆stica para estimar par谩metros desconocidos de una poblaci贸n a partir de una muestra aleatoria. Un intervalo de confianza es una herramienta estad铆stica que se utiliza para estimar un par谩metro desconocido de una poblaci贸n a partir de una muestra aleatoria. Es un rango de valores dentro del cual se espera que est茅 el verdadero valor del par谩metro con cierto nivel de confianza.

En general, el intervalo de confianza se define como:

$[estimador - margen de error, estimador + margen de error]$

Donde el estimador es el valor que se utiliza para estimar el par谩metro de la poblaci贸n, y el margen de error es una medida de la precisi贸n de la estimaci贸n. El nivel de confianza se refiere a la probabilidad de que el intervalo de confianza contenga el verdadero valor del par谩metro.

Es importante tener en cuenta que el intervalo de confianza se basa en una muestra aleatoria de la poblaci贸n y, por lo tanto, puede variar de una muestra a otra. Adem谩s, el tama帽o de la muestra y el nivel de confianza afectan el ancho del intervalo de confianza.

En Python, se pueden calcular los intervalos de confianza utilizando diferentes m茅todos, como el m茅todo de la distribuci贸n `t` o el m茅todo de la distribuci贸n normal, dependiendo de si la varianza de la poblaci贸n es conocida o desconocida.

Tomamos una muestra aleatoria de tama帽o `30` y obtenemos una media muestral de `75`. El intervalo de confianza se puede calcular de la siguiente manera en Python:

```python
import numpy as np
from scipy.stats import norm

# Datos de la muestra
n = 30
mean = 75
std_dev = 5

# Nivel de confianza
alpha = 0.05

# Valor cr铆tico de z para el nivel de confianza
z_crit = norm.ppf(1 - alpha/2)

# Intervalo de confianza
lower = mean - z_crit * std_dev / np.sqrt(n)
upper = mean + z_crit * std_dev / np.sqrt(n)

print("Intervalo de confianza del 95% para la media poblacional:", (lower, upper))
```

## **Varianza poblacional desconocida. Valor T**

Cuando la varianza poblacional es desconocida, se utiliza la distribuci贸n `T` de `Student` para calcular intervalos de confianza. **Donde:**

Cuando la varianza poblacional es desconocida, se utiliza la distribuci贸n `t` de `Student` para calcular el intervalo de confianza. En este caso, el valor cr铆tico `t` se calcula a partir de los grados de libertad, que est谩n dados por el tama帽o de la muestra menos uno.

La f贸rmula para calcular el intervalo de confianza en este caso es similar a la del intervalo de confianza con varianza poblacional conocida, pero se utiliza la distribuci贸n `t` en lugar de la distribuci贸n normal est谩ndar:

donde `IC` es el intervalo de confianza, la media muestral y el error est谩ndar se calculan de manera similar al caso con varianza poblacional conocida, y `t` es el valor cr铆tico de la distribuci贸n `t` de `Student` correspondiente al nivel de confianza deseado y los grados de libertad.

```python
import numpy as np
from scipy.stats import t

# datos de la muestra
data = np.array([10.2, 11.5, 9.7, 12.3, 8.9, 10.8, 11.1, 9.5, 10.6, 12.0])

# tama帽o de la muestra y nivel de confianza
n = len(data)
alpha = 0.05

# calculamos la media y la desviaci贸n est谩ndar de la muestra
mean = np.mean(data)
std_dev = np.std(data, ddof=1)

# calculamos el error est谩ndar
se = std_dev / np.sqrt(n)

# calculamos el valor cr铆tico de t para el nivel de confianza y grados de libertad
t_crit = t.ppf(1 - alpha / 2, n - 1)

# calculamos el intervalo de confianza
lower = mean - t_crit * se
upper = mean + t_crit * se

print(f"Intervalo de confianza al {1-alpha:.0%}: ({lower:.2f}, {upper:.2f})")
```
