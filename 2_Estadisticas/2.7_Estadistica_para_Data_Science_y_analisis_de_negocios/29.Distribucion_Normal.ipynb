## **1. Introducci贸n a la distribuci贸n normal**

La distribuci贸n normal, tambi茅n conocida como distribuci贸n de Gauss, es una distribuci贸n de probabilidad continua que es ampliamente utilizada en estad铆stica debido a que muchos fen贸menos naturales y sociales se distribuyen de manera aproximadamente normal. La distribuci贸n normal se caracteriza por su media (渭) y su desviaci贸n est谩ndar (), y tiene una forma de campana sim茅trica alrededor de su media.

En Python, se puede generar una distribuci贸n normal utilizando la librer铆a NumPy y la funci贸n ```numpy.random.normal()```. A continuaci贸n se muestra un ejemplo de c贸mo generar una distribuci贸n normal con media 0 y desviaci贸n est谩ndar 1 y graficar su densidad de probabilidad utilizando la librer铆a Matplotlib:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribuci贸n normal con media 0 y desviaci贸n est谩ndar 1
mu, sigma = 0, 1
datos = np.random.normal(mu, sigma, 1000)

# Graficar la densidad de probabilidad de la distribuci贸n normal
count, bins, ignored = plt.hist(datos, 30, density=True, color='lightblue', edgecolor='black', alpha=0.7, label='Datos')
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp( - (bins - mu)**2 / (2 * sigma**2))
plt.plot(bins, y, color='red', linewidth=2, label='Distribuci贸n normal')

# Personalizar la gr谩fica
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribuci贸n Normal')
plt.legend()

plt.show()
````

Este c贸digo generar谩 una distribuci贸n normal con media 0 y desviaci贸n est谩ndar 1 utilizando la funci贸n `numpy.random.normal()`, y luego graficar谩 su densidad de probabilidad utilizando la funci贸n `plt.hist()` y la funci贸n `plt.plot()`. La densidad de probabilidad se grafica utilizando la f贸rmula matem谩tica de la distribuci贸n normal, que es



La funci贸n de densidad de probabilidad (PDF) de la distribuci贸n normal est谩 dada por la f贸rmula:

$f(x | \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x - \mu}{\sigma}\right)^2}$

donde:
- $(x)$ es la variable aleatoria.
- $(\mu)$ es la media de la distribuci贸n.
- $(\sigma)$ es la desviaci贸n est谩ndar de la distribuci贸n.
- $(e)$ es la base del logaritmo natural, aproximadamente 2.71828.

Esta f贸rmula describe la forma de la campana de la distribuci贸n normal y especifica c贸mo la probabilidad se distribuye en torno a la media. La distribuci贸n normal es sim茅trica respecto a su media y su forma es determinada por la media y la desviaci贸n est谩ndar. La constante $\frac{1}{\sigma \sqrt{2\pi}}$ garantiza que el 谩rea total bajo la curva de la distribuci贸n normal sea igual a 1.

## **3. Explicaci贸n Distribuci贸n Normal**

La distribuci贸n normal, tambi茅n conocida como distribuci贸n gaussiana, es una distribuci贸n de probabilidad continua que se utiliza en muchos campos de la estad铆stica y las ciencias naturales. Esta distribuci贸n es sim茅trica alrededor de su media y su forma es una campana.

La distribuci贸n normal se define por dos par谩metros: su media ($\mu$) y su desviaci贸n est谩ndar ($\sigma$). La media es el punto central de la distribuci贸n y la desviaci贸n est谩ndar mide la dispersi贸n de los datos alrededor de la media.

En Python, podemos generar una distribuci贸n normal utilizando la librer铆a ```numpy``` y la funci贸n ```numpy.random.normal()```. En el siguiente ejemplo, generaremos una muestra aleatoria de datos normalmente distribuidos con una media de 0 y una desviaci贸n est谩ndar de 1:

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar una muestra aleatoria de datos normalmente distribuidos
mu, sigma = 0, 1 # media y desviaci贸n est谩ndar
datos = np.random.normal(mu, sigma, 1000)

# Configurar el estilo del gr谩fico con Seaborn
sns.set_style("whitegrid")

# Crear una figura y ejes
fig, ax = plt.subplots()

# Graficar el histograma de la muestra con barras azules
count, bins, ignored = ax.hist(datos, 30, density=True, color='blue', alpha=0.7, edgecolor='black')

# Calcular la curva de la distribuci贸n normal
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu) ** 2 / (2 * sigma ** 2))

# Graficar la curva de la distribuci贸n normal en rojo
ax.plot(bins, y, linewidth=2, color='red')

# Etiquetar los ejes y agregar un t铆tulo
ax.set_xlabel('Valores')
ax.set_ylabel('Densidad de probabilidad')
ax.set_title('Distribuci贸n Normal')

# Mostrar la gr谩fica
plt.show()
````

En este ejemplo, la muestra de datos se genera utilizando la funci贸n ```numpy.random.normal()```. Luego, se grafica un histograma de la muestra y se superpone una curva de la distribuci贸n normal utilizando la f贸rmula matem谩tica de la distribuci贸n normal. Finalmente, se muestra la gr谩fica resultante.

El resultado ser谩 una gr谩fica con la forma t铆pica de una distribuci贸n normal: una curva en forma de campana, centrada en la media y con una dispersi贸n determinada por la desviaci贸n est谩ndar.

## **4. Regla Emp铆rica 68-95-99.7**

La regla emp铆rica 68-95-99.7, tambi茅n conocida como la regla de tres sigmas, es una regla emp铆rica que se aplica a una distribuci贸n normal. La regla establece que el 68% de los datos se encuentran dentro de una desviaci贸n est谩ndar de la media, el 95% de los datos se encuentran dentro de dos desviaciones est谩ndar de la media, y el 99.7% de los datos se encuentran dentro de tres desviaciones est谩ndar de la media.

A continuaci贸n se muestra un ejemplo de c贸mo se puede aplicar la regla emp铆rica en Python utilizando la librer铆a ```numpy``` para generar datos aleatorios distribuidos normalmente y la librer铆a ```matplotlib``` para graficar la distribuci贸n y las 谩reas bajo la curva correspondientes a cada intervalo de la regla emp铆rica:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios distribuidos normalmente con media de 0 y desviaci贸n est谩ndar de 1
datos = np.random.normal(0, 1, 100000)

# Calcular la media y la desviaci贸n est谩ndar de los datos
media = np.mean(datos)
desv_std = np.std(datos)

# Calcular los l铆mites de los intervalos de la regla emp铆rica
lim_inf_1 = media - desv_std
lim_sup_1 = media + desv_std

lim_inf_2 = media - 2 * desv_std
lim_sup_2 = media + 2 * desv_std

lim_inf_3 = media - 3 * desv_std
lim_sup_3 = media + 3 * desv_std

# Crear una figura y ejes
fig, ax = plt.subplots()

# Graficar la distribuci贸n de los datos y los intervalos de la regla emp铆rica
plt.hist(datos, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
plt.plot(x, 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2/(2 * desv_std**2)), color='black')
plt.axvline(lim_inf_1, color='red', linestyle='--', label='1 desv. est谩ndar')
plt.axvline(lim_sup_1, color='red', linestyle='--')
plt.axvline(lim_inf_2, color='green', linestyle='--', label='2 desv. est谩ndar')
plt.axvline(lim_sup_2, color='green', linestyle='--')
plt.axvline(lim_inf_3, color='purple', linestyle='--', label='3 desv. est谩ndar')
plt.axvline(lim_sup_3, color='purple', linestyle='--')

# Agregar leyenda y t铆tulo al gr谩fico
plt.legend()
plt.title('Regla emp铆rica 68-95-99.7')
ax.set_xlabel('Valores')
ax.set_ylabel('Densidad de probabilidad')

# Mostrar la gr谩fica
plt.show()
````

## **5. Ejemplificaci贸n de la Regla Emp铆rica**

La Regla Emp铆rica, tambi茅n conocida como la regla de 68-95-99.7, es una regla emp铆rica que describe c贸mo se distribuyen los datos en una distribuci贸n normal. Esta regla establece que alrededor del 68% de los datos se encuentran dentro de una desviaci贸n est谩ndar de la media, el 95% de los datos se encuentran dentro de dos desviaciones est谩ndar de la media, y el 99.7% de los datos se encuentran dentro de tres desviaciones est谩ndar de la media.

A continuaci贸n se presenta un ejemplo en Python que muestra c贸mo usar la Regla Emp铆rica para graficar una distribuci贸n normal y resaltar el porcentaje de datos que se encuentran dentro de cada una de las desviaciones est谩ndar.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribuci贸n normal
mu = 0
sigma = 1
x = np.random.normal(mu, sigma, size=10000)

# Crear una figura y ejes
fig, ax = plt.subplots()

# Graficar la distribuci贸n normal
count, bins, ignored = plt.hist(x, 30, density=True, color='lightblue', alpha=0.7, edgecolor='black')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='red', label='Distribuci贸n normal')

# Calcular los porcentajes de datos dentro de cada desviaci贸n est谩ndar
within_1std = np.sum((x >= mu - sigma) & (x <= mu + sigma)) / len(x) * 100
within_2std = np.sum((x >= mu - 2*sigma) & (x <= mu + 2*sigma)) / len(x) * 100
within_3std = np.sum((x >= mu - 3*sigma) & (x <= mu + 3*sigma)) / len(x) * 100

# Mostrar los porcentajes en la gr谩fica
plt.text(0.5, 0.9, f'68% de los datos\n(dentro de 1 desviaci贸n est谩ndar)', ha='center', va='center', transform=ax.transAxes)
plt.text(0.5, 0.7, f'95% de los datos\n(dentro de 2 desviaciones est谩ndar)', ha='center', va='center', transform=ax.transAxes)
plt.text(0.5, 0.5, f'99.7% de los datos\n(dentro de 3 desviaciones est谩ndar)', ha='center', va='center', transform=ax.transAxes)

# Agregar etiquetas y t铆tulo
ax.set_xlabel('Valores')
ax.set_ylabel('Densidad de probabilidad')
plt.title('Distribuci贸n Normal y Regla Emp铆rica')

# Mostrar la leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
````

Este c贸digo genera una muestra de 10000 datos de una distribuci贸n normal con media 0 y desviaci贸n est谩ndar 1, y luego grafica la distribuci贸n normal utilizando el histograma y la curva de densidad. Luego, utiliza las m谩scaras de NumPy para contar los datos que se encuentran dentro de cada una de las desviaciones est谩ndar y calcula los porcentajes correspondientes. Finalmente, muestra los porcentajes en la gr谩fica utilizando la funci贸n text de Matplotlib. El resultado es una gr谩fica que muestra la distribuci贸n normal y resalta el porcentaje de datos que se encuentran dentro de cada una de las desviaciones est谩ndar de la media.
