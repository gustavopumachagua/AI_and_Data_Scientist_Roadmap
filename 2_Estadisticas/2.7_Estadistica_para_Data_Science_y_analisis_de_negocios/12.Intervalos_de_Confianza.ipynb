# **12. Formaliza tus resultados con Intervalos de Confianza: Qu茅 son y c贸mo calcularlos**

## **Introducci贸n**

Los intervalos de confianza son un rango de valores alrededor de una estimaci贸n de un par谩metro poblacional, como la media o la proporci贸n, que se utiliza para proporcionar una medida de la incertidumbre asociada con la estimaci贸n. En otras palabras, un intervalo de confianza indica el rango de valores dentro del cual es probable que se encuentre el verdadero valor del par谩metro poblacional con cierto nivel de confianza.

Por ejemplo, si se desea estimar la media de una poblaci贸n a partir de una muestra de datos, se puede calcular un intervalo de confianza para esta media. Si se utiliza un nivel de confianza del 95%, el intervalo de confianza indicar谩 que es probable que el verdadero valor de la media poblacional se encuentre dentro del rango de valores calculados con un 95% de confianza.

Los intervalos de confianza son una herramienta 煤til en la estad铆stica inferencial, ya que permiten estimar el verdadero valor de un par谩metro poblacional a partir de una muestra de datos y proporcionan informaci贸n sobre la incertidumbre asociada con la estimaci贸n.

## **Definici贸n del Intervalo de Confianza**

El intervalo de confianza es un rango de valores que se calcula a partir de una muestra de datos y que tiene una probabilidad determinada de contener el verdadero valor de la poblaci贸n. En otras palabras, el intervalo de confianza nos indica qu茅 tan precisos son nuestros c谩lculos y cu谩nta variabilidad hay en los datos.

Por ejemplo, si se quiere estimar el promedio de altura de una poblaci贸n, se puede tomar una muestra aleatoria de individuos y calcular la media de esta muestra. Sin embargo, esta media muestral no ser谩 necesariamente igual al verdadero valor de la poblaci贸n. Por lo tanto, se puede construir un intervalo de confianza alrededor de la media muestral, que indique la precisi贸n de esta estimaci贸n y la probabilidad de contener el verdadero valor de la poblaci贸n.

A continuaci贸n, se muestra un ejemplo de c贸mo calcular e interpretar un intervalo de confianza para la media de una distribuci贸n normal:

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generar datos aleatorios
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la media y el error est谩ndar de la media
media = np.mean(datos)
error = stats.sem(datos)

# Calcular el intervalo de confianza al 95%
intervalo = stats.t.interval(0.95, len(datos)-1, loc=media, scale=error)

# Personalizar colores y estilos
plt.hist(datos, bins=30, alpha=0.5, color='blue', edgecolor='black')
plt.axvline(intervalo[0], color='red', linestyle='dashed', linewidth=2, label='Intervalo de confianza al 95%')
plt.axvline(intervalo[1], color='red', linestyle='dashed', linewidth=2)
plt.axvline(media, color='green', linestyle='dashed', linewidth=2, label='Media')

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Agregar etiquetas de ejes y un t铆tulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Intervalo de Confianza')

# Mostrar la gr谩fica
plt.show()
```

## **Interpretaci贸n del Nivel de Confianza**

El nivel de confianza se refiere a la probabilidad de que el intervalo de confianza contenga el verdadero valor del par谩metro poblacional. Por ejemplo, si se establece un nivel de confianza del 95%, se espera que el verdadero valor del par谩metro se encuentre dentro del intervalo de confianza el 95% del tiempo en que se realiza la prueba.

A continuaci贸n se muestra un ejemplo de c贸mo interpretar el nivel de confianza en un intervalo de confianza para la media poblacional.

Supongamos que se desea conocer el intervalo de confianza para la media de una poblaci贸n normalmente distribuida con desviaci贸n est谩ndar conocida. Se realiza una muestra de tama帽o 50 y se obtiene una media muestral de 25 y una desviaci贸n est谩ndar poblacional de 5. Se desea calcular un intervalo de confianza al 95%.

El c贸digo en Python para calcular y graficar el intervalo de confianza es el siguiente:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar los datos aleatorios
datos = np.random.normal(loc=25, scale=5, size=50)

# Calcular la media muestral y el error est谩ndar
media = np.mean(datos)
se = stats.sem(datos)

# Calcular el intervalo de confianza
ic = stats.norm.interval(0.95, loc=media, scale=se)

# Graficar el histograma y el intervalo de confianza con colores y estilos personalizados
plt.hist(datos, bins=10, alpha=0.5, color='blue', edgecolor='black', label='Datos')
plt.axvline(ic[0], color='red', linestyle='dashed', linewidth=2, label='Intervalo de confianza al 95%')
plt.axvline(ic[1], color='red', linestyle='dashed', linewidth=2)

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Agregar etiquetas de ejes y un t铆tulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Intervalo de Confianza')

# Mostrar la gr谩fica
plt.show()
```

## **F贸rmula del Intervalo de Confianza**

La f贸rmula general para el intervalo de confianza de una muestra con una distribuci贸n normal y una desviaci贸n est谩ndar poblacional conocida es:

$IC = x 卤 Z_{(伪/2)} * (/n)$

donde:

* IC es el intervalo de confianza
* $\bar{x}$ es la media muestral
* $Z_{\alpha/2}$ es el valor cr铆tico de la distribuci贸n normal est谩ndar que corresponde al nivel de confianza deseado ($\alpha$ es el nivel de significancia, por lo que $1-\alpha$ es el nivel de confianza)
* $\sigma$ es la desviaci贸n est谩ndar poblacional
* $n$ es el tama帽o de la muestra

## **Error Est谩ndar de la Media**

El Error Est谩ndar de la Media (EE) es una medida de la precisi贸n de la estimaci贸n de la media poblacional a partir de una muestra. Se define como la desviaci贸n est谩ndar de la distribuci贸n muestral de medias.

El $EE$ es 煤til para estimar el rango de valores dentro del cual se espera que se encuentre la media poblacional con cierto nivel de confianza.

La f贸rmula del $EE$ es:

$EE =  / n$

donde $$ es la desviaci贸n est谩ndar de la poblaci贸n y $n$ es el tama帽o de la muestra.

A continuaci贸n se presenta un ejemplo de c贸mo calcular y graficar el $EE$ utilizando Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con una distribuci贸n normal
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular el error est谩ndar de la media
ee = np.std(datos) / np.sqrt(len(datos))

# Graficar los datos y el error est谩ndar con colores y estilos personalizados
plt.hist(datos, bins=30, alpha=0.5, color='blue', edgecolor='black', label='Datos')
plt.axvline(np.mean(datos), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(np.mean(datos) - ee, color='green', linestyle='dashed', linewidth=2, label='Error Est谩ndar')
plt.axvline(np.mean(datos) + ee, color='green', linestyle='dashed', linewidth=2)

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Agregar etiquetas de ejes y un t铆tulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Media y Error Est谩ndar')

# Mostrar la gr谩fica
plt.show()
```

## **Valor Z conforme al Nivel de Confianza**

El valor Z es un valor cr铆tico de la distribuci贸n normal est谩ndar que se utiliza para calcular intervalos de confianza y probar hip贸tesis. El valor Z se relaciona con el nivel de confianza deseado y se utiliza para calcular el margen de error en una muestra aleatoria.

El valor Z depende del nivel de confianza deseado. Por ejemplo, para un nivel de confianza del 95%, el valor Z es de 1.96. Esto significa que el intervalo de confianza de la media estar谩 dentro de 1.96 errores est谩ndar de la media (SEM) alrededor de la media muestral.

El valor Z se puede encontrar en tablas de distribuci贸n normal est谩ndar o se puede calcular utilizando funciones estad铆sticas en lenguajes de programaci贸n como Python.

A continuaci贸n se muestra un ejemplo de c贸digo en Python para calcular el valor Z dado un nivel de confianza:

```python
import scipy.stats as st

confianza = 0.95

# Calcula el valor Z para el nivel de confianza dado
valor_z = st.norm.ppf((1 + confianza) / 2)

print("Valor Z para un nivel de confianza del {}%: {:.2f}".format(confianza * 100, valor_z))
```

## **Ejemplo de c谩lculo del Intervalo de Confianza**

Supongamos que tenemos un conjunto de datos con una distribuci贸n normal y queremos calcular el intervalo de confianza al 95% para la media poblacional. Primero, generamos los datos aleatorios:

```python
import numpy as np

np.random.seed(123) # para reproducibilidad
datos = np.random.normal(loc=10, scale=2, size=100)
```

Luego, calculamos la media y el error est谩ndar de la media:

```python
media_muestral = np.mean(datos)
sem = np.std(datos, ddof=1) / np.sqrt(len(datos))
```

El factor de correcci贸n `ddof=1` en la funci贸n `std()` indica que se est谩 calculando la desviaci贸n est谩ndar muestral. Luego, podemos calcular el valor cr铆tico z correspondiente al nivel de confianza del 95%:

```python
from scipy.stats import norm

alpha = 0.05
z = norm.ppf(1 - alpha/2)
```

Finalmente, podemos calcular el intervalo de confianza como:

```python
intervalo = (media_muestral - z*sem, media_muestral + z*sem)
print(f"El intervalo de confianza al 95% es: {intervalo}")
```

```python
import matplotlib.pyplot as plt

# Graficar el histograma de los datos y el intervalo de confianza
plt.hist(datos, bins=20, alpha=0.5, color='blue', edgecolor='black', label='Datos')
plt.axvline(intervalo[0], color='red', linestyle='dashed', linewidth=2, label='Intervalo de Confianza (95%)')
plt.axvline(intervalo[1], color='red', linestyle='dashed', linewidth=2)
plt.legend(loc='upper right')

# Etiquetas de ejes y t铆tulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Intervalo de Confianza al 95%')

# Mostrar la gr谩fica
plt.show()
```

## **Intervalo de Confianza con Python**

Ya hemos hablado previamente sobre el intervalo de confianza en las preguntas anteriores, pero de manera general se puede decir que el intervalo de confianza es un rango num茅rico que contiene el valor desconocido del par谩metro poblacional con un cierto nivel de confianza.

Para entender mejor el intervalo de confianza, aqu铆 hay un ejemplo con c贸digo y gr谩fica.

Suponga que tenemos una muestra aleatoria de 1000 observaciones de una distribuci贸n normal con una media desconocida pero desviaci贸n est谩ndar conocida de 10. Queremos encontrar el intervalo de confianza del 95% para la media poblacional.

Primero, importamos las bibliotecas necesarias:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

Luego, generamos datos aleatorios:

```python
np.random.seed(1234)
datos = np.random.normal(loc=50, scale=10, size=1000)
```

A continuaci贸n, calculamos el intervalo de confianza:

```python
media = np.mean(datos)
nivel_confianza = 0.95
z_score = stats.norm.ppf((1 + nivel_confianza) / 2)
error_estandar = 10 / np.sqrt(1000)
intervalo_confianza = (media - z_score * error_estandar, media + z_score * error_estandar)
```

Finalmente, imprimimos el resultado y graficamos el histograma y el intervalo de confianza:

```python
print("Intervalo de confianza del 95% para la media poblacional:", intervalo_confianza)

# Graficar el histograma de datos y el intervalo de confianza
plt.hist(datos, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black', label='Datos')
plt.axvline(intervalo_confianza[0], color='red', linestyle='dashed', linewidth=2, label='Intervalo de Confianza (95%)')
plt.axvline(intervalo_confianza[1], color='red', linestyle='dashed', linewidth=2)
plt.legend(loc='upper right')

# Etiquetas de ejes y t铆tulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Intervalo de Confianza al 95%')

# Mostrar la gr谩fica
plt.show()
```

## **Graficando el Intervalo de Confianza con Seaborn**

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra aleatoria de datos
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la media y el error est谩ndar
media = np.mean(datos)
error_estandar = np.std(datos) / np.sqrt(len(datos))

# Crear una figura
plt.figure(figsize=(8, 6))

# Graficar la distribuci贸n de la muestra con el intervalo de confianza
sns.histplot(datos, kde=True, color='lightblue', bins=20, element='step')
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media')
plt.fill_between([media - 1.96 * error_estandar, media + 1.96 * error_estandar], 0, 0.08, color='lightcoral', alpha=0.5, label='Intervalo de Confianza (95%)')

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.title('Distribuci贸n de Datos con Intervalo de Confianza al 95%')

# Agregar leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
```

Luego, se grafica la distribuci贸n de la muestra con un histograma y una estimaci贸n de densidad de kernel (kde), y se agrega una l铆nea roja que representa el intervalo de confianza del 95% para la media de la poblaci贸n. La l铆nea roja se traza utilizando los l铆mites inferior y superior del intervalo, que se calculan como la media muestral m谩s y menos 1.96 veces el error est谩ndar, respectivamente.

Este gr谩fico permite visualizar la precisi贸n de la estimaci贸n de la media de la poblaci贸n a partir de la muestra y c贸mo esta precisi贸n se ve afectada por el tama帽o de la muestra y la varianza de la poblaci贸n.
