Para obtener la puntuaci贸n Z de un valor espec铆fico, primero debemos restar la media de la distribuci贸n de los datos y luego dividir por la desviaci贸n est谩ndar. Esto nos dar谩 una medida de cu谩ntas desviaciones est谩ndar se encuentra el valor de inter茅s de la media de la distribuci贸n.

## **2. Distribuci贸n Normal**

La distribuci贸n normal es una distribuci贸n de probabilidad continua que se utiliza com煤nmente en estad铆stica para describir datos que tienen una distribuci贸n sim茅trica alrededor de la media. Tiene una forma de campana y se caracteriza por dos par谩metros: la media ($\mu$) y la desviaci贸n est谩ndar ($\sigma$).

En Python, la distribuci贸n normal se puede generar utilizando la funci贸n ```numpy.random.normal()``` de la librer铆a NumPy. La funci贸n toma como argumentos la media, la desviaci贸n est谩ndar y el tama帽o de la muestra. A continuaci贸n se muestra un ejemplo de c贸mo generar y graficar una muestra de datos con distribuci贸n normal:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos con distribuci贸n normal
mu = 0  # media
sigma = 1  # desviaci贸n est谩ndar
size = 10000  # tama帽o de la muestra
data = np.random.normal(mu, sigma, size)

# Crear una figura y ejes
fig, ax = plt.subplots()

# Graficar el histograma de la muestra
plt.hist(data, bins=50, density=True, alpha=0.6, color='lightblue', edgecolor='black', label='Datos')

# Graficar la curva de la distribuci贸n normal te贸rica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )
plt.plot(x, y, 'r--', linewidth=2, label='Distribuci贸n normal')

# Agregar etiquetas y t铆tulo
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribuci贸n Normal')

# Agregar leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
````

En este ejemplo, se gener贸 una muestra de 10000 datos con una media de cero y una desviaci贸n est谩ndar de uno. Luego se grafic贸 el histograma de la muestra y la curva de la distribuci贸n normal te贸rica utilizando la f贸rmula matem谩tica de la distribuci贸n normal est谩ndar. El resultado es una gr谩fica que muestra la distribuci贸n normal de la muestra de datos generada

La curva roja representa la distribuci贸n normal te贸rica, mientras que el histograma azul representa la distribuci贸n emp铆rica de la muestra generada. La curva te贸rica es sim茅trica alrededor de la media (cero) y tiene una desviaci贸n est谩ndar de uno, lo que indica que la mayor铆a de los datos de la muestra se encuentran dentro del rango de $\pm$ desviaciones est谩ndar alrededor de la media. Esta propiedad se puede utilizar para calcular las puntuaciones $Z$ y realizar pruebas de hip贸tesis estad铆sticas basadas en la distribuci贸n normal.

## **3. Ejemplo con 2 muestras**

Supongamos que queremos comparar la altura promedio de hombres y mujeres. Tenemos dos muestras independientes, una con 100 hombres y otra con 100 mujeres, y queremos ver si hay una diferencia significativa entre las alturas promedio de ambas muestras.

Primero, cargamos las librer铆as necesarias y generamos datos de altura aleatorios para cada muestra:

````python
import numpy as np
import scipy.stats as stats

# Generar datos aleatorios
np.random.seed(123)  # Para reproducibilidad
altura_hombres = np.random.normal(loc=175, scale=7, size=100)
altura_mujeres = np.random.normal(loc=162, scale=6, size=100)
````

Luego, podemos calcular la media, la desviaci贸n est谩ndar y el tama帽o de cada muestra:

````python
# Calcular estad铆sticas descriptivas
media_hombres = np.mean(altura_hombres)
media_mujeres = np.mean(altura_mujeres)
std_hombres = np.std(altura_hombres, ddof=1)
std_mujeres = np.std(altura_mujeres, ddof=1)
n_hombres = len(altura_hombres)
n_mujeres = len(altura_mujeres)
````

Podemos graficar la distribuci贸n de alturas para cada muestra usando un histograma:

````python
import matplotlib.pyplot as plt

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 8))

# Graficar histogramas con colores personalizados
ax.hist(altura_hombres, bins=15, alpha=0.6, color='blue', label='Hombres')
ax.hist(altura_mujeres, bins=15, alpha=0.6, color='orange', label='Mujeres')

# Agregar l铆neas verticales para las medias
ax.axvline(x=media_hombres, color='navy', linestyle='dashed', linewidth=2, label=f'Media hombres ({media_hombres:.2f})')
ax.axvline(x=media_mujeres, color='darkorange', linestyle='dashed', linewidth=2, label=f'Media mujeres ({media_mujeres:.2f})')

# Agregar etiquetas y t铆tulo
ax.set_xlabel('Altura (cm)', fontsize=12)
ax.set_ylabel('Frecuencia', fontsize=12)
ax.set_title('Distribuci贸n de Altura en Hombres y Mujeres', fontsize=14)

# Agregar leyenda
ax.legend(loc='upper right')

# Ajustar el estilo de la cuadr铆cula y el fondo
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_facecolor('whitesmoke')  # Cambia el color de fondo del gr谩fico

# Mostrar la gr谩fica
plt.show()
````

Esto deber铆a producir una gr谩fica con dos histogramas superpuestos, uno para la altura de los hombres y otro para la altura de las mujeres, y dos l铆neas verticales punteadas que representan la altura promedio de cada muestra.

Para hacer una comparaci贸n formal de las dos muestras, podemos realizar una prueba t de dos muestras. En este caso, asumimos que las dos muestras son independientes y que tienen varianzas iguales (aunque podemos usar una correcci贸n de Welch si no se cumple esta 煤ltima suposici贸n). Usando la funci贸n ```ttest_ind``` de la librer铆a ```scipy.stats```, podemos calcular el valor p de la prueba t:

````python
# Prueba t de dos muestras
t, p = stats.ttest_ind(altura_hombres, altura_mujeres)
print('Valor t: {:.2f}'.format(t))
print('Valor p: {:.4f}'.format(p))
````

La distribuci贸n normal est谩ndar es una distribuci贸n normal con una media de cero y una desviaci贸n est谩ndar de uno. Esta distribuci贸n se utiliza com煤nmente en la estad铆stica para estandarizar los datos y hacerlos comparables.

En Python, se puede generar una distribuci贸n normal est谩ndar utilizando la librer铆a ```numpy``` y la funci贸n ```random.normal()``` especificando una media de cero y una desviaci贸n est谩ndar de uno:

## **5. Estandarizaci贸n de las muestras**

La estandarizaci贸n de una muestra es un proceso mediante el cual se transforma la distribuci贸n de los datos de la muestra en una distribuci贸n normal est谩ndar, es decir, una distribuci贸n normal con media cero y desviaci贸n est谩ndar igual a uno. Esto se logra restando la media de la muestra a cada uno de los valores de la muestra y luego dividiendo el resultado entre la desviaci贸n est谩ndar de la muestra.

A continuaci贸n se presenta un ejemplo con c贸digo y gr谩fica para ilustrar la estandarizaci贸n de una muestra:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos con media 10 y desviaci贸n est谩ndar 3
np.random.seed(1)
muestra = np.random.normal(loc=10, scale=3, size=100)

# Calcular la media y desviaci贸n est谩ndar de la muestra
media = np.mean(muestra)
desviacion = np.std(muestra)

# Estandarizar la muestra

estandarizada = (muestra - media) / desviacion

# Crear una figura con dos subgr谩ficos
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Personalizar los histogramas
hist_kw = dict(bins=20, alpha=0.7, edgecolor='black')

# Graficar la distribuci贸n original y la estandarizada
axs[0].hist(muestra, **hist_kw, color='blue')
axs[0].set_title('Distribuci贸n Original')
axs[0].set_xlabel('Valores')
axs[0].set_ylabel('Frecuencia')

axs[1].hist(estandarizada, **hist_kw, color='green')
axs[1].set_title('Distribuci贸n Estandarizada')
axs[1].set_xlabel('Valores')
axs[1].set_ylabel('Frecuencia')

plt.show()
````

## **6. Valores o Puntuaciones Z**

Los valores o puntuaciones Z representan el n煤mero de desviaciones est谩ndar que una observaci贸n est谩 por encima o por debajo de la media de una distribuci贸n. Es decir, cuantifican la distancia de una observaci贸n respecto a la media en t茅rminos de la desviaci贸n est谩ndar de la distribuci贸n. La estandarizaci贸n de las muestras es 煤til para comparar observaciones de diferentes distribuciones, ya que permite expresarlas en una misma escala.

La f贸rmula para calcular el valor Z de una observaci贸n $x$ es:


 $Z = (x - 渭) / $

Donde $\mu$ es la media de la distribuci贸n y $\sigma$ es la desviaci贸n est谩ndar.

A continuaci贸n se muestra un ejemplo de c贸mo calcular los valores Z y graficar su distribuci贸n:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra aleatoria de una distribuci贸n normal
np.random.seed(1)
mu = 50
sigma = 10
sample = np.random.normal(mu, sigma, size=100)

# Calcular los valores Z de la muestra
z_scores = (sample - mu) / sigma

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))

# Personalizar el histograma
ax.hist(z_scores, bins=10, alpha=0.5, color='blue', edgecolor='black')

# Personalizar los ejes y el t铆tulo
ax.set_xlabel('Valores Z')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribuci贸n de los valores Z')

# Mostrar la gr谩fica
plt.show()
````

## **7. Propiedades**

Las principales propiedades de los valores o puntuaciones Z son:

* Media igual a cero: la media de los valores Z siempre ser谩 cero.
* Desviaci贸n est谩ndar igual a uno: la desviaci贸n est谩ndar de los valores Z siempre ser谩 uno.
* Distribuci贸n normal est谩ndar: los valores Z se distribuyen normalmente alrededor de cero con una desviaci贸n est谩ndar de uno.
* La transformaci贸n Z es lineal: si se aplican las mismas transformaciones lineales a la variable original y a los valores Z, los valores Z seguir谩n siendo los mismos. Esto permite comparar directamente valores entre diferentes distribuciones.

## **8. Distribuci贸n Normal Est谩ndar y Probabilidad**

La distribuci贸n normal est谩ndar es una distribuci贸n normal con media cero y desviaci贸n est谩ndar uno. Se utiliza con frecuencia para estandarizar una variable normal y convertirla en una variable normal est谩ndar. La probabilidad asociada con una puntuaci贸n Z determinada en la distribuci贸n normal est谩ndar se puede obtener utilizando una tabla de valores Z o una funci贸n de distribuci贸n acumulativa.

A continuaci贸n se presenta un ejemplo de c贸mo calcular la probabilidad asociada con una puntuaci贸n Z utilizando la funci贸n de distribuci贸n acumulativa de la distribuci贸n normal est谩ndar en Python:

````python
import scipy.stats as st

# Calcular la probabilidad asociada con una puntuaci贸n Z de 1.5
z_score = 1.5
prob = st.norm.cdf(z_score)

print("La probabilidad asociada con una puntuaci贸n Z de 1.5 es:", prob)
````

Esto significa que la probabilidad de obtener una puntuaci贸n Z de 1.5 o menor en la distribuci贸n normal est谩ndar es del 93.32%.

A continuaci贸n, se muestra c贸mo graficar la distribuci贸n normal est谩ndar y sombrear la probabilidad asociada con una puntuaci贸n Z espec铆fica:

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# Crear un array de valores Z desde -3 hasta 3
z_values = np.linspace(-3, 3, 1000)

# Calcular la densidad de probabilidad para los valores Z
pdf = st.norm.pdf(z_values, loc=0, scale=1)

# Calcular la probabilidad asociada con una puntuaci贸n Z de 1.5
z_score = 1.5
prob = st.norm.cdf(z_score)

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar la densidad de probabilidad de la distribuci贸n normal est谩ndar
ax.plot(z_values, pdf, color='black', linewidth=2)

# Sombrear la probabilidad asociada con una puntuaci贸n Z de 1.5
ax.fill_between(z_values[z_values <= z_score], pdf[z_values <= z_score], color='red', alpha=0.5)

# Personalizar los ejes y el t铆tulo
ax.set_xlabel('Valores Z')
ax.set_ylabel('Densidad de probabilidad')
ax.set_title('Distribuci贸n normal est谩ndar')

# Mostrar la gr谩fica
plt.show()
````
