## **Variables num茅ricas. Usar una tabla de distribuci贸n de frecuencia**

Las variables num茅ricas son aquellas que contienen valores num茅ricos y pueden ser medidas y analizadas en t茅rminos de estad铆sticas descriptivas, como la media, la mediana, la moda, la varianza y la desviaci贸n est谩ndar. Estas variables se pueden dividir en dos tipos: variables continuas y variables discretas.

* Las variables continuas pueden tomar cualquier valor dentro de un rango determinado, como la edad, la altura o el peso. Las variables discretas, por otro lado, solo pueden tomar valores enteros espec铆ficos, como el n煤mero de hijos o el n煤mero de estudiantes en una clase.

* Para analizar una variable num茅rica, una herramienta 煤til es la tabla de distribuci贸n de frecuencia. Esta tabla muestra la frecuencia de cada valor de la variable y se utiliza para resumir grandes conjuntos de datos num茅ricos. La tabla de distribuci贸n de frecuencia se divide en varias columnas, que incluyen los valores de la variable, la frecuencia absoluta (n煤mero de veces que aparece cada valor en el conjunto de datos), la frecuencia relativa (proporci贸n de cada valor en relaci贸n con el n煤mero total de observaciones), y la frecuencia acumulada (suma acumulada de las frecuencias relativas).

A continuaci贸n se muestra un ejemplo de c贸mo crear una tabla de distribuci贸n de frecuencia para una variable num茅rica en Python:

```python
import pandas as pd
import numpy as np

# Creamos un conjunto de datos aleatorio de edades
edades = np.random.randint(18, 60, size=50)

# Creamos una tabla de frecuencia
tabla_frecuencia = pd.DataFrame({"Edad": edades})
tabla_frecuencia = tabla_frecuencia.groupby("Edad").size().reset_index(name="Frecuencia")

# Agregamos las columnas de frecuencia relativa y frecuencia acumulada
tabla_frecuencia["Frecuencia relativa"] = tabla_frecuencia["Frecuencia"] / len(edades)
tabla_frecuencia["Frecuencia acumulada"] = tabla_frecuencia["Frecuencia relativa"].cumsum()

# Imprimimos la tabla de frecuencia
print(tabla_frecuencia)
```

Este c贸digo crear谩 una tabla de distribuci贸n de frecuencia para un conjunto de datos de edades aleatorias. La tabla mostrar谩 cada valor de edad y su frecuencia correspondiente, as铆 como la frecuencia relativa y acumulada.
