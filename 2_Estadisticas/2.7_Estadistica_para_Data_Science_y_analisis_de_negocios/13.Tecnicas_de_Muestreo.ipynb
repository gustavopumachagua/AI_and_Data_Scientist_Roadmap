## **Introducci贸n**

* Muestreo por conglomerados: En esta t茅cnica, la poblaci贸n se divide en grupos m谩s grandes (conglomerados) y se seleccionan algunos de estos conglomerados al azar para formar la muestra. * Muestreo sistem谩tico: En esta t茅cnica, se elige un n煤mero aleatorio para comenzar y luego se selecciona cada elemento k de la poblaci贸n a partir de ese n煤mero. ## **Muestreo por Conveniencia**

El muestreo por conveniencia es una t茅cnica de muestreo no probabil铆stico en la que se seleccionan los elementos de la muestra de acuerdo con la conveniencia del investigador o del encuestado. Es decir, se seleccionan aquellos elementos que est谩n disponibles y que son f谩ciles de conseguir.

```python
import random
import matplotlib.pyplot as plt

# Crear una lista de valores de edad
edades = [23, 32, 27, 19, 41, 38, 29, 25, 34, 31, 37, 22, 26, 39]

# Realizar un muestreo por conveniencia de 5 participantes
muestra = random.sample(edades, 5)

# Crear un histograma de las edades
plt.hist(edades, bins=10, alpha=0.5, label='Poblaci贸n', color='blue', edgecolor='black')
plt.hist(muestra, bins=5, alpha=0.5, label='Muestra', color='orange', edgecolor='black')

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades de la Poblaci贸n y la Muestra')

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

Es importante tener en cuenta que el muestreo por conveniencia no garantiza una muestra representativa de la poblaci贸n, por lo que los resultados pueden estar sesgados y no ser generalizables a la poblaci贸n completa.

## **Muestreo Aleatorio Simple**

El muestreo aleatorio simple es una t茅cnica de muestreo probabil铆stico en la que cada elemento de la poblaci贸n tiene la misma probabilidad de ser seleccionado para formar parte de la muestra. A continuaci贸n, te presento un ejemplo de c贸digo en Python para realizar un muestreo aleatorio simple y graficar los resultados:

```python
import random
import matplotlib.pyplot as plt

# Crear una lista de valores de edad
edades = [23, 32, 27, 19, 41, 38, 29, 25, 34, 31, 37, 22, 26, 39]

# Realizar un muestreo aleatorio simple de 5 participantes
muestra = random.sample(edades, 5)

# Crear un histograma de las edades
plt.hist(edades, bins=10, alpha=0.5, label='Poblaci贸n', color='blue', edgecolor='black')
plt.hist(muestra, bins=5, alpha=0.5, label='Muestra', color='orange', edgecolor='black')

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades de la Poblaci贸n y la Muestra')

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, se crea una lista de edades y se realiza un muestreo aleatorio simple de 5 participantes utilizando la funci贸n `random.sample()` de Python. La poblaci贸n completa se representa con una distribuci贸n de frecuencia de color azul claro, mientras que la muestra se representa con una distribuci贸n de frecuencia de color naranja.

Es importante destacar que el muestreo aleatorio simple proporciona una muestra representativa de la poblaci贸n, siempre y cuando se seleccione al azar y sin reemplazo, lo que significa que cada elemento de la poblaci贸n tiene la misma probabilidad de ser seleccionado y que solo se selecciona una vez.

## **Muestreo Sistem谩tico**

El muestreo sistem谩tico es una t茅cnica de muestreo probabil铆stico en la que se selecciona cada k-茅simo elemento de la poblaci贸n para formar parte de la muestra, comenzando por un elemento aleatorio. Es una t茅cnica 煤til cuando la poblaci贸n est谩 ordenada de alguna manera, por ejemplo, por orden alfab茅tico o por fecha.

A continuaci贸n, te presento un ejemplo de c贸digo en Python para realizar un muestreo sistem谩tico y graficar los resultados:

```python
import random
import matplotlib.pyplot as plt

# Crear una lista de valores de edad
edades = [23, 32, 27, 19, 41, 38, 29, 25, 34, 31, 37, 22, 26, 39]

# Definir el tama帽o de la muestra y el intervalo de selecci贸n
tam_muestra = 5
k = len(edades) // tam_muestra

# Seleccionar un elemento aleatorio como punto de partida
punto_partida = random.randint(0, k-1)

# Realizar un muestreo sistem谩tico de 5 participantes
muestra = [edades[i] for i in range(punto_partida, len(edades), k)]

# Crear un histograma de las edades
plt.hist(edades, bins=10, alpha=0.5, label='Poblaci贸n', color='blue', edgecolor='black')
plt.hist(muestra, bins=5, alpha=0.5, label='Muestra', color='orange', edgecolor='black')

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades de la Poblaci贸n y la Muestra')

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, se crea una lista de edades y se define el tama帽o de la muestra y el intervalo de selecci贸n. Luego se selecciona un elemento aleatorio como punto de partida y se realiza un muestreo sistem谩tico de 5 participantes utilizando un bucle for y una lista de comprensi贸n. Finalmente, se imprime la muestra y se grafican los resultados utilizando un histograma. La poblaci贸n completa se representa con una distribuci贸n de frecuencia de color azul claro, mientras que la muestra se representa con una distribuci贸n de frecuencia de color naranja.

Es importante destacar que el muestreo sistem谩tico puede proporcionar una muestra representativa de la poblaci贸n si se selecciona el intervalo de manera adecuada y si se garantiza que la poblaci贸n est茅 bien ordenada.





## **Muestreo por Conglomerados o Clusters**

El muestreo por conglomerados o clusters es una t茅cnica de muestreo probabil铆stico en la que la poblaci贸n se divide en grupos o conglomerados, y luego se seleccionan algunos de estos conglomerados al azar para formar la muestra. Una vez seleccionados los conglomerados, se toman mediciones de todos los elementos dentro de los conglomerados seleccionados. Esta t茅cnica se utiliza com煤nmente cuando los elementos de la poblaci贸n est谩n agrupados geogr谩ficamente o en alg煤n otro sentido.

A continuaci贸n, te presento un ejemplo de c贸digo en Python para realizar un muestreo por conglomerados y graficar los resultados:

```python
import random
import matplotlib.pyplot as plt

# Crear una lista de edades por ciudad
ciudades = {'Bogot谩': [23, 32, 27, 19, 41, 38, 29, 25, 34, 31], 
            'Medell铆n': [37, 22, 26, 39, 36, 24, 33, 28, 30]}

# Seleccionar 2 conglomerados al azar
conglomerados = random.sample(list(ciudades.keys()), 2)

# Tomar mediciones de todos los elementos dentro de los conglomerados seleccionados
muestra = []
for c in conglomerados:
    muestra.extend(ciudades[c])

# Crear un histograma de las edades
edades = ciudades['Bogot谩'] + ciudades['Medell铆n']

plt.hist(edades, bins=10, alpha=0.5, label='Poblaci贸n', color='blue', edgecolor='black')
plt.hist(muestra, bins=10, alpha=0.5, label='Muestra', color='orange', edgecolor='black')

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades de la Poblaci贸n y la Muestra')

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, se crea una lista de edades por ciudad y se seleccionan 2 conglomerados al azar utilizando la funci贸n `random.sample()` de Python. Luego se toman mediciones de todos los elementos dentro de los conglomerados seleccionados utilizando un bucle for y la funci贸n `extend()` de Python. Finalmente, se imprime la muestra y se grafican los resultados utilizando un histograma. La poblaci贸n completa se representa con una distribuci贸n de frecuencia de color azul claro, mientras que la muestra se representa con una distribuci贸n de frecuencia de color naranja.

Es importante destacar que el muestreo por conglomerados puede ser 煤til cuando la poblaci贸n se encuentra agrupada geogr谩ficamente o en alg煤n otro sentido, pero se debe tener cuidado en la selecci贸n de los conglomerados para asegurar que la muestra sea representativa de la poblaci贸n.

## **Funci贸n Choice de Numpy**

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un array de edades
edades = np.array([19, 21, 23, 24, 26, 27, 29, 31, 32, 33, 35, 37, 38, 39, 41])

# Seleccionar una muestra aleatoria de tama帽o 5
muestra = np.random.choice(edades, size=5, replace=False)

# Crear un histograma de las edades
plt.hist(edades, bins=[0, 20, 30, 40, 50], alpha=0.5, label='Poblaci贸n', color='blue', edgecolor='black', linewidth=1.2)
plt.hist(muestra, bins=[0, 20, 30, 40, 50], alpha=0.7, label='Muestra', color='orange', edgecolor='black', linewidth=1.2)

# Agregar etiquetas de ejes y t铆tulo
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Histograma de Edades de la Poblaci贸n y la Muestra', fontsize=14)

# Agregar una leyenda en la esquina superior derecha
plt.legend(loc='upper right')

# Personalizar la cuadr铆cula
plt.grid(axis='y', linestyle='--', alpha=0.7)

# A帽adir etiquetas sobre las barras
for i, freq in enumerate(np.histogram(edades, bins=[0, 20, 30, 40, 50])[0]):
    plt.text(edades.min() + i * 10, freq + 0.1, str(freq), ha='center', va='bottom')

# Mostrar la gr谩fica
plt.show()
```

En este ejemplo, se crea un array de edades utilizando la funci贸n `numpy.array()`. Luego, se utiliza la funci贸n `numpy.random.choice()` para seleccionar una muestra aleatoria de tama帽o 5. Se imprime la muestra y se grafican los resultados utilizando un histograma. La poblaci贸n completa se representa con una distribuci贸n de frecuencia de color azul claro, mientras que la muestra se representa con una distribuci贸n de frecuencia de color naranja.

Es importante destacar que la funci贸n `numpy.random.choice()` permite especificar si se permite la selecci贸n repetida de elementos `(replace=True)` o si se requiere que cada elemento sea seleccionado solo una vez `(replace=False)`. En el ejemplo anterior, se especific贸 `replace=False` para seleccionar una muestra sin reemplazo.
