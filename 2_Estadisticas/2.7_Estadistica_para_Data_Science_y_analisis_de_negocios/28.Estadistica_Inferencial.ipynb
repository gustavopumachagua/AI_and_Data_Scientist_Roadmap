## **1. Introducci贸n**

La estad铆stica inferencial es la rama de la estad铆stica que se encarga de realizar inferencias sobre una poblaci贸n a partir de una muestra de datos. En otras palabras, permite generalizar las conclusiones obtenidas de un conjunto limitado de datos a toda la poblaci贸n.

Por ejemplo, supongamos que queremos estimar la altura promedio de los estudiantes de una universidad. Ser铆a imposible medir la altura de todos los estudiantes, por lo que se puede tomar una muestra aleatoria de ellos y utilizar esa informaci贸n para hacer una inferencia sobre la altura promedio de todos los estudiantes de la universidad.

Para realizar inferencias estad铆sticas, se utilizan t茅cnicas como la estimaci贸n de par谩metros y la prueba de hip贸tesis.

A continuaci贸n se muestra un ejemplo de c贸mo realizar una prueba de hip贸tesis en Python. Supongamos que se tiene una muestra de notas de ex谩menes y se quiere saber si la media de las notas en la poblaci贸n es igual a 70.

````python
import numpy as np
from scipy import stats

# Datos de muestra
notas = np.array([60, 70, 80, 90, 75, 85, 72, 78, 82, 68])

# Prueba de hip贸tesis
resultado = stats.ttest_1samp(notas, 70)

# Imprimir resultado
if resultado.pvalue < 0.05:
    print('La media de notas es significativamente diferente de 70')
else:
    print('No se puede rechazar la hip贸tesis de que la media de notas es igual a 70')
````

En este ejemplo, la hip贸tesis nula es que la media de notas es igual a 70, y la alternativa es que es diferente de 70. La prueba de hip贸tesis devuelve un valor de pvalue que indica la probabilidad de obtener una media muestral igual o m谩s extrema que la observada, si la hip贸tesis nula es cierta. Si este valor es menor que el nivel de significancia elegido (en este caso, 0.05), se rechaza la hip贸tesis nula y se concluye que la media de notas es significativamente diferente de 70.

En cuanto a la representaci贸n gr谩fica de la estad铆stica inferencial, una forma com煤n de hacerlo es mediante intervalos de confianza. Estos intervalos indican el rango de valores en el que se espera que se encuentre el verdadero valor del par谩metro con un cierto nivel de confianza.

Por ejemplo, si se quiere estimar la media de notas en la poblaci贸n, se puede calcular un intervalo de confianza del 95% para esta media a partir de la muestra de notas. En Python, esto se puede hacer de la siguiente manera:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'notas'

# Personalizar la paleta de colores
mi_paleta = sns.color_palette("pastel")

# Crear el gr谩fico de barras con intervalo de confianza del 95%
ax = sns.barplot(data=notas, errorbar=('ci', 95), palette=mi_paleta)

# Personalizar el t铆tulo y los ejes
ax.set_title('Notas con Intervalo de Confianza del 95%', fontsize=16)
ax.set_xlabel('Asignaturas', fontsize=14)
ax.set_ylabel('Media de Notas', fontsize=14)

# Rotar las etiquetas del eje x para mayor legibilidad
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize=12)

# A帽adir etiquetas en las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)

# Mostrar la leyenda (si es aplicable)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title='Categor铆as', title_fontsize='14', fontsize='12', loc='upper right')

# Ajustar el dise帽o para evitar cortar las etiquetas
plt.tight_layout()

# Mostrar el gr谩fico
plt.show()
````

## **2. Estad铆stica descriptiva e inferencial**

La estad铆stica es una disciplina que se divide en dos 谩reas principales: la estad铆stica descriptiva y la estad铆stica inferencial.

La estad铆stica descriptiva se enfoca en resumir y describir las caracter铆sticas de un conjunto de datos, mientras que la estad铆stica inferencial se enfoca en hacer inferencias sobre una poblaci贸n a partir de una muestra de datos.

A continuaci贸n, se presenta un ejemplo de c贸mo se pueden utilizar ambas 谩reas de la estad铆stica en Python para analizar un conjunto de datos.

Primero, se importan las bibliotecas necesarias:

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
````

Este conjunto de datos contiene informaci贸n sobre el monto de la propina que los clientes dejaron en un restaurante, junto con informaci贸n sobre la factura total, el g茅nero del servidor, el tama帽o del grupo, etc.

Se puede utilizar la estad铆stica descriptiva para obtener informaci贸n sobre las caracter铆sticas del conjunto de datos, como la media, la mediana, la desviaci贸n est谩ndar, etc. Por ejemplo, para obtener un resumen de las variables num茅ricas, se puede utilizar la funci贸n ```describe()```:

````python
df.describe()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' con una columna 'total_bill'

# Personalizar el color del histograma
color = 'skyblue'

# Crear el gr谩fico de histograma sin KDE
sns.histplot(df['total_bill'], kde=False, color=color, bins=20)  # Puedes ajustar el n煤mero de bins seg煤n tus preferencias

# Personalizar el t铆tulo y los ejes
plt.title('Distribuci贸n de Total Bill', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# A帽adir l铆neas verticales o cualquier otra personalizaci贸n adicional
# Por ejemplo, a帽adir una l铆nea vertical para la media
mean_value = df['total_bill'].mean()
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Media')

# Mostrar la leyenda
plt.legend()

# Ajustar el dise帽o
plt.tight_layout()

# Mostrar el gr谩fico
plt.show()
````

Para visualizar las relaciones entre variables num茅ricas, se pueden utilizar gr谩ficos de dispersi贸n:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' con columnas 'total_bill' y 'tip'

# Personalizar el color y el tama帽o de los puntos
color = 'skyblue'
size = 50

# Crear el gr谩fico de dispersi贸n
sns.scatterplot(x='total_bill', y='tip', data=df, color=color, s=size)

# Personalizar el t铆tulo y los ejes
plt.title('Relaci贸n entre Total Bill y Tip', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)

# A帽adir l铆neas, anotaciones u otras personalizaciones adicionales
# Por ejemplo, a帽adir una l铆nea de regresi贸n
sns.regplot(x='total_bill', y='tip', data=df, scatter=False, color='red')

# Ajustar el dise帽o
plt.tight_layout()

# Mostrar el gr谩fico
plt.show()
````

Para utilizar la estad铆stica inferencial, se puede realizar una prueba de hip贸tesis para determinar si hay una relaci贸n significativa entre dos variables num茅ricas. Por ejemplo, se puede realizar una prueba de correlaci贸n de Pearson entre el total de la factura y la propina:

````python
from scipy.stats import pearsonr

r, p = pearsonr(df['total_bill'], df['tip'])
print(f'Correlaci贸n de Pearson: {r:.3f}, valor p: {p:.3f}')
````

Esto devuelve un valor de correlaci贸n de Pearson de 0.676 y un valor p muy bajo, lo que sugiere que hay una relaci贸n significativa entre el total de la factura y la propina.

Para visualizar la matriz de correlaciones entre las variables num茅ricas, se puede utilizar un mapa de calor:

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos de iris
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                   header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Seleccionar solo las columnas num茅ricas (todas menos 'species')
iris_numeric = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Calcular la matriz de correlaciones de Pearson para las columnas num茅ricas
corr = iris_numeric.corr()

# Crear un heatmap para visualizar la matriz de correlaciones
plt.figure(figsize=(8, 6))  # Establecer el tama帽o del gr谩fico
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlaci贸n de Iris')
plt.show()
````

Esto muestra que hay una fuerte correlaci贸n positiva entre el total de la factura y la propina, as铆 como una correlaci贸n positiva moderada entre el tama帽o del grupo y el total de la factura y la propina.

En resumen, la estad铆stica descriptiva se puede utilizar para resumir y visualizar las caracter铆sticas de un conjunto de datos, mientras que la estad铆stica inferencial se puede utilizar para hacer inferencias sobre una poblaci贸n a partir de una muestra de datos.
