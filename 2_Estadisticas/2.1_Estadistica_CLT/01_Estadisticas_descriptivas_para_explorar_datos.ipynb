# **01 Estadísticas descriptivas para explorar datos**

Este módulo ofrece una visión general del curso y un repaso de las principales herramientas utilizadas en estadística descriptiva para visualizar la información.

## **1.1 Introducción**

La estadística descriptiva es fundamental para explorar, entender y describir las características principales de un conjunto de datos. Esta incluye la utilización de medidas de tendencia central (como la media, mediana y moda), medidas de dispersión (como el rango, varianza, desviación estándar) y la distribución de los datos (a través de histogramas, por ejemplo).

**Medidas de Tendencia Central**

Estas medidas nos dan una idea del valor central o típico de un conjunto de datos.

1. **Media**: El promedio de todos los valores.

2. **Mediana**: El valor medio cuando los datos están ordenados.

3. **Moda**: El valor que más se repite.

**Ejemplo**

Consideremos el siguiente conjunto de datos: 5, 3, 8, 9, 4, 2, 7.

- **Media**: $(5 + 3 + 8 + 9 + 4 + 2 + 7) / 7 = 5.43$

- **Mediana**: Ordenando los datos (2, 3, 4, 5, 7, 8, 9), la mediana es 5.

- **Moda**: No hay moda en este caso, ya que todos los valores son únicos.

**Medidas de Dispersión**

Estas medidas indican cuánto se dispersan los datos alrededor de un valor central.

1. **Rango**: La diferencia entre el valor máximo y mínimo.

2. **Varianza**: El promedio de las diferencias al cuadrado entre cada valor y la media.

3. **Desviación Estándar**: La raíz cuadrada de la varianza.

**Ejemplo**

Usando el mismo conjunto de datos: 5, 3, 8, 9, 4, 2, 7.

- **Rango**: 9 - 2 = 7

- **Varianza**: Vamos a calcularlo a continuación.

- **Desviación Estándar**: También la calcularemos a continuación.

**Histogramas**

Un histograma es una representación gráfica de la distribución de un conjunto de datos, que nos ayuda a visualizar cómo se distribuyen los valores.

**Ejemplo**

Vamos a generar un histograma para el conjunto de datos proporcionado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el conjunto de datos
datos = np.array([5, 3, 8, 9, 4, 2, 7])

# Calculamos las medidas de tendencia central
media = np.mean(datos)
mediana = np.median(datos)
modas = np.bincount(datos).argmax()  # En este caso puede no ser útil si todos los valores son únicos

# Calculamos las medidas de dispersión
rango = np.ptp(datos)
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Mostramos los resultados
media, mediana, modas, rango, varianza, desviacion_estandar
```

Los cálculos en Python para el conjunto de datos 5, 3, 8, 9, 4, 2, 7 nos dan los siguientes resultados:

- **Media**: 5.43

- **Mediana**: 5.0

- **Moda**: En este contexto, el cálculo directo de la moda no es útil debido a cómo se manejó con `np.bincount()`, que espera enteros no negativos y puede dar resultados engañosos en este contexto particular. En este conjunto, todos los valores son únicos, así que podemos decir que no hay moda.

- **Rango**: 7

- **Varianza**: 5.96

- **Desviación Estándar**: 2.44

Ahora, generaremos un histograma para visualizar la distribución de estos datos.

```python
# Generamos el histograma
plt.figure(figsize=(8, 6))
plt.hist(datos, bins='auto', color='skyblue', alpha=0.7, rwidth=0.85)
plt.title('Histograma de los Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)

# Mostramos el gráfico
plt.show()
```

Aquí tienes el histograma para el conjunto de datos proporcionado. Este gráfico muestra la distribución de los datos, donde el eje $X$ representa los valores de los datos y el eje $Y$ la frecuencia de estos valores. Como los datos son bastante variados y todos los valores son únicos (excepto si consideramos agrupaciones más amplias), el histograma nos ayuda a visualizar esta distribución uniforme de valores a lo largo del rango.

Con estos ejemplos y gráficos, hemos explorado las estadísticas descriptivas básicas que te permiten obtener una visión general y comprender mejor tus datos. Utilizar Python para estos cálculos y visualizaciones facilita mucho el análisis exploratorio, permitiéndote concentrarte en interpretar los resultados más que en los cálculos manuales.

## **1.2 Gráfico circular, gráfico de barras e histogramas**

Los gráficos son herramientas esenciales en la estadística para visualizar datos y facilitar su interpretación. Cada tipo de gráfico tiene usos específicos según la naturaleza de los datos y lo que se desea comunicar. Vamos a explorar el gráfico circular, el gráfico de barras y los histogramas, detallando sus usos, cómo se construyen y proporcionando ejemplos prácticos tanto aritméticamente como mediante código en Python.

**Gráfico Circular**

**¿Qué es y cuándo se utiliza?**
Un gráfico circular o pastel muestra la proporción de categorías como segmentos de un círculo, donde el todo representa el conjunto de datos. Es ideal para mostrar proporciones o porcentajes entre categorías cuando se tienen datos categóricos y se desea resaltar cómo cada categoría se compara con el total.

**Mejor uso**: Comparar partes de un todo en datos categóricos, especialmente cuando no hay muchas categorías, ya que muchas rebanadas pueden hacer que el gráfico sea difícil de leer.

**Fórmula para calcular el ángulo de cada segmento**:

$\text{Ángulo de cada segmento} = \frac{\text{Valor de la categoría}}{\text{Total de valores}} \times 360$

**Gráfico de Barras**

**¿Qué es y cuándo se utiliza?**
Un gráfico de barras representa datos con rectángulos verticales u horizontales (barras), donde la longitud de la barra es proporcional al valor que representa. Es útil para comparar diferentes grupos o para mostrar cambios en los datos a lo largo del tiempo, especialmente cuando los datos son discretos o categóricos.

**Mejor uso**: Comparación entre diferentes categorías o grupos, visualización de cambios o tendencias a lo largo del tiempo para categorías limitadas.

**Histogramas**

**¿Qué es y cuándo se utiliza?**
Un histograma es similar a un gráfico de barras, pero se utiliza para datos continuos agrupados en intervalos o "bins". Muestra la frecuencia de los datos dentro de cada intervalo, siendo una herramienta clave para entender la distribución de los datos.

**Mejor uso**: Visualizar la distribución de datos continuos, identificar modas, simetría/asimetría y detectar posibles outliers.

**Ejemplo Práctico**

Para proporcionar ejemplos prácticos, vamos a crear un DataFrame que contenga datos ficticios representativos para cada tipo de gráfico. Este DataFrame incluirá:

- Datos categóricos para los gráficos circular y de barras.

- Datos continuos para el histograma.

Vamos a proceder a crear este DataFrame y a generar los respectivos gráficos.

```python
import pandas as pd

# Creación de DataFrames separados
data_categorico = {
    'Categoria': ['A', 'B', 'C', 'D'],
    'Valor': [23, 17, 35, 25]
}
df_categorico = pd.DataFrame(data_categorico)

data_continuo = {
    'Datos Continuos': np.random.normal(50, 15, 100)  # 100 datos distribuidos normalmente
}
df_continuo = pd.DataFrame(data_continuo)

# Mostrar los DataFrames
df_categorico, df_continuo.head()
```

Para datos categóricos:

- Categoría: A, B, C, D

- Valor: 23, 17, 35, 25

Para datos continuos (se muestran los primeros 5 valores como ejemplo):

- Datos Continuos: 58.99, 79.76, 66.75, 72.81, 65.64, ...

Con estos DataFrames, procederemos a generar los gráficos solicitados:

1. **Gráfico Circular** para los datos categóricos, mostrando la proporción de cada categoría.

2. **Gráfico de Barras** también para los datos categóricos, comparando visualmente los valores de cada categoría.

3. **Histograma** para los datos continuos, para observar la distribución de estos datos.

Empezaremos generando el **Gráfico Circular** y el **Gráfico de Barras** para los datos categóricos.

```python
# Gráfico Circular
plt.figure(figsize=(7, 7))
plt.pie(df_categorico['Valor'], labels=df_categorico['Categoria'], autopct='%1.1f%%', startangle=140)
plt.title('Gráfico Circular de Categorías')

# Mostramos el gráfico circular
plt.show()

# Gráfico de Barras
plt.figure(figsize=(8, 6))
plt.bar(df_categorico['Categoria'], df_categorico['Valor'], color='skyblue')
plt.title('Gráfico de Barras de Categorías')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--')

# Mostramos el gráfico de barras
plt.show()
```

Aquí tienes dos representaciones visuales de los datos categóricos:

1. **Gráfico Circular**: Muestra la proporción de cada categoría en relación al total. Este gráfico es útil para entender rápidamente cómo se distribuyen los valores entre diferentes categorías.

2. **Gráfico de Barras**: Compara las cantidades de cada categoría. La altura de cada barra es proporcional al valor que representa, facilitando la comparación directa entre categorías.

Finalmente, vamos a generar un **Histograma** para visualizar la distribución de los datos continuos.

```python
# Histograma para los datos continuos
plt.figure(figsize=(10, 6))
plt.hist(df_continuo['Datos Continuos'], bins='auto', color='skyblue', edgecolor='black')
plt.title('Histograma de Datos Continuos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostramos el histograma
plt.show()
```

El **Histograma** muestra la distribución de los 100 datos continuos generados. Cada barra representa el número de datos que caen dentro de un rango específico de valores, lo que nos permite ver cómo se distribuyen los datos, identificar la centralidad, la dispersión, así como la simetría o asimetría de la distribución.

En resumen:

- **Gráfico Circular**: Útil para mostrar proporciones o porcentajes entre categorías. Ideal cuando los datos son categóricos y se quiere ver cómo cada parte se compara con el todo.
  
- **Gráfico de Barras**: Excelente para comparar datos entre diferentes categorías o ver cambios a lo largo del tiempo para un número limitado de grupos.
  
- **Histogramas**: Indispensables para visualizar la distribución de datos continuos, útil para entender la distribución general, identificar la forma de la distribución, y detectar posibles valores atípicos.

Cada uno de estos gráficos tiene su lugar dependiendo de lo que se desee comunicar y del tipo de datos con los que se esté trabajando.

## **1.3 Diagrama de caja y bigotes y diagrama de dispersión**

Los diagramas de caja y bigotes y los diagramas de dispersión son herramientas estadísticas poderosas para la visualización de datos, cada uno con aplicaciones específicas que nos permiten entender mejor las características y relaciones dentro de un conjunto de datos. A continuación, exploraremos ambos tipos de gráficos, sus usos, cómo se construyen y proporcionaremos ejemplos prácticos tanto aritméticamente como mediante código en Python.

**Diagrama de Caja y Bigotes (Box Plot)**

**¿Qué es y cuándo se utiliza?**

Un diagrama de caja y bigotes resume la distribución de un conjunto de datos cuantitativos mediante cuartiles, destacando la mediana, los cuartiles superior e inferior, y los valores atípicos. Es particularmente útil para visualizar la dispersión de los datos, la tendencia central y la presencia de valores atípicos.

**Mejor uso**: Comparar la distribución de varios conjuntos de datos, identificar valores atípicos y entender la dispersión de los datos.

**Componentes clave del diagrama**:

- **Mediana (Q2 / 50º percentil)**: El valor medio del conjunto de datos.

- **Cuartil inferior (Q1 / 25º percentil)**: El valor medio entre el menor número del conjunto de datos y su mediana.

- **Cuartil superior (Q3 / 75º percentil)**: El valor medio entre la mediana y el mayor número del conjunto de datos.

- **Rango intercuartílico (IQR)**: La diferencia entre el cuartil superior e inferior (Q3 - Q1).

- **Bigotes**: Extienden hasta el último dato dentro del rango `[Q1 - 1.5*IQR, Q3 + 1.5*IQR]`. Los datos fuera de este rango se consideran valores atípicos.

**Diagrama de Dispersión (Scatter Plot)**

**¿Qué es y cuándo se utiliza?**

Un diagrama de dispersión muestra la relación entre dos variables cuantitativas, representando cada punto de datos como un punto en el plano cartesiano. Es útil para identificar correlaciones, tendencias o patrones potenciales entre dos variables.

**Mejor uso**: Explorar relaciones entre dos variables, identificar correlaciones y detectar agrupaciones o patrones anómalos.

**Ejemplo Práctico**

Para ilustrar ambos tipos de diagramas, crearemos un DataFrame que contenga dos variables cuantitativas. Una variable se utilizará para el diagrama de caja y bigotes, representando la distribución de una única variable. La segunda y tercera variables se utilizarán para el diagrama de dispersión, explorando la relación entre ellas.

Crearemos este DataFrame y generaremos los respectivos gráficos.

```python
# Re-importando las librerías necesarias después del restablecimiento del entorno de ejecución
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creación del DataFrame para ejemplos
data_ejemplo = {
    'Variable A': np.random.normal(50, 10, 100),  # 100 datos distribuidos normalmente para el box plot
    'Variable B': np.random.uniform(20, 60, 100),  # 100 datos distribuidos uniformemente para el scatter plot
    'Variable C': np.random.uniform(10, 30, 100)   # Otra variable para el scatter plot
}
df_ejemplo = pd.DataFrame(data_ejemplo)

# Mostrar los primeros datos del DataFrame
df_ejemplo.head()
```

Tenemos un DataFrame con tres variables:

- **Variable A**: Datos distribuidos normalmente, que utilizaremos para el diagrama de caja y bigotes.

- **Variable B y Variable C**: Datos distribuidos uniformemente, que utilizaremos para crear el diagrama de dispersión.

Empezaremos generando el **Diagrama de Caja y Bigotes** para la Variable A, para visualizar su distribución.

```python
# Diagrama de Caja y Bigotes para la Variable A
plt.figure(figsize=(8, 6))
plt.boxplot(df_ejemplo['Variable A'])
plt.title('Diagrama de Caja y Bigotes para la Variable A')
plt.ylabel('Valores')

# Mostramos el diagrama
plt.show()
```

El **Diagrama de Caja y Bigotes** para la Variable A muestra la distribución de los datos. Los elementos clave que podemos observar incluyen la mediana (línea dentro de la caja), los cuartiles superior e inferior (los bordes de la caja), y los bigotes que se extienden desde la caja hasta los valores máximos y mínimos dentro del rango intercuartílico ajustado. Los puntos fuera de los bigotes representan valores atípicos, que son valores que difieren significativamente de la mayoría de los datos.

Ahora, generaremos un **Diagrama de Dispersión** utilizando las Variables B y C para explorar la relación entre estas dos variables.

```python
# Diagrama de Dispersión entre Variable B y Variable C
plt.figure(figsize=(10, 6))
plt.scatter(df_ejemplo['Variable B'], df_ejemplo['Variable C'], color='skyblue', edgecolor='black')
plt.title('Diagrama de Dispersión entre Variable B y Variable C')
plt.xlabel('Variable B')
plt.ylabel('Variable C')

# Mostramos el diagrama
plt.show()
```

El **Diagrama de Dispersión** entre la Variable B y la Variable C muestra cómo se relacionan estas dos variables entre sí. Cada punto en el gráfico representa una observación en el conjunto de datos, con su posición en el eje $X$ dada por el valor de la Variable B y su posición en el eje $Y$ por el valor de la Variable C. Este tipo de gráfico es útil para identificar cualquier relación o patrón entre las dos variables, como correlaciones positivas o negativas, o la ausencia de una relación clara.

A partir del gráfico, parece que no hay una relación lineal clara entre la Variable B y la Variable C, lo cual es esperado dada la forma aleatoria en que generamos los datos. Los diagramas de dispersión son especialmente útiles en análisis exploratorios de datos para identificar relaciones entre variables que podrían no ser evidentes a primera vista.

En resumen, el **diagrama de caja y bigotes** es excelente para visualizar la distribución de un conjunto de datos, destacando la mediana, los cuartiles y los valores atípicos. Por otro lado, el **diagrama de dispersión** es fundamental para explorar y visualizar las relaciones entre dos variables continuas, ayudando a identificar patrones, tendencias o correlaciones.

## **1.4 Proporcionar contexto es clave para los análisis estadísticos**

Proporcionar contexto es esencial en el análisis estadístico porque los números y datos brutos, por sí solos, a menudo no cuentan la historia completa o pueden ser mal interpretados. El contexto ayuda a entender mejor la importancia de los datos, las comparaciones y las conclusiones que se derivan de ellos. Veamos algunas maneras en que el contexto es clave, con ejemplos y gráficos para ilustrar estos puntos.

**Ejemplo 1: Tendencias a lo largo del tiempo**

**Contexto**: Cambios en las tendencias de ventas a lo largo del tiempo.

**Sin contexto**: Observamos un incremento en las ventas en marzo.

**Con contexto**: El incremento en las ventas en marzo se compara con los datos de los últimos 5 años, donde típicamente marzo es un mes lento. Este año, el incremento coincide con una campaña de marketing lanzada a finales de febrero.

**Visualización**:
- Podríamos usar un gráfico de líneas para mostrar las ventas mensuales a lo largo de los últimos cinco años, destacando el pico inusual de marzo.

**Ejemplo 2: Comparación entre grupos**

**Contexto**: Comparación de la efectividad de dos medicamentos.

**Sin contexto**: El Medicamento A tuvo un 70% de efectividad, y el Medicamento B, un 75%.

**Con contexto**: El Medicamento A se administró a una población con casos más severos de la enfermedad. Además, el Medicamento B tiene efectos secundarios significativos no presentes en el Medicamento A.

**Visualización**:
- Un diagrama de caja y bigotes podría mostrar la distribución de la efectividad de ambos medicamentos, con notas sobre la severidad de la enfermedad y los efectos secundarios.

**Ejemplo 3: Densidad poblacional y área geográfica**

**Contexto**: Comparar la densidad poblacional de diferentes países.

**Sin contexto**: País A tiene una densidad poblacional mayor que País B.

**Con contexto**: País A es geográficamente mucho más pequeño que País B y tiene una concentración urbana más alta, lo que contribuye a su mayor densidad poblacional.

**Visualización**:
- Un mapa de calor o un gráfico de barras que muestre la densidad poblacional junto con el tamaño geográfico de los países puede proporcionar una imagen más completa.

Ahora, generaremos gráficos para ilustrar cómo el contexto puede cambiar la interpretación de los datos. Empezaremos con el **Ejemplo 1** sobre las tendencias de ventas a lo largo del tiempo, creando datos ficticios y un gráfico de líneas.

```python
# Generación de datos ficticios para el Ejemplo 1: Tendencias de ventas a lo largo del tiempo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas_2023 = [200, 180, 250, 220, 210]  # Ventas para el año 2023
ventas_promedio_5_años = [190, 175, 180, 200, 195]  # Promedio de ventas de los últimos 5 años

# Creación del gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(meses, ventas_2023, label='Ventas 2023', marker='o')
plt.plot(meses, ventas_promedio_5_años, label='Promedio Ventas Últimos 5 Años', linestyle='--', marker='x')
plt.title('Tendencia de Ventas Marzo 2023 vs. Promedio Últimos 5 Años')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

El gráfico de líneas muestra la tendencia de ventas para marzo de 2023 en comparación con el promedio de ventas de los últimos 5 años. Aquí, el contexto proporcionado por el promedio de ventas de años anteriores nos ayuda a entender que el pico en marzo de 2023 es inusual y significativo, especialmente si consideramos la información adicional de que marzo típicamente es un mes lento y que hubo una campaña de marketing reciente. Este tipo de análisis contextual puede ayudar a las empresas a evaluar el impacto de sus estrategias y a planificar futuras acciones.

Este ejemplo ilustra cómo el contexto transforma la interpretación de los datos brutos, permitiendo una comprensión más profunda y matizada de las tendencias y eventos subyacentes. Al proporcionar contexto, no solo presentamos datos, sino que también ofrecemos insights y valor agregado, facilitando la toma de decisiones informada basada en una comprensión completa de las circunstancias.

## **1.5 Errores al visualizar la información**

Visualizar la información de manera incorrecta o engañosa puede llevar a interpretaciones erróneas, decisiones mal informadas y, en el peor de los casos, a conclusiones completamente falsas. Algunos errores comunes incluyen el uso inapropiado de tipos de gráficos, la manipulación del eje de las coordenadas, la omisión de datos relevantes y la elección de escalas inadecuadas. Vamos a explorar algunos de estos errores con ejemplos y gráficos.

**Ejemplo 1: Manipulación del Eje $Y$**

**Problema**: Comenzar el eje $Y$ en un valor distinto de cero en un gráfico de barras o de líneas puede exagerar o minimizar visualmente las diferencias entre las barras o los puntos de datos.

**Consecuencia**: Se pueden malinterpretar las magnitudes de las diferencias o cambios.

**Solución**: Siempre iniciar el eje $Y$ en cero para representaciones precisas, a menos que haya una buena razón para hacer lo contrario y se aclare al espectador.

**Ejemplo 2: Uso Inadecuado de Gráficos Circulares**

**Problema**: Usar gráficos circulares cuando hay muchas categorías o las diferencias entre las partes son mínimas.

**Consecuencia**: Hace difícil la comparación directa entre segmentos, llevando a interpretaciones erróneas.

**Solución**: Optar por gráficos de barras o gráficos de líneas, que permiten comparaciones más claras.

**Ejemplo 3: Escalas no Lineales sin Indicación**

**Problema**: Utilizar escalas logarítmicas o no lineales sin indicarlo claramente.

**Consecuencia**: Puede confundir a los espectadores, haciendo que interpreten incorrectamente las relaciones entre los datos.

**Solución**: Marcar claramente cuando se usen escalas no estándar y proporcionar una breve explicación si es posible.

**Visualizaciones**

A continuación, generaremos visualizaciones para ilustrar el **Ejemplo 1 (Manipulación del Eje Y)** y el **Ejemplo 2 (Uso Inadecuado de Gráficos Circulares)**, utilizando datos ficticios para mostrar cómo estos errores pueden afectar la interpretación de la información.

```python
# Ejemplo 1: Manipulación del Eje Y
ventas = [100, 105, 103, 110, 115]  # Datos de ventas ficticios
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May']

# Gráfico Correcto
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(meses, ventas, color='skyblue')
plt.title('Eje Y Comenzando en 0')
plt.ylim(0)  # Asegurando que el eje Y comience en 0

# Gráfico Incorrecto
plt.subplot(1, 2, 2)
plt.bar(meses, ventas, color='lightcoral')
plt.title('Eje Y No Comenzando en 0')
plt.ylim(100)  # Comenzando el eje Y en 100 para exagerar diferencias

plt.tight_layout()
plt.show()

# Ejemplo 2: Uso Inadecuado de Gráficos Circulares
porcentajes = [30, 25, 20, 15, 10]  # Datos de porcentajes ficticios
categorias = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E']

plt.figure(figsize=(5, 5))
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Gráfico Circular con Muchas Categorías')
plt.show()
```

En las visualizaciones presentadas, podemos ver claramente cómo los errores en la visualización de la información pueden llevar a interpretaciones incorrectas:

**Ejemplo 1: Manipulación del Eje Y**

- **Gráfico Correcto (Eje Y Comenzando en 0)**: Muestra una representación precisa de las ventas a lo largo de los meses, donde las diferencias entre las barras son fáciles de comparar y entender.
- **Gráfico Incorrecto (Eje Y No Comenzando en 0)**: Exagera las diferencias entre las ventas mensuales. Aunque las variaciones reales son menores, el gráfico sugiere cambios dramáticos, lo cual puede llevar a interpretaciones erróneas sobre la performance.

**Ejemplo 2: Uso Inadecuado de Gráficos Circulares**

- **Gráfico Circular con Muchas Categorías**: Aunque los gráficos circulares son útiles para mostrar proporciones, en este caso, la presencia de muchas categorías y la proximidad de sus valores hacen que sea difícil discernir diferencias exactas y comparar eficazmente las secciones.

Estos ejemplos destacan la importancia de elegir el tipo correcto de gráfico y configurar los ejes $y$ escalas de manera adecuada para evitar interpretaciones incorrectas. Al visualizar datos, es crucial ser preciso y claro, proporcionando el contexto adecuado y evitando manipulaciones que puedan llevar a conclusiones erróneas. La integridad en la visualización de datos no solo mejora la comprensión sino que también fomenta la confianza en el análisis presentado.

## **1.6 Media y mediana**

**Datos No Agrupados**

- **Media (Promedio)**: Es la suma de todos los valores en el conjunto de datos dividida entre el número total de valores.

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

donde $(x_i)$ son los valores individuales y $(n)$ es el número total de valores.

- **Mediana**: Es el valor que se encuentra en el medio del conjunto de datos una vez que ha sido ordenado. Si hay un número impar de observaciones, es el valor medio. Si hay un número par, es el promedio de los dos valores medios.

  - Para un número impar de observaciones:

  $\text{Mediana} = x_{(\frac{n+1}{2})}$

  - Para un número par de observaciones:

  $\text{Mediana} = \frac{x_{(\frac{n}{2})} + x_{(\frac{n}{2} + 1)}}{2}$

**Datos Agrupados**

Los datos agrupados están organizados en clases o intervalos.

- **Media de Datos Agrupados**:

$\text{Media} = \frac{\sum_{i=1}^{k} f_i \cdot x_i}{\sum_{i=1}^{k} f_i}$

donde $(f_i)$ es la frecuencia de cada intervalo, $(x_i)$ es el punto medio del intervalo, y $(k)$ es el número total de intervalos.

- **Mediana de Datos Agrupados**:

 Se encuentra primero determinando el intervalo mediano (el intervalo que contiene a la mediana). La fórmula para calcular la mediana en datos agrupados es:

$\text{Mediana} = L + \left(\frac{\frac{N}{2} - F}{f}\right)W$

donde $(L)$ es el límite inferior del intervalo mediano, $(N)$ es el número total de datos, $(F)$ es la frecuencia acumulada antes del intervalo mediano, $(f)$ es la frecuencia del intervalo mediano, y $(W)$ es la amplitud del intervalo.

Ahora, procederemos a crear un DataFrame para datos no agrupados, calcular y visualizar la media y la mediana. Luego, agruparemos esos datos y haremos lo mismo para los datos agrupados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generación de datos no agrupados
np.random.seed(0)  # Para reproducibilidad
datos_no_agrupados = np.random.normal(50, 15, 100)  # 100 datos distribuidos normalmente

# Creación de un DataFrame para datos no agrupados
df_no_agrupados = pd.DataFrame(datos_no_agrupados, columns=['Valores'])

# Cálculo de la media y la mediana para datos no agrupados
media_no_agrupados = np.mean(datos_no_agrupados)
mediana_no_agrupados = np.median(datos_no_agrupados)

media_no_agrupados, mediana_no_agrupados
```

Para nuestro conjunto de **datos no agrupados**, la media es aproximadamente 50.90 y la mediana es aproximadamente 51.41, lo que indica una distribución relativamente simétrica alrededor del centro.

A continuación, agruparemos estos datos y realizaremos los cálculos para datos agrupados. Para ello, dividiremos los datos en intervalos y calcularemos la media y la mediana utilizando las fórmulas apropiadas para datos agrupados. Además, visualizaremos los datos utilizando histogramas y gráficos para cada caso.

```python
# Agrupación de los datos en intervalos
bins = np.arange(0, 101, 10)
df_no_agrupados['Intervalo'] = pd.cut(df_no_agrupados['Valores'], bins=bins, include_lowest=True)

# Datos agrupados: calculando la frecuencia y el punto medio de cada intervalo
datos_agrupados = df_no_agrupados.groupby('Intervalo')['Valores'].agg(['count', 'mean']).reset_index()
datos_agrupados.columns = ['Intervalo', 'Frecuencia', 'Punto Medio']

# Cálculo de la media para datos agrupados
media_agrupados = np.sum(datos_agrupados['Punto Medio'] * datos_agrupados['Frecuencia']) / np.sum(datos_agrupados['Frecuencia'])

# Preparación para calcular la mediana de datos agrupados
N = np.sum(datos_agrupados['Frecuencia'])
datos_agrupados['Frecuencia Acumulada'] = datos_agrupados['Frecuencia'].cumsum()
intervalo_mediana = datos_agrupados[datos_agrupados['Frecuencia Acumulada'] >= N/2].iloc[0]
L = intervalo_mediana['Intervalo'].left
F = datos_agrupados[datos_agrupados['Intervalo'] == intervalo_mediana['Intervalo']]['Frecuencia Acumulada'].shift(1).fillna(0).iloc[0]
f = intervalo_mediana['Frecuencia']
W = intervalo_mediana['Intervalo'].right - intervalo_mediana['Intervalo'].left

# Cálculo de la mediana para datos agrupados
mediana_agrupados = L + ((N/2 - F) / f) * W

media_agrupados, mediana_agrupados
```

Para los **datos agrupados**, la media calculada es aproximadamente 50.90, similar a la obtenida para los datos no agrupados, lo que refleja la consistencia de la media como medida de tendencia central. Sin embargo, el cálculo de la mediana nos da aproximadamente 67.86, lo cual difiere del cálculo directo de los datos no agrupados. Esta diferencia se debe a la naturaleza de los datos agrupados y la estimación involucrada en el cálculo de la mediana a partir de intervalos.

Ahora, visualizaremos ambos conjuntos de datos para comparar cómo se ven los datos no agrupados y agrupados.

**Visualización de Datos No Agrupados: Histograma**

```python
plt.figure(figsize=(10, 6))
plt.hist(df_no_agrupados['Valores'], bins=bins, edgecolor='black', alpha=0.7)
plt.axvline(media_no_agrupados, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media_no_agrupados:.2f}')
plt.axvline(mediana_no_agrupados, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana: {mediana_no_agrupados:.2f}')
plt.title('Histograma de Datos No Agrupados')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

El histograma muestra la distribución de los **datos no agrupados** con la media y la mediana indicadas. La cercanía de la media y la mediana sugiere una distribución simétrica de los datos alrededor del centro.

**Visualización de Datos Agrupados: Histograma**

Para visualizar los datos agrupados, usaremos el histograma de los datos originales pero con las marcas de la media y la mediana calculadas para los datos agrupados, para comparar.

```python
plt.figure(figsize=(10, 6))
plt.hist(df_no_agrupados['Valores'], bins=bins, edgecolor='black', alpha=0.7)
plt.axvline(media_agrupados, color='red', linestyle='dashed', linewidth=1, label=f'Media Agrupada: {media_agrupados:.2f}')
plt.axvline(mediana_agrupados, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana Agrupada: {mediana_agrupados:.2f}')
plt.title('Histograma de Datos Agrupados (con Media y Mediana Agrupadas)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

En el histograma de datos agrupados, se representan las mismas observaciones pero con las líneas que indican la media y la mediana calculadas a partir de los datos agrupados. Aquí, la línea de la mediana parece estar más alejada de la media en comparación con el histograma de datos no agrupados, reflejando la diferencia en el cálculo de la mediana para datos agrupados.

## **1.7 Percentiles, resumen de cinco números y desviación típica**

**Percentiles**

Los **percentiles** son medidas que dividen un conjunto de datos ordenado en cien partes iguales. Son útiles para entender la distribución de los datos. Por ejemplo, el 25º percentil (también conocido como primer cuartil) divide el 25% de los datos más bajos del 75% más alto, mientras que el 50º percentil (la mediana) divide el conjunto de datos en dos mitades.

Fórmula general para calcular el percentil $(P_k)$ en un conjunto de datos ordenado:

$P_k = \left( \frac{k}{100} \right) \times (N + 1)$

donde:

- $(P_k)$ es el k-ésimo percentil,
- $(k)$ es el percentil deseado,
- $(N)$ es el número total de observaciones en el conjunto de datos.

**Resumen de Cinco Números**

El **resumen de cinco números** consiste en cinco valores que proporcionan una visión completa de la distribución de un conjunto de datos:

1. **Mínimo**: El valor más bajo.

2. **Primer cuartil (Q1 / 25º percentil)**: Divide el 25% de los datos más bajos del resto.

3. **Mediana (Q2 / 50º percentil)**: El valor medio.

4. **Tercer cuartil (Q3 / 75º percentil)**: Divide el 75% de los datos más bajos del 25% más alto.

5. **Máximo**: El valor más alto.

**Desviación Estándar**

La **desviación estándar** es una medida que indica cuánto se desvían los valores en un conjunto de datos respecto a la media. Muestra la dispersión o variabilidad de los datos.

Fórmula para la desviación estándar $(s)$ de una muestra:

$s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N - 1}}$

donde:

- $(x_i)$ son los valores individuales,
- $(\bar{x})$ es la media de los datos,
- $(N)$ es el número total de observaciones.

Para ilustrar estos conceptos, vamos a generar un DataFrame con datos aleatorios y calcular el resumen de cinco números, varios percentiles específicos y la desviación estándar. Además, visualizaremos estos conceptos con gráficos apropiados.

```python
# Generación de datos aleatorios para el ejemplo
np.random.seed(42)  # Para reproducibilidad
datos_ejemplo = np.random.normal(loc=50, scale=10, size=100)  # 100 datos con media 50 y desv. estándar 10

# Creando un DataFrame para los datos
df_ejemplo = pd.DataFrame(datos_ejemplo, columns=['Datos'])

# Cálculo del resumen de cinco números y percentiles específicos
resumen_cinco_numeros = df_ejemplo['Datos'].describe()[['min', '25%', '50%', '75%', 'max']]
percentil_10 = np.percentile(datos_ejemplo, 10)  # 10º percentil
percentil_90 = np.percentile(datos_ejemplo, 90)  # 90º percentil

# Cálculo de la desviación estándar
desviacion_estandar = df_ejemplo['Datos'].std()

resumen_cinco_numeros, percentil_10, percentil_90, desviacion_estandar
```

Para nuestro conjunto de datos generado, tenemos el siguiente análisis:

- **Resumen de Cinco Números**:

  - Mínimo: ~23.80

  - Primer Cuartil (Q1 / 25%): ~43.99

  - Mediana (Q2 / 50%): ~48.73

  - Tercer Cuartil (Q3 / 75%): ~54.06

  - Máximo: ~68.52

- **Percentiles Específicos**:

  - 10º Percentil: ~36.63
  
  - 90º Percentil: ~60.06

- **Desviación Estándar**: ~9.08

Estos resultados nos dan una visión detallada de la distribución de los datos: el resumen de cinco números nos muestra la distribución de los datos desde el valor más bajo hasta el más alto, pasando por los cuartiles que dividen el conjunto en cuatro partes iguales. Los percentiles específicos ofrecen puntos de referencia adicionales dentro de la distribución, y la desviación estándar nos dice cuánto se desvían los valores de la media en promedio.

Para visualizar estos conceptos, generaremos un **diagrama de caja (box plot)**, que ilustra el resumen de cinco números y los percentiles, y un **histograma**, que nos ayudará a ver la dispersión y forma de la distribución de los datos.

```python
# Diagrama de caja (box plot) para visualizar el resumen de cinco números
plt.figure(figsize=(10, 6))
plt.boxplot(df_ejemplo['Datos'], vert=False, patch_artist=True)
plt.title('Diagrama de Caja')
plt.xlabel('Datos')

# Mostrando el diagrama de caja
plt.show()

# Histograma para visualizar la distribución de los datos
plt.figure(figsize=(10, 6))
plt.hist(df_ejemplo['Datos'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de los Datos')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')

# Mostrando el histograma
plt.show()
```

En las visualizaciones proporcionadas, podemos observar lo siguiente:

- **Diagrama de Caja**: Ilustra claramente el resumen de cinco números de nuestro conjunto de datos. Podemos ver el rango de los datos a través del mínimo y máximo, la distribución interna a través de los cuartiles Q1, Q2 (mediana), y Q3, y cualquier dato atípico que se desvíe significativamente de los cuartiles.

- **Histograma**: Muestra la distribución y dispersión de los datos, con la desviación estándar reflejada en la amplitud de la distribución. La forma del histograma también nos da una idea de la simetría de los datos respecto a la media.

Estas herramientas estadísticas y gráficas proporcionan una comprensión profunda y detallada de la distribución de un conjunto de datos, permitiendo análisis más informados y decisiones basadas en datos.

## **1.8 Introducción y estadísticas descriptivas para explorar los datos**

La **introducción a las estadísticas descriptivas** para explorar datos es un paso crucial en el análisis de datos, permitiendo resumir y comprender las características principales de un conjunto de datos antes de proceder a análisis más complejos. Las estadísticas descriptivas proporcionan información sobre la tendencia central, la dispersión y la forma de la distribución de los datos.

**Componentes Clave de las Estadísticas Descriptivas**

**Medidas de Tendencia Central**:

   - **Media** (Promedio): Suma de todos los valores dividida por el número de valores.

   - **Mediana**: El valor central en un conjunto de datos ordenados.

   - **Moda**: El valor que aparece con mayor frecuencia en un conjunto de datos.

**Medidas de Dispersión**:

   - **Varianza**: El promedio de los cuadrados de las desviaciones de la media.

   - **Desviación Estándar**: La raíz cuadrada de la varianza, mide la dispersión de los datos respecto a la media.

   - **Rango**: Diferencia entre el máximo y el mínimo valor en el conjunto de datos.

   - **Cuartiles**: Valores que dividen los datos en cuatro partes iguales.

**Forma de la Distribución**:

   - **Sesgo (Skewness)**: Medida de la asimetría de la distribución de los datos.

   - **Curtosis**: Medida del pico de la distribución de los datos.

**Ejemplo Práctico: Análisis de Datos de Altura**

Supongamos que tenemos un conjunto de datos que representa las alturas (en cm) de un grupo de estudiantes:

$\text{Alturas} = [160, 165, 167, 165, 170, 172, 168, 162, 174, 168, 165, 173]$

**Cálculo de Estadísticas Descriptivas**

Calcularemos las medidas de tendencia central, dispersión y exploraremos la forma de la distribución de estos datos.

**Implementación en Python**

Vamos a utilizar Python para calcular estas estadísticas y visualizar los datos:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de alturas
alturas = np.array([160, 165, 167, 165, 170, 172, 168, 162, 174, 168, 165, 173])

# Cálculos de estadísticas descriptivas
media = np.mean(alturas)
mediana = np.median(alturas)
moda = np.bincount(alturas).argmax()  # Simplificación para demostración
varianza = np.var(alturas)
desviacion_std = np.std(alturas)
rango = np.ptp(alturas)  # Peak to peak

# Visualización de los datos
plt.figure(figsize=(10, 6))
sns.histplot(alturas, kde=True, color='blue', bins=5)  # Histograma con estimación de densidad
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.title('Distribución de Alturas de Estudiantes')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda} (Más común)")
print(f"Varianza: {varianza:.2f}")
print(f"Desviación Estándar: {desviacion_std:.2f}")
print(f"Rango: {rango}")
```

**Interpretación**

- **Histograma**: Muestra la distribución de las alturas. La línea roja representa la media y la verde la mediana, proporcionando una visualización directa de la tendencia central.

- **Media y Mediana**: Valores centrales que resumen la tendencia de las alturas. La cercanía de la media y la mediana sugiere una distribución relativamente simétrica.

- **Moda**: La altura más frecuente.

- **Varianza y Desviación Estándar**: Indican que las alturas están dispersas alrededor de la media por aproximadamente 4.33 cm en promedio.

Este análisis de estadísticas descriptivas provee una comprensión

 fundamental de los datos antes de proceder con análisis más avanzados, como correlación o modelos predictivos.
