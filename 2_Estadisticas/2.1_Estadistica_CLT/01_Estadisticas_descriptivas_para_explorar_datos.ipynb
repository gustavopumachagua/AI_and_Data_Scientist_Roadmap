# **01 Estad√≠sticas descriptivas para explorar datos**

Este m√≥dulo ofrece una visi√≥n general del curso y un repaso de las principales herramientas utilizadas en estad√≠stica descriptiva para visualizar la informaci√≥n.

## **1.1 Introducci√≥n**

La estad√≠stica descriptiva es fundamental para explorar, entender y describir las caracter√≠sticas principales de un conjunto de datos. Esta incluye la utilizaci√≥n de medidas de tendencia central (como la media, mediana y moda), medidas de dispersi√≥n (como el rango, varianza, desviaci√≥n est√°ndar) y la distribuci√≥n de los datos (a trav√©s de histogramas, por ejemplo).

**Medidas de Tendencia Central**

Estas medidas nos dan una idea del valor central o t√≠pico de un conjunto de datos.

1. **Media**: El promedio de todos los valores.

2. **Mediana**: El valor medio cuando los datos est√°n ordenados.

3. **Moda**: El valor que m√°s se repite.

**Ejemplo**

Consideremos el siguiente conjunto de datos: 5, 3, 8, 9, 4, 2, 7.

- **Media**: $(5 + 3 + 8 + 9 + 4 + 2 + 7) / 7 = 5.43$

- **Mediana**: Ordenando los datos (2, 3, 4, 5, 7, 8, 9), la mediana es 5.

- **Moda**: No hay moda en este caso, ya que todos los valores son √∫nicos.

**Medidas de Dispersi√≥n**

Estas medidas indican cu√°nto se dispersan los datos alrededor de un valor central.

1. **Rango**: La diferencia entre el valor m√°ximo y m√≠nimo.

2. **Varianza**: El promedio de las diferencias al cuadrado entre cada valor y la media.

3. **Desviaci√≥n Est√°ndar**: La ra√≠z cuadrada de la varianza.

**Ejemplo**

Usando el mismo conjunto de datos: 5, 3, 8, 9, 4, 2, 7.

- **Rango**: 9 - 2 = 7

- **Varianza**: Vamos a calcularlo a continuaci√≥n.

- **Desviaci√≥n Est√°ndar**: Tambi√©n la calcularemos a continuaci√≥n.

**Histogramas**

Un histograma es una representaci√≥n gr√°fica de la distribuci√≥n de un conjunto de datos, que nos ayuda a visualizar c√≥mo se distribuyen los valores.

**Ejemplo**

Vamos a generar un histograma para el conjunto de datos proporcionado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el conjunto de datos
datos = np.array([5, 3, 8, 9, 4, 2, 7])

# Calculamos las medidas de tendencia central
media = np.mean(datos)
mediana = np.median(datos)
modas = np.bincount(datos).argmax()  # En este caso puede no ser √∫til si todos los valores son √∫nicos

# Calculamos las medidas de dispersi√≥n
rango = np.ptp(datos)
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Mostramos los resultados
media, mediana, modas, rango, varianza, desviacion_estandar
```

Los c√°lculos en Python para el conjunto de datos 5, 3, 8, 9, 4, 2, 7 nos dan los siguientes resultados:

- **Media**: 5.43

- **Mediana**: 5.0

- **Moda**: En este contexto, el c√°lculo directo de la moda no es √∫til debido a c√≥mo se manej√≥ con `np.bincount()`, que espera enteros no negativos y puede dar resultados enga√±osos en este contexto particular. En este conjunto, todos los valores son √∫nicos, as√≠ que podemos decir que no hay moda.

- **Rango**: 7

- **Varianza**: 5.96

- **Desviaci√≥n Est√°ndar**: 2.44

Ahora, generaremos un histograma para visualizar la distribuci√≥n de estos datos.

```python
# Generamos el histograma
plt.figure(figsize=(8, 6))
plt.hist(datos, bins='auto', color='skyblue', alpha=0.7, rwidth=0.85)
plt.title('Histograma de los Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)

# Mostramos el gr√°fico
plt.show()
```

Aqu√≠ tienes el histograma para el conjunto de datos proporcionado. Este gr√°fico muestra la distribuci√≥n de los datos, donde el eje $X$ representa los valores de los datos y el eje $Y$ la frecuencia de estos valores. Como los datos son bastante variados y todos los valores son √∫nicos (excepto si consideramos agrupaciones m√°s amplias), el histograma nos ayuda a visualizar esta distribuci√≥n uniforme de valores a lo largo del rango.

Con estos ejemplos y gr√°ficos, hemos explorado las estad√≠sticas descriptivas b√°sicas que te permiten obtener una visi√≥n general y comprender mejor tus datos. Utilizar Python para estos c√°lculos y visualizaciones facilita mucho el an√°lisis exploratorio, permiti√©ndote concentrarte en interpretar los resultados m√°s que en los c√°lculos manuales.

## **1.2 Gr√°fico circular, gr√°fico de barras e histogramas**

Los gr√°ficos son herramientas esenciales en la estad√≠stica para visualizar datos y facilitar su interpretaci√≥n. Cada tipo de gr√°fico tiene usos espec√≠ficos seg√∫n la naturaleza de los datos y lo que se desea comunicar. Vamos a explorar el gr√°fico circular, el gr√°fico de barras y los histogramas, detallando sus usos, c√≥mo se construyen y proporcionando ejemplos pr√°cticos tanto aritm√©ticamente como mediante c√≥digo en Python.

**Gr√°fico Circular**

**¬øQu√© es y cu√°ndo se utiliza?**
Un gr√°fico circular o pastel muestra la proporci√≥n de categor√≠as como segmentos de un c√≠rculo, donde el todo representa el conjunto de datos. Es ideal para mostrar proporciones o porcentajes entre categor√≠as cuando se tienen datos categ√≥ricos y se desea resaltar c√≥mo cada categor√≠a se compara con el total.

**Mejor uso**: Comparar partes de un todo en datos categ√≥ricos, especialmente cuando no hay muchas categor√≠as, ya que muchas rebanadas pueden hacer que el gr√°fico sea dif√≠cil de leer.

**F√≥rmula para calcular el √°ngulo de cada segmento**:

$\text{√Ångulo de cada segmento} = \frac{\text{Valor de la categor√≠a}}{\text{Total de valores}} \times 360$

**Gr√°fico de Barras**

**¬øQu√© es y cu√°ndo se utiliza?**
Un gr√°fico de barras representa datos con rect√°ngulos verticales u horizontales (barras), donde la longitud de la barra es proporcional al valor que representa. Es √∫til para comparar diferentes grupos o para mostrar cambios en los datos a lo largo del tiempo, especialmente cuando los datos son discretos o categ√≥ricos.

**Mejor uso**: Comparaci√≥n entre diferentes categor√≠as o grupos, visualizaci√≥n de cambios o tendencias a lo largo del tiempo para categor√≠as limitadas.

**Histogramas**

**¬øQu√© es y cu√°ndo se utiliza?**
Un histograma es similar a un gr√°fico de barras, pero se utiliza para datos continuos agrupados en intervalos o "bins". Muestra la frecuencia de los datos dentro de cada intervalo, siendo una herramienta clave para entender la distribuci√≥n de los datos.

**Mejor uso**: Visualizar la distribuci√≥n de datos continuos, identificar modas, simetr√≠a/asimetr√≠a y detectar posibles outliers.

**Ejemplo Pr√°ctico**

Para proporcionar ejemplos pr√°cticos, vamos a crear un DataFrame que contenga datos ficticios representativos para cada tipo de gr√°fico. Este DataFrame incluir√°:

- Datos categ√≥ricos para los gr√°ficos circular y de barras.

- Datos continuos para el histograma.

Vamos a proceder a crear este DataFrame y a generar los respectivos gr√°ficos.

```python
import pandas as pd

# Creaci√≥n de DataFrames separados
data_categorico = {
    'Categoria': ['A', 'B', 'C', 'D'],
    'Valor': [23, 17, 35, 25]
}
df_categorico = pd.DataFrame(data_categorico)

data_continuo = {
    'Datos Continuos': np.random.normal(50, 15, 100)  # 100 datos distribuidos normalmente
}
df_continuo = pd.DataFrame(data_continuo)

# Mostrar los DataFrames
df_categorico, df_continuo.head()
```

Para datos categ√≥ricos:

- Categor√≠a: A, B, C, D

- Valor: 23, 17, 35, 25

Para datos continuos (se muestran los primeros 5 valores como ejemplo):

- Datos Continuos: 58.99, 79.76, 66.75, 72.81, 65.64, ...

Con estos DataFrames, procederemos a generar los gr√°ficos solicitados:

1. **Gr√°fico Circular** para los datos categ√≥ricos, mostrando la proporci√≥n de cada categor√≠a.

2. **Gr√°fico de Barras** tambi√©n para los datos categ√≥ricos, comparando visualmente los valores de cada categor√≠a.

3. **Histograma** para los datos continuos, para observar la distribuci√≥n de estos datos.

Empezaremos generando el **Gr√°fico Circular** y el **Gr√°fico de Barras** para los datos categ√≥ricos.

```python
# Gr√°fico Circular
plt.figure(figsize=(7, 7))
plt.pie(df_categorico['Valor'], labels=df_categorico['Categoria'], autopct='%1.1f%%', startangle=140)
plt.title('Gr√°fico Circular de Categor√≠as')

# Mostramos el gr√°fico circular
plt.show()

# Gr√°fico de Barras
plt.figure(figsize=(8, 6))
plt.bar(df_categorico['Categoria'], df_categorico['Valor'], color='skyblue')
plt.title('Gr√°fico de Barras de Categor√≠as')
plt.xlabel('Categor√≠a')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--')

# Mostramos el gr√°fico de barras
plt.show()
```

Aqu√≠ tienes dos representaciones visuales de los datos categ√≥ricos:

1. **Gr√°fico Circular**: Muestra la proporci√≥n de cada categor√≠a en relaci√≥n al total. Este gr√°fico es √∫til para entender r√°pidamente c√≥mo se distribuyen los valores entre diferentes categor√≠as.

2. **Gr√°fico de Barras**: Compara las cantidades de cada categor√≠a. La altura de cada barra es proporcional al valor que representa, facilitando la comparaci√≥n directa entre categor√≠as.

Finalmente, vamos a generar un **Histograma** para visualizar la distribuci√≥n de los datos continuos.

```python
# Histograma para los datos continuos
plt.figure(figsize=(10, 6))
plt.hist(df_continuo['Datos Continuos'], bins='auto', color='skyblue', edgecolor='black')
plt.title('Histograma de Datos Continuos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostramos el histograma
plt.show()
```

El **Histograma** muestra la distribuci√≥n de los 100 datos continuos generados. Cada barra representa el n√∫mero de datos que caen dentro de un rango espec√≠fico de valores, lo que nos permite ver c√≥mo se distribuyen los datos, identificar la centralidad, la dispersi√≥n, as√≠ como la simetr√≠a o asimetr√≠a de la distribuci√≥n.

En resumen:

- **Gr√°fico Circular**: √ötil para mostrar proporciones o porcentajes entre categor√≠as. Ideal cuando los datos son categ√≥ricos y se quiere ver c√≥mo cada parte se compara con el todo.
  
- **Gr√°fico de Barras**: Excelente para comparar datos entre diferentes categor√≠as o ver cambios a lo largo del tiempo para un n√∫mero limitado de grupos.
  
- **Histogramas**: Indispensables para visualizar la distribuci√≥n de datos continuos, √∫til para entender la distribuci√≥n general, identificar la forma de la distribuci√≥n, y detectar posibles valores at√≠picos.

Cada uno de estos gr√°ficos tiene su lugar dependiendo de lo que se desee comunicar y del tipo de datos con los que se est√© trabajando.

## **1.3 Diagrama de caja y bigotes y diagrama de dispersi√≥n**

Los diagramas de caja y bigotes y los diagramas de dispersi√≥n son herramientas estad√≠sticas poderosas para la visualizaci√≥n de datos, cada uno con aplicaciones espec√≠ficas que nos permiten entender mejor las caracter√≠sticas y relaciones dentro de un conjunto de datos. A continuaci√≥n, exploraremos ambos tipos de gr√°ficos, sus usos, c√≥mo se construyen y proporcionaremos ejemplos pr√°cticos tanto aritm√©ticamente como mediante c√≥digo en Python.

**Diagrama de Caja y Bigotes (Box Plot)**

**¬øQu√© es y cu√°ndo se utiliza?**

Un diagrama de caja y bigotes resume la distribuci√≥n de un conjunto de datos cuantitativos mediante cuartiles, destacando la mediana, los cuartiles superior e inferior, y los valores at√≠picos. Es particularmente √∫til para visualizar la dispersi√≥n de los datos, la tendencia central y la presencia de valores at√≠picos.

**Mejor uso**: Comparar la distribuci√≥n de varios conjuntos de datos, identificar valores at√≠picos y entender la dispersi√≥n de los datos.

**Componentes clave del diagrama**:

- **Mediana (Q2 / 50¬∫ percentil)**: El valor medio del conjunto de datos.

- **Cuartil inferior (Q1 / 25¬∫ percentil)**: El valor medio entre el menor n√∫mero del conjunto de datos y su mediana.

- **Cuartil superior (Q3 / 75¬∫ percentil)**: El valor medio entre la mediana y el mayor n√∫mero del conjunto de datos.

- **Rango intercuart√≠lico (IQR)**: La diferencia entre el cuartil superior e inferior (Q3 - Q1).

- **Bigotes**: Extienden hasta el √∫ltimo dato dentro del rango `[Q1 - 1.5*IQR, Q3 + 1.5*IQR]`. Los datos fuera de este rango se consideran valores at√≠picos.

**Diagrama de Dispersi√≥n (Scatter Plot)**

**¬øQu√© es y cu√°ndo se utiliza?**

Un diagrama de dispersi√≥n muestra la relaci√≥n entre dos variables cuantitativas, representando cada punto de datos como un punto en el plano cartesiano. Es √∫til para identificar correlaciones, tendencias o patrones potenciales entre dos variables.

**Mejor uso**: Explorar relaciones entre dos variables, identificar correlaciones y detectar agrupaciones o patrones an√≥malos.

**Ejemplo Pr√°ctico**

Para ilustrar ambos tipos de diagramas, crearemos un DataFrame que contenga dos variables cuantitativas. Una variable se utilizar√° para el diagrama de caja y bigotes, representando la distribuci√≥n de una √∫nica variable. La segunda y tercera variables se utilizar√°n para el diagrama de dispersi√≥n, explorando la relaci√≥n entre ellas.

Crearemos este DataFrame y generaremos los respectivos gr√°ficos.

```python
# Re-importando las librer√≠as necesarias despu√©s del restablecimiento del entorno de ejecuci√≥n
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creaci√≥n del DataFrame para ejemplos
data_ejemplo = {
    'Variable A': np.random.normal(50, 10, 100),  # 100 datos distribuidos normalmente para el box plot
    'Variable B': np.random.uniform(20, 60, 100),  # 100 datos distribuidos uniformemente para el scatter plot
    'Variable C': np.random.uniform(10, 30, 100)   # Otra variable para el scatter plot
}
df_ejemplo = pd.DataFrame(data_ejemplo)

# Mostrar los primeros datos del DataFrame
df_ejemplo.head()
```

Tenemos un DataFrame con tres variables:

- **Variable A**: Datos distribuidos normalmente, que utilizaremos para el diagrama de caja y bigotes.

- **Variable B y Variable C**: Datos distribuidos uniformemente, que utilizaremos para crear el diagrama de dispersi√≥n.

Empezaremos generando el **Diagrama de Caja y Bigotes** para la Variable A, para visualizar su distribuci√≥n.

```python
# Diagrama de Caja y Bigotes para la Variable A
plt.figure(figsize=(8, 6))
plt.boxplot(df_ejemplo['Variable A'])
plt.title('Diagrama de Caja y Bigotes para la Variable A')
plt.ylabel('Valores')

# Mostramos el diagrama
plt.show()
```

El **Diagrama de Caja y Bigotes** para la Variable A muestra la distribuci√≥n de los datos. Los elementos clave que podemos observar incluyen la mediana (l√≠nea dentro de la caja), los cuartiles superior e inferior (los bordes de la caja), y los bigotes que se extienden desde la caja hasta los valores m√°ximos y m√≠nimos dentro del rango intercuart√≠lico ajustado. Los puntos fuera de los bigotes representan valores at√≠picos, que son valores que difieren significativamente de la mayor√≠a de los datos.

Ahora, generaremos un **Diagrama de Dispersi√≥n** utilizando las Variables B y C para explorar la relaci√≥n entre estas dos variables.

```python
# Diagrama de Dispersi√≥n entre Variable B y Variable C
plt.figure(figsize=(10, 6))
plt.scatter(df_ejemplo['Variable B'], df_ejemplo['Variable C'], color='skyblue', edgecolor='black')
plt.title('Diagrama de Dispersi√≥n entre Variable B y Variable C')
plt.xlabel('Variable B')
plt.ylabel('Variable C')

# Mostramos el diagrama
plt.show()
```

El **Diagrama de Dispersi√≥n** entre la Variable B y la Variable C muestra c√≥mo se relacionan estas dos variables entre s√≠. Cada punto en el gr√°fico representa una observaci√≥n en el conjunto de datos, con su posici√≥n en el eje $X$ dada por el valor de la Variable B y su posici√≥n en el eje $Y$ por el valor de la Variable C. Este tipo de gr√°fico es √∫til para identificar cualquier relaci√≥n o patr√≥n entre las dos variables, como correlaciones positivas o negativas, o la ausencia de una relaci√≥n clara.

A partir del gr√°fico, parece que no hay una relaci√≥n lineal clara entre la Variable B y la Variable C, lo cual es esperado dada la forma aleatoria en que generamos los datos. Los diagramas de dispersi√≥n son especialmente √∫tiles en an√°lisis exploratorios de datos para identificar relaciones entre variables que podr√≠an no ser evidentes a primera vista.

En resumen, el **diagrama de caja y bigotes** es excelente para visualizar la distribuci√≥n de un conjunto de datos, destacando la mediana, los cuartiles y los valores at√≠picos. Por otro lado, el **diagrama de dispersi√≥n** es fundamental para explorar y visualizar las relaciones entre dos variables continuas, ayudando a identificar patrones, tendencias o correlaciones.

## **1.4 Proporcionar contexto es clave para los an√°lisis estad√≠sticos**

Proporcionar contexto es esencial en el an√°lisis estad√≠stico porque los n√∫meros y datos brutos, por s√≠ solos, a menudo no cuentan la historia completa o pueden ser mal interpretados. El contexto ayuda a entender mejor la importancia de los datos, las comparaciones y las conclusiones que se derivan de ellos. Veamos algunas maneras en que el contexto es clave, con ejemplos y gr√°ficos para ilustrar estos puntos.

**Ejemplo 1: Tendencias a lo largo del tiempo**

**Contexto**: Cambios en las tendencias de ventas a lo largo del tiempo.

**Sin contexto**: Observamos un incremento en las ventas en marzo.

**Con contexto**: El incremento en las ventas en marzo se compara con los datos de los √∫ltimos 5 a√±os, donde t√≠picamente marzo es un mes lento. Este a√±o, el incremento coincide con una campa√±a de marketing lanzada a finales de febrero.

**Visualizaci√≥n**:
- Podr√≠amos usar un gr√°fico de l√≠neas para mostrar las ventas mensuales a lo largo de los √∫ltimos cinco a√±os, destacando el pico inusual de marzo.

**Ejemplo 2: Comparaci√≥n entre grupos**

**Contexto**: Comparaci√≥n de la efectividad de dos medicamentos.

**Sin contexto**: El Medicamento A tuvo un 70% de efectividad, y el Medicamento B, un 75%.

**Con contexto**: El Medicamento A se administr√≥ a una poblaci√≥n con casos m√°s severos de la enfermedad. Adem√°s, el Medicamento B tiene efectos secundarios significativos no presentes en el Medicamento A.

**Visualizaci√≥n**:
- Un diagrama de caja y bigotes podr√≠a mostrar la distribuci√≥n de la efectividad de ambos medicamentos, con notas sobre la severidad de la enfermedad y los efectos secundarios.

**Ejemplo 3: Densidad poblacional y √°rea geogr√°fica**

**Contexto**: Comparar la densidad poblacional de diferentes pa√≠ses.

**Sin contexto**: Pa√≠s A tiene una densidad poblacional mayor que Pa√≠s B.

**Con contexto**: Pa√≠s A es geogr√°ficamente mucho m√°s peque√±o que Pa√≠s B y tiene una concentraci√≥n urbana m√°s alta, lo que contribuye a su mayor densidad poblacional.

**Visualizaci√≥n**:
- Un mapa de calor o un gr√°fico de barras que muestre la densidad poblacional junto con el tama√±o geogr√°fico de los pa√≠ses puede proporcionar una imagen m√°s completa.

Ahora, generaremos gr√°ficos para ilustrar c√≥mo el contexto puede cambiar la interpretaci√≥n de los datos. Empezaremos con el **Ejemplo 1** sobre las tendencias de ventas a lo largo del tiempo, creando datos ficticios y un gr√°fico de l√≠neas.

```python
# Generaci√≥n de datos ficticios para el Ejemplo 1: Tendencias de ventas a lo largo del tiempo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas_2023 = [200, 180, 250, 220, 210]  # Ventas para el a√±o 2023
ventas_promedio_5_a√±os = [190, 175, 180, 200, 195]  # Promedio de ventas de los √∫ltimos 5 a√±os

# Creaci√≥n del gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(meses, ventas_2023, label='Ventas 2023', marker='o')
plt.plot(meses, ventas_promedio_5_a√±os, label='Promedio Ventas √öltimos 5 A√±os', linestyle='--', marker='x')
plt.title('Tendencia de Ventas Marzo 2023 vs. Promedio √öltimos 5 A√±os')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

El gr√°fico de l√≠neas muestra la tendencia de ventas para marzo de 2023 en comparaci√≥n con el promedio de ventas de los √∫ltimos 5 a√±os. Aqu√≠, el contexto proporcionado por el promedio de ventas de a√±os anteriores nos ayuda a entender que el pico en marzo de 2023 es inusual y significativo, especialmente si consideramos la informaci√≥n adicional de que marzo t√≠picamente es un mes lento y que hubo una campa√±a de marketing reciente. Este tipo de an√°lisis contextual puede ayudar a las empresas a evaluar el impacto de sus estrategias y a planificar futuras acciones.

Este ejemplo ilustra c√≥mo el contexto transforma la interpretaci√≥n de los datos brutos, permitiendo una comprensi√≥n m√°s profunda y matizada de las tendencias y eventos subyacentes. Al proporcionar contexto, no solo presentamos datos, sino que tambi√©n ofrecemos insights y valor agregado, facilitando la toma de decisiones informada basada en una comprensi√≥n completa de las circunstancias.

## **1.5 Errores al visualizar la informaci√≥n**

Visualizar la informaci√≥n de manera incorrecta o enga√±osa puede llevar a interpretaciones err√≥neas, decisiones mal informadas y, en el peor de los casos, a conclusiones completamente falsas. Algunos errores comunes incluyen el uso inapropiado de tipos de gr√°ficos, la manipulaci√≥n del eje de las coordenadas, la omisi√≥n de datos relevantes y la elecci√≥n de escalas inadecuadas. Vamos a explorar algunos de estos errores con ejemplos y gr√°ficos.

**Ejemplo 1: Manipulaci√≥n del Eje $Y$**

**Problema**: Comenzar el eje $Y$ en un valor distinto de cero en un gr√°fico de barras o de l√≠neas puede exagerar o minimizar visualmente las diferencias entre las barras o los puntos de datos.

**Consecuencia**: Se pueden malinterpretar las magnitudes de las diferencias o cambios.

**Soluci√≥n**: Siempre iniciar el eje $Y$ en cero para representaciones precisas, a menos que haya una buena raz√≥n para hacer lo contrario y se aclare al espectador.

**Ejemplo 2: Uso Inadecuado de Gr√°ficos Circulares**

**Problema**: Usar gr√°ficos circulares cuando hay muchas categor√≠as o las diferencias entre las partes son m√≠nimas.

**Consecuencia**: Hace dif√≠cil la comparaci√≥n directa entre segmentos, llevando a interpretaciones err√≥neas.

**Soluci√≥n**: Optar por gr√°ficos de barras o gr√°ficos de l√≠neas, que permiten comparaciones m√°s claras.

**Ejemplo 3: Escalas no Lineales sin Indicaci√≥n**

**Problema**: Utilizar escalas logar√≠tmicas o no lineales sin indicarlo claramente.

**Consecuencia**: Puede confundir a los espectadores, haciendo que interpreten incorrectamente las relaciones entre los datos.

**Soluci√≥n**: Marcar claramente cuando se usen escalas no est√°ndar y proporcionar una breve explicaci√≥n si es posible.

**Visualizaciones**

A continuaci√≥n, generaremos visualizaciones para ilustrar el **Ejemplo 1 (Manipulaci√≥n del Eje Y)** y el **Ejemplo 2 (Uso Inadecuado de Gr√°ficos Circulares)**, utilizando datos ficticios para mostrar c√≥mo estos errores pueden afectar la interpretaci√≥n de la informaci√≥n.

```python
# Ejemplo 1: Manipulaci√≥n del Eje Y
ventas = [100, 105, 103, 110, 115]  # Datos de ventas ficticios
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May']

# Gr√°fico Correcto
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(meses, ventas, color='skyblue')
plt.title('Eje Y Comenzando en 0')
plt.ylim(0)  # Asegurando que el eje Y comience en 0

# Gr√°fico Incorrecto
plt.subplot(1, 2, 2)
plt.bar(meses, ventas, color='lightcoral')
plt.title('Eje Y No Comenzando en 0')
plt.ylim(100)  # Comenzando el eje Y en 100 para exagerar diferencias

plt.tight_layout()
plt.show()

# Ejemplo 2: Uso Inadecuado de Gr√°ficos Circulares
porcentajes = [30, 25, 20, 15, 10]  # Datos de porcentajes ficticios
categorias = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E']

plt.figure(figsize=(5, 5))
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Gr√°fico Circular con Muchas Categor√≠as')
plt.show()
```

En las visualizaciones presentadas, podemos ver claramente c√≥mo los errores en la visualizaci√≥n de la informaci√≥n pueden llevar a interpretaciones incorrectas:

**Ejemplo 1: Manipulaci√≥n del Eje Y**

- **Gr√°fico Correcto (Eje Y Comenzando en 0)**: Muestra una representaci√≥n precisa de las ventas a lo largo de los meses, donde las diferencias entre las barras son f√°ciles de comparar y entender.
- **Gr√°fico Incorrecto (Eje Y No Comenzando en 0)**: Exagera las diferencias entre las ventas mensuales. Aunque las variaciones reales son menores, el gr√°fico sugiere cambios dram√°ticos, lo cual puede llevar a interpretaciones err√≥neas sobre la performance.

**Ejemplo 2: Uso Inadecuado de Gr√°ficos Circulares**

- **Gr√°fico Circular con Muchas Categor√≠as**: Aunque los gr√°ficos circulares son √∫tiles para mostrar proporciones, en este caso, la presencia de muchas categor√≠as y la proximidad de sus valores hacen que sea dif√≠cil discernir diferencias exactas y comparar eficazmente las secciones.

Estos ejemplos destacan la importancia de elegir el tipo correcto de gr√°fico y configurar los ejes $y$ escalas de manera adecuada para evitar interpretaciones incorrectas. Al visualizar datos, es crucial ser preciso y claro, proporcionando el contexto adecuado y evitando manipulaciones que puedan llevar a conclusiones err√≥neas. La integridad en la visualizaci√≥n de datos no solo mejora la comprensi√≥n sino que tambi√©n fomenta la confianza en el an√°lisis presentado.

## **1.6 Media y mediana**

**Datos No Agrupados**

- **Media (Promedio)**: Es la suma de todos los valores en el conjunto de datos dividida entre el n√∫mero total de valores.

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

donde $(x_i)$ son los valores individuales y $(n)$ es el n√∫mero total de valores.

- **Mediana**: Es el valor que se encuentra en el medio del conjunto de datos una vez que ha sido ordenado. Si hay un n√∫mero impar de observaciones, es el valor medio. Si hay un n√∫mero par, es el promedio de los dos valores medios.

  - Para un n√∫mero impar de observaciones:

  $\text{Mediana} = x_{(\frac{n+1}{2})}$

  - Para un n√∫mero par de observaciones:

  $\text{Mediana} = \frac{x_{(\frac{n}{2})} + x_{(\frac{n}{2} + 1)}}{2}$

**Datos Agrupados**

Los datos agrupados est√°n organizados en clases o intervalos.

- **Media de Datos Agrupados**:

$\text{Media} = \frac{\sum_{i=1}^{k} f_i \cdot x_i}{\sum_{i=1}^{k} f_i}$

donde $(f_i)$ es la frecuencia de cada intervalo, $(x_i)$ es el punto medio del intervalo, y $(k)$ es el n√∫mero total de intervalos.

- **Mediana de Datos Agrupados**:

 Se encuentra primero determinando el intervalo mediano (el intervalo que contiene a la mediana). La f√≥rmula para calcular la mediana en datos agrupados es:

$\text{Mediana} = L + \left(\frac{\frac{N}{2} - F}{f}\right)W$

donde $(L)$ es el l√≠mite inferior del intervalo mediano, $(N)$ es el n√∫mero total de datos, $(F)$ es la frecuencia acumulada antes del intervalo mediano, $(f)$ es la frecuencia del intervalo mediano, y $(W)$ es la amplitud del intervalo.

Ahora, procederemos a crear un DataFrame para datos no agrupados, calcular y visualizar la media y la mediana. Luego, agruparemos esos datos y haremos lo mismo para los datos agrupados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generaci√≥n de datos no agrupados
np.random.seed(0)  # Para reproducibilidad
datos_no_agrupados = np.random.normal(50, 15, 100)  # 100 datos distribuidos normalmente

# Creaci√≥n de un DataFrame para datos no agrupados
df_no_agrupados = pd.DataFrame(datos_no_agrupados, columns=['Valores'])

# C√°lculo de la media y la mediana para datos no agrupados
media_no_agrupados = np.mean(datos_no_agrupados)
mediana_no_agrupados = np.median(datos_no_agrupados)

media_no_agrupados, mediana_no_agrupados
```

Para nuestro conjunto de **datos no agrupados**, la media es aproximadamente 50.90 y la mediana es aproximadamente 51.41, lo que indica una distribuci√≥n relativamente sim√©trica alrededor del centro.

A continuaci√≥n, agruparemos estos datos y realizaremos los c√°lculos para datos agrupados. Para ello, dividiremos los datos en intervalos y calcularemos la media y la mediana utilizando las f√≥rmulas apropiadas para datos agrupados. Adem√°s, visualizaremos los datos utilizando histogramas y gr√°ficos para cada caso.

```python
# Agrupaci√≥n de los datos en intervalos
bins = np.arange(0, 101, 10)
df_no_agrupados['Intervalo'] = pd.cut(df_no_agrupados['Valores'], bins=bins, include_lowest=True)

# Datos agrupados: calculando la frecuencia y el punto medio de cada intervalo
datos_agrupados = df_no_agrupados.groupby('Intervalo')['Valores'].agg(['count', 'mean']).reset_index()
datos_agrupados.columns = ['Intervalo', 'Frecuencia', 'Punto Medio']

# C√°lculo de la media para datos agrupados
media_agrupados = np.sum(datos_agrupados['Punto Medio'] * datos_agrupados['Frecuencia']) / np.sum(datos_agrupados['Frecuencia'])

# Preparaci√≥n para calcular la mediana de datos agrupados
N = np.sum(datos_agrupados['Frecuencia'])
datos_agrupados['Frecuencia Acumulada'] = datos_agrupados['Frecuencia'].cumsum()
intervalo_mediana = datos_agrupados[datos_agrupados['Frecuencia Acumulada'] >= N/2].iloc[0]
L = intervalo_mediana['Intervalo'].left
F = datos_agrupados[datos_agrupados['Intervalo'] == intervalo_mediana['Intervalo']]['Frecuencia Acumulada'].shift(1).fillna(0).iloc[0]
f = intervalo_mediana['Frecuencia']
W = intervalo_mediana['Intervalo'].right - intervalo_mediana['Intervalo'].left

# C√°lculo de la mediana para datos agrupados
mediana_agrupados = L + ((N/2 - F) / f) * W

media_agrupados, mediana_agrupados
```

Para los **datos agrupados**, la media calculada es aproximadamente 50.90, similar a la obtenida para los datos no agrupados, lo que refleja la consistencia de la media como medida de tendencia central. Sin embargo, el c√°lculo de la mediana nos da aproximadamente 67.86, lo cual difiere del c√°lculo directo de los datos no agrupados. Esta diferencia se debe a la naturaleza de los datos agrupados y la estimaci√≥n involucrada en el c√°lculo de la mediana a partir de intervalos.

Ahora, visualizaremos ambos conjuntos de datos para comparar c√≥mo se ven los datos no agrupados y agrupados.

**Visualizaci√≥n de Datos No Agrupados: Histograma**

```python
plt.figure(figsize=(10, 6))
plt.hist(df_no_agrupados['Valores'], bins=bins, edgecolor='black', alpha=0.7)
plt.axvline(media_no_agrupados, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media_no_agrupados:.2f}')
plt.axvline(mediana_no_agrupados, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana: {mediana_no_agrupados:.2f}')
plt.title('Histograma de Datos No Agrupados')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

El histograma muestra la distribuci√≥n de los **datos no agrupados** con la media y la mediana indicadas. La cercan√≠a de la media y la mediana sugiere una distribuci√≥n sim√©trica de los datos alrededor del centro.

**Visualizaci√≥n de Datos Agrupados: Histograma**

Para visualizar los datos agrupados, usaremos el histograma de los datos originales pero con las marcas de la media y la mediana calculadas para los datos agrupados, para comparar.

```python
plt.figure(figsize=(10, 6))
plt.hist(df_no_agrupados['Valores'], bins=bins, edgecolor='black', alpha=0.7)
plt.axvline(media_agrupados, color='red', linestyle='dashed', linewidth=1, label=f'Media Agrupada: {media_agrupados:.2f}')
plt.axvline(mediana_agrupados, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana Agrupada: {mediana_agrupados:.2f}')
plt.title('Histograma de Datos Agrupados (con Media y Mediana Agrupadas)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

En el histograma de datos agrupados, se representan las mismas observaciones pero con las l√≠neas que indican la media y la mediana calculadas a partir de los datos agrupados. Aqu√≠, la l√≠nea de la mediana parece estar m√°s alejada de la media en comparaci√≥n con el histograma de datos no agrupados, reflejando la diferencia en el c√°lculo de la mediana para datos agrupados.

## **1.7 Percentiles, resumen de cinco n√∫meros y desviaci√≥n t√≠pica**

**Percentiles**

Los **percentiles** son medidas que dividen un conjunto de datos ordenado en cien partes iguales. Son √∫tiles para entender la distribuci√≥n de los datos. Por ejemplo, el 25¬∫ percentil (tambi√©n conocido como primer cuartil) divide el 25% de los datos m√°s bajos del 75% m√°s alto, mientras que el 50¬∫ percentil (la mediana) divide el conjunto de datos en dos mitades.

F√≥rmula general para calcular el percentil $(P_k)$ en un conjunto de datos ordenado:

$P_k = \left( \frac{k}{100} \right) \times (N + 1)$

donde:

- $(P_k)$ es el k-√©simo percentil,
- $(k)$ es el percentil deseado,
- $(N)$ es el n√∫mero total de observaciones en el conjunto de datos.

**Resumen de Cinco N√∫meros**

El **resumen de cinco n√∫meros** consiste en cinco valores que proporcionan una visi√≥n completa de la distribuci√≥n de un conjunto de datos:

1. **M√≠nimo**: El valor m√°s bajo.

2. **Primer cuartil (Q1 / 25¬∫ percentil)**: Divide el 25% de los datos m√°s bajos del resto.

3. **Mediana (Q2 / 50¬∫ percentil)**: El valor medio.

4. **Tercer cuartil (Q3 / 75¬∫ percentil)**: Divide el 75% de los datos m√°s bajos del 25% m√°s alto.

5. **M√°ximo**: El valor m√°s alto.

**Desviaci√≥n Est√°ndar**

La **desviaci√≥n est√°ndar** es una medida que indica cu√°nto se desv√≠an los valores en un conjunto de datos respecto a la media. Muestra la dispersi√≥n o variabilidad de los datos.

F√≥rmula para la desviaci√≥n est√°ndar $(s)$ de una muestra:

$s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N - 1}}$

donde:

- $(x_i)$ son los valores individuales,
- $(\bar{x})$ es la media de los datos,
- $(N)$ es el n√∫mero total de observaciones.

Para ilustrar estos conceptos, vamos a generar un DataFrame con datos aleatorios y calcular el resumen de cinco n√∫meros, varios percentiles espec√≠ficos y la desviaci√≥n est√°ndar. Adem√°s, visualizaremos estos conceptos con gr√°ficos apropiados.

```python
# Generaci√≥n de datos aleatorios para el ejemplo
np.random.seed(42)  # Para reproducibilidad
datos_ejemplo = np.random.normal(loc=50, scale=10, size=100)  # 100 datos con media 50 y desv. est√°ndar 10

# Creando un DataFrame para los datos
df_ejemplo = pd.DataFrame(datos_ejemplo, columns=['Datos'])

# C√°lculo del resumen de cinco n√∫meros y percentiles espec√≠ficos
resumen_cinco_numeros = df_ejemplo['Datos'].describe()[['min', '25%', '50%', '75%', 'max']]
percentil_10 = np.percentile(datos_ejemplo, 10)  # 10¬∫ percentil
percentil_90 = np.percentile(datos_ejemplo, 90)  # 90¬∫ percentil

# C√°lculo de la desviaci√≥n est√°ndar
desviacion_estandar = df_ejemplo['Datos'].std()

resumen_cinco_numeros, percentil_10, percentil_90, desviacion_estandar
```

Para nuestro conjunto de datos generado, tenemos el siguiente an√°lisis:

- **Resumen de Cinco N√∫meros**:

  - M√≠nimo: ~23.80

  - Primer Cuartil (Q1 / 25%): ~43.99

  - Mediana (Q2 / 50%): ~48.73

  - Tercer Cuartil (Q3 / 75%): ~54.06

  - M√°ximo: ~68.52

- **Percentiles Espec√≠ficos**:

  - 10¬∫ Percentil: ~36.63
  
  - 90¬∫ Percentil: ~60.06

- **Desviaci√≥n Est√°ndar**: ~9.08

Estos resultados nos dan una visi√≥n detallada de la distribuci√≥n de los datos: el resumen de cinco n√∫meros nos muestra la distribuci√≥n de los datos desde el valor m√°s bajo hasta el m√°s alto, pasando por los cuartiles que dividen el conjunto en cuatro partes iguales. Los percentiles espec√≠ficos ofrecen puntos de referencia adicionales dentro de la distribuci√≥n, y la desviaci√≥n est√°ndar nos dice cu√°nto se desv√≠an los valores de la media en promedio.

Para visualizar estos conceptos, generaremos un **diagrama de caja (box plot)**, que ilustra el resumen de cinco n√∫meros y los percentiles, y un **histograma**, que nos ayudar√° a ver la dispersi√≥n y forma de la distribuci√≥n de los datos.

```python
# Diagrama de caja (box plot) para visualizar el resumen de cinco n√∫meros
plt.figure(figsize=(10, 6))
plt.boxplot(df_ejemplo['Datos'], vert=False, patch_artist=True)
plt.title('Diagrama de Caja')
plt.xlabel('Datos')

# Mostrando el diagrama de caja
plt.show()

# Histograma para visualizar la distribuci√≥n de los datos
plt.figure(figsize=(10, 6))
plt.hist(df_ejemplo['Datos'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de los Datos')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')

# Mostrando el histograma
plt.show()
```

En las visualizaciones proporcionadas, podemos observar lo siguiente:

- **Diagrama de Caja**: Ilustra claramente el resumen de cinco n√∫meros de nuestro conjunto de datos. Podemos ver el rango de los datos a trav√©s del m√≠nimo y m√°ximo, la distribuci√≥n interna a trav√©s de los cuartiles Q1, Q2 (mediana), y Q3, y cualquier dato at√≠pico que se desv√≠e significativamente de los cuartiles.

- **Histograma**: Muestra la distribuci√≥n y dispersi√≥n de los datos, con la desviaci√≥n est√°ndar reflejada en la amplitud de la distribuci√≥n. La forma del histograma tambi√©n nos da una idea de la simetr√≠a de los datos respecto a la media.

Estas herramientas estad√≠sticas y gr√°ficas proporcionan una comprensi√≥n profunda y detallada de la distribuci√≥n de un conjunto de datos, permitiendo an√°lisis m√°s informados y decisiones basadas en datos.

## **1.8 Introducci√≥n y estad√≠sticas descriptivas para explorar los datos**

La **introducci√≥n a las estad√≠sticas descriptivas** para explorar datos es un paso crucial en el an√°lisis de datos, permitiendo resumir y comprender las caracter√≠sticas principales de un conjunto de datos antes de proceder a an√°lisis m√°s complejos. Las estad√≠sticas descriptivas proporcionan informaci√≥n sobre la tendencia central, la dispersi√≥n y la forma de la distribuci√≥n de los datos.

**Componentes Clave de las Estad√≠sticas Descriptivas**

**Medidas de Tendencia Central**:

   - **Media** (Promedio): Suma de todos los valores dividida por el n√∫mero de valores.

   - **Mediana**: El valor central en un conjunto de datos ordenados.

   - **Moda**: El valor que aparece con mayor frecuencia en un conjunto de datos.

**Medidas de Dispersi√≥n**:

   - **Varianza**: El promedio de los cuadrados de las desviaciones de la media.

   - **Desviaci√≥n Est√°ndar**: La ra√≠z cuadrada de la varianza, mide la dispersi√≥n de los datos respecto a la media.

   - **Rango**: Diferencia entre el m√°ximo y el m√≠nimo valor en el conjunto de datos.

   - **Cuartiles**: Valores que dividen los datos en cuatro partes iguales.

**Forma de la Distribuci√≥n**:

   - **Sesgo (Skewness)**: Medida de la asimetr√≠a de la distribuci√≥n de los datos.

   - **Curtosis**: Medida del pico de la distribuci√≥n de los datos.

**Ejemplo Pr√°ctico: An√°lisis de Datos de Altura**

Supongamos que tenemos un conjunto de datos que representa las alturas (en cm) de un grupo de estudiantes:

$\text{Alturas} = [160, 165, 167, 165, 170, 172, 168, 162, 174, 168, 165, 173]$

**C√°lculo de Estad√≠sticas Descriptivas**

Calcularemos las medidas de tendencia central, dispersi√≥n y exploraremos la forma de la distribuci√≥n de estos datos.

**Implementaci√≥n en Python**

Vamos a utilizar Python para calcular estas estad√≠sticas y visualizar los datos:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de alturas
alturas = np.array([160, 165, 167, 165, 170, 172, 168, 162, 174, 168, 165, 173])

# C√°lculos de estad√≠sticas descriptivas
media = np.mean(alturas)
mediana = np.median(alturas)
moda = np.bincount(alturas).argmax()  # Simplificaci√≥n para demostraci√≥n
varianza = np.var(alturas)
desviacion_std = np.std(alturas)
rango = np.ptp(alturas)  # Peak to peak

# Visualizaci√≥n de los datos
plt.figure(figsize=(10, 6))
sns.histplot(alturas, kde=True, color='blue', bins=5)  # Histograma con estimaci√≥n de densidad
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.title('Distribuci√≥n de Alturas de Estudiantes')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda} (M√°s com√∫n)")
print(f"Varianza: {varianza:.2f}")
print(f"Desviaci√≥n Est√°ndar: {desviacion_std:.2f}")
print(f"Rango: {rango}")
```

**Interpretaci√≥n**

- **Histograma**: Muestra la distribuci√≥n de las alturas. La l√≠nea roja representa la media y la verde la mediana, proporcionando una visualizaci√≥n directa de la tendencia central.

- **Media y Mediana**: Valores centrales que resumen la tendencia de las alturas. La cercan√≠a de la media y la mediana sugiere una distribuci√≥n relativamente sim√©trica.

- **Moda**: La altura m√°s frecuente.

- **Varianza y Desviaci√≥n Est√°ndar**: Indican que las alturas est√°n dispersas alrededor de la media por aproximadamente 4.33 cm en promedio.

Este an√°lisis de estad√≠sticas descriptivas provee una comprensi√≥n

 fundamental de los datos antes de proceder con an√°lisis m√°s avanzados, como correlaci√≥n o modelos predictivos.
